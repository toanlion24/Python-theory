````python
# VÃ­ dá»¥: Táº¡o module mathutils (simulate trong cell)

# ========== mathutils.py ==========
"""
Math Utilities Module.

Provides common mathematical operations and utilities.

Example:
    >>> import mathutils
    >>> mathutils.add(2, 3)
    5
"""

__version__ = "1.0.0"
__author__ = "Python Developer"

# Constants
PI = 3.14159265359
E = 2.71828182846
GOLDEN_RATIO = 1.61803398875

# Private helper
def _validate_numbers(*args):
    """Validate all arguments are numbers."""
    for arg in args:
        if not isinstance(arg, (int, float)):
            raise TypeError(f"Expected number, got {type(arg).__name__}")

# Public functions
def add(a, b):
    """Add two numbers."""
    _validate_numbers(a, b)
    return a + b

def subtract(a, b):
    """Subtract b from a."""
    _validate_numbers(a, b)
    return a - b

def multiply(a, b):
    """Multiply two numbers."""
    _validate_numbers(a, b)
    return a * b

def divide(a, b):
    """Divide a by b."""
    _validate_numbers(a, b)
    if b == 0:
        raise ValueError("Cannot divide by zero")
    return a / b

def circle_area(radius):
    """Calculate circle area."""
    _validate_numbers(radius)
    if radius < 0:
        raise ValueError("Radius cannot be negative")
    return PI * radius ** 2

def factorial(n):
    """Calculate factorial of n."""
    if not isinstance(n, int) or n < 0:
        raise ValueError("n must be non-negative integer")
    if n <= 1:
        return 1
    return n * factorial(n - 1)

class Calculator:
    """Simple calculator class."""
    
    def __init__(self):
        self.result = 0
        self.history = []
    
    def add(self, x):
        self.result += x
        self.history.append(f"+ {x}")
        return self
    
    def subtract(self, x):
        self.result -= x
        self.history.append(f"- {x}")
        return self
    
    def clear(self):
        self.result = 0
        self.history = []
        return self
    
    def __str__(self):
        return f"Calculator(result={self.result})"

# ========== End of module ==========

# Test (giáº£ láº­p import)
print("Custom Module Demo:")
print("=" * 50)

print(f"\n__version__ = {__version__}")
print(f"PI = {PI}")
print(f"add(5, 3) = {add(5, 3)}")
print(f"circle_area(5) = {circle_area(5):.2f}")
print(f"factorial(5) = {factorial(5)}")

calc = Calculator()
calc.add(10).subtract(3).add(5)
print(f"\nCalculator: {calc}")
print(f"History: {calc.history}")
````

````python
# Pattern phá»• biáº¿n cho CLI tools

# ========== mytool.py ==========
"""
Command-line tool example.

Usage:
    python mytool.py --name John --greeting hello
"""

import sys

def parse_args(args):
    """Parse command line arguments."""
    # Simplified arg parsing
    result = {}
    i = 0
    while i < len(args):
        if args[i].startswith('--'):
            key = args[i][2:]
            value = args[i + 1] if i + 1 < len(args) else None
            result[key] = value
            i += 2
        else:
            i += 1
    return result

def greet(name, greeting="hello"):
    """Generate greeting message."""
    greetings = {
        "hello": f"Hello, {name}!",
        "goodbye": f"Goodbye, {name}!",
        "morning": f"Good morning, {name}!"
    }
    return greetings.get(greeting, f"Hi, {name}!")

def main():
    """Main entry point."""
    # Simulate command line args
    # In real use: args = parse_args(sys.argv[1:])
    args = {'name': 'Python Developer', 'greeting': 'hello'}
    
    name = args.get('name', 'World')
    greeting = args.get('greeting', 'hello')
    
    print(greet(name, greeting))

if __name__ == "__main__":
    main()
````

## 6. `__init__.py`

### 6.1 Vai trÃ² cá»§a `__init__.py`

1. **ÄÃ¡nh dáº¥u directory lÃ  package** (Python < 3.3 báº¯t buá»™c)
2. **Package initialization code** - cháº¡y khi import package
3. **Control exports** - Ä‘á»‹nh nghÄ©a `__all__`
4. **Simplify imports** - re-export items

### 6.2 CÃ¡c pattern phá»• biáº¿n

````python
# Pattern 1: Empty __init__.py
# - ÄÆ¡n giáº£n nháº¥t
# - Chá»‰ Ä‘Ã¡nh dáº¥u lÃ  package

# mypackage/__init__.py
# (empty file)

# Usage:
# from mypackage.module1 import something

print("Pattern 1: Empty __init__.py")
print("  - Simplest approach")
print("  - Must use full import path")
````

````python
# Pattern 2: Re-export items
# - Expose commonly used items at package level
# - Simpler imports for users

# ========== mypackage/__init__.py ==========
"""
My Package - A collection of utilities.

Usage:
    from mypackage import Calculator, add, PI
"""

__version__ = "1.0.0"
__author__ = "Python Developer"

# Re-export from submodules
# from .module1 import Calculator, add
# from .module2 import PI, E
# from .utils import format_currency

print("Pattern 2: Re-export items")
print("  - from mypackage import Calculator")
print("  - Instead of: from mypackage.module1 import Calculator")
````

````python
# Pattern 3: __all__ - Control what gets exported

# ========== mypackage/__init__.py ==========
"""
My Package with controlled exports.
"""

# Public API
__all__ = [
    'Calculator',
    'add',
    'subtract',
    'PI',
]

# from .module1 import Calculator, add, subtract, _internal_func
# from .module2 import PI, E, _private_constant

# _internal_func vÃ  _private_constant khÃ´ng cÃ³ trong __all__
# Sáº½ khÃ´ng Ä‘Æ°á»£c export khi: from mypackage import *

print("Pattern 3: __all__ for controlled exports")
print("  - Only items in __all__ are exported with 'from pkg import *'")
print("  - Internal items stay hidden")
````

````python
# Pattern 4: Lazy imports (for large packages)

# ========== mypackage/__init__.py ==========
"""
Large package with lazy imports for faster startup.
"""

def __getattr__(name):
    """
    Lazy import - only import when accessed.
    Improves import time for large packages.
    """
    if name == "Calculator":
        # from .module1 import Calculator
        # return Calculator
        print(f"  Lazy loading {name}...")
        return type('Calculator', (), {})
    
    if name == "HeavyModule":
        # from .heavy import HeavyModule
        # return HeavyModule
        print(f"  Lazy loading {name}...")
        return type('HeavyModule', (), {})
    
    raise AttributeError(f"module {__name__!r} has no attribute {name!r}")

print("Pattern 4: Lazy imports")
print("  - Only load modules when accessed")
print("  - Better startup performance")
````

### 9.5 collections - Specialized Containers

````python
from collections import Counter, defaultdict, namedtuple, deque, OrderedDict

print("collections Module:")
print("=" * 50)

# Counter
print("\n1. Counter:")
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
counter = Counter(words)
print(f"   Counter({words[:3]}...): {counter}")
print(f"   most_common(2): {counter.most_common(2)}")

# defaultdict
print("\n2. defaultdict:")
dd = defaultdict(list)
dd['fruits'].append('apple')
dd['fruits'].append('banana')
dd['vegetables'].append('carrot')
print(f"   defaultdict(list): {dict(dd)}")

# namedtuple
print("\n3. namedtuple:")
Point = namedtuple('Point', ['x', 'y'])
p = Point(3, 4)
print(f"   Point(3, 4): {p}")
print(f"   p.x = {p.x}, p.y = {p.y}")

# deque (double-ended queue)
print("\n4. deque:")
dq = deque([1, 2, 3], maxlen=5)
dq.append(4)
dq.appendleft(0)
print(f"   deque after append/appendleft: {dq}")
````

### 9.6 re - Regular Expressions

````python
import re

print("re Module (Regular Expressions):")
print("=" * 50)

text = "Contact us at: support@example.com or sales@company.org"

# Find all emails
email_pattern = r'[\w.-]+@[\w.-]+\.\w+'
emails = re.findall(email_pattern, text)
print(f"\nFind all emails:")
print(f"  Text: {text}")
print(f"  Found: {emails}")

# Search
match = re.search(r'\d+', "Order ID: 12345")
if match:
    print(f"\nSearch for digits: {match.group()}")

# Replace
result = re.sub(r'\s+', ' ', "Hello    World   Python")
print(f"\nReplace multiple spaces: '{result}'")

# Match
phone_pattern = r'^\+?\d{10,12}$'
print(f"\nValidate phone number:")
print(f"  '+1234567890': {bool(re.match(phone_pattern, '+1234567890'))}")
print(f"  'abc123': {bool(re.match(phone_pattern, 'abc123'))}")
````

## 12. Best Practices

### 12.1 Import Best Practices

| âœ… Do | âŒ Don't |
|-------|----------|
| Group imports (stdlib, third-party, local) | Mix import groups |
| Use absolute imports | Overuse relative imports |
| Import specific items | `from module import *` |
| One import per line (PEP 8) | Multiple imports on one line |
| Alphabetize imports | Random order |

````python
# âœ… GOOD Import Style

example_good = """
# ========== good_style.py ==========

# Standard library imports (alphabetized)
import json
import os
import sys
from datetime import datetime
from pathlib import Path

# Third-party imports (alphabetized)
import numpy as np
import pandas as pd
from flask import Flask, request

# Local application imports
from myapp.config import settings
from myapp.models import User
from myapp.utils import validate_email
"""

print("âœ… Good Import Style:")
print(example_good)
````
