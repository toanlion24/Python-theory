#### VÃ­ dá»¥: DÃ¹ng Inheritance (âœ… Good)

````python
# âœ… DÃ™NG Inheritance - DRY (Don't Repeat Yourself)

class Animal:
    """Parent class chá»©a cÃ¡c thuá»™c tÃ­nh/methods chung."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        return f"{self.name} is eating"
    
    def sleep(self):
        return f"{self.name} is sleeping"
    
    def info(self):
        return f"{self.name} is {self.age} years old"

class Dog(Animal):  # Káº¿ thá»«a tá»« Animal
    """Dog class - chá»‰ cáº§n Ä‘á»‹nh nghÄ©a method riÃªng."""
    
    def bark(self):
        return f"{self.name} says: Woof!"

class Cat(Animal):  # Káº¿ thá»«a tá»« Animal
    """Cat class - chá»‰ cáº§n Ä‘á»‹nh nghÄ©a method riÃªng."""
    
    def meow(self):
        return f"{self.name} says: Meow!"

class Bird(Animal):  # Káº¿ thá»«a tá»« Animal
    """Bird class - chá»‰ cáº§n Ä‘á»‹nh nghÄ©a method riÃªng."""
    
    def fly(self):
        return f"{self.name} is flying"

# Sá»­ dá»¥ng
dog = Dog("Buddy", 3)
cat = Cat("Whiskers", 2)
bird = Bird("Tweety", 1)

# Táº¥t cáº£ Ä‘á»u cÃ³ cÃ¡c method tá»« Animal
print(dog.info())   # Inherited
print(dog.eat())    # Inherited
print(dog.bark())   # Own method

print()
print(cat.info())   # Inherited
print(cat.meow())   # Own method

print()
print(bird.sleep()) # Inherited
print(bird.fly())   # Own method
````

### 2.2 VÃ­ dá»¥ chi tiáº¿t

````python
# Parent Class
class Vehicle:
    """Base class Ä‘áº¡i diá»‡n cho phÆ°Æ¡ng tiá»‡n giao thÃ´ng."""
    
    # Class attribute
    vehicle_count = 0
    
    def __init__(self, brand, model, year):
        """Constructor cá»§a Vehicle."""
        self.brand = brand
        self.model = model
        self.year = year
        self.is_running = False
        Vehicle.vehicle_count += 1
    
    def start(self):
        """Khá»Ÿi Ä‘á»™ng xe."""
        self.is_running = True
        return f"{self.brand} {self.model} is starting..."
    
    def stop(self):
        """Dá»«ng xe."""
        self.is_running = False
        return f"{self.brand} {self.model} has stopped."
    
    def info(self):
        """ThÃ´ng tin xe."""
        status = "Running" if self.is_running else "Stopped"
        return f"{self.year} {self.brand} {self.model} [{status}]"

# Child Class
class Car(Vehicle):
    """Car class káº¿ thá»«a tá»« Vehicle."""
    
    def __init__(self, brand, model, year, num_doors, fuel_type):
        """
        Constructor cá»§a Car.
        
        Args:
            brand, model, year: Tá»« parent class
            num_doors: Sá»‘ cá»­a (riÃªng cá»§a Car)
            fuel_type: Loáº¡i nhiÃªn liá»‡u (riÃªng cá»§a Car)
        """
        # Gá»i constructor cá»§a parent class
        super().__init__(brand, model, year)
        
        # Thuá»™c tÃ­nh riÃªng cá»§a Car
        self.num_doors = num_doors
        self.fuel_type = fuel_type
    
    # Method riÃªng cá»§a Car
    def honk(self):
        """Báº¥m cÃ²i."""
        return "Beep beep! ğŸš—"
    
    def open_trunk(self):
        """Má»Ÿ cá»‘p xe."""
        return "Trunk is now open."
    
    # Override method tá»« parent
    def info(self):
        """Override info() Ä‘á»ƒ thÃªm thÃ´ng tin chi tiáº¿t."""
        parent_info = super().info()  # Gá»i method cá»§a parent
        return f"{parent_info} | {self.num_doors} doors | Fuel: {self.fuel_type}"

# Sá»­ dá»¥ng
print("=" * 60)
print("VEHICLE CLASS (Parent)")
print("=" * 60)

vehicle = Vehicle("Generic", "Vehicle", 2020)
print(vehicle.info())
print(vehicle.start())
print(vehicle.info())

print("\n" + "=" * 60)
print("CAR CLASS (Child)")
print("=" * 60)

car = Car("Toyota", "Camry", 2023, 4, "Gasoline")
print(car.info())       # Overridden method
print(car.start())      # Inherited method
print(car.honk())       # Own method
print(car.open_trunk()) # Own method
print(car.info())       # Overridden method

print(f"\nTotal vehicles created: {Vehicle.vehicle_count}")
````

### 2.3 Child class cÃ³ thá»ƒ:

1. **Káº¿ thá»«a** táº¥t cáº£ attributes vÃ  methods tá»« parent
2. **ThÃªm** attributes vÃ  methods má»›i
3. **Override** (ghi Ä‘Ã¨) methods cá»§a parent
4. **Má»Ÿ rá»™ng** methods cá»§a parent báº±ng `super()`

````python
# Minh há»a cÃ¡c kháº£ nÄƒng cá»§a Child class

class Parent:
    def __init__(self):
        self.parent_attr = "I'm from Parent"
    
    def parent_method(self):
        return "Parent method called"
    
    def shared_method(self):
        return "Original from Parent"

class Child(Parent):
    def __init__(self):
        super().__init__()  # 1. Káº¿ thá»«a
        self.child_attr = "I'm from Child"  # 2. ThÃªm má»›i
    
    def child_method(self):  # 2. ThÃªm má»›i
        return "Child method called"
    
    def shared_method(self):  # 3. Override
        return "Overridden by Child"
    
    def extended_method(self):  # 4. Má»Ÿ rá»™ng
        parent_result = super().parent_method()
        return f"{parent_result} + Extended by Child"

# Test
child = Child()

print("1. Káº¿ thá»«a attribute:", child.parent_attr)
print("2. ThÃªm attribute má»›i:", child.child_attr)
print("3. Káº¿ thá»«a method:", child.parent_method())
print("4. ThÃªm method má»›i:", child.child_method())
print("5. Override method:", child.shared_method())
print("6. Má»Ÿ rá»™ng method:", child.extended_method())
````

### 3.1 Single Inheritance

````python
# SINGLE INHERITANCE: Má»™t child káº¿ thá»«a tá»« má»™t parent

class Animal:
    """Parent class."""
    
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound"

class Dog(Animal):  # Single inheritance
    """Child class káº¿ thá»«a tá»« Animal."""
    
    def speak(self):
        return f"{self.name} barks: Woof!"
    
    def fetch(self):
        return f"{self.name} is fetching the ball"

# Test
dog = Dog("Buddy")
print(dog.speak())  # Overridden
print(dog.fetch())  # Own method

# Kiá»ƒm tra inheritance
print(f"\nDog's bases: {Dog.__bases__}")  # Tuple chá»©a parent classes
````

### 3.2 Multilevel Inheritance

````python
# MULTILEVEL INHERITANCE: Chuá»—i káº¿ thá»«a A â†’ B â†’ C

class Animal:
    """Level 1: Grandparent class."""
    
    def __init__(self, name):
        self.name = name
        print(f"Animal.__init__({name})")
    
    def eat(self):
        return f"{self.name} is eating"
    
    def sleep(self):
        return f"{self.name} is sleeping"

class Mammal(Animal):
    """Level 2: Parent class - káº¿ thá»«a tá»« Animal."""
    
    def __init__(self, name, fur_color):
        super().__init__(name)
        self.fur_color = fur_color
        print(f"Mammal.__init__({name}, {fur_color})")
    
    def give_birth(self):
        return f"{self.name} gives birth to live young"
    
    def nurse(self):
        return f"{self.name} nurses its young"

class Dog(Mammal):
    """Level 3: Child class - káº¿ thá»«a tá»« Mammal."""
    
    def __init__(self, name, fur_color, breed):
        super().__init__(name, fur_color)
        self.breed = breed
        print(f"Dog.__init__({name}, {fur_color}, {breed})")
    
    def bark(self):
        return f"{self.name} the {self.breed} barks: Woof!"
    
    def fetch(self):
        return f"{self.name} is fetching"

# Test
print("Creating a Dog object...")
print("-" * 40)
dog = Dog("Buddy", "Golden", "Labrador")
print("-" * 40)

print("\nDog can access methods from all levels:")
print(f"From Animal: {dog.eat()}")        # Level 1
print(f"From Animal: {dog.sleep()}")      # Level 1
print(f"From Mammal: {dog.give_birth()}") # Level 2
print(f"From Mammal: {dog.nurse()}")      # Level 2
print(f"From Dog: {dog.bark()}")          # Level 3

print(f"\nDog's attributes:")
print(f"  name (Animal): {dog.name}")
print(f"  fur_color (Mammal): {dog.fur_color}")
print(f"  breed (Dog): {dog.breed}")
````

### 3.3 Hierarchical Inheritance

````python
# HIERARCHICAL INHERITANCE: Nhiá»u child káº¿ thá»«a tá»« má»™t parent

class Shape:
    """Parent class cho táº¥t cáº£ cÃ¡c hÃ¬nh."""
    
    def __init__(self, color="black"):
        self.color = color
    
    def describe(self):
        return f"A {self.color} shape"
    
    def area(self):
        raise NotImplementedError("Subclass must implement area()")

class Circle(Shape):  # Child 1
    def __init__(self, radius, color="black"):
        super().__init__(color)
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2
    
    def describe(self):
        return f"A {self.color} circle with radius {self.radius}"

class Rectangle(Shape):  # Child 2
    def __init__(self, width, height, color="black"):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def describe(self):
        return f"A {self.color} rectangle ({self.width}x{self.height})"

class Triangle(Shape):  # Child 3
    def __init__(self, base, height, color="black"):
        super().__init__(color)
        self.base = base
        self.height = height
    
    def area(self):
        return 0.5 * self.base * self.height
    
    def describe(self):
        return f"A {self.color} triangle (base={self.base}, h={self.height})"

# Test - Polymorphism in action
shapes = [
    Circle(5, "red"),
    Rectangle(4, 6, "blue"),
    Triangle(3, 8, "green")
]

print("Hierarchical Inheritance Demo:")
print("=" * 50)

for shape in shapes:
    print(shape.describe())
    print(f"  Area: {shape.area():.2f}")
    print()
````

````python
# Override vá»›i má»Ÿ rá»™ng - Gá»i parent method vÃ  thÃªm logic

class Employee:
    """Base Employee class."""
    
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def calculate_pay(self):
        """TÃ­nh lÆ°Æ¡ng cÆ¡ báº£n."""
        return self.salary
    
    def get_details(self):
        """Chi tiáº¿t nhÃ¢n viÃªn."""
        return f"Name: {self.name}, Salary: ${self.salary:,.2f}"

class Manager(Employee):
    """Manager cÃ³ thÃªm bonus."""
    
    def __init__(self, name, salary, bonus_rate=0.2):
        super().__init__(name, salary)
        self.bonus_rate = bonus_rate
    
    def calculate_pay(self):
        """Override: LÆ°Æ¡ng + Bonus."""
        base_pay = super().calculate_pay()  # Gá»i parent method
        bonus = base_pay * self.bonus_rate
        return base_pay + bonus
    
    def get_details(self):
        """Override vá»›i má»Ÿ rá»™ng."""
        base_details = super().get_details()  # Gá»i parent method
        return f"{base_details}, Role: Manager, Bonus: {self.bonus_rate:.0%}"

class SalesEmployee(Employee):
    """Sales cÃ³ thÃªm commission."""
    
    def __init__(self, name, salary, sales_amount, commission_rate=0.1):
        super().__init__(name, salary)
        self.sales_amount = sales_amount
        self.commission_rate = commission_rate
    
    def calculate_pay(self):
        """Override: LÆ°Æ¡ng + Commission."""
        base_pay = super().calculate_pay()
        commission = self.sales_amount * self.commission_rate
        return base_pay + commission
    
    def get_details(self):
        base_details = super().get_details()
        return f"{base_details}, Role: Sales, Commission: {self.commission_rate:.0%}"

# Test
employees = [
    Employee("John Doe", 50000),
    Manager("Jane Smith", 80000, 0.25),
    SalesEmployee("Bob Wilson", 45000, 100000, 0.15)
]

print("Employee Payment System:")
print("=" * 70)

for emp in employees:
    print(emp.get_details())
    print(f"  Total Pay: ${emp.calculate_pay():,.2f}")
    print()
````

````python
# super() trong __init__

class Person:
    """Person base class."""
    
    def __init__(self, name, age):
        print(f"  Person.__init__(name={name}, age={age})")
        self.name = name
        self.age = age
    
    def introduce(self):
        return f"Hi, I'm {self.name}, {self.age} years old."

class Student(Person):
    """Student inherits from Person."""
    
    def __init__(self, name, age, student_id, major):
        print(f"  Student.__init__(name={name}, age={age}, id={student_id}, major={major})")
        
        # Gá»i __init__ cá»§a parent - QUAN TRá»ŒNG!
        super().__init__(name, age)
        
        # ThÃªm attributes riÃªng
        self.student_id = student_id
        self.major = major
    
    def introduce(self):
        # Gá»i method cá»§a parent vÃ  má»Ÿ rá»™ng
        parent_intro = super().introduce()
        return f"{parent_intro} I study {self.major}."
    
    def study(self):
        return f"{self.name} is studying {self.major}"

class GraduateStudent(Student):
    """GraduateStudent inherits from Student (Multilevel)."""
    
    def __init__(self, name, age, student_id, major, thesis_topic):
        print(f"  GraduateStudent.__init__(...)")
        
        # Gá»i __init__ cá»§a Student (parent)
        super().__init__(name, age, student_id, major)
        
        self.thesis_topic = thesis_topic
    
    def introduce(self):
        parent_intro = super().introduce()
        return f"{parent_intro} My thesis is about '{self.thesis_topic}'."
    
    def research(self):
        return f"{self.name} is researching: {self.thesis_topic}"

# Test
print("Creating objects with super():")
print("=" * 60)

print("\n1. Creating Person:")
person = Person("John", 30)

print("\n2. Creating Student:")
student = Student("Alice", 20, "S001", "Computer Science")

print("\n3. Creating GraduateStudent:")
grad = GraduateStudent("Bob", 25, "G001", "AI", "Deep Learning for NLP")

print("\n" + "=" * 60)
print("Introductions:")
print("=" * 60)
print(f"Person: {person.introduce()}")
print(f"Student: {student.introduce()}")
print(f"GraduateStudent: {grad.introduce()}")
````

Python giáº£i quyáº¿t báº±ng **MRO (Method Resolution Order)**.

````python
# Diamond Problem Demo

class A:
    def __init__(self):
        print("A.__init__")
        self.value = "A"
    
    def greet(self):
        return "Hello from A"

class B(A):
    def __init__(self):
        print("B.__init__")
        super().__init__()
        self.value = "B"
    
    def greet(self):
        return "Hello from B"

class C(A):
    def __init__(self):
        print("C.__init__")
        super().__init__()
        self.value = "C"
    
    def greet(self):
        return "Hello from C"

class D(B, C):  # Diamond inheritance
    def __init__(self):
        print("D.__init__")
        super().__init__()

# Test
print("Creating D object:")
print("-" * 30)
d = D()
print("-" * 30)

print(f"\nd.value = {d.value}")  # Follows MRO
print(f"d.greet() = {d.greet()}")  # Follows MRO

print(f"\nMRO of D: {[cls.__name__ for cls in D.__mro__]}")
````

## 7. Method Resolution Order (MRO)

### 7.1 KhÃ¡i niá»‡m

**MRO** lÃ  thá»© tá»± mÃ  Python tÃ¬m kiáº¿m methods khi cÃ³ inheritance. Python sá»­ dá»¥ng thuáº­t toÃ¡n **C3 Linearization**.

### 7.2 CÃ¡ch xem MRO

```python
# CÃ¡ch 1: __mro__ attribute
ClassName.__mro__

# CÃ¡ch 2: mro() method
ClassName.mro()
```

### 7.3 VÃ­ dá»¥ chi tiáº¿t

````python
# MRO Demo

class A:
    def method(self):
        return "A.method()"

class B(A):
    def method(self):
        return "B.method()"

class C(A):
    def method(self):
        return "C.method()"

class D(B, C):
    pass  # KhÃ´ng override - sáº½ dÃ¹ng method theo MRO

class E(C, B):
    pass  # Thá»© tá»± parent khÃ¡c

# Xem MRO
print("Method Resolution Order:")
print("=" * 50)

print("\nD(B, C):")
print(f"  MRO: {[cls.__name__ for cls in D.__mro__]}")
print(f"  D().method() = {D().method()}")  # Follows MRO

print("\nE(C, B):")
print(f"  MRO: {[cls.__name__ for cls in E.__mro__]}")
print(f"  E().method() = {E().method()}")  # Follows MRO
````

````python
# MRO vá»›i super() chain

class A:
    def method(self):
        print("A.method() called")
        return "A"

class B(A):
    def method(self):
        print("B.method() called")
        result = super().method()  # Gá»i theo MRO
        return f"B -> {result}"

class C(A):
    def method(self):
        print("C.method() called")
        result = super().method()  # Gá»i theo MRO
        return f"C -> {result}"

class D(B, C):
    def method(self):
        print("D.method() called")
        result = super().method()  # Gá»i theo MRO
        return f"D -> {result}"

print("super() chain following MRO:")
print("=" * 50)
print(f"MRO: {[cls.__name__ for cls in D.__mro__]}")
print()

d = D()
result = d.method()
print(f"\nFinal result: {result}")
````

## 9. Abstract Base Class (ABC)

### 9.1 KhÃ¡i niá»‡m

**Abstract Base Class** lÃ  class khÃ´ng thá»ƒ táº¡o instance trá»±c tiáº¿p, chá»‰ dÃ¹ng lÃ m template cho cÃ¡c subclasses.

### 9.2 Sá»­ dá»¥ng abc module

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract Base Class cho cÃ¡c hÃ¬nh."""
    
    def __init__(self, color="black"):
        self.color = color
    
    @abstractmethod
    def area(self):
        """Pháº£i Ä‘Æ°á»£c implement bá»Ÿi subclass."""
        pass
    
    @abstractmethod
    def perimeter(self):
        """Pháº£i Ä‘Æ°á»£c implement bá»Ÿi subclass."""
        pass
    
    def describe(self):
        """Method thÃ´ng thÆ°á»ng - cÃ³ thá»ƒ override hoáº·c khÃ´ng."""
        return f"A {self.color} {self.__class__.__name__}"

class Rectangle(Shape):
    """Concrete class - pháº£i implement táº¥t cáº£ abstract methods."""
    
    def __init__(self, width, height, color="black"):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self):  # Implement abstract method
        return self.width * self.height
    
    def perimeter(self):  # Implement abstract method
        return 2 * (self.width + self.height)

class Circle(Shape):
    """Concrete class."""
    
    def __init__(self, radius, color="black"):
        super().__init__(color)
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        import math
        return 2 * math.pi * self.radius

# Test
print("Abstract Base Class Demo:")
print("=" * 50)

# âŒ KhÃ´ng thá»ƒ táº¡o instance cá»§a ABC
try:
    shape = Shape()
except TypeError as e:
    print(f"Cannot create Shape: {e}")

print()

# âœ… CÃ³ thá»ƒ táº¡o instance cá»§a concrete classes
rect = Rectangle(4, 5, "blue")
circle = Circle(3, "red")

for shape in [rect, circle]:
    print(shape.describe())
    print(f"  Area: {shape.area():.2f}")
    print(f"  Perimeter: {shape.perimeter():.2f}")
    print()
````

````python
from abc import ABC, abstractmethod
from datetime import datetime, date

class Employee(ABC):
    """
    Abstract Base Class cho Employee.
    
    Attributes:
        emp_id: MÃ£ nhÃ¢n viÃªn
        name: TÃªn nhÃ¢n viÃªn
        hire_date: NgÃ y vÃ o lÃ m
        base_salary: LÆ°Æ¡ng cÆ¡ báº£n
    """
    
    _employee_count = 0
    
    def __init__(self, name, base_salary, hire_date=None):
        Employee._employee_count += 1
        self.emp_id = f"EMP{Employee._employee_count:04d}"
        self.name = name
        self.base_salary = base_salary
        self.hire_date = hire_date or date.today()
    
    @property
    def years_employed(self):
        """Sá»‘ nÄƒm lÃ m viá»‡c."""
        return (date.today() - self.hire_date).days // 365
    
    @abstractmethod
    def calculate_pay(self):
        """TÃ­nh tá»•ng lÆ°Æ¡ng - pháº£i implement."""
        pass
    
    @abstractmethod
    def get_role(self):
        """Tráº£ vá» vai trÃ² - pháº£i implement."""
        pass
    
    def get_info(self):
        """ThÃ´ng tin cÆ¡ báº£n cá»§a nhÃ¢n viÃªn."""
        return (
            f"ID: {self.emp_id}\n"
            f"Name: {self.name}\n"
            f"Role: {self.get_role()}\n"
            f"Hire Date: {self.hire_date}\n"
            f"Years: {self.years_employed}\n"
            f"Total Pay: ${self.calculate_pay():,.2f}"
        )
    
    def __str__(self):
        return f"{self.name} ({self.get_role()})"

class FullTimeEmployee(Employee):
    """NhÃ¢n viÃªn toÃ n thá»i gian vá»›i bonus."""
    
    def __init__(self, name, base_salary, bonus_rate=0.1, hire_date=None):
        super().__init__(name, base_salary, hire_date)
        self.bonus_rate = bonus_rate
    
    def calculate_pay(self):
        bonus = self.base_salary * self.bonus_rate
        return self.base_salary + bonus
    
    def get_role(self):
        return "Full-Time Employee"

class PartTimeEmployee(Employee):
    """NhÃ¢n viÃªn bÃ¡n thá»i gian theo giá»."""
    
    def __init__(self, name, hourly_rate, hours_worked, hire_date=None):
        super().__init__(name, 0, hire_date)  # base_salary = 0
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked
    
    def calculate_pay(self):
        return self.hourly_rate * self.hours_worked
    
    def get_role(self):
        return "Part-Time Employee"

class Manager(FullTimeEmployee):
    """Manager vá»›i team vÃ  bonus cao hÆ¡n."""
    
    def __init__(self, name, base_salary, department, hire_date=None):
        super().__init__(name, base_salary, bonus_rate=0.2, hire_date=hire_date)
        self.department = department
        self.team = []
    
    def add_team_member(self, employee):
        if employee not in self.team:
            self.team.append(employee)
            return True
        return False
    
    def get_role(self):
        return f"Manager - {self.department}"
    
    def get_team_info(self):
        info = f"Team ({len(self.team)} members):\n"
        for emp in self.team:
            info += f"  - {emp}\n"
        return info

class Developer(FullTimeEmployee):
    """Developer vá»›i tech stack."""
    
    def __init__(self, name, base_salary, tech_stack, hire_date=None):
        super().__init__(name, base_salary, bonus_rate=0.15, hire_date=hire_date)
        self.tech_stack = tech_stack
    
    def get_role(self):
        return f"Developer ({', '.join(self.tech_stack)})"

class Intern(PartTimeEmployee):
    """Thá»±c táº­p sinh."""
    
    def __init__(self, name, hourly_rate, hours_worked, university, hire_date=None):
        super().__init__(name, hourly_rate, hours_worked, hire_date)
        self.university = university
    
    def get_role(self):
        return f"Intern from {self.university}"

# Demo
print("Employee Management System Demo")
print("=" * 60)

# Táº¡o employees
manager = Manager("Alice Johnson", 100000, "Engineering")
dev1 = Developer("Bob Smith", 80000, ["Python", "JavaScript"])
dev2 = Developer("Carol White", 75000, ["Java", "Go"])
intern = Intern("David Lee", 20, 80, "MIT")

# Manager thÃªm team
manager.add_team_member(dev1)
manager.add_team_member(dev2)
manager.add_team_member(intern)

# Hiá»ƒn thá»‹ thÃ´ng tin
employees = [manager, dev1, dev2, intern]

for emp in employees:
    print(emp.get_info())
    print("-" * 40)

print("\nManager's Team:")
print(manager.get_team_info())
````

### 10.2 E-Commerce Product System

````python
from abc import ABC, abstractmethod

class Product(ABC):
    """Abstract base class cho sáº£n pháº©m."""
    
    _product_count = 0
    
    def __init__(self, name, price, quantity):
        Product._product_count += 1
        self.sku = f"SKU{Product._product_count:05d}"
        self.name = name
        self._price = price
        self.quantity = quantity
    
    @property
    def price(self):
        return self._price
    
    @price.setter
    def price(self, value):
        if value < 0:
            raise ValueError("Price cannot be negative")
        self._price = value
    
    @abstractmethod
    def get_category(self):
        pass
    
    @abstractmethod
    def calculate_shipping(self):
        pass
    
    def get_total_value(self):
        return self.price * self.quantity
    
    def __str__(self):
        return f"[{self.sku}] {self.name} - ${self.price:.2f}"

class PhysicalProduct(Product):
    """Sáº£n pháº©m váº­t lÃ½ - cáº§n ship."""
    
    def __init__(self, name, price, quantity, weight, dimensions):
        super().__init__(name, price, quantity)
        self.weight = weight  # kg
        self.dimensions = dimensions  # (length, width, height) in cm
    
    def calculate_shipping(self):
        # $2 per kg + $0.01 per cubic cm
        volume = self.dimensions[0] * self.dimensions[1] * self.dimensions[2]
        return (self.weight * 2) + (volume * 0.01)

class DigitalProduct(Product):
    """Sáº£n pháº©m sá»‘ - khÃ´ng cáº§n ship."""
    
    def __init__(self, name, price, quantity, file_size, download_link):
        super().__init__(name, price, quantity)
        self.file_size = file_size  # MB
        self.download_link = download_link
    
    def calculate_shipping(self):
        return 0  # Digital products don't need shipping

class Electronics(PhysicalProduct):
    """Äá»“ Ä‘iá»‡n tá»­."""
    
    def __init__(self, name, price, quantity, weight, dimensions, warranty_months):
        super().__init__(name, price, quantity, weight, dimensions)
        self.warranty_months = warranty_months
    
    def get_category(self):
        return "Electronics"
    
    def calculate_shipping(self):
        # Electronics need careful handling - extra $5
        return super().calculate_shipping() + 5

class Clothing(PhysicalProduct):
    """Quáº§n Ã¡o."""
    
    def __init__(self, name, price, quantity, weight, dimensions, size, color):
        super().__init__(name, price, quantity, weight, dimensions)
        self.size = size
        self.color = color
    
    def get_category(self):
        return "Clothing"

class Software(DigitalProduct):
    """Pháº§n má»m."""
    
    def __init__(self, name, price, quantity, file_size, download_link, version, license_type):
        super().__init__(name, price, quantity, file_size, download_link)
        self.version = version
        self.license_type = license_type
    
    def get_category(self):
        return f"Software ({self.license_type})"

class Ebook(DigitalProduct):
    """SÃ¡ch Ä‘iá»‡n tá»­."""
    
    def __init__(self, name, price, quantity, file_size, download_link, author, pages):
        super().__init__(name, price, quantity, file_size, download_link)
        self.author = author
        self.pages = pages
    
    def get_category(self):
        return "E-Book"

# Demo
print("E-Commerce Product System Demo")
print("=" * 60)

products = [
    Electronics("iPhone 15", 999, 50, 0.2, (15, 7, 1), 12),
    Electronics("MacBook Pro", 2499, 30, 2.0, (35, 25, 2), 24),
    Clothing("Winter Jacket", 199, 100, 0.8, (40, 30, 5), "L", "Black"),
    Software("Adobe Photoshop", 299, 999, 2048, "https://adobe.com/download", "2024", "Annual"),
    Ebook("Python Programming", 29.99, 999, 15, "https://books.com/python", "John Doe", 450)
]

total_shipping = 0

for product in products:
    print(product)
    print(f"  Category: {product.get_category()}")
    print(f"  Stock: {product.quantity}")
    print(f"  Total Value: ${product.get_total_value():,.2f}")
    shipping = product.calculate_shipping()
    print(f"  Shipping: ${shipping:.2f}")
    total_shipping += shipping * product.quantity
    print()

print(f"Total estimated shipping revenue: ${total_shipping:,.2f}")
````

### 11.2 Liskov Substitution Principle (LSP)

> Objects cá»§a subclass pháº£i cÃ³ thá»ƒ thay tháº¿ objects cá»§a superclass mÃ  khÃ´ng lÃ m thay Ä‘á»•i tÃ­nh Ä‘Ãºng Ä‘áº¯n cá»§a chÆ°Æ¡ng trÃ¬nh.

````python
# âŒ Vi pháº¡m LSP

class Rectangle:
    def __init__(self, width, height):
        self._width = width
        self._height = height
    
    @property
    def width(self):
        return self._width
    
    @width.setter
    def width(self, value):
        self._width = value
    
    @property
    def height(self):
        return self._height
    
    @height.setter
    def height(self, value):
        self._height = value
    
    def area(self):
        return self._width * self._height

class Square(Rectangle):  # Square IS-A Rectangle vá» máº·t toÃ¡n há»c
    def __init__(self, side):
        super().__init__(side, side)
    
    @Rectangle.width.setter
    def width(self, value):
        self._width = value
        self._height = value  # Pháº£i giá»¯ width = height
    
    @Rectangle.height.setter
    def height(self, value):
        self._width = value
        self._height = value

# Váº¥n Ä‘á»: Square khÃ´ng thá»ƒ thay tháº¿ Rectangle hoÃ n toÃ n
def test_rectangle(rect):
    rect.width = 5
    rect.height = 4
    # Ká»³ vá»ng: area = 5 * 4 = 20
    expected = 20
    actual = rect.area()
    print(f"Expected: {expected}, Actual: {actual}, Pass: {actual == expected}")

print("Testing with Rectangle:")
test_rectangle(Rectangle(10, 10))

print("\nTesting with Square:")
test_rectangle(Square(10))  # âŒ Fails! Square khÃ´ng thá»ƒ thay tháº¿ Rectangle
````

### 11.3 Favor Composition over Inheritance

Khi khÃ´ng cháº¯c cháº¯n, Æ°u tiÃªn **Composition** (HAS-A) hÆ¡n **Inheritance** (IS-A).

````python
# Composition vs Inheritance

# âœ… Composition - Linh hoáº¡t hÆ¡n
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower
    
    def start(self):
        return f"Engine ({self.horsepower}HP) starting..."

class Wheels:
    def __init__(self, count):
        self.count = count
    
    def rotate(self):
        return f"{self.count} wheels rotating"

class Car:
    """Car HAS-A Engine, HAS Wheels (Composition)."""
    
    def __init__(self, brand, engine_hp):
        self.brand = brand
        self.engine = Engine(engine_hp)  # Composition
        self.wheels = Wheels(4)          # Composition
    
    def start(self):
        return f"{self.brand}: {self.engine.start()}"
    
    def drive(self):
        return f"{self.brand}: {self.wheels.rotate()}"

# Test
car = Car("Toyota", 200)
print(car.start())
print(car.drive())

# Æ¯u Ä‘iá»ƒm:
# - CÃ³ thá»ƒ thay Ä‘á»•i engine, wheels dá»… dÃ ng
# - KhÃ´ng bá»‹ rÃ ng buá»™c bá»Ÿi inheritance hierarchy
# - Linh hoáº¡t hÆ¡n trong viá»‡c tÃ¡i sá»­ dá»¥ng code
````

## 12. Common Mistakes

### 12.1 QuÃªn gá»i super().__init__()

````python
# âŒ MISTAKE: QuÃªn gá»i super().__init__()

class Parent:
    def __init__(self, name):
        self.name = name
        self.created_at = "2024-01-01"

class Child(Parent):
    def __init__(self, name, age):
        # âŒ QuÃªn gá»i super().__init__(name)
        self.age = age

try:
    child = Child("Alice", 10)
    print(child.name)  # AttributeError!
except AttributeError as e:
    print(f"Error: {e}")

print()

# âœ… CORRECT
class ChildCorrect(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # âœ… Gá»i super()
        self.age = age

child = ChildCorrect("Alice", 10)
print(f"Name: {child.name}, Age: {child.age}")
````

### 12.2 Override mÃ  khÃ´ng gá»i parent method khi cáº§n

````python
# âŒ MISTAKE: Override hoÃ n toÃ n thay vÃ¬ má»Ÿ rá»™ng

class Logger:
    def log(self, message):
        timestamp = "2024-01-01 10:00:00"
        return f"[{timestamp}] {message}"

class FileLogger(Logger):
    def log(self, message):
        # âŒ Máº¥t timestamp tá»« parent!
        return f"FILE: {message}"

logger = FileLogger()
print("Bad:", logger.log("Test message"))

# âœ… CORRECT: Má»Ÿ rá»™ng parent method
class FileLoggerCorrect(Logger):
    def log(self, message):
        parent_log = super().log(message)  # âœ… Giá»¯ timestamp
        return f"FILE: {parent_log}"

logger = FileLoggerCorrect()
print("Good:", logger.log("Test message"))
````

### 12.3 Sá»­ dá»¥ng sai Inheritance (IS-A vs HAS-A)

````python
# âŒ MISTAKE: Sai logic IS-A

class Stack(list):  # Stack IS-A list? KhÃ´ng há»£p lÃ½!
    """Stack káº¿ thá»«a list - nhÆ°ng expose quÃ¡ nhiá»u methods."""
    pass

stack = Stack()
stack.push = stack.append
stack.push(1)
stack.push(2)

# Váº¥n Ä‘á»: Stack cho phÃ©p insert á»Ÿ giá»¯a - vi pháº¡m LIFO!
stack.insert(0, 999)  # âŒ KhÃ´ng nÃªn cho phÃ©p!
print("Bad Stack:", stack)

print()

# âœ… CORRECT: DÃ¹ng Composition
class ProperStack:
    """Stack dÃ¹ng composition - chá»‰ expose push/pop."""
    
    def __init__(self):
        self._items = []  # Composition
    
    def push(self, item):
        self._items.append(item)
    
    def pop(self):
        if not self._items:
            raise IndexError("Stack is empty")
        return self._items.pop()
    
    def peek(self):
        if not self._items:
            raise IndexError("Stack is empty")
        return self._items[-1]
    
    def is_empty(self):
        return len(self._items) == 0
    
    def __len__(self):
        return len(self._items)

stack = ProperStack()
stack.push(1)
stack.push(2)
print("Good Stack - Pop:", stack.pop())
# stack.insert(0, 999)  # âŒ AttributeError - Ä‘Ãºng nhÆ° mong muá»‘n!
````
