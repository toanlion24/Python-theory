### 1.2 Táº¡i sao cáº§n Polymorphism?

| Lá»£i Ã­ch | MÃ´ táº£ |
|---------|-------|
| **Flexibility** | Code linh hoáº¡t, dá»… má»Ÿ rá»™ng |
| **Code Reuse** | Má»™t function hoáº¡t Ä‘á»™ng vá»›i nhiá»u types |
| **Loose Coupling** | Giáº£m sá»± phá»¥ thuá»™c giá»¯a cÃ¡c components |
| **Maintainability** | Dá»… báº£o trÃ¬, thÃªm types má»›i khÃ´ng cáº§n sá»­a code cÅ© |

````python
# VÃ­ dá»¥ Ä‘Æ¡n giáº£n vá» Polymorphism

# Polymorphism vá»›i built-in function len()
print("=" * 50)
print("Polymorphism vá»›i len()")
print("=" * 50)

# CÃ¹ng function len(), khÃ¡c type
my_string = "Hello World"
my_list = [1, 2, 3, 4, 5]
my_dict = {"a": 1, "b": 2, "c": 3}
my_tuple = (10, 20, 30)
my_set = {1, 2, 3, 4}

print(f"len('{my_string}') = {len(my_string)}")
print(f"len({my_list}) = {len(my_list)}")
print(f"len({my_dict}) = {len(my_dict)}")
print(f"len({my_tuple}) = {len(my_tuple)}")
print(f"len({my_set}) = {len(my_set)}")
````

````python
# Polymorphism vá»›i operators

print("=" * 50)
print("Polymorphism vá»›i + operator")
print("=" * 50)

# CÃ¹ng operator +, khÃ¡c hÃ nh vi
print(f"int:    5 + 3 = {5 + 3}")              # Addition
print(f"float:  5.5 + 3.2 = {5.5 + 3.2}")      # Addition
print(f"str:    'Hello' + ' World' = {'Hello' + ' World'}")  # Concatenation
print(f"list:   [1, 2] + [3, 4] = {[1, 2] + [3, 4]}")        # Concatenation
print(f"tuple:  (1, 2) + (3, 4) = {(1, 2) + (3, 4)}")        # Concatenation

print("\n" + "=" * 50)
print("Polymorphism vá»›i * operator")
print("=" * 50)

# CÃ¹ng operator *, khÃ¡c hÃ nh vi
print(f"int:    5 * 3 = {5 * 3}")              # Multiplication
print(f"str:    'Hi' * 3 = {'Hi' * 3}")        # Repetition
print(f"list:   [1, 2] * 3 = {[1, 2] * 3}")    # Repetition
````

````python
# âŒ KHÃ”NG cÃ³ Polymorphism - Code cá»©ng nháº¯c

def make_dog_speak():
    return "Woof!"

def make_cat_speak():
    return "Meow!"

def make_bird_speak():
    return "Tweet!"

def make_cow_speak():
    return "Moo!"

# Pháº£i gá»i tá»«ng function riÃªng
print("Without Polymorphism:")
print(make_dog_speak())
print(make_cat_speak())
print(make_bird_speak())

# Náº¿u thÃªm animal má»›i -> pháº£i thÃªm function má»›i
# Náº¿u cáº§n loop qua animals -> pháº£i dÃ¹ng if/else
````

````python
# âœ… CÃ“ Polymorphism - Code linh hoáº¡t

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Bird:
    def speak(self):
        return "Tweet!"

class Cow:
    def speak(self):
        return "Moo!"

# Polymorphic function - hoáº¡t Ä‘á»™ng vá»›i Báº¤T Ká»² animal nÃ o cÃ³ speak()
def make_animal_speak(animal):
    """Polymorphic function - khÃ´ng quan tÃ¢m type cá»¥ thá»ƒ."""
    return animal.speak()

# Dá»… dÃ ng iterate qua cÃ¡c animals
print("With Polymorphism:")
animals = [Dog(), Cat(), Bird(), Cow()]

for animal in animals:
    print(f"{animal.__class__.__name__}: {make_animal_speak(animal)}")

# ThÃªm animal má»›i? Chá»‰ cáº§n táº¡o class vá»›i speak() - KHÃ”NG cáº§n sá»­a code cÅ©!
````

### 2.2 Python's Approach

Python chá»§ yáº¿u sá»­ dá»¥ng **Runtime Polymorphism** vá»›i:
1. **Duck Typing** - KhÃ´ng quan tÃ¢m type, chá»‰ cáº§n cÃ³ methods cáº§n thiáº¿t
2. **Method Overriding** - Child class override parent methods
3. **Operator Overloading** - Magic methods cho operators

````python
# Python khÃ´ng cÃ³ Method Overloading truyá»n thá»‘ng
# NhÆ°ng cÃ³ thá»ƒ simulate báº±ng default arguments hoáº·c *args

class Calculator:
    # âŒ KhÃ´ng thá»ƒ cÃ³ 2 methods cÃ¹ng tÃªn vá»›i sá»‘ parameters khÃ¡c nhau
    # def add(self, a, b):
    #     return a + b
    # def add(self, a, b, c):  # Sáº½ override method trÃªn!
    #     return a + b + c
    
    # âœ… Sá»­ dá»¥ng *args Ä‘á»ƒ simulate method overloading
    def add(self, *args):
        """Cá»™ng báº¥t ká»³ sá»‘ lÆ°á»£ng arguments nÃ o."""
        return sum(args)
    
    # âœ… Hoáº·c dÃ¹ng default arguments
    def multiply(self, a, b=1, c=1):
        """NhÃ¢n 2 hoáº·c 3 sá»‘."""
        return a * b * c

calc = Calculator()

# Simulate overloading vá»›i *args
print(f"add(1, 2) = {calc.add(1, 2)}")
print(f"add(1, 2, 3) = {calc.add(1, 2, 3)}")
print(f"add(1, 2, 3, 4, 5) = {calc.add(1, 2, 3, 4, 5)}")

print()

# Simulate overloading vá»›i default args
print(f"multiply(5) = {calc.multiply(5)}")
print(f"multiply(5, 3) = {calc.multiply(5, 3)}")
print(f"multiply(5, 3, 2) = {calc.multiply(5, 3, 2)}")
````

### 3.2 VÃ­ dá»¥ Duck Typing

````python
# Duck Typing - Classic Example

class Duck:
    """Vá»‹t tháº­t."""
    
    def quack(self):
        return "Quack! Quack! ğŸ¦†"
    
    def fly(self):
        return "Duck is flying over the pond"

class Person:
    """NgÆ°á»i giáº£ lÃ m vá»‹t."""
    
    def __init__(self, name):
        self.name = name
    
    def quack(self):
        return f"{self.name} imitates: 'Quack quack!' ğŸ§‘"
    
    def fly(self):
        return f"{self.name} flaps arms pretending to fly"

class RobotDuck:
    """Robot vá»‹t."""
    
    def __init__(self, model):
        self.model = model
    
    def quack(self):
        return f"Robot {self.model}: *BEEP* Quack simulation *BEEP* ğŸ¤–"
    
    def fly(self):
        return f"Robot {self.model} activates propellers"

class Airplane:
    """MÃ¡y bay - CHá»ˆ cÃ³ fly(), KHÃ”NG cÃ³ quack()."""
    
    def fly(self):
        return "Airplane is flying at 30,000 feet âœˆï¸"

# Duck Typing function
def in_the_pond(duck):
    """
    Function nÃ y KHÃ”NG quan tÃ¢m type.
    Chá»‰ cáº§n object cÃ³ quack() vÃ  fly().
    """
    print(duck.quack())
    print(duck.fly())
    print()

# Test - Táº¥t cáº£ hoáº¡t Ä‘á»™ng vÃ¬ Ä‘á»u cÃ³ quack() vÃ  fly()
print("Duck Typing Demo:")
print("=" * 50)

in_the_pond(Duck())
in_the_pond(Person("John"))
in_the_pond(RobotDuck("DK-3000"))

# Airplane khÃ´ng cÃ³ quack() -> sáº½ lá»—i
print("Testing Airplane:")
try:
    in_the_pond(Airplane())
except AttributeError as e:
    print(f"Error: {e}")
    print("Airplane khÃ´ng pháº£i 'duck' vÃ¬ khÃ´ng cÃ³ quack()!")
````

### 3.3 Duck Typing vá»›i File-like Objects

````python
# Duck Typing thá»±c táº¿ - File-like objects

import io

class MockFile:
    """File giáº£ - cÃ³ cÃ¹ng interface vá»›i file tháº­t."""
    
    def __init__(self):
        self.content = ""
    
    def write(self, data):
        self.content += data
        return len(data)
    
    def read(self):
        return self.content

class Logger:
    """Logger - cÅ©ng cÃ³ write() method."""
    
    def __init__(self, prefix="LOG"):
        self.prefix = prefix
        self.logs = []
    
    def write(self, data):
        log_entry = f"[{self.prefix}] {data}"
        self.logs.append(log_entry)
        print(log_entry)
        return len(data)

def write_report(file_like, report_data):
    """
    Polymorphic function - hoáº¡t Ä‘á»™ng vá»›i báº¥t ká»³ object nÃ o cÃ³ write().
    - File tháº­t
    - StringIO
    - MockFile
    - Logger
    """
    file_like.write("=== REPORT START ===\n")
    file_like.write(report_data)
    file_like.write("\n=== REPORT END ===\n")

# Test vá»›i cÃ¡c "file-like" objects khÃ¡c nhau
print("Duck Typing vá»›i File-like Objects:")
print("=" * 50)

# 1. StringIO (built-in file-like object)
string_buffer = io.StringIO()
write_report(string_buffer, "Data from StringIO")
print("StringIO content:")
print(string_buffer.getvalue())

# 2. MockFile (custom file-like object)
mock_file = MockFile()
write_report(mock_file, "Data from MockFile")
print("\nMockFile content:")
print(mock_file.read())

# 3. Logger (cÃ³ write() method)
print("\nLogger output:")
logger = Logger("INFO")
write_report(logger, "Data from Logger")
````

### 3.4 EAFP - Easier to Ask for Forgiveness than Permission

Python's duck typing thÆ°á»ng Ä‘i kÃ¨m vá»›i nguyÃªn táº¯c **EAFP**:

```python
# LBYL (Look Before You Leap) - Kiá»ƒm tra trÆ°á»›c
if hasattr(obj, 'quack'):
    obj.quack()

# EAFP (Easier to Ask Forgiveness) - Thá»­ vÃ  báº¯t lá»—i
try:
    obj.quack()
except AttributeError:
    handle_error()
```

````python
# EAFP vs LBYL

class Duck:
    def quack(self):
        return "Quack!"

class Stone:
    pass

def make_it_quack_lbyl(thing):
    """LBYL - Look Before You Leap."""
    if hasattr(thing, 'quack') and callable(getattr(thing, 'quack')):
        return thing.quack()
    else:
        return "Cannot quack"

def make_it_quack_eafp(thing):
    """EAFP - Easier to Ask Forgiveness than Permission."""
    try:
        return thing.quack()
    except AttributeError:
        return "Cannot quack"

# Test
duck = Duck()
stone = Stone()

print("LBYL approach:")
print(f"  Duck: {make_it_quack_lbyl(duck)}")
print(f"  Stone: {make_it_quack_lbyl(stone)}")

print("\nEAFP approach (Pythonic):")
print(f"  Duck: {make_it_quack_eafp(duck)}")
print(f"  Stone: {make_it_quack_eafp(stone)}")
````

````python
# Method Overriding Polymorphism

class Animal:
    """Base class vá»›i methods chung."""
    
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        """Default implementation - sáº½ bá»‹ override."""
        return f"{self.name} makes a sound"
    
    def move(self):
        """Default implementation."""
        return f"{self.name} is moving"
    
    def describe(self):
        """Template method - dÃ¹ng cÃ¡c methods cÃ³ thá»ƒ bá»‹ override."""
        return f"{self.name} is a {self.__class__.__name__}. {self.speak()}. {self.move()}."

class Dog(Animal):
    def speak(self):  # Override
        return f"{self.name} barks: Woof! Woof! ğŸ•"
    
    def move(self):  # Override
        return f"{self.name} runs on 4 legs"
    
    def fetch(self):  # ThÃªm method má»›i
        return f"{self.name} fetches the ball"

class Cat(Animal):
    def speak(self):  # Override
        return f"{self.name} meows: Meow! ğŸ±"
    
    def move(self):  # Override
        return f"{self.name} walks gracefully"
    
    def purr(self):  # ThÃªm method má»›i
        return f"{self.name} purrs happily"

class Bird(Animal):
    def speak(self):  # Override
        return f"{self.name} tweets: Tweet! ğŸ¦"
    
    def move(self):  # Override
        return f"{self.name} flies through the sky"

class Fish(Animal):
    def speak(self):  # Override
        return f"{self.name} blubs: ... (fish don't speak!) ğŸŸ"
    
    def move(self):  # Override
        return f"{self.name} swims in the water"

# Polymorphism in action
print("Method Overriding Polymorphism:")
print("=" * 60)

animals = [
    Animal("Generic"),
    Dog("Buddy"),
    Cat("Whiskers"),
    Bird("Tweety"),
    Fish("Nemo")
]

# CÃ¹ng method call, khÃ¡c behavior
for animal in animals:
    print(f"\n{animal.__class__.__name__}:")
    print(f"  {animal.speak()}")
    print(f"  {animal.move()}")
````

````python
# Template Method Pattern vá»›i Polymorphism

class DataProcessor:
    """Base class vá»›i template method."""
    
    def process(self, data):
        """
        Template method - Ä‘á»‹nh nghÄ©a workflow.
        CÃ¡c bÆ°á»›c cá»¥ thá»ƒ Ä‘Æ°á»£c override bá»Ÿi subclasses.
        """
        print(f"\nProcessing with {self.__class__.__name__}...")
        validated_data = self.validate(data)
        transformed_data = self.transform(validated_data)
        result = self.save(transformed_data)
        return result
    
    def validate(self, data):
        """Override Ä‘á»ƒ custom validation."""
        return data
    
    def transform(self, data):
        """Override Ä‘á»ƒ custom transformation."""
        return data
    
    def save(self, data):
        """Override Ä‘á»ƒ custom save logic."""
        return f"Saved: {data}"

class JSONProcessor(DataProcessor):
    def validate(self, data):
        print("  Validating JSON format...")
        return data
    
    def transform(self, data):
        print("  Parsing JSON...")
        return f"{{json: {data}}}"
    
    def save(self, data):
        print("  Saving to .json file...")
        return f"Saved JSON: {data}"

class CSVProcessor(DataProcessor):
    def validate(self, data):
        print("  Validating CSV format...")
        return data
    
    def transform(self, data):
        print("  Parsing CSV rows...")
        return f"row1,row2,{data}"
    
    def save(self, data):
        print("  Saving to .csv file...")
        return f"Saved CSV: {data}"

class XMLProcessor(DataProcessor):
    def validate(self, data):
        print("  Validating XML schema...")
        return data
    
    def transform(self, data):
        print("  Parsing XML nodes...")
        return f"<data>{data}</data>"
    
    def save(self, data):
        print("  Saving to .xml file...")
        return f"Saved XML: {data}"

# Polymorphism - cÃ¹ng process() method
processors = [
    JSONProcessor(),
    CSVProcessor(),
    XMLProcessor()
]

for processor in processors:
    result = processor.process("sample data")
    print(f"  Result: {result}")
````

### 5.3 VÃ­ dá»¥: Vector Class

````python
import math

class Vector:
    """
    2D Vector vá»›i Ä‘áº§y Ä‘á»§ operator overloading.
    
    VÃ­ dá»¥:
        v1 = Vector(3, 4)
        v2 = Vector(1, 2)
        v3 = v1 + v2  # Vector(4, 6)
    """
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    # === Arithmetic Operators ===
    
    def __add__(self, other):
        """Vector + Vector hoáº·c Vector + sá»‘."""
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        elif isinstance(other, (int, float)):
            return Vector(self.x + other, self.y + other)
        return NotImplemented
    
    def __radd__(self, other):
        """Sá»‘ + Vector (reverse add)."""
        return self.__add__(other)
    
    def __sub__(self, other):
        """Vector - Vector."""
        if isinstance(other, Vector):
            return Vector(self.x - other.x, self.y - other.y)
        return NotImplemented
    
    def __mul__(self, scalar):
        """Vector * scalar (scalar multiplication)."""
        if isinstance(scalar, (int, float)):
            return Vector(self.x * scalar, self.y * scalar)
        return NotImplemented
    
    def __rmul__(self, scalar):
        """Scalar * Vector (reverse multiply)."""
        return self.__mul__(scalar)
    
    def __truediv__(self, scalar):
        """Vector / scalar."""
        if isinstance(scalar, (int, float)):
            if scalar == 0:
                raise ZeroDivisionError("Cannot divide vector by zero")
            return Vector(self.x / scalar, self.y / scalar)
        return NotImplemented
    
    def __neg__(self):
        """Unary minus: -Vector."""
        return Vector(-self.x, -self.y)
    
    def __pos__(self):
        """Unary plus: +Vector."""
        return Vector(self.x, self.y)
    
    def __abs__(self):
        """abs(Vector) - magnitude."""
        return math.sqrt(self.x ** 2 + self.y ** 2)
    
    # === Comparison Operators ===
    
    def __eq__(self, other):
        """Vector == Vector."""
        if isinstance(other, Vector):
            return self.x == other.x and self.y == other.y
        return NotImplemented
    
    def __ne__(self, other):
        """Vector != Vector."""
        result = self.__eq__(other)
        if result is NotImplemented:
            return result
        return not result
    
    def __lt__(self, other):
        """Vector < Vector (so sÃ¡nh magnitude)."""
        if isinstance(other, Vector):
            return abs(self) < abs(other)
        return NotImplemented
    
    def __le__(self, other):
        """Vector <= Vector."""
        return self == other or self < other
    
    def __gt__(self, other):
        """Vector > Vector."""
        if isinstance(other, Vector):
            return abs(self) > abs(other)
        return NotImplemented
    
    def __ge__(self, other):
        """Vector >= Vector."""
        return self == other or self > other
    
    # === Built-in Functions ===
    
    def __len__(self):
        """len(Vector) - tráº£ vá» 2 (sá»‘ chiá»u)."""
        return 2
    
    def __bool__(self):
        """bool(Vector) - False náº¿u zero vector."""
        return self.x != 0 or self.y != 0
    
    # === String Representation ===
    
    def __str__(self):
        """str(Vector) - human-readable."""
        return f"Vector({self.x}, {self.y})"
    
    def __repr__(self):
        """repr(Vector) - unambiguous."""
        return f"Vector({self.x!r}, {self.y!r})"
    
    # === Extra Methods ===
    
    def dot(self, other):
        """Dot product."""
        return self.x * other.x + self.y * other.y
    
    def normalize(self):
        """Return unit vector."""
        magnitude = abs(self)
        if magnitude == 0:
            raise ValueError("Cannot normalize zero vector")
        return self / magnitude

# Demo
print("Vector Class vá»›i Operator Overloading:")
print("=" * 50)

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print()

# Arithmetic
print(f"v1 + v2 = {v1 + v2}")
print(f"v1 - v2 = {v1 - v2}")
print(f"v1 * 2 = {v1 * 2}")
print(f"3 * v2 = {3 * v2}")
print(f"v1 / 2 = {v1 / 2}")
print(f"-v1 = {-v1}")
print()

# Built-in functions
print(f"abs(v1) = {abs(v1)}")  # Magnitude: sqrt(9 + 16) = 5
print(f"len(v1) = {len(v1)}")
print(f"bool(v1) = {bool(v1)}")
print(f"bool(Vector(0, 0)) = {bool(Vector(0, 0))}")
print()

# Comparisons
print(f"v1 == v2: {v1 == v2}")
print(f"v1 == Vector(3, 4): {v1 == Vector(3, 4)}")
print(f"v1 > v2: {v1 > v2}")
print()

# Extra methods
print(f"v1.dot(v2) = {v1.dot(v2)}")
print(f"v1.normalize() = {v1.normalize()}")
````

### 5.4 VÃ­ dá»¥: Money Class

````python
from functools import total_ordering

@total_ordering  # Tá»± Ä‘á»™ng táº¡o __le__, __gt__, __ge__ tá»« __eq__ vÃ  __lt__
class Money:
    """
    Money class vá»›i operator overloading.
    
    VÃ­ dá»¥:
        m1 = Money(100, "USD")
        m2 = Money(50, "USD")
        total = m1 + m2  # Money(150, "USD")
    """
    
    EXCHANGE_RATES = {
        "USD": 1.0,
        "EUR": 0.85,
        "GBP": 0.73,
        "VND": 24000.0,
        "JPY": 110.0
    }
    
    def __init__(self, amount, currency="USD"):
        if currency not in self.EXCHANGE_RATES:
            raise ValueError(f"Unknown currency: {currency}")
        self.amount = amount
        self.currency = currency
    
    def to_usd(self):
        """Convert to USD."""
        return self.amount / self.EXCHANGE_RATES[self.currency]
    
    def convert_to(self, target_currency):
        """Convert to another currency."""
        if target_currency not in self.EXCHANGE_RATES:
            raise ValueError(f"Unknown currency: {target_currency}")
        usd_amount = self.to_usd()
        new_amount = usd_amount * self.EXCHANGE_RATES[target_currency]
        return Money(new_amount, target_currency)
    
    # === Arithmetic Operators ===
    
    def __add__(self, other):
        """Money + Money (cÃ¹ng currency) hoáº·c Money + sá»‘."""
        if isinstance(other, Money):
            if self.currency != other.currency:
                # Convert other to self's currency
                other = other.convert_to(self.currency)
            return Money(self.amount + other.amount, self.currency)
        elif isinstance(other, (int, float)):
            return Money(self.amount + other, self.currency)
        return NotImplemented
    
    def __radd__(self, other):
        return self.__add__(other)
    
    def __sub__(self, other):
        if isinstance(other, Money):
            if self.currency != other.currency:
                other = other.convert_to(self.currency)
            return Money(self.amount - other.amount, self.currency)
        elif isinstance(other, (int, float)):
            return Money(self.amount - other, self.currency)
        return NotImplemented
    
    def __mul__(self, factor):
        """Money * factor."""
        if isinstance(factor, (int, float)):
            return Money(self.amount * factor, self.currency)
        return NotImplemented
    
    def __rmul__(self, factor):
        return self.__mul__(factor)
    
    def __truediv__(self, factor):
        """Money / factor."""
        if isinstance(factor, (int, float)):
            if factor == 0:
                raise ZeroDivisionError("Cannot divide money by zero")
            return Money(self.amount / factor, self.currency)
        return NotImplemented
    
    def __neg__(self):
        return Money(-self.amount, self.currency)
    
    def __abs__(self):
        return Money(abs(self.amount), self.currency)
    
    # === Comparison Operators ===
    
    def __eq__(self, other):
        if isinstance(other, Money):
            return abs(self.to_usd() - other.to_usd()) < 0.01
        return NotImplemented
    
    def __lt__(self, other):
        if isinstance(other, Money):
            return self.to_usd() < other.to_usd()
        return NotImplemented
    
    # === String Representation ===
    
    def __str__(self):
        return f"{self.currency} {self.amount:,.2f}"
    
    def __repr__(self):
        return f"Money({self.amount!r}, {self.currency!r})"
    
    # === Type Conversion ===
    
    def __float__(self):
        return float(self.amount)
    
    def __int__(self):
        return int(self.amount)

# Demo
print("Money Class vá»›i Operator Overloading:")
print("=" * 50)

m1 = Money(100, "USD")
m2 = Money(50, "USD")
m3 = Money(85, "EUR")
m4 = Money(2400000, "VND")

print(f"m1 = {m1}")
print(f"m2 = {m2}")
print(f"m3 = {m3}")
print(f"m4 = {m4}")
print()

# Arithmetic
print(f"m1 + m2 = {m1 + m2}")
print(f"m1 - m2 = {m1 - m2}")
print(f"m1 * 2 = {m1 * 2}")
print(f"m1 / 4 = {m1 / 4}")
print()

# Cross-currency operations (auto convert)
print(f"m1 (USD) + m3 (EUR) = {m1 + m3}")
print()

# Comparisons
print(f"m1 > m2: {m1 > m2}")
print(f"m3 (EUR) == USD 100: {m3 == Money(100, 'USD')}")
print()

# Conversion
print(f"m1.convert_to('EUR') = {m1.convert_to('EUR')}")
print(f"m4.convert_to('USD') = {m4.convert_to('USD')}")
````

## 6. Magic Methods chi tiáº¿t

### 6.1 Container Magic Methods

| Magic Method | MÃ´ táº£ |
|--------------|-------|
| `__len__(self)` | `len(obj)` |
| `__getitem__(self, key)` | `obj[key]` |
| `__setitem__(self, key, value)` | `obj[key] = value` |
| `__delitem__(self, key)` | `del obj[key]` |
| `__contains__(self, item)` | `item in obj` |
| `__iter__(self)` | `for item in obj` |
| `__next__(self)` | `next(obj)` |

````python
# Custom Container vá»›i Magic Methods

class Playlist:
    """
    Playlist class hoáº¡t Ä‘á»™ng nhÆ° má»™t container.
    
    VÃ­ dá»¥:
        playlist = Playlist("Rock Classics")
        playlist.add("Bohemian Rhapsody")
        print(playlist[0])  # "Bohemian Rhapsody"
        print(len(playlist))  # 1
    """
    
    def __init__(self, name):
        self.name = name
        self._songs = []
    
    def add(self, song):
        """ThÃªm bÃ i hÃ¡t."""
        self._songs.append(song)
    
    # === Container Magic Methods ===
    
    def __len__(self):
        """len(playlist)."""
        return len(self._songs)
    
    def __getitem__(self, index):
        """playlist[index] hoáº·c playlist[start:end]."""
        return self._songs[index]
    
    def __setitem__(self, index, value):
        """playlist[index] = value."""
        self._songs[index] = value
    
    def __delitem__(self, index):
        """del playlist[index]."""
        del self._songs[index]
    
    def __contains__(self, song):
        """song in playlist."""
        return song in self._songs
    
    def __iter__(self):
        """for song in playlist."""
        return iter(self._songs)
    
    def __reversed__(self):
        """reversed(playlist)."""
        return reversed(self._songs)
    
    # === Operators ===
    
    def __add__(self, other):
        """playlist1 + playlist2."""
        if isinstance(other, Playlist):
            new_playlist = Playlist(f"{self.name} + {other.name}")
            new_playlist._songs = self._songs + other._songs
            return new_playlist
        return NotImplemented
    
    def __str__(self):
        songs_str = "\n  ".join(f"{i+1}. {song}" for i, song in enumerate(self._songs))
        return f"Playlist: {self.name}\n  {songs_str}"

# Demo
print("Custom Container Demo:")
print("=" * 50)

rock = Playlist("Rock Classics")
rock.add("Bohemian Rhapsody")
rock.add("Stairway to Heaven")
rock.add("Hotel California")

pop = Playlist("Pop Hits")
pop.add("Billie Jean")
pop.add("Like a Prayer")

print(rock)
print()

# Container operations
print(f"len(rock) = {len(rock)}")
print(f"rock[0] = {rock[0]}")
print(f"rock[1:3] = {rock[1:3]}")
print(f"'Hotel California' in rock: {'Hotel California' in rock}")
print()

# Iteration
print("Iterating:")
for song in rock:
    print(f"  â™ª {song}")

print()

# Combine playlists
combined = rock + pop
print(combined)
````

### 6.2 Callable Magic Method

````python
# __call__ - LÃ m object cÃ³ thá»ƒ gá»i nhÆ° function

class Counter:
    """
    Counter cÃ³ thá»ƒ gá»i nhÆ° function.
    
    VÃ­ dá»¥:
        counter = Counter()
        counter()  # 1
        counter()  # 2
    """
    
    def __init__(self, start=0):
        self.count = start
    
    def __call__(self, increment=1):
        """Gá»i object nhÆ° function."""
        self.count += increment
        return self.count

class Multiplier:
    """Function factory - táº¡o cÃ¡c multiplier functions."""
    
    def __init__(self, factor):
        self.factor = factor
    
    def __call__(self, x):
        return x * self.factor

class Pipeline:
    """Function pipeline - chain cÃ¡c functions."""
    
    def __init__(self, *functions):
        self.functions = functions
    
    def __call__(self, x):
        result = x
        for func in self.functions:
            result = func(result)
        return result

# Demo
print("Callable Objects Demo:")
print("=" * 50)

# Counter
counter = Counter()
print(f"counter() = {counter()}")
print(f"counter() = {counter()}")
print(f"counter(5) = {counter(5)}")
print()

# Multiplier
double = Multiplier(2)
triple = Multiplier(3)
print(f"double(5) = {double(5)}")
print(f"triple(5) = {triple(5)}")
print()

# Pipeline
process = Pipeline(
    lambda x: x + 1,      # Add 1
    lambda x: x * 2,      # Multiply by 2
    lambda x: x ** 2      # Square
)
print(f"Pipeline: (5 + 1) * 2 = 12, then 12Â² = 144")
print(f"process(5) = {process(5)}")
````

### 6.3 Context Manager Magic Methods

````python
# __enter__ vÃ  __exit__ - Context Manager Protocol

import time

class Timer:
    """
    Timer context manager Ä‘á»ƒ Ä‘o thá»i gian.
    
    VÃ­ dá»¥:
        with Timer("My operation"):
            # code here
    """
    
    def __init__(self, name="Timer"):
        self.name = name
        self.start = None
        self.end = None
    
    def __enter__(self):
        """Called when entering 'with' block."""
        self.start = time.time()
        print(f"[{self.name}] Starting...")
        return self  # CÃ³ thá»ƒ return object Ä‘á»ƒ dÃ¹ng vá»›i 'as'
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        """Called when exiting 'with' block."""
        self.end = time.time()
        elapsed = self.end - self.start
        print(f"[{self.name}] Finished in {elapsed:.4f} seconds")
        
        # Return True Ä‘á»ƒ suppress exception, False Ä‘á»ƒ propagate
        return False
    
    @property
    def elapsed(self):
        if self.start and self.end:
            return self.end - self.start
        return None

class DatabaseConnection:
    """Simulated database connection."""
    
    def __init__(self, host):
        self.host = host
        self.connected = False
    
    def __enter__(self):
        self.connected = True
        print(f"Connected to {self.host}")
        return self
    
    def __exit__(self, exc_type, exc_val, exc_tb):
        self.connected = False
        print(f"Disconnected from {self.host}")
        return False
    
    def query(self, sql):
        if not self.connected:
            raise RuntimeError("Not connected!")
        return f"Results for: {sql}"

# Demo
print("Context Manager Demo:")
print("=" * 50)

# Timer
with Timer("Sleep operation"):
    time.sleep(0.1)
    print("  Working...")

print()

# Database
with DatabaseConnection("localhost:5432") as db:
    result = db.query("SELECT * FROM users")
    print(f"  {result}")
````

### 7.1 Abstract Base Classes (ABC)

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract base class cho cÃ¡c hÃ¬nh."""
    
    @abstractmethod
    def area(self):
        """Calculate area - MUST be implemented."""
        pass
    
    @abstractmethod
    def perimeter(self):
        """Calculate perimeter - MUST be implemented."""
        pass
    
    def describe(self):
        """Non-abstract method - optional override."""
        return f"{self.__class__.__name__}: Area={self.area():.2f}, Perimeter={self.perimeter():.2f}"

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        import math
        return math.pi * self.radius ** 2
    
    def perimeter(self):
        import math
        return 2 * math.pi * self.radius

# Cannot instantiate ABC
try:
    shape = Shape()
except TypeError as e:
    print(f"Cannot create Shape: {e}")

print()

# Polymorphism vá»›i ABC
shapes = [
    Rectangle(5, 3),
    Circle(4),
    Rectangle(10, 2)
]

for shape in shapes:
    print(shape.describe())
````

### 7.2 Protocol (Structural Subtyping) - Python 3.8+

````python
from typing import Protocol, runtime_checkable

@runtime_checkable
class Drawable(Protocol):
    """Protocol - Ä‘á»‹nh nghÄ©a interface theo structure."""
    
    def draw(self) -> str:
        """Draw the object."""
        ...

@runtime_checkable
class Resizable(Protocol):
    """Protocol cho objects cÃ³ thá»ƒ resize."""
    
    def resize(self, factor: float) -> None:
        ...

# CÃ¡c classes KHÃ”NG cáº§n explicitly inherit tá»« Protocol
class Square:
    """Square implements Drawable vÃ  Resizable implicitly."""
    
    def __init__(self, side):
        self.side = side
    
    def draw(self) -> str:
        return f"Drawing square with side {self.side}"
    
    def resize(self, factor: float) -> None:
        self.side *= factor

class Text:
    """Text chá»‰ implements Drawable."""
    
    def __init__(self, content):
        self.content = content
    
    def draw(self) -> str:
        return f"Drawing text: '{self.content}'"

def render(item: Drawable) -> None:
    """Function yÃªu cáº§u Drawable protocol."""
    print(item.draw())

def scale(item: Resizable, factor: float) -> None:
    """Function yÃªu cáº§u Resizable protocol."""
    item.resize(factor)

# Demo
print("Protocol Demo:")
print("=" * 50)

square = Square(10)
text = Text("Hello World")

# Cáº£ hai Ä‘á»u Drawable
print(f"isinstance(square, Drawable): {isinstance(square, Drawable)}")
print(f"isinstance(text, Drawable): {isinstance(text, Drawable)}")
print()

# Chá»‰ Square lÃ  Resizable
print(f"isinstance(square, Resizable): {isinstance(square, Resizable)}")
print(f"isinstance(text, Resizable): {isinstance(text, Resizable)}")
print()

# Render both
render(square)
render(text)

# Scale only square
scale(square, 2)
render(square)
````

````python
class Temperature:
    """
    Temperature class hoáº¡t Ä‘á»™ng vá»›i nhiá»u built-in functions.
    """
    
    def __init__(self, celsius):
        self._celsius = celsius
    
    @property
    def celsius(self):
        return self._celsius
    
    @property
    def fahrenheit(self):
        return self._celsius * 9/5 + 32
    
    @property
    def kelvin(self):
        return self._celsius + 273.15
    
    # === Built-in function support ===
    
    def __str__(self):
        """str(temp)."""
        return f"{self._celsius}Â°C"
    
    def __repr__(self):
        """repr(temp)."""
        return f"Temperature({self._celsius})"
    
    def __float__(self):
        """float(temp)."""
        return float(self._celsius)
    
    def __int__(self):
        """int(temp)."""
        return int(self._celsius)
    
    def __bool__(self):
        """bool(temp) - True if above freezing."""
        return self._celsius > 0
    
    def __abs__(self):
        """abs(temp)."""
        return Temperature(abs(self._celsius))
    
    def __round__(self, n=0):
        """round(temp, n)."""
        return Temperature(round(self._celsius, n))
    
    def __hash__(self):
        """hash(temp) - Ä‘á»ƒ cÃ³ thá»ƒ dÃ¹ng trong set/dict."""
        return hash(self._celsius)
    
    def __format__(self, format_spec):
        """format(temp, spec) hoáº·c f'{temp:spec}'."""
        if format_spec == 'f':
            return f"{self.fahrenheit:.1f}Â°F"
        elif format_spec == 'k':
            return f"{self.kelvin:.1f}K"
        else:
            return f"{self._celsius}Â°C"
    
    # === Comparison ===
    
    def __eq__(self, other):
        if isinstance(other, Temperature):
            return self._celsius == other._celsius
        return self._celsius == other
    
    def __lt__(self, other):
        if isinstance(other, Temperature):
            return self._celsius < other._celsius
        return self._celsius < other
    
    def __le__(self, other):
        return self == other or self < other
    
    def __gt__(self, other):
        if isinstance(other, Temperature):
            return self._celsius > other._celsius
        return self._celsius > other
    
    def __ge__(self, other):
        return self == other or self > other
    
    # === Arithmetic ===
    
    def __add__(self, other):
        if isinstance(other, Temperature):
            return Temperature(self._celsius + other._celsius)
        return Temperature(self._celsius + other)
    
    def __sub__(self, other):
        if isinstance(other, Temperature):
            return Temperature(self._celsius - other._celsius)
        return Temperature(self._celsius - other)

# Demo
print("Temperature Class vá»›i Built-in Functions:")
print("=" * 50)

temp = Temperature(25.7)

print(f"str(temp) = {str(temp)}")
print(f"repr(temp) = {repr(temp)}")
print(f"float(temp) = {float(temp)}")
print(f"int(temp) = {int(temp)}")
print(f"bool(temp) = {bool(temp)}")
print(f"bool(Temperature(-5)) = {bool(Temperature(-5))}")
print(f"abs(Temperature(-10)) = {abs(Temperature(-10))}")
print(f"round(temp) = {round(temp)}")
print()

# Format
print(f"f'{{temp}}' = {temp}")
print(f"f'{{temp:f}}' = {temp:f}")
print(f"f'{{temp:k}}' = {temp:k}")
print()

# Comparison
cold = Temperature(10)
hot = Temperature(35)
print(f"{cold} < {hot}: {cold < hot}")

# Hashable - cÃ³ thá»ƒ dÃ¹ng trong set
temp_set = {Temperature(20), Temperature(25), Temperature(20)}
print(f"Set of temperatures: {temp_set}")
````

### 9.1 Payment Processing System

````python
from abc import ABC, abstractmethod
from datetime import datetime

class PaymentProcessor(ABC):
    """Abstract base class cho payment processors."""
    
    @abstractmethod
    def validate(self) -> bool:
        """Validate payment details."""
        pass
    
    @abstractmethod
    def process(self, amount: float) -> dict:
        """Process the payment."""
        pass
    
    @abstractmethod
    def refund(self, transaction_id: str) -> dict:
        """Refund a payment."""
        pass
    
    def _generate_transaction_id(self) -> str:
        """Generate unique transaction ID."""
        import random
        return f"TXN{datetime.now().strftime('%Y%m%d%H%M%S')}{random.randint(1000, 9999)}"

class CreditCardProcessor(PaymentProcessor):
    """Credit card payment processor."""
    
    def __init__(self, card_number: str, cvv: str, expiry: str, holder: str):
        self.card_number = card_number
        self.cvv = cvv
        self.expiry = expiry
        self.holder = holder
    
    def validate(self) -> bool:
        """Validate card details."""
        is_valid = (
            len(self.card_number) == 16 and
            len(self.cvv) == 3 and
            len(self.expiry) == 5  # MM/YY
        )
        print(f"[CreditCard] Validating card ending in {self.card_number[-4:]}: {'Valid' if is_valid else 'Invalid'}")
        return is_valid
    
    def process(self, amount: float) -> dict:
        """Process credit card payment."""
        if not self.validate():
            return {"success": False, "error": "Invalid card"}
        
        txn_id = self._generate_transaction_id()
        print(f"[CreditCard] Charging ${amount:.2f} to card ***{self.card_number[-4:]}")
        
        return {
            "success": True,
            "transaction_id": txn_id,
            "amount": amount,
            "method": "credit_card",
            "last_four": self.card_number[-4:]
        }
    
    def refund(self, transaction_id: str) -> dict:
        print(f"[CreditCard] Refunding transaction {transaction_id}")
        return {"success": True, "refund_id": f"REF_{transaction_id}"}

class PayPalProcessor(PaymentProcessor):
    """PayPal payment processor."""
    
    def __init__(self, email: str):
        self.email = email
    
    def validate(self) -> bool:
        is_valid = "@" in self.email and "." in self.email
        print(f"[PayPal] Validating {self.email}: {'Valid' if is_valid else 'Invalid'}")
        return is_valid
    
    def process(self, amount: float) -> dict:
        if not self.validate():
            return {"success": False, "error": "Invalid PayPal email"}
        
        txn_id = self._generate_transaction_id()
        print(f"[PayPal] Charging ${amount:.2f} to {self.email}")
        
        return {
            "success": True,
            "transaction_id": txn_id,
            "amount": amount,
            "method": "paypal",
            "email": self.email
        }
    
    def refund(self, transaction_id: str) -> dict:
        print(f"[PayPal] Refunding transaction {transaction_id}")
        return {"success": True, "refund_id": f"PP_REF_{transaction_id}"}

class CryptoProcessor(PaymentProcessor):
    """Cryptocurrency payment processor."""
    
    def __init__(self, wallet_address: str, currency: str = "BTC"):
        self.wallet_address = wallet_address
        self.currency = currency
    
    def validate(self) -> bool:
        is_valid = len(self.wallet_address) >= 26
        print(f"[Crypto] Validating {self.currency} wallet: {'Valid' if is_valid else 'Invalid'}")
        return is_valid
    
    def process(self, amount: float) -> dict:
        if not self.validate():
            return {"success": False, "error": "Invalid wallet"}
        
        txn_id = self._generate_transaction_id()
        print(f"[Crypto] Sending ${amount:.2f} in {self.currency}")
        
        return {
            "success": True,
            "transaction_id": txn_id,
            "amount": amount,
            "method": "crypto",
            "currency": self.currency
        }
    
    def refund(self, transaction_id: str) -> dict:
        print(f"[Crypto] Refunding transaction {transaction_id}")
        return {"success": True, "refund_id": f"CRYPTO_REF_{transaction_id}"}

# Polymorphic function
def checkout(processor: PaymentProcessor, amount: float) -> dict:
    """
    Polymorphic checkout - hoáº¡t Ä‘á»™ng vá»›i báº¥t ká»³ PaymentProcessor nÃ o.
    """
    print(f"\nProcessing ${amount:.2f} payment...")
    print("-" * 40)
    return processor.process(amount)

# Demo
print("Payment Processing System Demo")
print("=" * 50)

# Different payment methods
processors = [
    CreditCardProcessor("1234567890123456", "123", "12/25", "John Doe"),
    PayPalProcessor("john@email.com"),
    CryptoProcessor("1A2b3C4d5E6f7G8h9I0jKlMnOpQrStUv", "ETH")
]

# Process payments polymorphically
for processor in processors:
    result = checkout(processor, 99.99)
    print(f"Result: {result}")
````

### 9.2 Notification System

````python
from abc import ABC, abstractmethod
from datetime import datetime

class Notification(ABC):
    """Abstract base class cho notifications."""
    
    def __init__(self, recipient: str, message: str):
        self.recipient = recipient
        self.message = message
        self.timestamp = datetime.now()
        self.sent = False
    
    @abstractmethod
    def send(self) -> bool:
        """Send the notification."""
        pass
    
    @abstractmethod
    def get_delivery_status(self) -> str:
        """Get delivery status."""
        pass
    
    def __str__(self):
        return f"{self.__class__.__name__} to {self.recipient}"

class EmailNotification(Notification):
    """Email notification."""
    
    def __init__(self, recipient: str, message: str, subject: str = "Notification"):
        super().__init__(recipient, message)
        self.subject = subject
    
    def send(self) -> bool:
        print(f"ğŸ“§ Sending email to {self.recipient}")
        print(f"   Subject: {self.subject}")
        print(f"   Body: {self.message[:50]}...")
        self.sent = True
        return True
    
    def get_delivery_status(self) -> str:
        return "Delivered to inbox" if self.sent else "Pending"

class SMSNotification(Notification):
    """SMS notification."""
    
    MAX_LENGTH = 160
    
    def __init__(self, recipient: str, message: str):
        # Truncate message if too long
        if len(message) > self.MAX_LENGTH:
            message = message[:self.MAX_LENGTH - 3] + "..."
        super().__init__(recipient, message)
    
    def send(self) -> bool:
        print(f"ğŸ“± Sending SMS to {self.recipient}")
        print(f"   Message: {self.message}")
        self.sent = True
        return True
    
    def get_delivery_status(self) -> str:
        return "Delivered" if self.sent else "Pending"

class PushNotification(Notification):
    """Mobile push notification."""
    
    def __init__(self, recipient: str, message: str, title: str, icon: str = "ğŸ””"):
        super().__init__(recipient, message)
        self.title = title
        self.icon = icon
    
    def send(self) -> bool:
        print(f"{self.icon} Sending push to {self.recipient}")
        print(f"   Title: {self.title}")
        print(f"   Body: {self.message}")
        self.sent = True
        return True
    
    def get_delivery_status(self) -> str:
        return "Received" if self.sent else "Pending"

class SlackNotification(Notification):
    """Slack notification."""
    
    def __init__(self, recipient: str, message: str, channel: str = "#general"):
        super().__init__(recipient, message)
        self.channel = channel
    
    def send(self) -> bool:
        print(f"ğŸ’¬ Sending Slack message to {self.channel}")
        print(f"   @{self.recipient}: {self.message}")
        self.sent = True
        return True
    
    def get_delivery_status(self) -> str:
        return "Posted" if self.sent else "Pending"

class NotificationService:
    """Service Ä‘á»ƒ gá»­i notifications."""
    
    def __init__(self):
        self.notifications = []
    
    def send(self, notification: Notification) -> bool:
        """Polymorphic send - hoáº¡t Ä‘á»™ng vá»›i báº¥t ká»³ Notification nÃ o."""
        result = notification.send()
        self.notifications.append(notification)
        return result
    
    def send_all(self, notifications: list) -> dict:
        """Send multiple notifications."""
        results = {"success": 0, "failed": 0}
        for notification in notifications:
            if self.send(notification):
                results["success"] += 1
            else:
                results["failed"] += 1
        return results
    
    def get_report(self) -> str:
        """Get notification report."""
        report = "Notification Report:\n" + "=" * 40 + "\n"
        for notif in self.notifications:
            report += f"{notif} - Status: {notif.get_delivery_status()}\n"
        return report

# Demo
print("Notification System Demo")
print("=" * 50)

service = NotificationService()

notifications = [
    EmailNotification("user@example.com", "Your order has been shipped!", "Order Update"),
    SMSNotification("+1234567890", "Your verification code is 123456"),
    PushNotification("device_token_123", "New message from John", "New Message", "ğŸ’¬"),
    SlackNotification("john.doe", "Build #123 completed successfully", "#ci-cd")
]

# Send all notifications polymorphically
for notif in notifications:
    service.send(notif)
    print()

# Report
print(service.get_report())
````

### 11.1 QuÃªn implement reverse operators

````python
# âŒ MISTAKE: Chá»‰ cÃ³ __mul__, thiáº¿u __rmul__

class Vector2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __mul__(self, scalar):
        return Vector2D(self.x * scalar, self.y * scalar)
    
    def __str__(self):
        return f"Vector2D({self.x}, {self.y})"

v = Vector2D(3, 4)

print(f"v * 2 = {v * 2}")  # Works

try:
    print(f"2 * v = {2 * v}")  # Fails!
except TypeError as e:
    print(f"Error: {e}")
````

````python
# âœ… CORRECT: CÃ³ cáº£ __mul__ vÃ  __rmul__

class Vector2DFixed:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __mul__(self, scalar):
        if isinstance(scalar, (int, float)):
            return Vector2DFixed(self.x * scalar, self.y * scalar)
        return NotImplemented
    
    def __rmul__(self, scalar):
        """Handle: scalar * vector."""
        return self.__mul__(scalar)
    
    def __str__(self):
        return f"Vector2D({self.x}, {self.y})"

v = Vector2DFixed(3, 4)

print(f"v * 2 = {v * 2}")  # Works
print(f"2 * v = {2 * v}")  # Also works!
````

### 11.2 KhÃ´ng implement `__ne__` khi override `__eq__`

````python
# Python 3 tá»± Ä‘á»™ng táº¡o __ne__ tá»« __eq__, nhÆ°ng nÃªn explicit

class Item:
    def __init__(self, id, name):
        self.id = id
        self.name = name
    
    def __eq__(self, other):
        if isinstance(other, Item):
            return self.id == other.id
        return NotImplemented
    
    def __ne__(self, other):
        """Explicit __ne__ cho clarity."""
        result = self.__eq__(other)
        if result is NotImplemented:
            return result
        return not result
    
    def __hash__(self):
        """PHáº¢I implement __hash__ náº¿u implement __eq__."""
        return hash(self.id)

item1 = Item(1, "Apple")
item2 = Item(1, "Apple Fresh")
item3 = Item(2, "Banana")

print(f"item1 == item2: {item1 == item2}")  # True (same id)
print(f"item1 != item3: {item1 != item3}")  # True

# CÃ³ thá»ƒ dÃ¹ng trong set/dict vÃ¬ cÃ³ __hash__
items = {item1, item2, item3}
print(f"Items in set: {len(items)}")  # 2 (item1 vÃ  item2 trÃ¹ng)
````
