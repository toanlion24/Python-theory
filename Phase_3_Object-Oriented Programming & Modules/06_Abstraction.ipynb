```
â”‚ â€¢ "What to do"              â”‚ â€¢ "How to protect"                â”‚
```

````python
# Abstraction + Encapsulation working together

from abc import ABC, abstractmethod

# ABSTRACTION: Äá»‹nh nghÄ©a WHAT
class Database(ABC):
    """Abstract interface - chá»‰ Ä‘á»‹nh nghÄ©a WHAT."""
    
    @abstractmethod
    def connect(self):
        """Connect to database."""
        pass
    
    @abstractmethod
    def query(self, sql):
        """Execute query."""
        pass

# ENCAPSULATION + Implementation: Äá»‹nh nghÄ©a HOW
class MySQLDatabase(Database):
    """Concrete implementation vá»›i encapsulation."""
    
    def __init__(self, host, port, user, password):
        # ENCAPSULATION: Báº£o vá»‡ data
        self.__host = host
        self.__port = port
        self.__user = user
        self.__password = password  # Private!
        self.__connection = None
    
    def connect(self):
        # Implementation details hidden
        self.__connection = f"MySQL://{self.__user}@{self.__host}:{self.__port}"
        return f"Connected to MySQL at {self.__host}"
    
    def query(self, sql):
        if not self.__connection:
            raise RuntimeError("Not connected")
        return f"MySQL executing: {sql}"

# User chá»‰ cáº§n biáº¿t interface
db = MySQLDatabase("localhost", 3306, "admin", "secret123")
print(db.connect())
print(db.query("SELECT * FROM users"))

# Cannot access internal details
try:
    print(db.__password)  # Hidden by encapsulation
except AttributeError as e:
    print(f"Cannot access private data: {e}")
````

### 2.2 Äáº·c Ä‘iá»ƒm cá»§a ABC

| Äáº·c Ä‘iá»ƒm | MÃ´ táº£ |
|----------|-------|
| **Cannot instantiate** | KhÃ´ng thá»ƒ táº¡o instance trá»±c tiáº¿p |
| **Force implementation** | Subclass PHáº¢I implement abstract methods |
| **Can have concrete methods** | CÃ³ thá»ƒ cÃ³ methods vá»›i implementation |
| **Can have attributes** | CÃ³ thá»ƒ cÃ³ attributes |

````python
from abc import ABC, abstractmethod

class Animal(ABC):
    """
    Abstract Base Class for animals.
    
    Äáº·c Ä‘iá»ƒm:
    - KhÃ´ng thá»ƒ táº¡o instance trá»±c tiáº¿p
    - CÃ³ abstract methods pháº£i Ä‘Æ°á»£c implement
    - CÃ³ concrete methods cÃ³ thá»ƒ dÃ¹ng ngay
    """
    
    def __init__(self, name):
        self.name = name
    
    # Abstract method - PHáº¢I implement
    @abstractmethod
    def speak(self):
        """Make animal sound - must be implemented."""
        pass
    
    @abstractmethod
    def move(self):
        """Move the animal - must be implemented."""
        pass
    
    # Concrete method - cÃ³ implementation sáºµn
    def sleep(self):
        """All animals can sleep - concrete method."""
        return f"{self.name} is sleeping... Zzz"
    
    def describe(self):
        """Describe the animal - concrete method."""
        return f"{self.name} is a {self.__class__.__name__}"

# Thá»­ táº¡o instance tá»« ABC
print("Trying to instantiate ABC:")
print("=" * 50)

try:
    animal = Animal("Generic")  # ERROR!
except TypeError as e:
    print(f"Cannot instantiate ABC: {e}")
````

````python
# Concrete class pháº£i implement Táº¤T Cáº¢ abstract methods

class Dog(Animal):
    """Concrete class - implements all abstract methods."""
    
    def speak(self):
        return f"{self.name} says: Woof! Woof! ğŸ•"
    
    def move(self):
        return f"{self.name} runs on 4 legs"
    
    def fetch(self):
        """Dog-specific method."""
        return f"{self.name} fetches the ball!"

class Cat(Animal):
    def speak(self):
        return f"{self.name} says: Meow! ğŸ±"
    
    def move(self):
        return f"{self.name} walks gracefully"
    
    def purr(self):
        return f"{self.name} purrs happily"

class Bird(Animal):
    def speak(self):
        return f"{self.name} says: Tweet! ğŸ¦"
    
    def move(self):
        return f"{self.name} flies through the sky"

# Sá»­ dá»¥ng
print("\nConcrete classes:")
print("=" * 50)

animals = [
    Dog("Buddy"),
    Cat("Whiskers"),
    Bird("Tweety")
]

for animal in animals:
    print(f"\n{animal.describe()}")
    print(f"  {animal.speak()}")
    print(f"  {animal.move()}")
    print(f"  {animal.sleep()}")  # Tá»« concrete method
````

````python
# Incomplete implementation - KHÃ”NG thá»ƒ instantiate

class IncompleteAnimal(Animal):
    """Chá»‰ implement speak(), thiáº¿u move()."""
    
    def speak(self):
        return "Some sound"
    # Thiáº¿u move()!

print("Incomplete implementation:")
print("=" * 50)

try:
    incomplete = IncompleteAnimal("Incomplete")
except TypeError as e:
    print(f"Error: {e}")
````

### 3.2 Abstract Method vá»›i Default Implementation

````python
from abc import ABC, abstractmethod

class Logger(ABC):
    """
    Abstract Logger vá»›i default implementation.
    
    Abstract method CÃ“ THá»‚ cÃ³ default implementation,
    subclass váº«n PHáº¢I override nhÆ°ng cÃ³ thá»ƒ gá»i super().
    """
    
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def log(self, message, level="INFO"):
        """
        Abstract method vá»›i default implementation.
        
        Subclass PHáº¢I override, nhÆ°ng cÃ³ thá»ƒ dÃ¹ng super()
        Ä‘á»ƒ láº¥y formatted message.
        """
        from datetime import datetime
        timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        return f"[{timestamp}] [{level}] [{self.name}] {message}"
    
    @abstractmethod
    def close(self):
        """Close logger resources."""
        pass

class ConsoleLogger(Logger):
    """Log to console."""
    
    def log(self, message, level="INFO"):
        # Gá»i super() Ä‘á»ƒ láº¥y formatted message
        formatted = super().log(message, level)
        print(f"CONSOLE: {formatted}")
        return formatted
    
    def close(self):
        print("Console logger closed")

class FileLogger(Logger):
    """Log to file."""
    
    def __init__(self, name, filename):
        super().__init__(name)
        self.filename = filename
        self._logs = []  # Simulate file
    
    def log(self, message, level="INFO"):
        formatted = super().log(message, level)
        self._logs.append(formatted)
        print(f"FILE ({self.filename}): {formatted}")
        return formatted
    
    def close(self):
        print(f"File logger closed, wrote {len(self._logs)} entries")

# Test
print("Abstract Method with Default Implementation:")
print("=" * 60)

console = ConsoleLogger("App")
file_log = FileLogger("App", "app.log")

console.log("Application started")
file_log.log("User logged in", "DEBUG")

print()
console.close()
file_log.close()
````

### 3.3 Abstract Static vÃ  Class Methods

````python
from abc import ABC, abstractmethod

class Serializer(ABC):
    """
    Abstract class vá»›i abstract static vÃ  class methods.
    
    LÆ°u Ã½: @abstractmethod pháº£i Ä‘áº·t SAU @staticmethod/@classmethod
    """
    
    @classmethod
    @abstractmethod
    def get_format(cls):
        """Return format name."""
        pass
    
    @staticmethod
    @abstractmethod
    def serialize(data):
        """Serialize data to string."""
        pass
    
    @staticmethod
    @abstractmethod
    def deserialize(text):
        """Deserialize string to data."""
        pass

class JSONSerializer(Serializer):
    @classmethod
    def get_format(cls):
        return "JSON"
    
    @staticmethod
    def serialize(data):
        import json
        return json.dumps(data, indent=2)
    
    @staticmethod
    def deserialize(text):
        import json
        return json.loads(text)

class XMLSerializer(Serializer):
    @classmethod
    def get_format(cls):
        return "XML"
    
    @staticmethod
    def serialize(data):
        # Simplified XML
        result = "<data>\n"
        for key, value in data.items():
            result += f"  <{key}>{value}</{key}>\n"
        result += "</data>"
        return result
    
    @staticmethod
    def deserialize(text):
        # Simplified parsing
        return {"parsed": "from XML"}

# Test
print("Abstract Static/Class Methods:")
print("=" * 50)

data = {"name": "John", "age": 30, "city": "NYC"}

for SerializerClass in [JSONSerializer, XMLSerializer]:
    print(f"\n{SerializerClass.get_format()}:")
    print(SerializerClass.serialize(data))
````

````python
from abc import ABC, abstractmethod

class Vehicle(ABC):
    """
    Abstract Vehicle vá»›i abstract properties.
    """
    
    def __init__(self, brand, model):
        self._brand = brand
        self._model = model
    
    # Concrete properties
    @property
    def brand(self):
        return self._brand
    
    @property
    def model(self):
        return self._model
    
    # Abstract properties - pháº£i implement
    @property
    @abstractmethod
    def max_speed(self):
        """Maximum speed in km/h."""
        pass
    
    @property
    @abstractmethod
    def fuel_type(self):
        """Type of fuel used."""
        pass
    
    @property
    @abstractmethod
    def num_wheels(self):
        """Number of wheels."""
        pass
    
    # Abstract method
    @abstractmethod
    def start(self):
        pass
    
    # Concrete method sá»­ dá»¥ng abstract properties
    def get_specs(self):
        return (
            f"{self._brand} {self._model}\n"
            f"  Max Speed: {self.max_speed} km/h\n"
            f"  Fuel: {self.fuel_type}\n"
            f"  Wheels: {self.num_wheels}"
        )

class Car(Vehicle):
    @property
    def max_speed(self):
        return 220
    
    @property
    def fuel_type(self):
        return "Gasoline"
    
    @property
    def num_wheels(self):
        return 4
    
    def start(self):
        return f"{self._brand} {self._model} engine starting... Vroom!"

class Motorcycle(Vehicle):
    @property
    def max_speed(self):
        return 180
    
    @property
    def fuel_type(self):
        return "Gasoline"
    
    @property
    def num_wheels(self):
        return 2
    
    def start(self):
        return f"{self._brand} {self._model} engine revving... Braaap!"

class ElectricScooter(Vehicle):
    @property
    def max_speed(self):
        return 25
    
    @property
    def fuel_type(self):
        return "Electric"
    
    @property
    def num_wheels(self):
        return 2
    
    def start(self):
        return f"{self._brand} {self._model} powering on... *beep*"

# Test
print("Abstract Properties Demo:")
print("=" * 50)

vehicles = [
    Car("Toyota", "Camry"),
    Motorcycle("Harley-Davidson", "Iron 883"),
    ElectricScooter("Xiaomi", "Mi Pro 2")
]

for vehicle in vehicles:
    print(f"\n{vehicle.get_specs()}")
    print(f"  {vehicle.start()}")
````

### 4.2 Abstract Property vá»›i Setter

````python
from abc import ABC, abstractmethod

class ConfigurableDevice(ABC):
    """
    Abstract class vá»›i abstract property cÃ³ getter vÃ  setter.
    """
    
    @property
    @abstractmethod
    def volume(self):
        """Get current volume."""
        pass
    
    @volume.setter
    @abstractmethod
    def volume(self, value):
        """Set volume."""
        pass

class Speaker(ConfigurableDevice):
    def __init__(self):
        self._volume = 50
    
    @property
    def volume(self):
        return self._volume
    
    @volume.setter
    def volume(self, value):
        if not 0 <= value <= 100:
            raise ValueError("Volume must be 0-100")
        self._volume = value
        print(f"Speaker volume set to {value}%")

class Headphones(ConfigurableDevice):
    MAX_VOLUME = 80  # Safety limit
    
    def __init__(self):
        self._volume = 30
    
    @property
    def volume(self):
        return self._volume
    
    @volume.setter
    def volume(self, value):
        if not 0 <= value <= self.MAX_VOLUME:
            raise ValueError(f"Headphones volume limited to 0-{self.MAX_VOLUME}")
        self._volume = value
        print(f"Headphones volume set to {value}%")

# Test
print("Abstract Property with Setter:")
print("=" * 50)

speaker = Speaker()
headphones = Headphones()

print(f"\nSpeaker initial: {speaker.volume}%")
speaker.volume = 75

print(f"\nHeadphones initial: {headphones.volume}%")
headphones.volume = 60

# Test limit
try:
    headphones.volume = 100  # Over limit!
except ValueError as e:
    print(f"\nError: {e}")
````

````python
from abc import ABC, abstractmethod

# Interface: Chá»‰ cÃ³ abstract methods
class Drawable(ABC):
    """Interface for drawable objects."""
    
    @abstractmethod
    def draw(self):
        """Draw the object."""
        pass

class Resizable(ABC):
    """Interface for resizable objects."""
    
    @abstractmethod
    def resize(self, factor):
        """Resize by factor."""
        pass

class Movable(ABC):
    """Interface for movable objects."""
    
    @abstractmethod
    def move(self, x, y):
        """Move to position."""
        pass

class Clickable(ABC):
    """Interface for clickable objects."""
    
    @abstractmethod
    def on_click(self):
        """Handle click event."""
        pass

# Multiple interface implementation
class Button(Drawable, Resizable, Movable, Clickable):
    """Button implements multiple interfaces."""
    
    def __init__(self, x, y, width, height, label):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.label = label
    
    def draw(self):
        return f"Drawing button '{self.label}' at ({self.x}, {self.y}) size {self.width}x{self.height}"
    
    def resize(self, factor):
        self.width *= factor
        self.height *= factor
        return f"Button resized to {self.width}x{self.height}"
    
    def move(self, x, y):
        self.x = x
        self.y = y
        return f"Button moved to ({x}, {y})"
    
    def on_click(self):
        return f"Button '{self.label}' clicked!"

class Icon(Drawable, Movable):
    """Icon implements fewer interfaces."""
    
    def __init__(self, x, y, name):
        self.x = x
        self.y = y
        self.name = name
    
    def draw(self):
        return f"Drawing icon '{self.name}' at ({self.x}, {self.y})"
    
    def move(self, x, y):
        self.x = x
        self.y = y
        return f"Icon moved to ({x}, {y})"

# Test
print("Interface Pattern Demo:")
print("=" * 50)

button = Button(10, 20, 100, 30, "Submit")
icon = Icon(50, 50, "Settings")

# All drawables
drawables = [button, icon]
print("\nDrawing all:")
for d in drawables:
    print(f"  {d.draw()}")

# Only button is clickable
print(f"\n{button.on_click()}")

# Check interface
print(f"\nInterface checks:")
print(f"  button is Drawable: {isinstance(button, Drawable)}")
print(f"  button is Clickable: {isinstance(button, Clickable)}")
print(f"  icon is Clickable: {isinstance(icon, Clickable)}")
````

### 5.2 Protocol (Python 3.8+) - Structural Subtyping

Protocol lÃ  cÃ¡ch khÃ¡c Ä‘á»ƒ define interfaces, dá»±a trÃªn **structural typing** (duck typing vá»›i type hints).

````python
from typing import Protocol, runtime_checkable

@runtime_checkable
class Printable(Protocol):
    """
    Protocol - Structural subtyping.
    
    Class khÃ´ng cáº§n explicitly inherit,
    chá»‰ cáº§n cÃ³ method Ä‘Ãºng signature.
    """
    
    def print_content(self) -> str:
        ...

@runtime_checkable
class Saveable(Protocol):
    def save(self, filename: str) -> bool:
        ...

# Class KHÃ”NG cáº§n inherit tá»« Protocol
class Document:
    """Document class - khÃ´ng inherit tá»« Printable nhÆ°ng váº«n compatible."""
    
    def __init__(self, content):
        self.content = content
    
    def print_content(self) -> str:
        return f"Document: {self.content}"
    
    def save(self, filename: str) -> bool:
        print(f"Saved to {filename}")
        return True

class Image:
    """Image class - only implements Printable."""
    
    def __init__(self, path):
        self.path = path
    
    def print_content(self) -> str:
        return f"Image: {self.path}"

# Test
print("Protocol Demo:")
print("=" * 50)

doc = Document("Hello World")
img = Image("photo.jpg")

# isinstance works with @runtime_checkable
print(f"doc is Printable: {isinstance(doc, Printable)}")
print(f"doc is Saveable: {isinstance(doc, Saveable)}")
print(f"img is Printable: {isinstance(img, Printable)}")
print(f"img is Saveable: {isinstance(img, Saveable)}")

# Polymorphic function
def print_all(items: list[Printable]):
    for item in items:
        print(f"  {item.print_content()}")

print("\nPrinting all:")
print_all([doc, img])
````

### 6.2 Khi nÃ o dÃ¹ng gÃ¬?

```
â”‚   â€¢ Classes share common state (attributes)                     â”‚

â”‚                                                                 â”‚

â”‚   â€¢ Multiple inheritance needed                                 â”‚

â”‚                                                                 â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
```

````python
from abc import ABC, abstractmethod

# ABSTRACT CLASS: Animal vá»›i shared state vÃ  behavior
class Animal(ABC):
    """Abstract class - IS-A relationship."""
    
    def __init__(self, name, age):
        self.name = name  # Shared state
        self.age = age
    
    @abstractmethod
    def speak(self):
        pass
    
    def sleep(self):  # Concrete method
        return f"{self.name} is sleeping..."

# INTERFACES: Capabilities
class Flyable(ABC):
    """Interface - CAN-DO capability."""
    
    @abstractmethod
    def fly(self):
        pass

class Swimmable(ABC):
    """Interface - CAN-DO capability."""
    
    @abstractmethod
    def swim(self):
        pass

# Concrete classes
class Dog(Animal):
    """Dog IS-A Animal."""
    
    def speak(self):
        return f"{self.name} says: Woof!"

class Duck(Animal, Flyable, Swimmable):
    """Duck IS-A Animal, CAN Fly, CAN Swim."""
    
    def speak(self):
        return f"{self.name} says: Quack!"
    
    def fly(self):
        return f"{self.name} is flying!"
    
    def swim(self):
        return f"{self.name} is swimming!"

class Penguin(Animal, Swimmable):
    """Penguin IS-A Animal, CAN Swim (but NOT fly)."""
    
    def speak(self):
        return f"{self.name} says: Honk!"
    
    def swim(self):
        return f"{self.name} is swimming in cold water!"

# Test
print("Abstract Class + Interface:")
print("=" * 50)

dog = Dog("Buddy", 3)
duck = Duck("Donald", 2)
penguin = Penguin("Pingu", 4)

animals = [dog, duck, penguin]

for animal in animals:
    print(f"\n{animal.__class__.__name__} ({animal.name}):")
    print(f"  {animal.speak()}")
    print(f"  {animal.sleep()}")
    
    if isinstance(animal, Flyable):
        print(f"  {animal.fly()}")
    if isinstance(animal, Swimmable):
        print(f"  {animal.swim()}")
````

## 7. Design Patterns vá»›i Abstraction

### 7.1 Template Method Pattern

````python
from abc import ABC, abstractmethod

class DataProcessor(ABC):
    """
    Template Method Pattern.
    
    process() lÃ  template method Ä‘á»‹nh nghÄ©a skeleton.
    Subclasses implement cÃ¡c bÆ°á»›c cá»¥ thá»ƒ.
    """
    
    def process(self, data):
        """
        Template method - Ä‘á»‹nh nghÄ©a workflow.
        KHÃ”NG Ä‘Æ°á»£c override.
        """
        print(f"\n{'='*50}")
        print(f"Processing with {self.__class__.__name__}")
        print(f"{'='*50}")
        
        # Step 1: Load
        loaded = self.load(data)
        print(f"1. Loaded: {loaded}")
        
        # Step 2: Validate
        if not self.validate(loaded):
            raise ValueError("Validation failed")
        print(f"2. Validated: OK")
        
        # Step 3: Transform
        transformed = self.transform(loaded)
        print(f"3. Transformed: {transformed}")
        
        # Step 4: Save
        result = self.save(transformed)
        print(f"4. Saved: {result}")
        
        return result
    
    # Abstract methods - pháº£i implement
    @abstractmethod
    def load(self, data):
        pass
    
    @abstractmethod
    def validate(self, data):
        pass
    
    @abstractmethod
    def transform(self, data):
        pass
    
    @abstractmethod
    def save(self, data):
        pass

class JSONProcessor(DataProcessor):
    def load(self, data):
        return f"{{json: {data}}}"
    
    def validate(self, data):
        return "json" in data.lower()
    
    def transform(self, data):
        return data.upper()
    
    def save(self, data):
        return f"Saved to data.json"

class CSVProcessor(DataProcessor):
    def load(self, data):
        return f"row1,row2,{data}"
    
    def validate(self, data):
        return "," in data
    
    def transform(self, data):
        return data.split(",")
    
    def save(self, data):
        return f"Saved to data.csv ({len(data)} rows)"

# Test
json_proc = JSONProcessor()
csv_proc = CSVProcessor()

json_proc.process("user data")
csv_proc.process("value1,value2,value3")
````

### 7.2 Strategy Pattern

````python
from abc import ABC, abstractmethod

# Strategy Interface
class PaymentStrategy(ABC):
    """Interface for payment strategies."""
    
    @abstractmethod
    def pay(self, amount: float) -> str:
        pass
    
    @abstractmethod
    def validate(self) -> bool:
        pass

# Concrete Strategies
class CreditCardPayment(PaymentStrategy):
    def __init__(self, card_number, cvv):
        self.card_number = card_number
        self.cvv = cvv
    
    def validate(self):
        return len(self.card_number) == 16 and len(self.cvv) == 3
    
    def pay(self, amount):
        if not self.validate():
            raise ValueError("Invalid card")
        return f"Paid ${amount:.2f} with Credit Card ending in {self.card_number[-4:]}"

class PayPalPayment(PaymentStrategy):
    def __init__(self, email):
        self.email = email
    
    def validate(self):
        return "@" in self.email
    
    def pay(self, amount):
        if not self.validate():
            raise ValueError("Invalid email")
        return f"Paid ${amount:.2f} with PayPal ({self.email})"

class CryptoPayment(PaymentStrategy):
    def __init__(self, wallet_address, currency="BTC"):
        self.wallet = wallet_address
        self.currency = currency
    
    def validate(self):
        return len(self.wallet) >= 26
    
    def pay(self, amount):
        if not self.validate():
            raise ValueError("Invalid wallet")
        return f"Paid ${amount:.2f} in {self.currency} to {self.wallet[:8]}..."

# Context
class ShoppingCart:
    """Context class that uses PaymentStrategy."""
    
    def __init__(self):
        self.items = []
        self._payment_strategy = None
    
    def add_item(self, name, price):
        self.items.append({"name": name, "price": price})
    
    @property
    def total(self):
        return sum(item["price"] for item in self.items)
    
    def set_payment_strategy(self, strategy: PaymentStrategy):
        """Set payment strategy at runtime."""
        self._payment_strategy = strategy
    
    def checkout(self):
        if not self._payment_strategy:
            raise ValueError("Payment method not set")
        return self._payment_strategy.pay(self.total)

# Test
print("Strategy Pattern Demo:")
print("=" * 50)

cart = ShoppingCart()
cart.add_item("Laptop", 999.99)
cart.add_item("Mouse", 29.99)

print(f"Cart total: ${cart.total:.2f}")

# Change strategy at runtime
strategies = [
    CreditCardPayment("1234567890123456", "123"),
    PayPalPayment("user@email.com"),
    CryptoPayment("1A2b3C4d5E6f7G8h9I0jKlMnOpQrSt", "ETH")
]

for strategy in strategies:
    cart.set_payment_strategy(strategy)
    print(f"\n{cart.checkout()}")
````

### 7.3 Factory Pattern

````python
from abc import ABC, abstractmethod

# Product interface
class Notification(ABC):
    @abstractmethod
    def send(self, message: str) -> str:
        pass

# Concrete products
class EmailNotification(Notification):
    def __init__(self, recipient):
        self.recipient = recipient
    
    def send(self, message):
        return f"ğŸ“§ Email to {self.recipient}: {message}"

class SMSNotification(Notification):
    def __init__(self, phone):
        self.phone = phone
    
    def send(self, message):
        return f"ğŸ“± SMS to {self.phone}: {message}"

class PushNotification(Notification):
    def __init__(self, device_id):
        self.device_id = device_id
    
    def send(self, message):
        return f"ğŸ”” Push to {self.device_id}: {message}"

# Factory
class NotificationFactory:
    """Factory to create notifications."""
    
    @staticmethod
    def create(notification_type: str, **kwargs) -> Notification:
        factories = {
            "email": lambda: EmailNotification(kwargs.get("recipient")),
            "sms": lambda: SMSNotification(kwargs.get("phone")),
            "push": lambda: PushNotification(kwargs.get("device_id"))
        }
        
        if notification_type not in factories:
            raise ValueError(f"Unknown notification type: {notification_type}")
        
        return factories[notification_type]()

# Test
print("Factory Pattern Demo:")
print("=" * 50)

# Client code khÃ´ng cáº§n biáº¿t concrete classes
notifications = [
    NotificationFactory.create("email", recipient="user@example.com"),
    NotificationFactory.create("sms", phone="+1234567890"),
    NotificationFactory.create("push", device_id="device_abc123")
]

for notif in notifications:
    print(notif.send("Hello World!"))
````

````python
from abc import ABC, abstractmethod

# Interfaces
class Database(ABC):
    @abstractmethod
    def query(self, sql: str):
        pass

class Cache(ABC):
    @abstractmethod
    def get(self, key: str):
        pass
    
    @abstractmethod
    def set(self, key: str, value):
        pass

class Logger(ABC):
    @abstractmethod
    def log(self, message: str):
        pass

# Concrete implementations
class MySQLDatabase(Database):
    def query(self, sql):
        return f"MySQL: {sql}"

class PostgreSQLDatabase(Database):
    def query(self, sql):
        return f"PostgreSQL: {sql}"

class RedisCache(Cache):
    def __init__(self):
        self._store = {}
    
    def get(self, key):
        return self._store.get(key)
    
    def set(self, key, value):
        self._store[key] = value

class ConsoleLogger(Logger):
    def log(self, message):
        print(f"[LOG] {message}")

# Mock for testing
class MockDatabase(Database):
    def query(self, sql):
        return f"MOCK: {sql}"

# Service with Dependency Injection
class UserService:
    """
    Service vá»›i injected dependencies.
    Dá»… test, dá»… thay Ä‘á»•i implementation.
    """
    
    def __init__(self, db: Database, cache: Cache, logger: Logger):
        # Dependencies Ä‘Æ°á»£c inject tá»« bÃªn ngoÃ i
        self._db = db
        self._cache = cache
        self._logger = logger
    
    def get_user(self, user_id: int):
        self._logger.log(f"Getting user {user_id}")
        
        # Check cache
        cached = self._cache.get(f"user:{user_id}")
        if cached:
            self._logger.log("Cache hit!")
            return cached
        
        # Query database
        result = self._db.query(f"SELECT * FROM users WHERE id = {user_id}")
        
        # Store in cache
        self._cache.set(f"user:{user_id}", result)
        
        return result

# Test with real implementations
print("Dependency Injection Demo:")
print("=" * 50)

# Production setup
print("\n1. Production setup:")
prod_service = UserService(
    db=MySQLDatabase(),
    cache=RedisCache(),
    logger=ConsoleLogger()
)
print(prod_service.get_user(123))
print(prod_service.get_user(123))  # Cache hit

# Testing setup
print("\n2. Testing setup:")
test_service = UserService(
    db=MockDatabase(),  # Mock for testing!
    cache=RedisCache(),
    logger=ConsoleLogger()
)
print(test_service.get_user(456))
````

### 9.1 Plugin System

````python
from abc import ABC, abstractmethod
from typing import Dict, Type

# Plugin Interface
class Plugin(ABC):
    """Interface for plugins."""
    
    @property
    @abstractmethod
    def name(self) -> str:
        pass
    
    @property
    @abstractmethod
    def version(self) -> str:
        pass
    
    @abstractmethod
    def initialize(self) -> bool:
        pass
    
    @abstractmethod
    def execute(self, data: dict) -> dict:
        pass
    
    @abstractmethod
    def cleanup(self):
        pass

# Concrete plugins
class ValidationPlugin(Plugin):
    @property
    def name(self):
        return "Validator"
    
    @property
    def version(self):
        return "1.0.0"
    
    def initialize(self):
        print(f"  [{self.name}] Initialized")
        return True
    
    def execute(self, data):
        data["validated"] = True
        print(f"  [{self.name}] Data validated")
        return data
    
    def cleanup(self):
        print(f"  [{self.name}] Cleaned up")

class TransformPlugin(Plugin):
    @property
    def name(self):
        return "Transformer"
    
    @property
    def version(self):
        return "2.1.0"
    
    def initialize(self):
        print(f"  [{self.name}] Initialized")
        return True
    
    def execute(self, data):
        data["transformed"] = True
        if "value" in data:
            data["value"] = data["value"].upper()
        print(f"  [{self.name}] Data transformed")
        return data
    
    def cleanup(self):
        print(f"  [{self.name}] Cleaned up")

class LoggingPlugin(Plugin):
    @property
    def name(self):
        return "Logger"
    
    @property
    def version(self):
        return "1.5.0"
    
    def initialize(self):
        print(f"  [{self.name}] Initialized")
        return True
    
    def execute(self, data):
        print(f"  [{self.name}] Logging: {data}")
        data["logged"] = True
        return data
    
    def cleanup(self):
        print(f"  [{self.name}] Cleaned up")

# Plugin Manager
class PluginManager:
    """Manages and executes plugins."""
    
    def __init__(self):
        self._plugins: list[Plugin] = []
    
    def register(self, plugin: Plugin):
        self._plugins.append(plugin)
        print(f"Registered: {plugin.name} v{plugin.version}")
    
    def initialize_all(self):
        print("\nInitializing plugins...")
        for plugin in self._plugins:
            plugin.initialize()
    
    def process(self, data: dict) -> dict:
        print("\nProcessing data through pipeline...")
        result = data.copy()
        for plugin in self._plugins:
            result = plugin.execute(result)
        return result
    
    def cleanup_all(self):
        print("\nCleaning up plugins...")
        for plugin in reversed(self._plugins):
            plugin.cleanup()

# Test
print("Plugin System Demo:")
print("=" * 50)

manager = PluginManager()

# Register plugins
manager.register(ValidationPlugin())
manager.register(TransformPlugin())
manager.register(LoggingPlugin())

# Initialize
manager.initialize_all()

# Process data
data = {"value": "hello world", "id": 123}
result = manager.process(data)

print(f"\nFinal result: {result}")

# Cleanup
manager.cleanup_all()
````

````python
# âœ… GOOD: Segregated interfaces

class Workable(ABC):
    @abstractmethod
    def work(self): pass

class Eatable(ABC):
    @abstractmethod
    def eat(self): pass

class Sleepable(ABC):
    @abstractmethod
    def sleep(self): pass

class Programmable(ABC):
    @abstractmethod
    def program(self): pass

# Human implements nhiá»u interfaces
class Human(Workable, Eatable, Sleepable):
    def work(self):
        return "Human working"
    
    def eat(self):
        return "Human eating"
    
    def sleep(self):
        return "Human sleeping"

class Programmer(Human, Programmable):
    def program(self):
        return "Programmer coding"

# Robot chá»‰ implement nhá»¯ng gÃ¬ cáº§n
class Robot(Workable, Programmable):
    def work(self):
        return "Robot working 24/7"
    
    def program(self):
        return "Robot executing code"

# Test
print("\nâœ… Segregated interfaces:")

programmer = Programmer()
robot = Robot()

print(f"\nProgrammer: {programmer.work()}, {programmer.program()}")
print(f"Robot: {robot.work()}, {robot.program()}")

# Type checking
print(f"\nProgrammer is Sleepable: {isinstance(programmer, Sleepable)}")
print(f"Robot is Sleepable: {isinstance(robot, Sleepable)}")
````
