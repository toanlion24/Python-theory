## 1. L·∫≠p tr√¨nh h∆∞·ªõng ƒë·ªëi t∆∞·ª£ng (OOP) l√† g√¨?

### 1.1 ƒê·ªãnh nghƒ©a

**Object-Oriented Programming (OOP)** l√† m·ªôt m√¥ h√¨nh l·∫≠p tr√¨nh (programming paradigm) d·ª±a tr√™n kh√°i ni·ªám **"objects"** - c√°c th·ª±c th·ªÉ ch·ª©a:

- **Data** (d·ªØ li·ªáu) ‚Üí g·ªçi l√† **attributes** ho·∫∑c **properties**
- **Code** (h√†nh vi) ‚Üí g·ªçi l√† **methods** ho·∫∑c **functions**

### 1.2 B·ªën tr·ª• c·ªôt c·ªßa OOP

| Tr·ª• c·ªôt | Ti·∫øng Vi·ªát | M√¥ t·∫£ |
|---------|------------|-------|
| **Encapsulation** | ƒê√≥ng g√≥i | G√≥i data v√† methods v√†o m·ªôt ƒë∆°n v·ªã, ·∫©n implementation |
| **Abstraction** | Tr·ª´u t∆∞·ª£ng h√≥a | ·∫®n chi ti·∫øt ph·ª©c t·∫°p, ch·ªâ expose interface c·∫ßn thi·∫øt |
| **Inheritance** | K·∫ø th·ª´a | Class con th·ª´a h∆∞·ªüng t·ª´ class cha |
| **Polymorphism** | ƒêa h√¨nh | C√πng interface, nhi·ªÅu implementation kh√°c nhau |

### 1.3 So s√°nh: Procedural vs OOP

#### Procedural Programming (L·∫≠p tr√¨nh th·ªß t·ª•c)

````python
# === PROCEDURAL APPROACH ===
# Data v√† functions t√°ch bi·ªát

# Data: d√πng dictionary ho·∫∑c c√°c bi·∫øn ri√™ng l·∫ª
student1 = {
    "name": "Nguyen Van A",
    "age": 20,
    "grades": [8.5, 9.0, 7.5]
}

student2 = {
    "name": "Tran Thi B",
    "age": 21,
    "grades": [9.0, 8.5, 9.5]
}

# Functions: x·ª≠ l√Ω data t·ª´ b√™n ngo√†i
def calculate_average(student):
    if not student["grades"]:
        return 0
    return sum(student["grades"]) / len(student["grades"])

def add_grade(student, grade):
    student["grades"].append(grade)

def display_student(student):
    avg = calculate_average(student)
    print(f"{student['name']}, Age: {student['age']}, Average: {avg:.2f}")

# S·ª≠ d·ª•ng
add_grade(student1, 8.0)
display_student(student1)
display_student(student2)
````

#### V·∫•n ƒë·ªÅ v·ªõi Procedural:

1. **Data kh√¥ng ƒë∆∞·ª£c b·∫£o v·ªá**: Ai c≈©ng c√≥ th·ªÉ tr·ª±c ti·∫øp thay ƒë·ªïi `student1["grades"] = "invalid"`
2. **Kh√≥ qu·∫£n l√Ω**: Khi c√≥ nhi·ªÅu lo·∫°i ƒë·ªëi t∆∞·ª£ng (Student, Teacher, Course...), code tr·ªü n√™n ph·ª©c t·∫°p
3. **Kh√≥ t√°i s·ª≠ d·ª•ng**: Ph·∫£i copy functions khi c·∫ßn m·ªü r·ªông

#### Object-Oriented Programming

````python
# === OOP APPROACH ===
# Data v√† methods ƒë∆∞·ª£c g√≥i chung trong class

class Student:
    """Class ƒë·∫°i di·ªán cho m·ªôt sinh vi√™n."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self._grades = []  # Protected: kh√¥ng n√™n truy c·∫≠p tr·ª±c ti·∫øp
    
    def add_grade(self, grade):
        """Th√™m ƒëi·ªÉm v·ªõi validation."""
        if 0 <= grade <= 10:
            self._grades.append(grade)
            return True
        return False
    
    def calculate_average(self):
        """T√≠nh ƒëi·ªÉm trung b√¨nh."""
        if not self._grades:
            return 0
        return sum(self._grades) / len(self._grades)
    
    def display(self):
        """Hi·ªÉn th·ªã th√¥ng tin."""
        print(f"{self.name}, Age: {self.age}, Average: {self.calculate_average():.2f}")

# S·ª≠ d·ª•ng: m·ªçi th·ª© ƒë∆∞·ª£c g√≥i g·ªçn trong object
student1 = Student("Nguyen Van A", 20)
student2 = Student("Tran Thi B", 21)

student1.add_grade(8.5)
student1.add_grade(9.0)
student1.add_grade(7.5)

student2.add_grade(9.0)
student2.add_grade(8.5)

student1.display()
student2.display()
````

#### ∆Øu ƒëi·ªÉm c·ªßa OOP:

| ∆Øu ƒëi·ªÉm | Gi·∫£i th√≠ch |
|---------|------------|
| **T·ªï ch·ª©c code** | Data v√† methods li√™n quan ƒë∆∞·ª£c gom l·∫°i |
| **B·∫£o v·ªá data** | C√≥ th·ªÉ ki·ªÉm so√°t c√°ch data ƒë∆∞·ª£c truy c·∫≠p |
| **T√°i s·ª≠ d·ª•ng** | K·∫ø th·ª´a cho ph√©p m·ªü r·ªông class c√≥ s·∫µn |
| **D·ªÖ b·∫£o tr√¨** | Thay ƒë·ªïi trong class kh√¥ng ·∫£nh h∆∞·ªüng code b√™n ngo√†i |
| **M√¥ h√¨nh h√≥a th·ª±c t·∫ø** | Objects ph·∫£n √°nh c√°c ƒë·ªëi t∆∞·ª£ng trong th·∫ø gi·ªõi th·ª±c |

## 3. Object (Instance) - Th·ª±c th·ªÉ

### 3.1 ƒê·ªãnh nghƒ©a Object

**Object** (c√≤n g·ªçi l√† **instance**) l√† m·ªôt th·ª±c th·ªÉ c·ª• th·ªÉ ƒë∆∞·ª£c t·∫°o t·ª´ class.

- **Class** = Khu√¥n b√°nh
- **Object** = Nh·ªØng chi·∫øc b√°nh c·ª• th·ªÉ ƒë∆∞·ª£c l√†m t·ª´ khu√¥n

### 3.2 T·∫°o Objects

````python
class Car:
    """Class ƒë·∫°i di·ªán cho xe h∆°i."""
    
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        self.is_running = False
    
    def start(self):
        self.is_running = True
        return f"{self.brand} {self.model} started!"
    
    def stop(self):
        self.is_running = False
        return f"{self.brand} {self.model} stopped!"

# T·∫°o nhi·ªÅu objects t·ª´ c√πng m·ªôt class
car1 = Car("Toyota", "Camry", 2023)
car2 = Car("Honda", "Civic", 2022)
car3 = Car("BMW", "X5", 2024)

# M·ªói object c√≥ data ri√™ng
print(f"Car 1: {car1.brand} {car1.model} ({car1.year})")
print(f"Car 2: {car2.brand} {car2.model} ({car2.year})")
print(f"Car 3: {car3.brand} {car3.model} ({car3.year})")
````

### 3.3 Object Identity (ƒê·ªãnh danh)

M·ªói object c√≥ m·ªôt **identity** duy nh·∫•t (ƒë·ªãa ch·ªâ trong b·ªô nh·ªõ).

````python
# M·ªói object c√≥ id ri√™ng
print(f"car1 id: {id(car1)}")
print(f"car2 id: {id(car2)}")
print(f"car3 id: {id(car3)}")

# So s√°nh identity
print(f"\ncar1 is car2: {car1 is car2}")  # False - kh√°c object

# G√°n reference
car4 = car1  # car4 tr·ªè ƒë·∫øn c√πng object v·ªõi car1
print(f"car1 is car4: {car1 is car4}")  # True - c√πng object
````

### 3.4 isinstance() - Ki·ªÉm tra ki·ªÉu

````python
# isinstance(object, class) - ki·ªÉm tra object c√≥ ph·∫£i instance c·ªßa class kh√¥ng

print(f"car1 l√† Car: {isinstance(car1, Car)}")
print(f"car1 l√† object: {isinstance(car1, object)}")
print(f"car1 l√† str: {isinstance(car1, str)}")

# C√≥ th·ªÉ ki·ªÉm tra nhi·ªÅu types
x = 42
print(f"\n42 l√† int ho·∫∑c float: {isinstance(x, (int, float))}")
````

## 4. Attributes (Thu·ªôc t√≠nh) - Chi ti·∫øt

### 4.1 Instance Attributes

**Instance attributes** l√† thu·ªôc t√≠nh ri√™ng c·ªßa m·ªói object, ƒë∆∞·ª£c t·∫°o trong `__init__`.

````python
class Person:
    def __init__(self, name, age, email):
        # Instance attributes - m·ªói person c√≥ gi√° tr·ªã ri√™ng
        self.name = name
        self.age = age
        self.email = email

# T·∫°o 2 persons v·ªõi data kh√°c nhau
person1 = Person("Alice", 25, "alice@email.com")
person2 = Person("Bob", 30, "bob@email.com")

# M·ªói object c√≥ data ri√™ng
print(f"Person 1: {person1.name}, {person1.age}, {person1.email}")
print(f"Person 2: {person2.name}, {person2.age}, {person2.email}")

# Thay ƒë·ªïi attribute c·ªßa m·ªôt object kh√¥ng ·∫£nh h∆∞·ªüng object kh√°c
person1.age = 26
print(f"\nAfter changing person1.age:")
print(f"Person 1 age: {person1.age}")
print(f"Person 2 age: {person2.age}")  # Kh√¥ng thay ƒë·ªïi
````

### 4.2 Class Attributes

**Class attributes** l√† thu·ªôc t√≠nh chung cho t·∫•t c·∫£ objects, ƒë∆∞·ª£c ƒë·ªãnh nghƒ©a b√™n ngo√†i `__init__`.

````python
class Employee:
    # Class attributes - chung cho t·∫•t c·∫£ employees
    company = "TechCorp"
    employee_count = 0
    
    def __init__(self, name, position):
        # Instance attributes - ri√™ng cho m·ªói employee
        self.name = name
        self.position = position
        
        # TƒÉng counter khi t·∫°o employee m·ªõi
        Employee.employee_count += 1

# T·∫°o employees
emp1 = Employee("Alice", "Developer")
emp2 = Employee("Bob", "Designer")
emp3 = Employee("Charlie", "Manager")

# Class attribute gi·ªëng nhau cho t·∫•t c·∫£
print(f"{emp1.name} works at: {emp1.company}")
print(f"{emp2.name} works at: {emp2.company}")

# Truy c·∫≠p qua class name
print(f"\nCompany name: {Employee.company}")
print(f"Total employees: {Employee.employee_count}")
````

### 4.3 So s√°nh Class Attributes vs Instance Attributes

````python
class Dog:
    # Class attribute
    species = "Canis familiaris"
    
    def __init__(self, name):
        # Instance attribute
        self.name = name

dog1 = Dog("Buddy")
dog2 = Dog("Max")

print("=== Before modification ===")
print(f"dog1.species: {dog1.species}")
print(f"dog2.species: {dog2.species}")
print(f"Dog.species: {Dog.species}")

# Thay ƒë·ªïi class attribute qua class
Dog.species = "Canis lupus familiaris"
print("\n=== After Dog.species = 'Canis lupus familiaris' ===")
print(f"dog1.species: {dog1.species}")  # Thay ƒë·ªïi
print(f"dog2.species: {dog2.species}")  # Thay ƒë·ªïi

# ‚ö†Ô∏è Ch√∫ √Ω: G√°n qua instance t·∫°o instance attribute m·ªõi!
dog1.species = "Wolf"  # T·∫°o instance attribute cho dog1
print("\n=== After dog1.species = 'Wolf' ===")
print(f"dog1.species: {dog1.species}")  # Instance attribute
print(f"dog2.species: {dog2.species}")  # V·∫´n d√πng class attribute
print(f"Dog.species: {Dog.species}")
````

### 4.4 Truy c·∫≠p Attributes

| C√°ch truy c·∫≠p | C√∫ ph√°p | M√¥ t·∫£ |
|---------------|---------|-------|
| Dot notation | `obj.attr` | C√°ch ph·ªï bi·∫øn nh·∫•t |
| getattr() | `getattr(obj, 'attr')` | Dynamic access |
| setattr() | `setattr(obj, 'attr', value)` | Dynamic set |
| hasattr() | `hasattr(obj, 'attr')` | Ki·ªÉm tra t·ªìn t·∫°i |
| delattr() | `delattr(obj, 'attr')` | X√≥a attribute |

````python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

product = Product("Laptop", 999)

# Dot notation
print(f"Name: {product.name}")

# getattr() - c√≥ th·ªÉ set default value
print(f"Price: {getattr(product, 'price')}")
print(f"Stock: {getattr(product, 'stock', 'N/A')}")

# hasattr() - ki·ªÉm tra attribute c√≥ t·ªìn t·∫°i
print(f"\nHas 'name': {hasattr(product, 'name')}")
print(f"Has 'stock': {hasattr(product, 'stock')}")

# setattr() - dynamic set
setattr(product, 'stock', 50)
print(f"\nStock after setattr: {product.stock}")

# delattr() - x√≥a attribute
delattr(product, 'stock')
print(f"Has 'stock' after delattr: {hasattr(product, 'stock')}")
````

## 5. Methods (Ph∆∞∆°ng th·ª©c) - Chi ti·∫øt

### 5.1 Instance Methods

**Instance methods** l√† methods ho·∫°t ƒë·ªông tr√™n instance c·ª• th·ªÉ, c√≥ `self` l√† parameter ƒë·∫ßu ti√™n.

````python
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    
    # Instance methods - ho·∫°t ƒë·ªông tr√™n self
    def deposit(self, amount):
        """N·∫°p ti·ªÅn v√†o t√†i kho·∫£n."""
        if amount > 0:
            self.balance += amount
            return f"Deposited ${amount}. Balance: ${self.balance}"
        return "Invalid amount!"
    
    def withdraw(self, amount):
        """R√∫t ti·ªÅn t·ª´ t√†i kho·∫£n."""
        if 0 < amount <= self.balance:
            self.balance -= amount
            return f"Withdrew ${amount}. Balance: ${self.balance}"
        return "Invalid amount or insufficient funds!"
    
    def get_balance(self):
        """Xem s·ªë d∆∞."""
        return f"{self.owner}'s balance: ${self.balance}"

# S·ª≠ d·ª•ng instance methods
account = BankAccount("John", 1000)
print(account.deposit(500))
print(account.withdraw(200))
print(account.get_balance())
````

### 5.2 Keyword `self` - Gi·∫£i th√≠ch chi ti·∫øt

`self` l√† reference ƒë·∫øn object hi·ªán t·∫°i ƒëang g·ªçi method.

- Python t·ª± ƒë·ªông truy·ªÅn object v√†o `self`
- T√™n `self` l√† convention, c√≥ th·ªÉ ƒë·∫∑t t√™n kh√°c (nh∆∞ng kh√¥ng n√™n)

````python
class Demo:
    def show_self(self):
        print(f"self = {self}")
        print(f"self id = {id(self)}")

obj1 = Demo()
obj2 = Demo()

print(f"obj1 = {obj1}")
print(f"obj1 id = {id(obj1)}")
obj1.show_self()  # self tr·ªè ƒë·∫øn obj1

print(f"\nobj2 = {obj2}")
print(f"obj2 id = {id(obj2)}")
obj2.show_self()  # self tr·ªè ƒë·∫øn obj2
````

````python
# C√°ch Python g·ªçi method th·ª±c s·ª±
class Greeter:
    def say_hello(self, name):
        return f"Hello, {name}!"

g = Greeter()

# Hai c√°ch g·ªçi t∆∞∆°ng ƒë∆∞∆°ng:
print(g.say_hello("Alice"))        # C√°ch th√¥ng th∆∞·ªùng
print(Greeter.say_hello(g, "Bob"))  # C√°ch Python th·ª±c hi·ªán b√™n trong
````

### 5.3 Class Methods (@classmethod)

**Class methods** ho·∫°t ƒë·ªông tr√™n class, kh√¥ng ph·∫£i instance. D√πng decorator `@classmethod`.

````python
from datetime import date

class Person:
    species = "Homo sapiens"
    count = 0
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.count += 1
    
    # Class method - nh·∫≠n cls (class) thay v√¨ self (instance)
    @classmethod
    def from_birth_year(cls, name, birth_year):
        """Alternative constructor: t·∫°o Person t·ª´ nƒÉm sinh."""
        age = date.today().year - birth_year
        return cls(name, age)  # cls = Person
    
    @classmethod
    def get_count(cls):
        """Tr·∫£ v·ªÅ s·ªë l∆∞·ª£ng Person ƒë√£ t·∫°o."""
        return cls.count
    
    @classmethod
    def get_species(cls):
        """Tr·∫£ v·ªÅ species."""
        return cls.species

# S·ª≠ d·ª•ng class method nh∆∞ alternative constructor
person1 = Person("Alice", 25)
person2 = Person.from_birth_year("Bob", 1995)  # Alternative constructor

print(f"{person1.name}: {person1.age} years old")
print(f"{person2.name}: {person2.age} years old")
print(f"Total persons: {Person.get_count()}")
print(f"Species: {Person.get_species()}")
````

### 5.4 Static Methods (@staticmethod)

**Static methods** kh√¥ng nh·∫≠n `self` hay `cls`. Gi·ªëng nh∆∞ function th∆∞·ªùng, ch·ªâ ƒë∆∞·ª£c ƒë·∫∑t trong class v√¨ li√™n quan logic.

````python
class MathUtils:
    """Utility class v·ªõi c√°c h√†m to√°n h·ªçc."""
    
    @staticmethod
    def add(a, b):
        return a + b
    
    @staticmethod
    def multiply(a, b):
        return a * b
    
    @staticmethod
    def is_even(n):
        return n % 2 == 0
    
    @staticmethod
    def factorial(n):
        if n <= 1:
            return 1
        return n * MathUtils.factorial(n - 1)

# G·ªçi static methods - kh√¥ng c·∫ßn t·∫°o instance
print(f"5 + 3 = {MathUtils.add(5, 3)}")
print(f"5 * 3 = {MathUtils.multiply(5, 3)}")
print(f"Is 4 even? {MathUtils.is_even(4)}")
print(f"5! = {MathUtils.factorial(5)}")
````

### 5.5 So s√°nh c√°c lo·∫°i Methods

| Lo·∫°i | Decorator | Parameter ƒë·∫ßu ti√™n | Use case |
|------|-----------|-------------------|----------|
| Instance method | (kh√¥ng) | `self` | Thao t√°c v·ªõi instance data |
| Class method | `@classmethod` | `cls` | Alternative constructors, class-level operations |
| Static method | `@staticmethod` | (kh√¥ng c√≥) | Utility functions li√™n quan ƒë·∫øn class |

````python
class Example:
    class_attr = "I am class attribute"
    
    def __init__(self, value):
        self.value = value
    
    def instance_method(self):
        """Can access both instance and class attributes."""
        return f"Instance: {self.value}, Class: {self.class_attr}"
    
    @classmethod
    def class_method(cls):
        """Can only access class attributes."""
        return f"Class: {cls.class_attr}"
    
    @staticmethod
    def static_method():
        """Cannot access instance or class attributes directly."""
        return "I'm just a regular function inside a class"

obj = Example(42)
print(obj.instance_method())
print(Example.class_method())
print(Example.static_method())
````

### 6.2 __str__ vs __repr__

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __str__(self):
        """Human-readable string (cho print, str())."""
        return f"Point at ({self.x}, {self.y})"
    
    def __repr__(self):
        """Developer-readable string (c√≥ th·ªÉ d√πng ƒë·ªÉ recreate object)."""
        return f"Point({self.x}, {self.y})"

p = Point(3, 4)

print(f"print(p): {p}")           # G·ªçi __str__
print(f"str(p): {str(p)}")
print(f"repr(p): {repr(p)}")

# Trong list, __repr__ ƒë∆∞·ª£c d√πng
points = [Point(1, 2), Point(3, 4)]
print(f"List: {points}")
````

### 6.3 Comparison Magic Methods

````python
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa
    
    def __eq__(self, other):
        """student1 == student2"""
        if isinstance(other, Student):
            return self.gpa == other.gpa
        return False
    
    def __lt__(self, other):
        """student1 < student2"""
        if isinstance(other, Student):
            return self.gpa < other.gpa
        return NotImplemented
    
    def __le__(self, other):
        """student1 <= student2"""
        return self == other or self < other
    
    def __repr__(self):
        return f"Student('{self.name}', {self.gpa})"

s1 = Student("Alice", 3.5)
s2 = Student("Bob", 3.8)
s3 = Student("Charlie", 3.5)

print(f"{s1.name} == {s3.name}: {s1 == s3}")  # True (same GPA)
print(f"{s1.name} < {s2.name}: {s1 < s2}")    # True
print(f"{s2.name} > {s1.name}: {s2 > s1}")    # True (Python infers from __lt__)

# C√≥ th·ªÉ sort!
students = [s2, s1, s3]
students.sort()
print(f"\nSorted: {students}")
````

### 6.4 Operator Overloading

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """v1 + v2"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """v1 - v2"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """v * scalar"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __abs__(self):
        """abs(v) - magnitude"""
        return (self.x ** 2 + self.y ** 2) ** 0.5
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")
print(f"v1 - v2 = {v1 - v2}")
print(f"v1 * 2 = {v1 * 2}")
print(f"|v1| = {abs(v1)}")
````

### 6.5 Container Magic Methods

````python
class Playlist:
    def __init__(self, name):
        self.name = name
        self._songs = []
    
    def add(self, song):
        self._songs.append(song)
    
    def __len__(self):
        """len(playlist)"""
        return len(self._songs)
    
    def __getitem__(self, index):
        """playlist[index]"""
        return self._songs[index]
    
    def __setitem__(self, index, value):
        """playlist[index] = value"""
        self._songs[index] = value
    
    def __contains__(self, song):
        """song in playlist"""
        return song in self._songs
    
    def __iter__(self):
        """for song in playlist"""
        return iter(self._songs)
    
    def __repr__(self):
        return f"Playlist('{self.name}', {len(self)} songs)"

# S·ª≠ d·ª•ng
playlist = Playlist("My Favorites")
playlist.add("Song A")
playlist.add("Song B")
playlist.add("Song C")

print(f"Playlist: {playlist}")
print(f"Length: {len(playlist)}")
print(f"First song: {playlist[0]}")
print(f"'Song B' in playlist: {'Song B' in playlist}")

print("\nAll songs:")
for song in playlist:
    print(f"  - {song}")
````

### 7.1 Student Management System

````python
from datetime import date

class Student:
    """Class qu·∫£n l√Ω th√¥ng tin sinh vi√™n."""
    
    # Class attributes
    school_name = "Python University"
    total_students = 0
    
    def __init__(self, name, student_id, major, birth_year=2000):
        # Validation
        if not name or not student_id:
            raise ValueError("Name and student_id are required")
        
        # Instance attributes
        self.name = name
        self.student_id = student_id
        self.major = major
        self.birth_year = birth_year
        self._grades = {}  # Protected: subject -> grade
        self.enrollment_date = date.today()
        
        # Update class counter
        Student.total_students += 1
    
    # Alternative constructor
    @classmethod
    def from_dict(cls, data):
        """T·∫°o Student t·ª´ dictionary."""
        return cls(
            name=data['name'],
            student_id=data['student_id'],
            major=data.get('major', 'Undeclared'),
            birth_year=data.get('birth_year', 2000)
        )
    
    @classmethod
    def get_total_students(cls):
        return cls.total_students
    
    # Instance methods
    def add_grade(self, subject, grade):
        """Th√™m ƒëi·ªÉm cho m√¥n h·ªçc."""
        if not 0 <= grade <= 10:
            raise ValueError(f"Grade must be 0-10, got {grade}")
        self._grades[subject] = grade
        return self  # Method chaining
    
    def get_gpa(self):
        """T√≠nh ƒëi·ªÉm trung b√¨nh."""
        if not self._grades:
            return 0.0
        return sum(self._grades.values()) / len(self._grades)
    
    def get_age(self):
        """T√≠nh tu·ªïi."""
        return date.today().year - self.birth_year
    
    def get_classification(self):
        """X·∫øp lo·∫°i h·ªçc l·ª±c."""
        gpa = self.get_gpa()
        if gpa >= 9.0:
            return "Xu·∫•t s·∫Øc"
        elif gpa >= 8.0:
            return "Gi·ªèi"
        elif gpa >= 6.5:
            return "Kh√°"
        elif gpa >= 5.0:
            return "Trung b√¨nh"
        return "Y·∫øu"
    
    # Static method
    @staticmethod
    def validate_student_id(student_id):
        """Ki·ªÉm tra ƒë·ªãnh d·∫°ng m√£ sinh vi√™n."""
        return len(student_id) == 6 and student_id[:2].isalpha() and student_id[2:].isdigit()
    
    # Magic methods
    def __str__(self):
        return f"{self.name} ({self.student_id}) - {self.major}"
    
    def __repr__(self):
        return f"Student('{self.name}', '{self.student_id}', '{self.major}')"
    
    def __eq__(self, other):
        if isinstance(other, Student):
            return self.student_id == other.student_id
        return False
    
    def __lt__(self, other):
        if isinstance(other, Student):
            return self.get_gpa() < other.get_gpa()
        return NotImplemented

# === Demo ===
print("=" * 50)
print("STUDENT MANAGEMENT SYSTEM DEMO")
print("=" * 50)

# T·∫°o students
s1 = Student("Nguyen Van A", "SV0001", "Computer Science", 2003)
s2 = Student.from_dict({
    "name": "Tran Thi B",
    "student_id": "SV0002",
    "major": "Data Science"
})

# Th√™m ƒëi·ªÉm (method chaining)
s1.add_grade("Math", 8.5).add_grade("Physics", 9.0).add_grade("Programming", 9.5)
s2.add_grade("Math", 7.5).add_grade("Statistics", 8.0).add_grade("ML", 8.5)

# Hi·ªÉn th·ªã th√¥ng tin
for student in [s1, s2]:
    print(f"\n{student}")
    print(f"  Age: {student.get_age()}")
    print(f"  GPA: {student.get_gpa():.2f}")
    print(f"  Classification: {student.get_classification()}")

# So s√°nh students
print(f"\n{s1.name} > {s2.name} (by GPA): {s1 > s2}")

# Class methods
print(f"\nTotal students: {Student.get_total_students()}")
print(f"School: {Student.school_name}")

# Static method
print(f"\nValidate 'SV0001': {Student.validate_student_id('SV0001')}")
print(f"Validate 'invalid': {Student.validate_student_id('invalid')}")
````

### 7.2 Shopping Cart System

````python
class Product:
    """Class ƒë·∫°i di·ªán cho s·∫£n ph·∫©m."""
    
    def __init__(self, name, price, category="General"):
        self.name = name
        self.price = price
        self.category = category
    
    def __str__(self):
        return f"{self.name} - ${self.price:.2f}"
    
    def __repr__(self):
        return f"Product('{self.name}', {self.price})"


class CartItem:
    """Class ƒë·∫°i di·ªán cho item trong gi·ªè h√†ng."""
    
    def __init__(self, product, quantity=1):
        self.product = product
        self.quantity = quantity
    
    @property
    def subtotal(self):
        return self.product.price * self.quantity
    
    def __str__(self):
        return f"{self.product.name} x{self.quantity} = ${self.subtotal:.2f}"


class ShoppingCart:
    """Class qu·∫£n l√Ω gi·ªè h√†ng."""
    
    discount_threshold = 100  # ƒê∆°n h√†ng >= $100 ƒë∆∞·ª£c gi·∫£m gi√°
    discount_rate = 0.1  # Gi·∫£m 10%
    
    def __init__(self, customer_name):
        self.customer_name = customer_name
        self._items = {}  # product_name -> CartItem
    
    def add(self, product, quantity=1):
        """Th√™m s·∫£n ph·∫©m v√†o gi·ªè."""
        if product.name in self._items:
            self._items[product.name].quantity += quantity
        else:
            self._items[product.name] = CartItem(product, quantity)
        return self
    
    def remove(self, product_name):
        """X√≥a s·∫£n ph·∫©m kh·ªèi gi·ªè."""
        if product_name in self._items:
            del self._items[product_name]
        return self
    
    def update_quantity(self, product_name, quantity):
        """C·∫≠p nh·∫≠t s·ªë l∆∞·ª£ng."""
        if product_name in self._items:
            if quantity <= 0:
                self.remove(product_name)
            else:
                self._items[product_name].quantity = quantity
        return self
    
    @property
    def subtotal(self):
        """T·ªïng ti·ªÅn tr∆∞·ªõc gi·∫£m gi√°."""
        return sum(item.subtotal for item in self._items.values())
    
    @property
    def discount(self):
        """S·ªë ti·ªÅn ƒë∆∞·ª£c gi·∫£m."""
        if self.subtotal >= self.discount_threshold:
            return self.subtotal * self.discount_rate
        return 0
    
    @property
    def total(self):
        """T·ªïng ti·ªÅn sau gi·∫£m gi√°."""
        return self.subtotal - self.discount
    
    def checkout(self):
        """Hi·ªÉn th·ªã h√≥a ƒë∆°n."""
        print(f"\n{'=' * 40}")
        print(f"INVOICE - {self.customer_name}")
        print(f"{'=' * 40}")
        
        for item in self._items.values():
            print(f"  {item}")
        
        print(f"{'-' * 40}")
        print(f"  Subtotal: ${self.subtotal:.2f}")
        
        if self.discount > 0:
            print(f"  Discount ({self.discount_rate*100:.0f}%): -${self.discount:.2f}")
        
        print(f"  TOTAL: ${self.total:.2f}")
        print(f"{'=' * 40}")
    
    def __len__(self):
        return sum(item.quantity for item in self._items.values())
    
    def __repr__(self):
        return f"ShoppingCart('{self.customer_name}', {len(self)} items)"


# === Demo ===
print("SHOPPING CART DEMO")

# T·∫°o products
laptop = Product("Laptop", 999.99, "Electronics")
mouse = Product("Mouse", 29.99, "Electronics")
keyboard = Product("Keyboard", 79.99, "Electronics")
headphones = Product("Headphones", 149.99, "Electronics")

# T·∫°o cart v√† th√™m products (method chaining)
cart = ShoppingCart("John Doe")
cart.add(laptop).add(mouse, 2).add(keyboard).add(headphones)

print(f"Cart: {cart}")
print(f"Total items: {len(cart)}")

# Checkout
cart.checkout()
````

### 8.2 Common Mistakes

````python
# ‚ùå MISTAKE 1: Mutable default argument
class BadClass:
    def __init__(self, items=[]):  # ‚ùå Shared between instances!
        self.items = items

# ‚úÖ FIX:
class GoodClass:
    def __init__(self, items=None):
        self.items = items if items is not None else []

# Demo the problem
b1 = BadClass()
b1.items.append(1)
b2 = BadClass()
print(f"b2.items (should be empty): {b2.items}")  # [1] - Wrong!

g1 = GoodClass()
g1.items.append(1)
g2 = GoodClass()
print(f"g2.items (correct): {g2.items}")  # [] - Correct!
````

````python
# ‚ùå MISTAKE 2: Forgetting self
class Calculator:
    def __init__(self):
        self.result = 0
    
    # ‚ùå Missing self - this won't work as expected
    # def add(value):
    #     result += value  # NameError: result not defined
    
    # ‚úÖ Correct
    def add(self, value):
        self.result += value
        return self
````

````python
# ‚ùå MISTAKE 3: Modifying class attribute through instance
class Counter:
    count = 0  # Class attribute
    
    def __init__(self):
        # ‚ùå This creates instance attribute, doesn't modify class attribute
        # self.count += 1
        
        # ‚úÖ Correct way
        Counter.count += 1

c1 = Counter()
c2 = Counter()
print(f"Total count: {Counter.count}")
````
