## Câu 2: Thứ Tự Thực Thi if-elif-else

**Đáp án đúng: B) `B`**

**Giải thích:**

Python đánh giá các câu lệnh `if-elif-else` từ trên xuống dưới và chỉ thực thi khối ĐẦU TIÊN có điều kiện là `True`:

```python
score = 85
if score >= 90:      # 85 >= 90 là False, bỏ qua
    print("A")
elif score >= 80:    # 85 >= 80 là True, THỰC THI KHỐI NÀY
    print("B")       # In "B" và dừng
elif score >= 70:    # Không được đánh giá (đã tìm thấy match)
    print("C")
else:                # Không được đánh giá (đã tìm thấy match)
    print("F")
```

**Điểm quan trọng:**
- Python kiểm tra điều kiện tuần tự: `if` trước, sau đó các `elif` theo thứ tự
- Khi một điều kiện là `True`, khối đó được thực thi và Python BỎ QUA tất cả các khối `elif` và `else` còn lại
- Mặc dù `score >= 70` cũng là `True`, nó không bao giờ được kiểm tra vì điều kiện khớp đầu tiên (`score >= 80`) đã được thực thi
- Chỉ MỘT khối được thực thi, ngay cả khi nhiều điều kiện là `True`

**Kết quả:** `B`

## Câu 3: Toán Tử Logic - and

**Đáp án đúng: B) `At least one condition is False`**

**Giải thích:**

Toán tử `and` yêu cầu CẢ HAI điều kiện phải là `True` để toàn bộ biểu thức là `True`:

```python
x = 10
y = 5
if x > 5 and y > 10:  # (10 > 5) and (5 > 10)
                      # True and False = False
    print("Both conditions are True")
else:
    print("At least one condition is False")  # Khối này được thực thi
```

**Đánh giá:**
- `x > 5` → `10 > 5` → `True`
- `y > 10` → `5 > 10` → `False`
- `True and False` → `False`
- Vì điều kiện là `False`, khối `else` được thực thi

**Điểm quan trọng:**
- `and` trả về `True` chỉ khi TẤT CẢ điều kiện là `True`
- Nếu bất kỳ điều kiện nào là `False`, toàn bộ biểu thức `and` là `False`
- Python sử dụng short-circuit evaluation: nếu điều kiện đầu tiên là `False`, nó không đánh giá điều kiện thứ hai

**Kết quả:** `At least one condition is False`

## Câu 4: Toán Tử Logic - or

**Đáp án đúng: A) `At least one condition is True`**

**Giải thích:**

Toán tử `or` yêu cầu ÍT NHẤT MỘT điều kiện là `True` để toàn bộ biểu thức là `True`:

```python
x = 10
y = 5
if x > 5 or y > 10:  # (10 > 5) or (5 > 10)
                     # True or False = True
    print("At least one condition is True")  # Khối này được thực thi
else:
    print("Both conditions are False")
```

**Đánh giá:**
- `x > 5` → `10 > 5` → `True`
- `y > 10` → `5 > 10` → `False`
- `True or False` → `True`
- Vì điều kiện là `True`, khối `if` được thực thi

**Điểm quan trọng:**
- `or` trả về `True` nếu BẤT KỲ điều kiện nào là `True`
- Nó chỉ trả về `False` khi TẤT CẢ điều kiện đều là `False`
- Python sử dụng short-circuit evaluation: nếu điều kiện đầu tiên là `True`, nó không đánh giá điều kiện thứ hai

**Kết quả:** `At least one condition is True`

## Câu 5: Giá Trị Truthy và Falsy

**Đáp án đúng: C) `3`**

**Giải thích:**

Trong Python, các giá trị được đánh giá là `True` (truthy) hoặc `False` (falsy) trong ngữ cảnh boolean. Các giá trị falsy bao gồm: `0`, `[]`, `""`, `None`, `False`:

```python
values = [0, [], "", None, False, 1, "hello", [1, 2]]
count = 0
for val in values:
    if val:           # Kiểm tra xem val có truthy không
        count += 1    # Chỉ tăng nếu truthy
print(count)
```

**Đánh giá từng giá trị:**
- `0` → Falsy → Không tăng
- `[]` → Falsy (danh sách rỗng) → Không tăng
- `""` → Falsy (chuỗi rỗng) → Không tăng
- `None` → Falsy → Không tăng
- `False` → Falsy → Không tăng
- `1` → Truthy → Tăng (count = 1)
- `"hello"` → Truthy (chuỗi không rỗng) → Tăng (count = 2)
- `[1, 2]` → Truthy (danh sách không rỗng) → Tăng (count = 3)

**Điểm quan trọng:**
- Giá trị falsy: `False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`
- Tất cả các giá trị khác là truthy
- Collection rỗng là falsy, collection không rỗng là truthy

**Kết quả:** `3`

## Câu 6: Toán Tử So Sánh - == vs is

**Đáp án đúng: B) `True False True`**

**Giải thích:**

Toán tử `==` kiểm tra **sự bằng nhau về giá trị**, trong khi `is` kiểm tra **danh tính** (cùng một đối tượng trong bộ nhớ):

```python
a = [1, 2, 3]
b = [1, 2, 3]  # Danh sách mới với cùng giá trị
c = a          # c trỏ đến cùng đối tượng với a

print(a == b, a is b, a is c)
#           ==       is       is
#           giá trị  danh tính danh tính
```

**Phân tích:**
- `a == b`: `True` - Cả hai danh sách có cùng giá trị `[1, 2, 3]`
- `a is b`: `False` - Chúng là các đối tượng khác nhau trong bộ nhớ (danh tính khác nhau)
- `a is c`: `True` - `c` tham chiếu đến cùng đối tượng với `a` (cùng danh tính)

**Điểm quan trọng:**
- `==` so sánh giá trị/nội dung
- `is` so sánh danh tính đối tượng (địa chỉ bộ nhớ)
- Hai đối tượng có thể có cùng giá trị nhưng danh tính khác nhau
- Khi bạn gán `c = a`, cả hai biến đều tham chiếu đến cùng một đối tượng

**Kết quả:** `True False True`

## Câu 7: Toán Tử Ternary

**Đáp án đúng: B) `"even"`**

**Giải thích:**

Toán tử ternary (biểu thức điều kiện) cung cấp cách viết if-else ngắn gọn trong một dòng:

```python
x = 10
result = "even" if x % 2 == 0 else "odd"
#           giá_trị_True    điều_kiện   giá_trị_False
```

**Đánh giá:**
- `x % 2 == 0` → `10 % 2 == 0` → `0 == 0` → `True`
- Vì điều kiện là `True`, trả về giá trị đầu tiên: `"even"`

**Điểm quan trọng:**
- Cú pháp: `giá_trị_True if điều_kiện else giá_trị_False`
- Tương đương với: `if điều_kiện: result = giá_trị_True else: result = giá_trị_False`
- Điều kiện được đánh giá trước, sau đó giá trị phù hợp được trả về

**Kết quả:** `"even"`

## Câu 8: Câu Lệnh Điều Kiện Lồng Nhau

**Đáp án đúng: B) `B`**

**Giải thích:**

Câu lệnh điều kiện lồng nhau là các câu lệnh điều kiện bên trong các câu lệnh điều kiện khác. Python đánh giá điều kiện ngoài trước, sau đó điều kiện trong nếu áp dụng:

```python
x = 15
if x > 10:              # Điều kiện ngoài: 15 > 10 là True
    if x % 2 == 0:      # Điều kiện trong: 15 % 2 == 0 là False
        print("A")
    else:               # else trong được thực thi
        print("B")      # In "B"
else:
    print("C")          # Không được thực thi
```

**Từng bước:**
1. Kiểm tra điều kiện ngoài: `x > 10` → `15 > 10` → `True`
2. Vào khối `if`
3. Kiểm tra điều kiện trong: `x % 2 == 0` → `15 % 2 == 0` → `0 == 0` → `False`
4. Vào khối `else` trong
5. In `"B"`

**Điểm quan trọng:**
- Điều kiện ngoài phải là `True` để các điều kiện trong được đánh giá
- Mỗi cấp độ lồng nhau có logic if-else riêng
- Thụt lề rất quan trọng đối với câu lệnh điều kiện lồng nhau

**Kết quả:** `B`

## Câu 9: Short-Circuit Evaluation với and

**Đáp án đúng: B) `False`**

**Giải thích:**

Short-circuit evaluation có nghĩa là Python dừng đánh giá ngay khi kết quả được xác định. Với `and`, nếu điều kiện đầu tiên là `False`, điều kiện thứ hai không bao giờ được đánh giá:

```python
x = 0
result = x != 0 and (10 / x) > 1
#        x != 0: 0 != 0 là False
#        Vì False and ... = False, Python không đánh giá (10 / x)
#        Không có ZeroDivisionError!
```

**Đánh giá:**
- `x != 0` → `0 != 0` → `False`
- Với `and`, nếu toán hạng đầu tiên là `False`, toàn bộ biểu thức là `False`
- Python không đánh giá `(10 / x)` vì không cần thiết
- Kết quả: `False` (không có lỗi xảy ra)

**Điểm quan trọng:**
- Short-circuit ngăn chặn việc đánh giá không cần thiết và các lỗi tiềm ẩn
- `and`: Nếu cái đầu tiên là `False`, trả về `False` mà không kiểm tra cái thứ hai
- `or`: Nếu cái đầu tiên là `True`, trả về `True` mà không kiểm tra cái thứ hai

**Kết quả:** `False`

## Câu 10: Nhiều Điều Kiện elif

- Chỉ MỘT khối được thực thi trong chuỗi if-elif-else

## Câu 12: So Sánh với None

**Đáp án đúng: A) `A`**

**Giải thích:**

Khi so sánh với `None`, luôn sử dụng `is` hoặc `is not`, không phải `==`:

```python
x = None
if x is None:       # Cách được khuyến nghị: True
    print("A")      # Khối này được thực thi
elif x == None:     # Không được thực thi (đã khớp điều kiện đầu tiên)
    print("B")
else:
    print("C")
```

**Điểm quan trọng:**

- `== None` hoạt động nhưng không được khuyến nghị (theo hướng dẫn phong cách PEP 8)
- `is` kiểm tra danh tính đối tượng, `==` kiểm tra sự bằng nhau về giá trị
- Vì điều kiện đầu tiên khớp, `elif` và `else` không được đánh giá

## Câu 15: Biểu Thức Logic Phức Tạp

**Đáp án đúng: A) `A`**

**Giải thích:**

Độ ưu tiên toán tử quan trọng trong biểu thức logic. `and` có độ ưu tiên cao hơn `or`:

```python
x = 10
y = 5
z = 15
if (x > y) and (z > x) or (y > z):
    # Tương đương với: ((x > y) and (z > x)) or (y > z)
    # ((10 > 5) and (15 > 10)) or (5 > 15)
    # (True and True) or False
    # True or False = True
    print("A")     # Khối này được thực thi
else:
    print("B")
```

**Đánh giá (thứ tự quan trọng):**
1. `(x > y) and (z > x)` → `(10 > 5) and (15 > 10)` → `True and True` → `True`
2. `(y > z)` → `(5 > 15)` → `False`
3. `True or False` → `True`

**Điểm quan trọng:**
- Độ ưu tiên toán tử: `and` > `or`
- Biểu thức được đánh giá như: `((x > y) and (z > x)) or (y > z)`
- Sử dụng dấu ngoặc để làm rõ thứ tự nếu cần

**Kết quả:** `A`

## Câu 16: So Sánh Chuỗi

**Đáp án đúng: A) `A`**

Python so sánh chuỗi theo thứ tự từ điển (lexicographic) dựa trên giá trị Unicode/ASCII:

```python
str1 = "apple"
str2 = "banana"
if str1 < str2:    # "apple" < "banana"
                   # So sánh ký tự đầu tiên 'a' vs 'b'
                   # 'a' (97) < 'b' (98) trong ASCII
                   # True
    print("A")     # Khối này được thực thi
else:
    print("B")
```

**Đánh giá:**
- Chuỗi được so sánh từng ký tự một từ trái sang phải
- Ký tự đầu tiên: `'a'` vs `'b'`
- Trong ASCII: `'a'` (97) < `'b'` (98)
- Vì `'a' < 'b'`, `"apple" < "banana"` → `True`

**Điểm quan trọng:**
- So sánh từ điển: so sánh từng ký tự một
- Ký tự được so sánh theo code point Unicode/ASCII của chúng
- Nếu ký tự đầu tiên giống nhau, so sánh ký tự tiếp theo

**Kết quả:** `A`

## Câu 18: Toán Tử Ternary Lồng Nhau

## Câu 19: Short-Circuit Evaluation với or

**Đáp án đúng: A) `True`**

**Giải thích:**

Với `or`, nếu điều kiện đầu tiên là `True`, Python trả về `True` ngay lập tức mà không đánh giá điều kiện thứ hai:

```python
x = 5
result = x > 10 or x < 20
#         x > 10: 5 > 10 là False
#         Tiếp tục kiểm tra điều kiện thứ hai: x < 20
#         x < 20: 5 < 20 là True
#         False or True = True
```

**Đánh giá:**
- `x > 10` → `5 > 10` → `False`
- Vì điều kiện đầu tiên là `False`, đánh giá điều kiện thứ hai: `x < 20` → `5 < 20` → `True`
- `False or True` → `True`

**Điểm quan trọng:**
- `or` trả về `True` nếu BẤT KỲ toán hạng nào là `True`
- Short-circuit: nếu điều kiện đầu tiên là `True`, trả về `True` ngay lập tức
- Chỉ đánh giá điều kiện thứ hai nếu điều kiện đầu tiên là `False`

**Kết quả:** `True`

## Câu 20: Nhiều Điều Kiện với elif

**Đáp án đúng: A) `A`**

**Giải thích:**

Khi nhiều điều kiện được kiểm tra trong if-elif-else, chỉ điều kiện KHỚP ĐẦU TIÊN được thực thi:

```python
age = 25
has_license = True
if age >= 18 and has_license:  # 25 >= 18 and True = True and True = True
    print("A")                 # Khối này được thực thi và dừng
elif age >= 18:                # Không được kiểm tra (đã tìm thấy match)
    print("B")
elif has_license:              # Không được kiểm tra (đã tìm thấy match)
    print("C")
else:                          # Không được kiểm tra (đã tìm thấy match)
    print("D")
```

**Đánh giá:**
- Điều kiện đầu tiên: `age >= 18 and has_license` → `25 >= 18 and True` → `True and True` → `True`
- Vì điều kiện đầu tiên là `True`, nó được thực thi và tất cả các khối `elif`/`else` tiếp theo bị bỏ qua

**Điểm quan trọng:**
- Python chỉ thực thi điều kiện KHỚP ĐẦU TIÊN trong chuỗi if-elif-else
- Mặc dù `age >= 18` một mình cũng là `True`, nó không được kiểm tra vì điều kiện đầu tiên đã khớp
- Điều này khác với nhiều câu lệnh `if` riêng biệt sẽ đều được thực thi

**Kết quả:** `A`
