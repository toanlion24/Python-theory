## Question 20

What will be the output of the following code?

```python
class Factory:
    total_created = 0
    
    @classmethod
    def create(cls, name):
        cls.total_created += 1
        return cls(name)
    
    def __init__(self, name):
        self.name = name

f1 = Factory.create("A")
f2 = Factory.create("B")
f3 = Factory("C")
print(Factory.total_created)
```

**A.** 2

**B.** 3

**C.** 0

**D.** 1

**E.** Error: classmethod cannot create instances
