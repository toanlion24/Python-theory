## Câu 1 - Đáp án: A

### Đề bài
```python
x = 5
y = 3
result = f"{x=}, {y=}, {x+y=}"
print(result)
```

### Đáp án đúng: A. x=5, y=3, x+y=8

### Giải thích chi tiết

**Tính năng f-string debugging (Python 3.8+):**
- Thêm dấu `=` sau biến/biểu thức trong f-string
- Python tự động in tên biến + giá trị
- Rất hữu ích để debug nhanh!

**Cú pháp:**
```python
f"{biến=}"     → "biến=giá_trị"
f"{biểu_thức=}" → "biểu_thức=kết_quả"
```

**Phân tích:**
- `{x=}` → `"x=5"`
- `{y=}` → `"y=3"`
- `{x+y=}` → `"x+y=8"`

**Kết quả:** `"x=5, y=3, x+y=8"`

````python
# Minh họa
x = 5
y = 3
result = f"{x=}, {y=}, {x+y=}"
print(result)

# Các ví dụ khác
name = "Python"
print(f"{name=}")
print(f"{len(name)=}")
print(f"{name.upper()=}")
````

## Câu 2 - Đáp án: B

### Đề bài
```python
items = ['a', 'b', 'c']
result = list(enumerate(items, start=1))
print(result)
```

### Đáp án đúng: B. [(1, 'a'), (2, 'b'), (3, 'c')]

### Giải thích chi tiết

**Hàm `enumerate()`:**
- Trả về iterator với các cặp (index, phần_tử)
- Tham số `start` quyết định chỉ mục bắt đầu (mặc định 0)

**Cú pháp:**
```python
enumerate(iterable, start=0)
```

**Phân tích:**
- `start=1` → chỉ mục bắt đầu từ 1
- 'a' → (1, 'a')
- 'b' → (2, 'b')
- 'c' → (3, 'c')

**Ứng dụng thực tế:**
- Đánh số thứ tự từ 1 (thân thiện với người dùng)
- Tạo menu với số lựa chọn

````python
# Minh họa
items = ['a', 'b', 'c']

print("Mặc định (start=0):")
print(list(enumerate(items)))

print("\nVới start=1:")
print(list(enumerate(items, start=1)))

# Ứng dụng
print("\nTạo menu:")
menu = ['Pizza', 'Burger', 'Salad']
for num, item in enumerate(menu, start=1):
    print(f"{num}. {item}")
````

## Câu 3 - Đáp án: B

### Đề bài
```python
def greet(a, b, c):
    return f"{a}-{b}-{c}"

args = [1, 2, 3]
print(greet(*args))
```

### Đáp án đúng: B. 1-2-3

### Giải thích chi tiết

**Toán tử `*` trong lời gọi hàm:**
- Giải nén (unpack) list/tuple thành các đối số vị trí
- `func(*[1, 2, 3])` tương đương `func(1, 2, 3)`

**So sánh `*` trong định nghĩa vs gọi hàm:**

| Vị trí | Ý nghĩa |
|--------|--------|
| Định nghĩa: `def func(*args)` | Thu thập nhiều đối số vào tuple |
| Gọi hàm: `func(*list)` | Giải nén list thành nhiều đối số |

**Phân tích:**
```python
greet(*[1, 2, 3])
= greet(1, 2, 3)   # Giải nén
→ a=1, b=2, c=3
→ "1-2-3"
```

````python
# Minh họa
def greet(a, b, c):
    return f"{a}-{b}-{c}"

args = [1, 2, 3]
print(f"greet(*args) = {greet(*args)}")
print(f"Tương đương greet(1, 2, 3) = {greet(1, 2, 3)}")

# Ứng dụng khác
nums = [10, 20, 30, 40]
print(f"\nmax(*nums) = {max(*nums)}")
print(f"Tương đương max(10, 20, 30, 40) = {max(10, 20, 30, 40)}")
````

## Câu 4 - Đáp án: B

### Đề bài
```python
d = {'a': 1, 'b': 2}
x = d.setdefault('c', 3)
y = d.setdefault('a', 100)
print(x, y, d['a'])
```

### Đáp án đúng: B. 3 1 1

### Giải thích chi tiết

**Phương thức `dict.setdefault(key, default)`:**

| Key tồn tại? | Hành động | Trả về |
|--------------|-----------|--------|
| Không | Thêm key với giá trị default | default |
| Có | Không làm gì | Giá trị hiện có |

**Phân tích:**

1. `d.setdefault('c', 3)`:
   - 'c' KHÔNG có trong d
   - Thêm 'c': 3 vào d
   - Trả về **3**

2. `d.setdefault('a', 100)`:
   - 'a' ĐÃ CÓ trong d (giá trị 1)
   - Không thay đổi gì
   - Trả về **1** (giá trị hiện có)

3. `d['a']` = **1** (không bị đổi thành 100)

````python
# Minh họa
d = {'a': 1, 'b': 2}
print(f"Ban đầu: {d}")

x = d.setdefault('c', 3)  # Key mới
print(f"\nSau setdefault('c', 3):")
print(f"  x = {x}, d = {d}")

y = d.setdefault('a', 100)  # Key đã có
print(f"\nSau setdefault('a', 100):")
print(f"  y = {y}, d['a'] = {d['a']}")
print(f"  d = {d}")
````

## Câu 5 - Đáp án: A

### Đề bài
```python
a = b = [1, 2, 3]
a = a + [4]
b += [5]
print(a, b)
```

### Đáp án đúng: A. [1, 2, 3, 4] [1, 2, 3, 5]

### Giải thích chi tiết

**Gán chuỗi (chained assignment):**
- `a = b = [1, 2, 3]` → Cả a và b trỏ đến CÙNG list

**Phân tích từng bước:**

```
1. a = b = [1, 2, 3]
   a ──► [1, 2, 3] ◄── b

2. a = a + [4]  (tạo list MỚI cho a)
   a ──► [1, 2, 3, 4]  (mới)
   b ──► [1, 2, 3]     (gốc)

3. b += [5]  (sửa list gốc TẠI CHỖ)
   a ──► [1, 2, 3, 4]
   b ──► [1, 2, 3, 5]
```

**Điểm mấu chốt:**
- `a = a + [4]` tạo list mới, a trỏ đến list mới
- `b += [5]` sửa list gốc tại chỗ

````python
# Minh họa
a = b = [1, 2, 3]
print(f"Ban đầu: a is b = {a is b}")

a = a + [4]  # Tạo list mới
print(f"\nSau a = a + [4]:")
print(f"  a is b = {a is b}")
print(f"  a = {a}")
print(f"  b = {b}")

b += [5]  # Sửa tại chỗ
print(f"\nSau b += [5]:")
print(f"  a = {a}")
print(f"  b = {b}")
````

## Câu 6 - Đáp án: B

### Đề bài
```python
result = -7 % 3
print(result)
```

### Đáp án đúng: B. 2

### Giải thích chi tiết

**Quy tắc modulo trong Python:**
- Kết quả **CÙNG DẤU** với **số chia** (divisor)
- Công thức: `a = (a // b) * b + (a % b)`

**Tính toán:**
```
-7 % 3:
  - Số chia = 3 (dương) → kết quả dương
  - -7 // 3 = -3 (floor division, về âm vô cực)
  - Kiểm tra: (-3) * 3 + 2 = -9 + 2 = -7 ✓
  - Kết quả: 2
```

**So sánh với C/Java:**
- Python: -7 % 3 = **2** (theo dấu số chia)
- C/Java: -7 % 3 = **-1** (theo dấu số bị chia)

````python
# Minh họa
print(f"-7 % 3 = {-7 % 3}")   # 2 (số chia dương)
print(f"7 % -3 = {7 % -3}")   # -2 (số chia âm)
print(f"-7 % -3 = {-7 % -3}") # -1 (số chia âm)

# Kiểm chứng công thức
a, b = -7, 3
q = a // b
r = a % b
print(f"\nKiểm chứng: {a} = {q} * {b} + {r} = {q * b + r}")
````

## Câu 7 - Đáp án: B

### Đề bài
```python
result = -7 // 3
print(result)
```

### Đáp án đúng: B. -3

### Giải thích chi tiết

**Floor division (`//`) trong Python:**
- Làm tròn về **ÂM VÔ CỰC** (floor)
- KHÔNG phải làm tròn về 0 (truncation)

**Tính toán:**
```
-7 / 3 = -2.333...
floor(-2.333...) = -3 (số nguyên nhỏ hơn hoặc bằng)
```

**So sánh `//` và `int()`:**

| Phép tính | `//` (floor) | `int()` (truncation) |
|-----------|--------------|----------------------|
| -7 / 3 = -2.33 | -3 | -2 |
| 7 / 3 = 2.33 | 2 | 2 |

**Nhớ:** `//` luôn làm tròn XUỐNG (về âm vô cực)

````python
# Minh họa
print(f"-7 / 3 = {-7 / 3}")
print(f"-7 // 3 = {-7 // 3}")
print(f"int(-7 / 3) = {int(-7 / 3)}")

import math
print(f"\nmath.floor(-2.33) = {math.floor(-2.33)}")
print(f"math.trunc(-2.33) = {math.trunc(-2.33)}")
````

## Câu 8 - Đáp án: A

### Đề bài
```python
values = [0, "", None, "hello", [], 42]
result = any(values)
print(result)
```

### Đáp án đúng: A. True

### Giải thích chi tiết

**Hàm `any()`:**
- Trả về `True` nếu **ÍT NHẤT MỘT** phần tử là truthy
- Đánh giá ngắn mạch: dừng ngay khi gặp truthy đầu tiên
- Trả về `False` nếu tất cả falsy hoặc iterable rỗng

**Phân tích:**
```
values = [0, "", None, "hello", [], 42]
          ↓   ↓    ↓      ↓      ↓   ↓
        falsy falsy falsy TRUTHY falsy truthy
```

- `"hello"` là truthy → `any()` trả về **True**
- `any()` trả về boolean, KHÔNG phải giá trị truthy

````python
# Minh họa
values = [0, "", None, "hello", [], 42]

print("Kiểm tra từng phần tử:")
for v in values:
    print(f"  {repr(v):10} → bool = {bool(v)}")

print(f"\nany(values) = {any(values)}")
print(f"any([0, False, None]) = {any([0, False, None])}")
print(f"any([]) = {any([])}")
````

## Câu 9 - Đáp án: B

### Đề bài
```python
values = [1, 2, "", 3]
result = all(values)
print(result)
```

### Đáp án đúng: B. False

### Giải thích chi tiết

**Hàm `all()`:**
- Trả về `True` chỉ khi **TẤT CẢ** phần tử là truthy
- Đánh giá ngắn mạch: dừng ngay khi gặp falsy đầu tiên
- Trả về `True` nếu iterable rỗng (vacuous truth)

**Phân tích:**
```
values = [1, 2, "", 3]
          ↓  ↓  ↓   ↓
       truthy truthy FALSY truthy
```

- `""` (chuỗi rỗng) là falsy → `all()` trả về **False**

**So sánh `any()` và `all()`:**

| Hàm | Điều kiện True | Iterable rỗng |
|-----|----------------|---------------|
| `any()` | Ít nhất 1 truthy | False |
| `all()` | Tất cả truthy | True |

````python
# Minh họa
values = [1, 2, "", 3]

print("Kiểm tra từng phần tử:")
for v in values:
    print(f"  {repr(v):5} → bool = {bool(v)}")

print(f"\nall(values) = {all(values)}")
print(f"all([1, 2, 3]) = {all([1, 2, 3])}")
print(f"all([]) = {all([])}  ← vacuous truth")
````

## Câu 10 - Đáp án: B

### Đề bài
```python
result = list(filter(None, [0, 1, False, 2, '', 3, None]))
print(result)
```

### Đáp án đúng: B. [1, 2, 3]

### Giải thích chi tiết

**Hàm `filter()` với function=None:**
- Khi `function` là `None`, `filter()` dùng **hàm đồng nhất** (identity)
- Chỉ giữ lại các phần tử **truthy**
- Tương đương: `[x for x in iterable if x]`

**Phân tích:**
```
[0, 1, False, 2, '', 3, None]
 ↓  ↓    ↓    ↓  ↓   ↓   ↓
bỏ GIỮ  bỏ  GIỮ bỏ GIỮ  bỏ
```

**Kết quả:** [1, 2, 3]

````python
# Minh họa
data = [0, 1, False, 2, '', 3, None]

print("Kiểm tra từng phần tử:")
for v in data:
    status = "GIỮ" if v else "BỎ"
    print(f"  {repr(v):7} → {status}")

result = list(filter(None, data))
print(f"\nfilter(None, data) = {result}")

# Tương đương
result2 = [x for x in data if x]
print(f"List comprehension = {result2}")
````

print(x, y, z)

a, b = b, a  # Hoán đổi giá trị

## Câu 13 - Đáp án: A

### Đề bài
```python
text = "hello world"
words = text.split()
result = " ".join(word.capitalize() for word in words)
print(result)
```

### Đáp án đúng: A. Hello World

### Giải thích chi tiết

**Phương thức `str.capitalize()`:**
- Viết HOA ký tự đầu tiên
- Viết thường tất cả ký tự còn lại

**Phân tích:**
```
text = "hello world"
words = ["hello", "world"]

"hello".capitalize() → "Hello"
"world".capitalize() → "World"

" ".join(["Hello", "World"]) → "Hello World"
```

**So sánh các phương thức:**

| Phương thức | Ví dụ với "hELLO" |
|-------------|------------------|
| `capitalize()` | "Hello" |
| `title()` | "Hello" (cho mỗi từ) |
| `upper()` | "HELLO" |
| `lower()` | "hello" |

````python
# Minh họa
text = "hello world"
words = text.split()
print(f"words = {words}")

for word in words:
    print(f"  '{word}'.capitalize() = '{word.capitalize()}'")

result = " ".join(word.capitalize() for word in words)
print(f"\nKết quả: '{result}'")

# Cách ngắn hơn
print(f"text.title() = '{text.title()}'")
````

## Câu 14 - Đáp án: B

### Đề bài
```python
d = {'x': 1, 'y': 2}
keys = d.keys()
d['z'] = 3
print(list(keys))
```

### Đáp án đúng: B. ['x', 'y', 'z']

### Giải thích chi tiết

**Dictionary views (`keys()`, `values()`, `items()`):**
- Trả về **view object**, KHÔNG phải bản sao
- View là **động** - phản ánh thay đổi của dict
- Tiết kiệm bộ nhớ vì không tạo bản sao

**Phân tích:**
```
d = {'x': 1, 'y': 2}
keys = d.keys()  ← view, không phải copy

d['z'] = 3  ← thêm key mới

list(keys) → ['x', 'y', 'z']  ← view thấy cả 'z'
```

**Nếu muốn bản sao tĩnh:**
```python
keys_copy = list(d.keys())  # Tạo bản sao
```

````python
# Minh họa
d = {'x': 1, 'y': 2}
keys = d.keys()  # View object
print(f"Trước: list(keys) = {list(keys)}")
print(f"Kiểu: {type(keys)}")

d['z'] = 3  # Thêm key mới
print(f"\nSau thêm 'z': list(keys) = {list(keys)}")

# So sánh với bản sao
d2 = {'a': 1}
keys_copy = list(d2.keys())  # Bản sao
d2['b'] = 2
print(f"\nBản sao không thay đổi: {keys_copy}")
````

## Câu 15 - Đáp án: A

### Đề bài
```python
s1 = {1, 2, 3}
s2 = frozenset([2, 3, 4])
result = s1 - s2
print(result, type(result).__name__)
```

### Đáp án đúng: A. {1} set

### Giải thích chi tiết

**Phép toán giữa `set` và `frozenset`:**
- Có thể thực hiện các phép toán tập hợp
- **Kiểu kết quả** = kiểu của **toán hạng BÊN TRÁI**

**Phân tích:**
```
s1 = {1, 2, 3}      ← set (bên trái)
s2 = frozenset([2, 3, 4])

s1 - s2 = {1, 2, 3} - {2, 3, 4} = {1}

Kiểu kết quả: set (theo toán hạng trái)
```

**Quy tắc:**
- `set - frozenset` → kết quả là `set`
- `frozenset - set` → kết quả là `frozenset`

````python
# Minh họa
s1 = {1, 2, 3}
s2 = frozenset([2, 3, 4])

result1 = s1 - s2
print(f"s1 - s2 = {result1}")
print(f"Kiểu: {type(result1).__name__}")

result2 = s2 - s1
print(f"\ns2 - s1 = {result2}")
print(f"Kiểu: {type(result2).__name__}")
````

## Câu 16 - Đáp án: C

### Đề bài
```python
text = "abcde"
print(text[10:20])
```

### Đáp án đúng: C. ""

### Giải thích chi tiết

**Slicing và chỉ mục vượt giới hạn:**

| Phép toán | Khi vượt giới hạn |
|-----------|-------------------|
| Slicing `text[10:20]` | Trả về chuỗi rỗng `""` |
| Indexing `text[10]` | **IndexError** |

**Phân tích:**
```
text = "abcde"  (độ dài 5, index 0-4)

text[10:20] → slice bắt đầu từ index 10
           → không có phần tử nào trong phạm vi
           → trả về chuỗi rỗng ""
```

**Quy tắc:** Slicing **AN TOÀN** - không bao giờ gây IndexError

````python
# Minh họa
text = "abcde"
print(f"text = '{text}', độ dài = {len(text)}")

print(f"\ntext[10:20] = '{text[10:20]}' (chuỗi rỗng)")
print(f"text[2:100] = '{text[2:100]}' (một phần)")

# Indexing gây lỗi
try:
    print(text[10])
except IndexError as e:
    print(f"\ntext[10] gây lỗi: {e}")
````

## Câu 17 - Đáp án: B

### Đề bài
```python
a = [1, 2, 3]
b = a.copy()
c = list(a)
d = a[:]
print(a is b, a is c, a is d, b == c == d)
```

### Đáp án đúng: B. False False False True

### Giải thích chi tiết

**3 cách sao chép list (shallow copy):**

| Cách | Cú pháp |
|------|--------|
| 1 | `list.copy()` |
| 2 | `list(original)` |
| 3 | `original[:]` |

Cả 3 đều tạo **đối tượng MỚI** với cùng giá trị.

**Phân tích:**
- `a is b` → False (khác đối tượng)
- `a is c` → False (khác đối tượng)
- `a is d` → False (khác đối tượng)
- `b == c == d` → True (cùng giá trị [1, 2, 3])

````python
# Minh họa
a = [1, 2, 3]
b = a.copy()
c = list(a)
d = a[:]

print(f"id(a) = {id(a)}")
print(f"id(b) = {id(b)}")
print(f"id(c) = {id(c)}")
print(f"id(d) = {id(d)}")

print(f"\na is b: {a is b}")
print(f"a is c: {a is c}")
print(f"a is d: {a is d}")
print(f"b == c == d: {b == c == d}")
````

## Câu 18 - Đáp án: B

### Đề bài
```python
def func(**kwargs):
    return list(kwargs.keys())

data = {'a': 1, 'b': 2, 'c': 3}
print(func(**data))
```

### Đáp án đúng: B. ['a', 'b', 'c']

### Giải thích chi tiết

**Toán tử `**` với dictionary:**

| Vị trí | Ý nghĩa |
|--------|--------|
| Định nghĩa: `def func(**kwargs)` | Thu thập keyword args vào dict |
| Gọi hàm: `func(**dict)` | Giải nén dict thành keyword args |

**Phân tích:**
```python
func(**{'a': 1, 'b': 2, 'c': 3})
= func(a=1, b=2, c=3)   # Giải nén

Trong func:
  kwargs = {'a': 1, 'b': 2, 'c': 3}
  kwargs.keys() = ['a', 'b', 'c']
```

````python
# Minh họa
def func(**kwargs):
    print(f"kwargs = {kwargs}")
    return list(kwargs.keys())

data = {'a': 1, 'b': 2, 'c': 3}
print(f"func(**data) = {func(**data)}")

# Tương đương
print(f"\nfunc(a=1, b=2, c=3) = {func(a=1, b=2, c=3)}")
````

## Câu 19 - Đáp án: B

### Đề bài
```python
x = [1, 2]
y = [3, 4]
z = [*x, *y]
print(z)
```

### Đáp án đúng: B. [1, 2, 3, 4]

### Giải thích chi tiết

**Unpacking trong list literal (Python 3.5+):**
- `*iterable` giải nén các phần tử vào list/tuple
- Có thể kết hợp nhiều unpack trong cùng literal

**Phân tích:**
```python
[*x, *y]
= [*[1, 2], *[3, 4]]
= [1, 2, 3, 4]
```

**Ưu điểm so với `x + y`:**
- Có thể kết hợp với phần tử khác: `[0, *x, 2.5, *y, 5]`
- Hoạt động với bất kỳ iterable

````python
# Minh họa
x = [1, 2]
y = [3, 4]

z = [*x, *y]
print(f"[*x, *y] = {z}")

# Kết hợp với phần tử khác
w = [0, *x, 2.5, *y, 5]
print(f"[0, *x, 2.5, *y, 5] = {w}")

# Với tuple
t = (*x, *y)
print(f"(*x, *y) = {t}")
````

## Câu 20 - Đáp án: B

### Đề bài
```python
result = bool(""), bool(" "), bool("0"), bool("False")
print(result)
```

### Đáp án đúng: B. (False, True, True, True)

### Giải thích chi tiết

**Quy tắc truthiness của string:**
- Chuỗi **RỖNG** (`""`) → **False**
- **BẤT KỲ** chuỗi không rỗng → **True**

**Phân tích:**

| Chuỗi | Rỗng? | bool() |
|-------|-------|--------|
| `""` | Có | False |
| `" "` | Không (có dấu cách) | True |
| `"0"` | Không (ký tự '0') | True |
| `"False"` | Không (chữ 'False') | True |

**⚠️ Bẫy thường gặp:**
- `"0"` là **chuỗi** chứa ký tự '0', KHÔNG phải số 0
- `"False"` là **chuỗi** chứa chữ 'False', KHÔNG phải boolean

````python
# Minh họa
strings = ["", " ", "0", "False"]

print("Kiểm tra từng chuỗi:")
for s in strings:
    print(f"  {repr(s):10} → len={len(s)}, bool={bool(s)}")

result = bool(""), bool(" "), bool("0"), bool("False")
print(f"\nKết quả: {result}")

# So sánh với giá trị thực
print(f"\nSo sánh:")
print(f"  bool('0') = {bool('0')} (chuỗi)")
print(f"  bool(0) = {bool(0)} (số)")
print(f"  bool('False') = {bool('False')} (chuỗi)")
print(f"  bool(False) = {bool(False)} (boolean)")
````
