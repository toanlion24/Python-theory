# Python Basic Syntax - Cú Pháp Cơ Bản

## Giới Thiệu

Cú pháp cơ bản của Python là nền tảng để bạn có thể viết và hiểu các chương trình Python. Python được thiết kế với triết lý "đẹp hơn xấu, rõ ràng hơn mơ hồ, đơn giản hơn phức tạp". Điều này thể hiện rõ trong cú pháp của ngôn ngữ.

## 1. Comments - Ghi Chú

Ghi chú là những dòng mã không được thực thi, giúp bạn giải thích logic của chương trình cho chính bạn và những người khác đọc mã.

### Cú pháp:
- Ghi chú một dòng: Sử dụng dấu `#`
- Ghi chú nhiều dòng: Sử dụng ba dấu nháy kép `"""` hoặc ba dấu nháy đơn `'''`

````python
# Đây là một ghi chú một dòng
# Python sẽ bỏ qua dòng này khi chạy chương trình

# Ví dụ: Tính diện tích hình chữ nhật
chieu_dai = 10  # Chiều dài của hình chữ nhật (đơn vị: mét)
chieu_rong = 5   # Chiều rộng của hình chữ nhật (đơn vị: mét)
dien_tich = chieu_dai * chieu_rong  # Công thức tính diện tích
print(f"Diện tích hình chữ nhật là: {dien_tich} mét vuông")

"""
Đây là ghi chú nhiều dòng
Bạn có thể viết nhiều dòng giải thích ở đây
Python sẽ bỏ qua toàn bộ khối này
"""

'''
Bạn cũng có thể sử dụng ba dấu nháy đơn
để tạo ghi chú nhiều dòng
Cả hai cách đều hoạt động giống nhau
'''
````

## 2. Variables - Biến

Biến là tên được sử dụng để lưu trữ dữ liệu trong bộ nhớ. Trong Python, bạn không cần khai báo kiểu dữ liệu trước khi sử dụng biến.

### Đặc điểm:
- Biến được tạo khi bạn gán giá trị cho nó lần đầu tiên
- Tên biến có thể chứa chữ cái, số và dấu gạch dưới
- Tên biến phải bắt đầu bằng chữ cái hoặc dấu gạch dưới
- Tên biến phân biệt chữ hoa và chữ thường
- Python tự động xác định kiểu dữ liệu dựa trên giá trị được gán

````python
# Ví dụ 1: Tạo biến và gán giá trị
ten_sinh_vien = "Nguyễn Văn An"
tuoi = 20
diem_trung_binh = 8.5
da_tot_nghiep = False

print(f"Sinh viên: {ten_sinh_vien}")
print(f"Tuổi: {tuoi}")
print(f"Điểm trung bình: {diem_trung_binh}")
print(f"Đã tốt nghiệp: {da_tot_nghiep}")

# Ví dụ 2: Thay đổi giá trị biến
so_luong_sach = 10
print(f"Ban đầu có {so_luong_sach} quyển sách")

so_luong_sach = so_luong_sach + 5  # Mua thêm 5 quyển
print(f"Sau khi mua thêm, có {so_luong_sach} quyển sách")

# Ví dụ 3: Gán nhiều giá trị cùng lúc
x, y, z = 1, 2, 3
print(f"x = {x}, y = {y}, z = {z}")

# Ví dụ 4: Gán cùng một giá trị cho nhiều biến
a = b = c = 100
print(f"a = {a}, b = {b}, c = {c}")

# Ví dụ 5: Kiểm tra kiểu dữ liệu
print(f"Kiểu của ten_sinh_vien: {type(ten_sinh_vien)}")
print(f"Kiểu của tuoi: {type(tuoi)}")
print(f"Kiểu của diem_trung_binh: {type(diem_trung_binh)}")
print(f"Kiểu của da_tot_nghiep: {type(da_tot_nghiep)}")
````

## 3. Indentation - Thụt Lề

Thụt lề là đặc điểm quan trọng nhất của Python. Khác với các ngôn ngữ khác sử dụng dấu ngoặc nhọn `{}` để xác định khối mã, Python sử dụng thụt lề (khoảng trắng hoặc tab) để xác định cấu trúc mã.

### Quy tắc:
- Mỗi cấp độ thụt lề thường là 4 khoảng trắng (khuyến nghị)
- Tất cả các dòng trong cùng một khối mã phải có cùng mức độ thụt lề
- Thụt lề sai sẽ gây ra lỗi `IndentationError`

````python
# Ví dụ 1: Sử dụng thụt lề trong câu lệnh if
so_diem = 85

if so_diem >= 90:
    print("Xếp loại: Xuất sắc")
    print("Bạn đã đạt thành tích cao!")
elif so_diem >= 80:
    print("Xếp loại: Giỏi")
    print("Bạn đã làm tốt!")
else:
    print("Xếp loại: Khá")
    print("Hãy cố gắng thêm!")

# Ví dụ 2: Thụt lề trong vòng lặp
print("\nĐếm từ 1 đến 5:")
for i in range(1, 6):
    print(f"Số thứ {i}")
    if i == 3:
        print("  Đây là số ở giữa!")
    print("  Đã in xong số này")

# Ví dụ 3: Thụt lề lồng nhau
print("\nBảng cửu chương 2:")
for so_nhan in range(1, 6):
    ket_qua = 2 * so_nhan
    print(f"2 x {so_nhan} = {ket_qua}")
    if ket_qua > 5:
        print("  Kết quả lớn hơn 5")
        if ket_qua > 8:
            print("    Kết quả còn lớn hơn 8 nữa!")
````

## 4. Print Statement - Câu Lệnh In

Câu lệnh `print()` được sử dụng để hiển thị thông tin ra màn hình. Đây là một trong những hàm được sử dụng nhiều nhất trong Python.

### Cú pháp cơ bản:
- `print(value)` - In một giá trị
- `print(value1, value2, ...)` - In nhiều giá trị, cách nhau bởi khoảng trắng
- `print(value, end='')` - Thay đổi ký tự kết thúc (mặc định là xuống dòng)
- `print(value, sep='')` - Thay đổi ký tự phân cách giữa các giá trị

````python
# Ví dụ 1: In một giá trị đơn giản
print("Xin chào, tôi đang học Python!")

# Ví dụ 2: In nhiều giá trị
ten = "Mai"
tuoi = 25
print("Tên:", ten, "Tuổi:", tuoi)

# Ví dụ 3: In với f-string (cách hiện đại và khuyến nghị)
mon_hoc = "Python"
diem_so = 9.5
print(f"Môn học: {mon_hoc}, Điểm số: {diem_so}")

# Ví dụ 4: In với format() method
san_pham = "Laptop"
gia = 15000000
print("Sản phẩm: {}, Giá: {:,} VNĐ".format(san_pham, gia))

# Ví dụ 5: Thay đổi ký tự kết thúc
print("Dòng 1", end=" ")
print("Dòng 2", end=" ")
print("Dòng 3")

# Ví dụ 6: Thay đổi ký tự phân cách
print("Nguyễn", "Văn", "An", sep="-")
print("2024", "01", "15", sep="/")

# Ví dụ 7: In nhiều dòng
print("Dòng đầu tiên")
print("Dòng thứ hai")
print("Dòng thứ ba")

# Ví dụ 8: In với escape sequences
print("Tên tôi là \"Python\"")
print("Đường dẫn: C:\\Users\\Documents")
print("Dòng 1\nDòng 2\nDòng 3")
print("Tab\tGiữa\tCác\tTừ")
````

## 5. Input Statement - Câu Lệnh Nhập

Câu lệnh `input()` được sử dụng để nhận dữ liệu từ người dùng thông qua bàn phím. Tất cả dữ liệu nhập vào đều được xử lý như chuỗi ký tự (string).

### Cú pháp:
- `input()` - Nhận dữ liệu không có thông báo
- `input("Thông báo: ")` - Hiển thị thông báo trước khi nhận dữ liệu

````python
# Ví dụ 1: Nhập tên người dùng
ten_nguoi_dung = input("Vui lòng nhập tên của bạn: ")
print(f"Xin chào, {ten_nguoi_dung}!")

# Ví dụ 2: Nhập số và chuyển đổi kiểu
# Lưu ý: input() luôn trả về chuỗi, cần chuyển đổi sang số
tuoi_string = input("Vui lòng nhập tuổi của bạn: ")
tuoi_so = int(tuoi_string)  # Chuyển đổi sang số nguyên
print(f"Bạn {tuoi_so} tuổi, sau 5 năm nữa bạn sẽ {tuoi_so + 5} tuổi")

# Ví dụ 3: Nhập số thực
diem_so_string = input("Vui lòng nhập điểm số của bạn: ")
diem_so = float(diem_so_string)  # Chuyển đổi sang số thực
print(f"Điểm số của bạn là: {diem_so}")

# Ví dụ 4: Tính toán với dữ liệu nhập vào
print("\n=== Tính diện tích hình vuông ===")
canh = float(input("Nhập độ dài cạnh hình vuông (cm): "))
dien_tich = canh * canh
print(f"Diện tích hình vuông là: {dien_tich} cm²")

# Ví dụ 5: Nhập nhiều giá trị
print("\n=== Thông tin sinh viên ===")
ho_ten = input("Họ và tên: ")
ma_sinh_vien = input("Mã sinh viên: ")
lop = input("Lớp: ")
print(f"\nThông tin sinh viên:")
print(f"  Họ tên: {ho_ten}")
print(f"  Mã SV: {ma_sinh_vien}")
print(f"  Lớp: {lop}")
````

## 6. Data Types - Kiểu Dữ Liệu

Python có nhiều kiểu dữ liệu tích hợp sẵn. Kiểu dữ liệu xác định loại giá trị mà biến có thể lưu trữ và các phép toán có thể thực hiện.

### Các kiểu dữ liệu cơ bản:
1. **int** - Số nguyên (Integer)
2. **float** - Số thực (Floating point)
3. **str** - Chuỗi ký tự (String)
4. **bool** - Giá trị logic (Boolean: True/False)
5. **NoneType** - Giá trị rỗng (None)

````python
# Ví dụ 1: Kiểu int - Số nguyên
so_nguyen_duong = 42
so_nguyen_am = -15
so_khong = 0
so_lon = 1000000

print("=== Kiểu int (Số nguyên) ===")
print(f"Số nguyên dương: {so_nguyen_duong}, kiểu: {type(so_nguyen_duong)}")
print(f"Số nguyên âm: {so_nguyen_am}, kiểu: {type(so_nguyen_am)}")
print(f"Số không: {so_khong}, kiểu: {type(so_khong)}")

# Ví dụ 2: Kiểu float - Số thực
so_thuc_duong = 3.14
so_thuc_am = -2.5
so_thuc_lon = 123.456789
so_khoa_hoc = 1.5e3  # 1.5 x 10^3 = 1500.0

print("\n=== Kiểu float (Số thực) ===")
print(f"Số thực dương: {so_thuc_duong}, kiểu: {type(so_thuc_duong)}")
print(f"Số thực âm: {so_thuc_am}, kiểu: {type(so_thuc_am)}")
print(f"Số khoa học: {so_khoa_hoc}, kiểu: {type(so_khoa_hoc)}")

# Ví dụ 3: Kiểu str - Chuỗi ký tự
chuoi_don = 'Xin chào'
chuoi_kep = "Tôi đang học Python"
chuoi_nhieu_dong = """Đây là chuỗi
nhiều dòng
có thể viết
như thế này"""

print("\n=== Kiểu str (Chuỗi ký tự) ===")
print(f"Chuỗi đơn: {chuoi_don}, kiểu: {type(chuoi_don)}")
print(f"Chuỗi kép: {chuoi_kep}, kiểu: {type(chuoi_kep)}")
print(f"Chuỗi nhiều dòng:\n{chuoi_nhieu_dong}")

# Ví dụ 4: Kiểu bool - Giá trị logic
dung = True
sai = False
ket_qua_so_sanh = 5 > 3  # True
ket_qua_bang = 2 == 3    # False

print("\n=== Kiểu bool (Giá trị logic) ===")
print(f"True: {dung}, kiểu: {type(dung)}")
print(f"False: {sai}, kiểu: {type(sai)}")
print(f"5 > 3: {ket_qua_so_sanh}")
print(f"2 == 3: {ket_qua_bang}")

# Ví dụ 5: Kiểu NoneType - Giá trị rỗng
gia_tri_rong = None

print("\n=== Kiểu NoneType (Giá trị rỗng) ===")
print(f"None: {gia_tri_rong}, kiểu: {type(gia_tri_rong)}")

# Ví dụ 6: Chuyển đổi kiểu dữ liệu
print("\n=== Chuyển đổi kiểu dữ liệu ===")
so_chuoi = "123"
so_nguyen = int(so_chuoi)
so_thuc = float(so_chuoi)
print(f"Chuỗi '{so_chuoi}' -> Số nguyên: {so_nguyen}, kiểu: {type(so_nguyen)}")
print(f"Chuỗi '{so_chuoi}' -> Số thực: {so_thuc}, kiểu: {type(so_thuc)}")

so_nguyen_2 = 456
chuoi_2 = str(so_nguyen_2)
print(f"Số nguyên {so_nguyen_2} -> Chuỗi: '{chuoi_2}', kiểu: {type(chuoi_2)}")
````

## 7. Operators - Toán Tử

Toán tử là các ký hiệu đặc biệt được sử dụng để thực hiện các phép toán hoặc thao tác trên các giá trị và biến.

### Các loại toán tử:
1. **Arithmetic Operators** - Toán tử số học
2. **Comparison Operators** - Toán tử so sánh
3. **Logical Operators** - Toán tử logic
4. **Assignment Operators** - Toán tử gán
5. **Identity Operators** - Toán tử định danh
6. **Membership Operators** - Toán tử thành viên

````python
# === TOÁN TỬ SỐ HỌC ===
print("=== Toán tử số học ===")
a = 10
b = 3

print(f"a = {a}, b = {b}")
print(f"a + b (cộng) = {a + b}")
print(f"a - b (trừ) = {a - b}")
print(f"a * b (nhân) = {a * b}")
print(f"a / b (chia) = {a / b}")
print(f"a // b (chia lấy phần nguyên) = {a // b}")
print(f"a % b (chia lấy phần dư) = {a % b}")
print(f"a ** b (lũy thừa) = {a ** b}")

# === TOÁN TỬ SO SÁNH ===
print("\n=== Toán tử so sánh ===")
x = 5
y = 8

print(f"x = {x}, y = {y}")
print(f"x == y (bằng) : {x == y}")
print(f"x != y (khác) : {x != y}")
print(f"x < y (nhỏ hơn) : {x < y}")
print(f"x > y (lớn hơn) : {x > y}")
print(f"x <= y (nhỏ hơn hoặc bằng) : {x <= y}")
print(f"x >= y (lớn hơn hoặc bằng) : {x >= y}")

# === TOÁN TỬ LOGIC ===
print("\n=== Toán tử logic ===")
p = True
q = False

print(f"p = {p}, q = {q}")
print(f"p and q (và) : {p and q}")
print(f"p or q (hoặc) : {p or q}")
print(f"not p (phủ định) : {not p}")
print(f"not q (phủ định) : {not q}")

# Ví dụ thực tế với toán tử logic
diem_so = 85
co_lam_bai_tap = True
co_di_hoc_du = True

dat_yeu_cau = (diem_so >= 80) and co_lam_bai_tap and co_di_hoc_du
print(f"\nĐiểm số: {diem_so}, Làm bài tập: {co_lam_bai_tap}, Đi học đủ: {co_di_hoc_du}")
print(f"Đạt yêu cầu: {dat_yeu_cau}")

# === TOÁN TỬ GÁN ===
print("\n=== Toán tử gán ===")
so = 10
print(f"Giá trị ban đầu: {so}")

so += 5  # Tương đương: so = so + 5
print(f"Sau += 5: {so}")

so -= 3  # Tương đương: so = so - 3
print(f"Sau -= 3: {so}")

so *= 2  # Tương đương: so = so * 2
print(f"Sau *= 2: {so}")

so /= 4  # Tương đương: so = so / 4
print(f"Sau /= 4: {so}")

# === TOÁN TỬ ĐỊNH DANH ===
print("\n=== Toán tử định danh ===")
list1 = [1, 2, 3]
list2 = [1, 2, 3]
list3 = list1

print(f"list1 = {list1}")
print(f"list2 = {list2}")
print(f"list3 = {list3}")
print(f"list1 is list2 (cùng đối tượng?) : {list1 is list2}")
print(f"list1 is list3 (cùng đối tượng?) : {list1 is list3}")
print(f"list1 == list2 (cùng giá trị?) : {list1 == list2}")

# === TOÁN TỬ THÀNH VIÊN ===
print("\n=== Toán tử thành viên ===")
chuoi = "Python Programming"
danh_sach = [1, 2, 3, 4, 5]

print(f"Chuỗi: '{chuoi}'")
print(f"'Python' in chuoi : {'Python' in chuoi}")
print(f"'Java' in chuoi : {'Java' in chuoi}")
print(f"'P' in chuoi : {'P' in chuoi}")

print(f"\nDanh sách: {danh_sach}")
print(f"3 in danh_sach : {3 in danh_sach}")
print(f"10 in danh_sach : {10 in danh_sach}")
print(f"6 not in danh_sach : {6 not in danh_sach}")
````

## 8. String Operations - Thao Tác Với Chuỗi

Chuỗi ký tự (string) là một trong những kiểu dữ liệu quan trọng nhất trong Python. Python cung cấp nhiều cách để làm việc với chuỗi.

### Các thao tác cơ bản:
- Nối chuỗi (Concatenation)
- Nhân chuỗi (Repetition)
- Truy cập ký tự (Indexing)
- Cắt chuỗi (Slicing)
- Định dạng chuỗi (String Formatting)

````python
# Ví dụ 1: Nối chuỗi
ho = "Nguyễn"
ten = "Văn An"
ho_ten = ho + " " + ten
print(f"Nối chuỗi: '{ho}' + ' ' + '{ten}' = '{ho_ten}'")

# Ví dụ 2: Nhân chuỗi
dau_gach = "-"
duong_ke = dau_gach * 30
print(f"\nNhân chuỗi: '-' * 30 = '{duong_ke}'")

# Ví dụ 3: Truy cập ký tự (Indexing)
chuoi = "Python"
print(f"\nChuỗi: '{chuoi}'")
print(f"Ký tự đầu tiên (index 0): '{chuoi[0]}'")
print(f"Ký tự thứ hai (index 1): '{chuoi[1]}'")
print(f"Ký tự cuối cùng (index -1): '{chuoi[-1]}'")
print(f"Ký tự gần cuối (index -2): '{chuoi[-2]}'")

# Ví dụ 4: Cắt chuỗi (Slicing)
chuoi_dai = "Python Programming"
print(f"\nChuỗi: '{chuoi_dai}'")
print(f"Từ đầu đến index 6: '{chuoi_dai[0:6]}'")
print(f"Từ index 7 đến cuối: '{chuoi_dai[7:]}'")
print(f"Từ đầu đến index 10: '{chuoi_dai[:11]}'")
print(f"Toàn bộ chuỗi: '{chuoi_dai[:]}'")
print(f"Đảo ngược: '{chuoi_dai[::-1]}'")
print(f"Mỗi 2 ký tự: '{chuoi_dai[::2]}'")

# Ví dụ 5: Định dạng chuỗi với f-string
ten_mon = "Python"
so_tin_chi = 3
diem_so = 9.5
print(f"\n=== Định dạng với f-string ===")
print(f"Môn học: {ten_mon}, Số tín chỉ: {so_tin_chi}, Điểm: {diem_so}")

# Ví dụ 6: Định dạng với format()
print("\n=== Định dạng với format() ===")
print("Tên: {}, Tuổi: {}, Điểm: {:.2f}".format("Mai", 20, 8.756))

# Ví dụ 7: Các phương thức chuỗi cơ bản
chuoi_mau = "  Python Programming  "
print(f"\nChuỗi mẫu: '{chuoi_mau}'")
print(f"Chữ hoa: '{chuoi_mau.upper()}'")
print(f"Chữ thường: '{chuoi_mau.lower()}'")
print(f"Viết hoa chữ cái đầu: '{chuoi_mau.strip().capitalize()}'")
print(f"Loại bỏ khoảng trắng hai đầu: '{chuoi_mau.strip()}'")
print(f"Thay thế: '{chuoi_mau.replace('Python', 'Java')}'")
print(f"Độ dài: {len(chuoi_mau.strip())}")
print(f"Đếm 'm': {chuoi_mau.count('m')}")
````

## 9. Conditional Statements - Câu Lệnh Điều Kiện

Câu lệnh điều kiện cho phép chương trình thực hiện các hành động khác nhau dựa trên các điều kiện khác nhau.

### Cú pháp:
- `if` - Nếu điều kiện đúng
- `elif` - Nếu điều kiện trước đó sai và điều kiện này đúng
- `else` - Nếu tất cả điều kiện đều sai

````python
# Ví dụ 1: Câu lệnh if đơn giản
tuoi = 18
if tuoi >= 18:
    print("Bạn đã đủ tuổi trưởng thành")

# Ví dụ 2: if - else
diem_so = 75
if diem_so >= 50:
    print("Bạn đã đậu kỳ thi")
else:
    print("Bạn chưa đậu, cần thi lại")

# Ví dụ 3: if - elif - else (nhiều điều kiện)
diem_trung_binh = 85
print(f"\nĐiểm trung bình: {diem_trung_binh}")

if diem_trung_binh >= 90:
    xep_loai = "Xuất sắc"
    phan_thuong = "Học bổng toàn phần"
elif diem_trung_binh >= 80:
    xep_loai = "Giỏi"
    phan_thuong = "Học bổng một phần"
elif diem_trung_binh >= 70:
    xep_loai = "Khá"
    phan_thuong = "Giấy khen"
elif diem_trung_binh >= 50:
    xep_loai = "Trung bình"
    phan_thuong = "Không có"
else:
    xep_loai = "Yếu"
    phan_thuong = "Cần cố gắng"

print(f"Xếp loại: {xep_loai}")
print(f"Phần thưởng: {phan_thuong}")

# Ví dụ 4: Điều kiện lồng nhau
nhiet_do = 25
troi_mua = False

print(f"\nNhiệt độ: {nhiet_do}°C, Trời mưa: {troi_mua}")

if nhiet_do > 30:
    if troi_mua:
        print("Trời nóng và mưa, nên ở trong nhà")
    else:
        print("Trời nóng, nên đi bơi")
elif nhiet_do > 20:
    if troi_mua:
        print("Trời mát và mưa, nên mang áo mưa")
    else:
        print("Thời tiết đẹp, nên đi dạo")
else:
    print("Trời lạnh, nên mặc ấm")

# Ví dụ 5: Điều kiện với toán tử logic
co_tien = True
co_thoi_gian = True
co_ban_be = False

if co_tien and co_thoi_gian:
    if co_ban_be:
        print("\nĐi xem phim với bạn bè")
    else:
        print("\nĐi xem phim một mình")
elif co_tien or co_thoi_gian:
    print("\nCó thể làm việc khác")
else:
    print("\nỞ nhà nghỉ ngơi")
````

## 10. Loops - Vòng Lặp

Vòng lặp cho phép thực hiện một khối mã nhiều lần. Python có hai loại vòng lặp chính: `for` và `while`.

### Vòng lặp for:
- Sử dụng khi biết trước số lần lặp
- Duyệt qua các phần tử trong một dãy (sequence)

### Vòng lặp while:
- Sử dụng khi số lần lặp phụ thuộc vào điều kiện
- Lặp lại cho đến khi điều kiện trở thành False

````python
# === VÒNG LẶP FOR ===
print("=== Vòng lặp for ===")

# Ví dụ 1: Lặp với range()
print("\n1. Đếm từ 1 đến 5:")
for i in range(1, 6):
    print(f"  Số: {i}")

# Ví dụ 2: Lặp với range() và bước nhảy
print("\n2. Đếm từ 0 đến 10, mỗi 2 số:")
for i in range(0, 11, 2):
    print(f"  Số chẵn: {i}")

# Ví dụ 3: Lặp qua chuỗi
print("\n3. Lặp qua từng ký tự trong chuỗi:")
chuoi = "Python"
for ky_tu in chuoi:
    print(f"  Ký tự: {ky_tu}")

# Ví dụ 4: Lặp qua danh sách
print("\n4. Lặp qua danh sách môn học:")
danh_sach_mon = ["Toán", "Lý", "Hóa", "Sinh", "Anh"]
for mon in danh_sach_mon:
    print(f"  Môn: {mon}")

# Ví dụ 5: Lặp với enumerate() để lấy cả index và giá trị
print("\n5. Lặp với index và giá trị:")
danh_sach_diem = [8.5, 9.0, 7.5, 8.0, 9.5]
for index, diem in enumerate(danh_sach_diem, start=1):
    print(f"  Môn {index}: {diem} điểm")

# Ví dụ 6: Vòng lặp lồng nhau
print("\n6. Bảng cửu chương 2:")
for so_nhan in range(1, 6):
    ket_qua = 2 * so_nhan
    print(f"  2 x {so_nhan} = {ket_qua}")

# === VÒNG LẶP WHILE ===
print("\n\n=== Vòng lặp while ===")

# Ví dụ 7: Đếm ngược
print("\n7. Đếm ngược từ 5 về 1:")
dem = 5
while dem > 0:
    print(f"  Còn {dem} giây...")
    dem = dem - 1
print("  Bắt đầu!")

# Ví dụ 8: Tính tổng cho đến khi đạt mục tiêu
print("\n8. Tính tổng các số từ 1 đến khi tổng >= 20:")
tong = 0
so = 1
while tong < 20:
    tong = tong + so
    print(f"  Số: {so}, Tổng hiện tại: {tong}")
    so = so + 1
print(f"  Tổng cuối cùng: {tong}")

# Ví dụ 9: Vòng lặp với break (thoát sớm)
print("\n9. Tìm số đầu tiên chia hết cho 7 trong khoảng 1-50:")
for i in range(1, 51):
    if i % 7 == 0:
        print(f"  Tìm thấy: {i}")
        break

# Ví dụ 10: Vòng lặp với continue (bỏ qua lần lặp hiện tại)
print("\n10. In các số lẻ từ 1 đến 10:")
for i in range(1, 11):
    if i % 2 == 0:  # Nếu là số chẵn
        continue     # Bỏ qua, không in
    print(f"  Số lẻ: {i}")

# Ví dụ 11: Vòng lặp với else (chạy khi vòng lặp kết thúc bình thường)
print("\n11. Kiểm tra số nguyên tố:")
so_kiem_tra = 7
for i in range(2, so_kiem_tra):
    if so_kiem_tra % i == 0:
        print(f"  {so_kiem_tra} không phải số nguyên tố")
        break
else:
    print(f"  {so_kiem_tra} là số nguyên tố")
````

## 11. Functions - Hàm (Cơ Bản)

Hàm là một khối mã có thể tái sử dụng, được định nghĩa một lần và có thể được gọi nhiều lần. Hàm giúp mã nguồn gọn gàng, dễ đọc và dễ bảo trì hơn.

### Cú pháp định nghĩa hàm:
```python
def ten_ham(tham_so):
    # Khối mã của hàm
    return gia_tri_tra_ve
```

````python
# Ví dụ 1: Hàm đơn giản không có tham số
def chao_hoi():
    print("Xin chào! Chào mừng bạn đến với Python")
    print("Hãy bắt đầu học tập ngay hôm nay!")

# Gọi hàm
print("=== Ví dụ 1: Hàm không tham số ===")
chao_hoi()

# Ví dụ 2: Hàm có tham số
def in_thong_tin(ten, tuoi):
    print(f"Tên: {ten}")
    print(f"Tuổi: {tuoi}")

print("\n=== Ví dụ 2: Hàm có tham số ===")
in_thong_tin("Nguyễn Văn An", 20)
in_thong_tin("Trần Thị Bình", 22)

# Ví dụ 3: Hàm có giá trị trả về
def tinh_dien_tich_hinh_chu_nhat(chieu_dai, chieu_rong):
    dien_tich = chieu_dai * chieu_rong
    return dien_tich

print("\n=== Ví dụ 3: Hàm có giá trị trả về ===")
dien_tich_1 = tinh_dien_tich_hinh_chu_nhat(10, 5)
dien_tich_2 = tinh_dien_tich_hinh_chu_nhat(8, 6)
print(f"Diện tích hình chữ nhật 1: {dien_tich_1} m²")
print(f"Diện tích hình chữ nhật 2: {dien_tich_2} m²")

# Ví dụ 4: Hàm với nhiều giá trị trả về
def tinh_toan(so_a, so_b):
    tong = so_a + so_b
    hieu = so_a - so_b
    tich = so_a * so_b
    thuong = so_a / so_b
    return tong, hieu, tich, thuong

print("\n=== Ví dụ 4: Hàm trả về nhiều giá trị ===")
ket_qua = tinh_toan(10, 2)
tong, hieu, tich, thuong = ket_qua
print(f"Với a=10, b=2:")
print(f"  Tổng: {tong}")
print(f"  Hiệu: {hieu}")
print(f"  Tích: {tich}")
print(f"  Thương: {thuong}")

# Ví dụ 5: Hàm với tham số mặc định
def gioi_thieu(ten, tuoi=18, que_quan="Việt Nam"):
    print(f"Tên: {ten}")
    print(f"Tuổi: {tuoi}")
    print(f"Quê quán: {que_quan}")

print("\n=== Ví dụ 5: Hàm với tham số mặc định ===")
gioi_thieu("Mai", 20, "Hà Nội")
print()
gioi_thieu("Lan")  # Chỉ truyền tên, tuổi và quê quán dùng giá trị mặc định
print()
gioi_thieu("Hùng", 25)  # Truyền tên và tuổi, quê quán dùng giá trị mặc định

# Ví dụ 6: Hàm với tham số có tên (keyword arguments)
def tao_hoa_don(ten_san_pham, so_luong, don_gia, giam_gia=0):
    thanh_tien = so_luong * don_gia
    tong_tien = thanh_tien - giam_gia
    print(f"\n=== HÓA ĐƠN ===")
    print(f"Sản phẩm: {ten_san_pham}")
    print(f"Số lượng: {so_luong}")
    print(f"Đơn giá: {don_gia:,} VNĐ")
    print(f"Thành tiền: {thanh_tien:,} VNĐ")
    if giam_gia > 0:
        print(f"Giảm giá: {giam_gia:,} VNĐ")
    print(f"Tổng cộng: {tong_tien:,} VNĐ")

print("\n=== Ví dụ 6: Hàm với tham số có tên ===")
tao_hoa_don(ten_san_pham="Laptop", so_luong=1, don_gia=15000000, giam_gia=500000)
tao_hoa_don("Chuột máy tính", 2, 500000)  # Không giảm giá
````

## 12. Scope - Phạm Vi Biến

Phạm vi (scope) xác định nơi một biến có thể được truy cập trong chương trình. Python có các loại phạm vi: local (cục bộ), global (toàn cục), và nonlocal.

### Các loại phạm vi:
- **Local scope**: Biến được định nghĩa bên trong hàm
- **Global scope**: Biến được định nghĩa ở cấp độ module (ngoài hàm)
- **Nonlocal scope**: Biến trong hàm lồng nhau

````python
# === PHẠM VI TOÀN CỤC (GLOBAL SCOPE) ===
bien_toan_cuc = "Tôi là biến toàn cục"

def ham_1():
    print(f"Trong ham_1: {bien_toan_cuc}")

print("=== Phạm vi toàn cục ===")
print(f"Ngoài hàm: {bien_toan_cuc}")
ham_1()

# === PHẠM VI CỤC BỘ (LOCAL SCOPE) ===
def ham_2():
    bien_cuc_bo = "Tôi là biến cục bộ"
    print(f"Trong ham_2: {bien_cuc_bo}")

print("\n=== Phạm vi cục bộ ===")
ham_2()
# print(bien_cuc_bo)  # Lỗi! Không thể truy cập biến cục bộ từ ngoài hàm

# === THAY ĐỔI BIẾN TOÀN CỤC TRONG HÀM ===
dem_toan_cuc = 0

def tang_dem():
    global dem_toan_cuc  # Khai báo sử dụng biến toàn cục
    dem_toan_cuc = dem_toan_cuc + 1
    print(f"Trong hàm, đếm = {dem_toan_cuc}")

print("\n=== Thay đổi biến toàn cục ===")
print(f"Ban đầu: {dem_toan_cuc}")
tang_dem()
tang_dem()
print(f"Sau khi gọi hàm: {dem_toan_cuc}")

# === PHẠM VI NONLOCAL (Hàm lồng nhau) ===
def ham_ngoai():
    bien_nonlocal = "Giá trị ban đầu"
    
    def ham_trong():
        nonlocal bien_nonlocal  # Khai báo sử dụng biến nonlocal
        bien_nonlocal = "Giá trị đã thay đổi"
        print(f"Trong ham_trong: {bien_nonlocal}")
    
    print(f"Trước khi gọi ham_trong: {bien_nonlocal}")
    ham_trong()
    print(f"Sau khi gọi ham_trong: {bien_nonlocal}")

print("\n=== Phạm vi nonlocal ===")
ham_ngoai()
````

## 13. Error Handling - Xử Lý Lỗi (Cơ Bản)

Khi chương trình gặp lỗi, Python sẽ dừng lại và hiển thị thông báo lỗi. Chúng ta có thể xử lý lỗi bằng cách sử dụng `try` và `except` để chương trình tiếp tục chạy.

### Cú pháp:
```python
try:
    # Mã có thể gây lỗi
except LoaiLoi:
    # Xử lý khi có lỗi
```

````python
# Ví dụ 1: Xử lý lỗi chia cho 0
print("=== Ví dụ 1: Xử lý lỗi chia cho 0 ===")
try:
    so_a = 10
    so_b = 0
    ket_qua = so_a / so_b
    print(f"Kết quả: {ket_qua}")
except ZeroDivisionError:
    print("Lỗi: Không thể chia cho 0!")

# Ví dụ 2: Xử lý lỗi chuyển đổi kiểu
print("\n=== Ví dụ 2: Xử lý lỗi chuyển đổi kiểu ===")
try:
    chuoi_so = input("Nhập một số: ")
    so = int(chuoi_so)
    print(f"Số bạn nhập: {so}")
except ValueError:
    print("Lỗi: Bạn phải nhập một số nguyên hợp lệ!")

# Ví dụ 3: Xử lý nhiều loại lỗi
print("\n=== Ví dụ 3: Xử lý nhiều loại lỗi ===")
try:
    danh_sach = [1, 2, 3]
    print(f"Phần tử thứ 5: {danh_sach[5]}")
except IndexError:
    print("Lỗi: Chỉ số vượt quá phạm vi của danh sách!")

# Ví dụ 4: Xử lý lỗi với else (chạy khi không có lỗi)
print("\n=== Ví dụ 4: Xử lý với else ===")
try:
    so_1 = 10
    so_2 = 2
    ket_qua = so_1 / so_2
except ZeroDivisionError:
    print("Lỗi: Không thể chia cho 0!")
else:
    print(f"Phép chia thành công! Kết quả: {ket_qua}")

# Ví dụ 5: Xử lý lỗi với finally (luôn chạy)
print("\n=== Ví dụ 5: Xử lý với finally ===")
try:
    file = open("file_khong_ton_tai.txt", "r")
    noi_dung = file.read()
except FileNotFoundError:
    print("Lỗi: Không tìm thấy file!")
finally:
    print("Khối finally luôn được thực thi, dù có lỗi hay không")
````
