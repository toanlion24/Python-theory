## Câu 1 - Đáp án: A

### Đề bài
```python
text = "abcdefgh"
print(text[-3::-2])
```

### Đáp án đúng: A. fdb

### Giải thích chi tiết

**Cú pháp slicing:** `[start:end:step]`

**Phân tích chuỗi:**
```
Chuỗi:    a   b   c   d   e   f   g   h
Index:    0   1   2   3   4   5   6   7
Index âm: -8  -7  -6  -5  -4  -3  -2  -1
```

**Phân tích `text[-3::-2]`:**
- `start = -3`: Bắt đầu từ chỉ mục -3, tức là ký tự 'f' (index 5)
- `end` không được chỉ định: Đi đến hết (về đầu chuỗi vì step âm)
- `step = -2`: Lùi 2 bước mỗi lần (đi ngược)

**Quá trình lấy ký tự:**
1. Bắt đầu tại index 5 → 'f'
2. Lùi 2 bước → index 3 → 'd'
3. Lùi 2 bước → index 1 → 'b'
4. Lùi 2 bước → index -1 (ngoài phạm vi) → dừng

**Kết quả:** `"fdb"`

### Quy tắc nhớ
- **Step âm** = đi ngược từ phải sang trái
- Khi step âm và không có end, Python đi đến đầu chuỗi

````python
# Minh họa
text = "abcdefgh"
print(f"Chuỗi gốc: {text}")
print(f"text[-3] = '{text[-3]}' (ký tự 'f', index 5)")
print(f"text[-3::-2] = '{text[-3::-2]}'")

# Phân tích từng bước
print("\nPhân tích từng bước:")
print(f"  Index 5 → '{text[5]}'")
print(f"  Index 3 → '{text[3]}'")
print(f"  Index 1 → '{text[1]}'")
````

## Câu 2 - Đáp án: B

### Đề bài
```python
x = 10

def outer():
    x = 20
    def inner():
        x = 30
        print(x, end=" ")
    inner()
    print(x, end=" ")

outer()
print(x)
```

### Đáp án đúng: B. 30 20 10

### Giải thích chi tiết

**Quy tắc LEGB (Phạm vi biến trong Python):**
- **L**ocal: Biến cục bộ trong hàm hiện tại
- **E**nclosing: Biến trong hàm bao ngoài
- **G**lobal: Biến toàn cục
- **B**uilt-in: Biến/hàm có sẵn của Python

**Phân tích code:**

1. **`x = 10`**: Biến toàn cục x = 10

2. **Trong `outer()`**: 
   - `x = 20` tạo biến CỤC BỘ mới x = 20 (không ảnh hưởng x toàn cục)
   
3. **Trong `inner()`**:
   - `x = 30` tạo biến CỤC BỘ mới x = 30 (không ảnh hưởng x của outer)
   - `print(x)` in **30** (x cục bộ của inner)

4. **Sau khi inner() chạy xong**:
   - Quay lại outer(), `print(x)` in **20** (x cục bộ của outer)

5. **Sau khi outer() chạy xong**:
   - `print(x)` in **10** (x toàn cục)

### Điểm quan trọng
- Mỗi hàm có **phạm vi riêng biệt**
- Phép gán `x = ...` tạo biến CỤC BỘ mới
- Để thay đổi biến ngoài, cần dùng `global` hoặc `nonlocal`

````python
# Minh họa
x = 10  # Biến toàn cục

def outer():
    x = 20  # Biến cục bộ của outer
    print(f"outer - x cục bộ = {x}")
    
    def inner():
        x = 30  # Biến cục bộ của inner
        print(f"inner - x cục bộ = {x}")
    
    inner()
    print(f"outer sau inner - x vẫn = {x}")

outer()
print(f"toàn cục - x = {x}")
````

## Câu 3 - Đáp án: B

### Đề bài
```python
def outer():
    x = 10
    def inner():
        nonlocal x
        x = x + 5
        return x
    return inner()

print(outer())
```

### Đáp án đúng: B. 15

### Giải thích chi tiết

**Từ khóa `nonlocal`:**
- Cho phép hàm bên trong truy cập và **SỬA ĐỔI** biến của hàm bao ngoài
- Khác với `global` (truy cập biến toàn cục)

**Phân tích code:**

1. `x = 10` trong outer() tạo biến cục bộ
2. `nonlocal x` trong inner() khai báo: "Tôi muốn dùng x của hàm outer"
3. `x = x + 5` → x = 10 + 5 = **15**
4. `return x` trả về 15
5. `outer()` trả về kết quả của `inner()` = **15**

### So sánh `global` và `nonlocal`

| Từ khóa | Dùng để | Phạm vi |
|---------|---------|--------|
| `global` | Truy cập biến toàn cục | Cấp module |
| `nonlocal` | Truy cập biến hàm bao ngoài | Hàm enclosing |

````python
# Minh họa
def outer():
    x = 10
    print(f"outer - x ban đầu = {x}")
    
    def inner():
        nonlocal x  # Truy cập x của outer
        x = x + 5
        print(f"inner - x sau khi +5 = {x}")
        return x
    
    result = inner()
    print(f"outer - x sau inner = {x}")  # x đã bị thay đổi!
    return result

print(f"Kết quả: {outer()}")
````

## Câu 4 - Đáp án: B

### Đề bài
```python
nums = [1, 2, 3]
result = nums * 2
nums += [4]
print(result, nums)
```

### Đáp án đúng: B. [1, 2, 3, 1, 2, 3] [1, 2, 3, 4]

### Giải thích chi tiết

**Toán tử `*` với list:**
- `list * n` tạo **list MỚI** với các phần tử được lặp lại n lần
- List gốc không bị ảnh hưởng

**Toán tử `+=` với list:**
- `list += [...]` sửa đổi list **TẠI CHỖ** (in-place)
- Không tạo list mới

**Phân tích từng bước:**

1. `nums = [1, 2, 3]` → nums trỏ đến list [1, 2, 3]
2. `result = nums * 2` → Tạo list MỚI [1, 2, 3, 1, 2, 3], gán cho result
   - nums vẫn là [1, 2, 3]
   - result và nums là **2 đối tượng khác nhau**
3. `nums += [4]` → Sửa nums tại chỗ thành [1, 2, 3, 4]
   - result **KHÔNG bị ảnh hưởng** (vì là đối tượng khác)

**Kết quả:**
- result = [1, 2, 3, 1, 2, 3]
- nums = [1, 2, 3, 4]

````python
# Minh họa
nums = [1, 2, 3]
print(f"nums ban đầu: {nums}, id = {id(nums)}")

result = nums * 2  # Tạo list MỚI
print(f"result = nums * 2: {result}, id = {id(result)}")
print(f"nums và result là cùng đối tượng? {nums is result}")

nums += [4]  # Sửa nums tại chỗ
print(f"\nSau nums += [4]:")
print(f"result (không đổi): {result}")
print(f"nums (đã thay đổi): {nums}")
````

## Câu 5 - Đáp án: B

### Đề bài
```python
a = [1, 2]
b = [3, 4]
c = a
a = a + b
print(c)
```

### Đáp án đúng: B. [1, 2]

### Giải thích chi tiết

**Điểm mấu chốt:** `a = a + b` tạo **đối tượng MỚI**

**Phân tích từng bước:**

```
Bước 1: a = [1, 2]
        a ──► [1, 2]

Bước 2: c = a
        a ──► [1, 2] ◄── c
        (c trỏ đến CÙNG list với a)

Bước 3: a = a + b
        a ──► [1, 2, 3, 4]  (list MỚI)
        c ──► [1, 2]        (list GỐC, không đổi)
```

**Giải thích:**
- `a + b` tính toán và tạo list mới [1, 2, 3, 4]
- `a = ...` gán list mới này cho biến a
- c vẫn trỏ đến list gốc [1, 2]

### So sánh với Câu 6
- Câu 5: `a = a + b` → tạo đối tượng mới, gán lại a
- Câu 6: `a += b` → sửa đổi tại chỗ, a và c vẫn cùng trỏ đến 1 list

````python
# Minh họa
a = [1, 2]
b = [3, 4]
c = a
print(f"Trước: a is c = {a is c}")  # True - cùng đối tượng
print(f"id(a) = {id(a)}, id(c) = {id(c)}")

a = a + b  # Tạo list MỚI, gán cho a
print(f"\nSau a = a + b:")
print(f"a is c = {a is c}")  # False - khác đối tượng
print(f"a = {a}")
print(f"c = {c} (vẫn là list gốc!)")
````

## Câu 6 - Đáp án: A

### Đề bài
```python
a = [1, 2]
b = [3, 4]
c = a
a += b
print(c)
```

### Đáp án đúng: A. [1, 2, 3, 4]

### Giải thích chi tiết

**Điểm mấu chốt:** `a += b` sửa đổi **TẠI CHỖ** (in-place)

**Phân tích từng bước:**

```
Bước 1: a = [1, 2]
        a ──► [1, 2]

Bước 2: c = a
        a ──► [1, 2] ◄── c
        (c trỏ đến CÙNG list với a)

Bước 3: a += b
        a ──► [1, 2, 3, 4] ◄── c
        (CÙNG list, bị sửa đổi tại chỗ)
```

**Giải thích:**
- `+=` gọi phương thức `__iadd__()` (in-place add)
- Với list, `+=` mở rộng list hiện có (tương tự `extend()`)
- a và c vẫn trỏ đến **CÙNG MỘT** list
- Cả hai đều thấy sự thay đổi

### Bảng so sánh quan trọng

| Cú pháp | Hành vi | a và c sau đó |
|---------|---------|---------------|
| `a = a + b` | Tạo list MỚI, gán lại a | Khác nhau |
| `a += b` | Sửa list tại chỗ | Giống nhau |

````python
# Minh họa
a = [1, 2]
b = [3, 4]
c = a
print(f"Trước: a is c = {a is c}")
print(f"id(a) = {id(a)}")

a += b  # Sửa tại chỗ
print(f"\nSau a += b:")
print(f"a is c = {a is c}")  # Vẫn True!
print(f"id(a) = {id(a)} (cùng id!)")
print(f"a = {a}")
print(f"c = {c} (cũng bị thay đổi!)")
````

## Câu 7 - Đáp án: A

### Đề bài
```python
result = [i * j for i in range(1, 4) for j in range(1, 3)]
print(result)
```

### Đáp án đúng: A. [1, 2, 2, 4, 3, 6]

### Giải thích chi tiết

**Quy tắc list comprehension lồng nhau:**
- Đọc từ **TRÁI sang PHẢI** = từ **NGOÀI vào TRONG**
- Vòng lặp bên trái là vòng ngoài
- Vòng lặp bên phải là vòng trong

**Tương đương với:**
```python
result = []
for i in range(1, 4):      # i = 1, 2, 3
    for j in range(1, 3):  # j = 1, 2
        result.append(i * j)
```

**Tính toán từng bước:**

| i | j | i * j |
|---|---|-------|
| 1 | 1 | 1 |
| 1 | 2 | 2 |
| 2 | 1 | 2 |
| 2 | 2 | 4 |
| 3 | 1 | 3 |
| 3 | 2 | 6 |

**Kết quả:** [1, 2, 2, 4, 3, 6]

````python
# Minh họa
result = [i * j for i in range(1, 4) for j in range(1, 3)]
print(f"Kết quả: {result}")

# Phân tích từng bước
print("\nPhân tích chi tiết:")
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j} → i*j = {i*j}")
````

## Câu 8 - Đáp án: B

### Đề bài
```python
x = {1: 'a', 2: 'b'}
y = {2: 'c', 3: 'd'}
z = {**x, **y}
print(z)
```

### Đáp án đúng: B. {1: 'a', 2: 'c', 3: 'd'}

### Giải thích chi tiết

**Toán tử `**` với dictionary:**
- Giải nén (unpack) dictionary
- `{**x, **y}` kết hợp x và y thành dict mới

**Quy tắc khi có key trùng:**
- Giá trị từ dict **SAU** sẽ ghi đè dict trước
- Thứ tự quan trọng!

**Phân tích:**
```
x = {1: 'a', 2: 'b'}
y = {2: 'c', 3: 'd'}

z = {**x, **y}
  = {1: 'a', 2: 'b', 2: 'c', 3: 'd'}
                     ↑ key 2 bị ghi đè
  = {1: 'a', 2: 'c', 3: 'd'}
```

**Kết quả:** `{1: 'a', 2: 'c', 3: 'd'}`

````python
# Minh họa
x = {1: 'a', 2: 'b'}
y = {2: 'c', 3: 'd'}

z = {**x, **y}
print(f"z = {z}")

# Thử đảo thứ tự
z2 = {**y, **x}
print(f"z2 = {{**y, **x}} = {z2}")
print("→ Giá trị của key 2 thay đổi tùy thứ tự!")
````

## Câu 9 - Đáp án: B

### Đề bài
```python
def foo():
    try:
        return 1
    finally:
        return 2

print(foo())
```

### Đáp án đúng: B. 2

### Giải thích chi tiết

**Quy tắc của `finally`:**
- Khối `finally` **LUÔN LUÔN** được thực thi
- Chạy ngay cả khi có `return`, `break`, exception
- Nếu `finally` có `return`, nó **GHI ĐÈ** return của try/except

**Phân tích:**
1. Thực thi `try` block
2. Gặp `return 1` → chuẩn bị trả về 1
3. **NHƯNG** `finally` luôn chạy trước khi hàm kết thúc
4. `finally` có `return 2` → **GHI ĐÈ** return 1
5. Hàm trả về **2**

### ⚠️ Cảnh báo quan trọng
- Đây là **thực hành TỆ** - không nên dùng return trong finally
- Gây nhầm lẫn, khó debug
- Có thể che giấu exception

````python
# Minh họa
def foo():
    try:
        print("try: chuẩn bị return 1")
        return 1
    finally:
        print("finally: GHI ĐÈ với return 2")
        return 2  # Không nên làm thế này!

print(f"Kết quả: {foo()}")
````

## Câu 10 - Đáp án: B

### Đề bài
```python
values = [0, 1, 2, 3]
result = [v for v in values if v] 
print(sum(result) * len(result))
```

### Đáp án đúng: B. 18

### Giải thích chi tiết

**Giá trị Truthy và Falsy trong Python:**
- **Falsy:** `0`, `0.0`, `''`, `None`, `[]`, `{}`, `False`
- **Truthy:** Tất cả các giá trị khác

**Phân tích `[v for v in values if v]`:**
- `if v` kiểm tra v có truthy không
- 0 là **falsy** → bị loại
- 1, 2, 3 là **truthy** → được giữ
- result = [1, 2, 3]

**Tính toán:**
- `sum(result)` = 1 + 2 + 3 = 6
- `len(result)` = 3
- 6 * 3 = **18**

````python
# Minh họa
values = [0, 1, 2, 3]
print(f"values = {values}")

# Kiểm tra truthy/falsy
for v in values:
    print(f"  {v} → bool = {bool(v)}")

result = [v for v in values if v]
print(f"\nresult = {result}")
print(f"sum = {sum(result)}, len = {len(result)}")
print(f"sum * len = {sum(result) * len(result)}")
````

## Câu 11 - Đáp án: A

### Đề bài
```python
s = "hello"
t = s.replace("l", "L", 1)
u = s.replace("l", "L")
print(t, u)
```

### Đáp án đúng: A. heLlo heLLo

### Giải thích chi tiết

**Cú pháp `str.replace(old, new, count)`:**
- `old`: chuỗi cần thay
- `new`: chuỗi thay thế
- `count` (tùy chọn): số lần thay thế tối đa

**Phân tích:**

1. `s.replace("l", "L", 1)`:
   - Chỉ thay **1 lần** (l đầu tiên)
   - "hello" → "he**L**lo"

2. `s.replace("l", "L")`:
   - Thay **TẤT CẢ** (không có count)
   - "hello" → "he**LL**o"

**Lưu ý:** Chuỗi là bất biến (immutable)
- `replace()` trả về chuỗi MỚI
- Chuỗi gốc `s` không bị thay đổi

````python
# Minh họa
s = "hello"
t = s.replace("l", "L", 1)  # Thay 1 lần
u = s.replace("l", "L")     # Thay tất cả

print(f"Chuỗi gốc s = '{s}'")
print(f"t = s.replace('l', 'L', 1) = '{t}'")
print(f"u = s.replace('l', 'L') = '{u}'")
````

## Câu 12 - Đáp án: B

### Đề bài
```python
a = (1, 2, [3, 4])
a[2].append(5)
print(a)
```

### Đáp án đúng: B. (1, 2, [3, 4, 5])

### Giải thích chi tiết

**Tuple và tính bất biến:**
- Tuple **BẤT BIẾN** = không thể thay đổi các phần tử MÀ NÓ THAM CHIẾU
- NHƯNG nếu phần tử là mutable (như list), ta có thể sửa **NỘI DUNG** của nó

**Phân tích:**
```
a = (1, 2, [3, 4])
     |  |    |
     |  |    └── a[2] là một list (mutable)
     |  └─────── a[1] = 2
     └────────── a[0] = 1
```

- `a[2]` trả về list [3, 4]
- `a[2].append(5)` thêm 5 vào list
- List bị thay đổi: [3, 4, 5]
- Tuple vẫn tham chiếu đến **CÙNG list** (chỉ nội dung list thay đổi)

**Lỗi sẽ xảy ra nếu:**
```python
a[2] = [3, 4, 5]  # TypeError: tuple không hỗ trợ gán!
```

````python
# Minh họa
a = (1, 2, [3, 4])
print(f"Trước: a = {a}")
print(f"id(a[2]) = {id(a[2])}")

a[2].append(5)  # Sửa nội dung list, KHÔNG phải tuple
print(f"\nSau: a = {a}")
print(f"id(a[2]) = {id(a[2])} (cùng id!)")

# Thử gán sẽ lỗi:
try:
    a[2] = [3, 4, 5]
except TypeError as e:
    print(f"\nLỗi khi gán: {e}")
````

## Câu 13 - Đáp án: A

### Đề bài
```python
nums = [10, 20, 30, 40]
nums[1:3] = [200]
print(nums)
```

### Đáp án đúng: A. [10, 200, 40]

### Giải thích chi tiết

**Slice assignment (gán slice):**
- Thay thế một **PHẠM VI** các phần tử bằng iterable mới
- Số phần tử thay thế có thể **KHÁC** số phần tử bị thay

**Phân tích:**
```
nums = [10, 20, 30, 40]
         ↑   ↑
       [1] [2]

nums[1:3] = [20, 30]  ← slice này chứa 2 phần tử
nums[1:3] = [200]     ← thay bằng 1 phần tử

Kết quả: [10, 200, 40]  ← list co lại còn 3 phần tử
```

**Các trường hợp khác:**
- Thay 2 bằng 1 → list co lại
- Thay 1 bằng 3 → list mở rộng
- Thay 2 bằng 0 → xóa phần tử

## Câu 14 - Đáp án: B

### Đề bài
```python
def gen():
    yield 1
    yield 2
    yield 3

g = gen()
print(sum(g), sum(g))
```

### Đáp án đúng: B. 6 0

### Giải thích chi tiết

**Generator là gì?**
- Hàm dùng `yield` thay vì `return`
- Tạo giá trị "lười" (lazy) - chỉ khi được yêu cầu
- **CHỈ CÓ THỂ LẶP MỘT LẦN**

**Phân tích:**

1. `g = gen()` → tạo generator object
2. `sum(g)` lần 1:
   - Lặp qua g: 1, 2, 3
   - Tổng = 6
   - Generator bây giờ **CẠN KIỆT** (exhausted)
3. `sum(g)` lần 2:
   - Generator đã cạn kiệt
   - Không còn gì để lặp
   - Tổng = **0**

**Muốn dùng lại:** Phải tạo generator mới `g = gen()`

````python
# Minh họa
def gen():
    print("  yield 1")
    yield 1
    print("  yield 2")
    yield 2
    print("  yield 3")
    yield 3

g = gen()
print("Lần 1:")
print(f"sum(g) = {sum(g)}")

print("\nLần 2:")
print(f"sum(g) = {sum(g)}")  # Generator đã cạn!

# So sánh với list
print("\n--- So sánh với list ---")
lst = [1, 2, 3]
print(f"sum(lst) = {sum(lst)}, sum(lst) lần 2 = {sum(lst)}")
````

| Đặc điểm | `sorted(iterable)` | `list.sort()` |
|----------|-------------------|---------------|
| Kiểu | Hàm built-in | Phương thức của list |
| Trả về | List mới | None |

## Câu 16 - Đáp án: B

### Đề bài
```python
x = "hello"
y = "hello"
z = "".join(['h', 'e', 'l', 'l', 'o'])
print(x is y, x is z, x == z)
```

### Đáp án đúng: B. True False True

### Giải thích chi tiết

**String Interning (Nội hóa chuỗi):**
- Python tự động cache các chuỗi ngắn giống nhau
- Hai biến có cùng giá trị string literal → trỏ đến cùng đối tượng
- Chuỗi tạo động (như `join()`) → tạo đối tượng mới

**Phân tích:**

1. `x = "hello"` và `y = "hello"`:
   - Cả hai là string literal
   - Python intern chúng → cùng đối tượng
   - `x is y` → **True**

2. `z = "".join([...])`:
   - Tạo chuỗi động tại runtime
   - Đối tượng mới, không được intern
   - `x is z` → **False**

3. `x == z`:
   - So sánh GIÁ TRỊ
   - "hello" == "hello" → **True**

**So sánh `is` vs `==`:**
- `is`: So sánh ĐỊNH DANH (cùng đối tượng?)
- `==`: So sánh GIÁ TRỊ (nội dung giống nhau?)

````python
# Minh họa
x = "hello"
y = "hello"
z = "".join(['h', 'e', 'l', 'l', 'o'])

print(f"x = '{x}', id = {id(x)}")
print(f"y = '{y}', id = {id(y)}")
print(f"z = '{z}', id = {id(z)}")

print(f"\nx is y: {x is y} (cùng đối tượng, được intern)")
print(f"x is z: {x is z} (khác đối tượng, z tạo động)")
print(f"x == z: {x == z} (cùng giá trị)")
````

data = {'a': 1, 'b': 2, 'c': 3}

**Dict comprehension hoán đổi key-value:**

```python
{v: k for k, v in data.items()}

- Lấy từng cặp (k, v) từ data
- Tạo dict mới với v làm key, k làm value

```
data = {'a': 1, 'b': 2, 'c': 3}

Hoán đổi:
  'a': 1 → 1: 'a'

- Value phải hashable (để làm key)

## Câu 18 - Đáp án: B

### Đề bài
```python
funcs = []
for i in range(3):
    funcs.append(lambda: i)

print([f() for f in funcs])
```

### Đáp án đúng: B. [2, 2, 2]

### Giải thích chi tiết

**Bẫy Late Binding trong Closure:**
- Lambda bắt giữ **BIẾN** `i`, không phải **GIÁ TRỊ** của i
- Khi lambda được GỌI (không phải khi định nghĩa), nó mới đọc giá trị i
- Lúc đó vòng lặp đã kết thúc, i = 2

**Phân tích:**
```
Vòng lặp:
  i=0: tạo lambda tham chiếu đến biến i
  i=1: tạo lambda tham chiếu đến biến i
  i=2: tạo lambda tham chiếu đến biến i

Sau vòng lặp: i = 2

Gọi các lambda:
  f() → đọc i → 2
  f() → đọc i → 2
  f() → đọc i → 2
```

### Cách sửa
```python
# Dùng default argument để capture giá trị
funcs = [lambda i=i: i for i in range(3)]
# Kết quả: [0, 1, 2]
```

````python
# Minh họa - Vấn đề
funcs = []
for i in range(3):
    funcs.append(lambda: i)  # Bắt BIẾN i

print(f"Sai: {[f() for f in funcs]}")

# Cách sửa - capture giá trị
funcs2 = []
for i in range(3):
    funcs2.append(lambda i=i: i)  # i=i capture GIÁ TRỊ

print(f"Đúng: {[f() for f in funcs2]}")
````

## Câu 19 - Đáp án: A

### Đề bài
```python
x = [1, 2, 3]
y = x
x = x + [4, 5]
y = y + [6, 7]
print(len(x), len(y))
```

### Đáp án đúng: A. 5 5

### Giải thích chi tiết

**Phân tích từng bước:**

```
1. x = [1, 2, 3]
   x ──► [1, 2, 3]

2. y = x
   x ──► [1, 2, 3] ◄── y
   (cùng trỏ đến 1 list)

3. x = x + [4, 5]  (tạo list MỚI)
   x ──► [1, 2, 3, 4, 5]  (mới)
   y ──► [1, 2, 3]        (gốc)

4. y = y + [6, 7]  (tạo list MỚI từ list gốc)
   x ──► [1, 2, 3, 4, 5]
   y ──► [1, 2, 3, 6, 7]  (mới)
```

**Kết quả:**
- len(x) = 5
- len(y) = 5

**Nhớ:** `= +` tạo đối tượng mới, `+=` sửa tại chỗ

````python
# Minh họa
x = [1, 2, 3]
y = x
print(f"Ban đầu: x is y = {x is y}")

x = x + [4, 5]  # Tạo list mới
print(f"Sau x = x + [4,5]: x is y = {x is y}")
print(f"  y vẫn = {y}")

y = y + [6, 7]  # Tạo list mới từ y gốc
print(f"\nSau y = y + [6,7]:")
print(f"  x = {x}, len = {len(x)}")
print(f"  y = {y}, len = {len(y)}")
````

## Câu 20 - Đáp án: C

### Đề bài
```python
class Counter:
    count = 0
    def __init__(self):
        Counter.count += 1

a = Counter()
b = Counter()
c = Counter()
print(a.count, b.count, Counter.count)
```

### Đáp án đúng: C. 3 3 3

### Giải thích chi tiết

**Class variable vs Instance variable:**

| Loại | Định nghĩa | Chia sẻ |
|------|------------|--------|
| Class variable | Ngoài `__init__`, trong class | Chung cho TẤT CẢ instance |
| Instance variable | Trong `__init__`, dùng `self.` | Riêng cho MỖI instance |

**Phân tích:**

1. `count = 0` là **class variable** - chia sẻ bởi tất cả
2. Mỗi `Counter()` gọi `__init__` → `Counter.count += 1`
3. Sau 3 lần tạo: count = 3
4. Truy cập qua instance (`a.count`) → đọc class variable
5. Tất cả đều thấy cùng giá trị: **3**

**Lưu ý:** Nếu gán `a.count = 5`, sẽ tạo **instance variable** mới cho a, không ảnh hưởng class variable

````python
# Minh họa
class Counter:
    count = 0  # Class variable
    
    def __init__(self):
        Counter.count += 1
        print(f"Tạo instance, count = {Counter.count}")

a = Counter()
b = Counter()
c = Counter()

print(f"\na.count = {a.count}")
print(f"b.count = {b.count}")
print(f"Counter.count = {Counter.count}")
print("→ Tất cả đều là 3 vì dùng chung class variable!")
````
