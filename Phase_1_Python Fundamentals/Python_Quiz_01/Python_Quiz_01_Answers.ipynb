## Question 1 - Answer: C

**Correct Answer:** C. Python uses indentation to define code blocks, and incorrect indentation causes `IndentationError`

**Explanation:**
- Python uses indentation (whitespace at the beginning of lines) to define blocks of code
- Unlike Java/C++ which use curly braces `{}`
- Indentation is NOT optional - it's a syntax requirement
- Mixing tabs and spaces can cause `IndentationError`
- Standard is 4 spaces per indentation level

**Giải thích chi tiết (Tiếng Việt):**
- Python sử dụng **thụt lề** (khoảng trắng đầu dòng) để xác định khối lệnh
- Khác với Java/C++ sử dụng dấu ngoặc nhọn `{}`
- Thụt lề là **bắt buộc** trong Python, không phải tùy chọn
- Trộn lẫn tab và space sẽ gây ra lỗi `IndentationError`
- Tiêu chuẩn: **4 dấu cách** cho mỗi cấp thụt lề
- Đây là điểm khác biệt quan trọng của Python so với các ngôn ngữ khác

````python
# Demonstration
x = 10
y = 3
result = x // y  # Floor division
print(result)  # Output: 3
````

## Question 2 - Answer: C

**Correct Answer:** C. 3

**Explanation:**
- `//` is the floor division operator in Python
- Floor division divides and rounds DOWN to the nearest integer
- 10 // 3 = 3.33... → rounded down = 3
- Regular division `/` would give 3.3333333333333335

**Giải thích chi tiết (Tiếng Việt):**
- `//` là toán tử **chia lấy phần nguyên** (floor division) trong Python
- Chia và làm tròn **xuống** đến số nguyên gần nhất
- 10 // 3 = 3.33... → làm tròn xuống = **3**
- Phép chia thông thường `/` sẽ cho kết quả 3.3333...
- **Lưu ý:** Floor division luôn làm tròn về phía âm vô cực, không phải về 0

````python
# Demonstration
x = 10
y = 3
print(f"x // y = {x // y}")  # 3 (floor division)
print(f"x / y = {x / y}")    # 3.333... (regular division)
````

## Question 9 - Answer: D

**Correct Answer:** D. 0.0

**Explanation:**
- **Falsy values in Python:** `False`, `0`, `0.0`, `""` (empty string), `[]`, `()`, `{}`, `None`
- `"False"` - Truthy (non-empty string)
- `[0]` - Truthy (non-empty list)

## Question 10 - Answer: B

**Correct Answer:** B. 9

**Explanation:**
- Loop through 1, 2, 3, 4, 5
- `continue` skips even numbers (2, 4)
- Only odd numbers are added: 1 + 3 + 5 = 9

**Giải thích chi tiết (Tiếng Việt):**
- Vòng lặp qua 1, 2, 3, 4, 5
- `continue` **bỏ qua** phần còn lại của vòng lặp hiện tại
- Khi `i % 2 == 0` (số chẵn): bỏ qua → không cộng 2, 4
- Chỉ số lẻ được cộng: 1 + 3 + 5 = **9**
- **Lưu ý:** `continue` không thoát vòng lặp, chỉ bỏ qua lần lặp hiện tại

````python
# Demonstration
total = 0
for i in range(1, 6):
    if i % 2 == 0:
        continue
    total += i
    print(f"Added {i}, total = {total}")
print(f"Final total: {total}")  # 9
````

## Question 18 - Answer: B

**Correct Answer:** B. To modify a variable defined outside the function from within the function

**Explanation:**
- `global` keyword allows a function to modify a variable in the global scope
- Without `global`, assigning to a variable inside a function creates a local variable

**Giải thích chi tiết (Tiếng Việt):**
- Từ khóa `global` cho phép hàm **sửa đổi biến** ở phạm vi toàn cục
- Nếu không có `global`, gán giá trị cho biến trong hàm sẽ tạo **biến cục bộ** mới
- **Ví dụ:**
  ```python
  counter = 0
  def increment():
      global counter  # Khai báo sử dụng biến toàn cục
      counter += 1    # Sửa đổi biến toàn cục
  ```
- **Lưu ý:** Nên hạn chế dùng `global`, thay vào đó dùng tham số và return

````python
# Demonstration
counter = 0

def increment():
    global counter
    counter += 1

increment()
print(f"counter = {counter}")  # 1
````

## Question 20 - Answer: C

**Correct Answer:** C. 5

**Explanation:**
- `extend()` adds EACH element from the iterable individually
- Original: ["apple", "banana", "cherry"] (3 elements)
- After extend: adds "date" and "elderberry" as separate elements
- Result: 3 + 2 = 5 elements

**Giải thích chi tiết (Tiếng Việt):**
- `extend()` thêm **TỪNG phần tử** từ iterable riêng lẻ
- Ban đầu: ["apple", "banana", "cherry"] (3 phần tử)
- Sau extend: thêm "date" và "elderberry" như các phần tử riêng
- Kết quả: 3 + 2 = **5 phần tử**
- **So sánh:**
  - `append([a, b])` → thêm 1 phần tử là list
  - `extend([a, b])` → thêm 2 phần tử riêng lẻ

````python
# Demonstration - Compare append vs extend
list1 = [1, 2, 3]
list1.append([4, 5])
print(f"After append: {list1}, len = {len(list1)}")  # 4

list2 = [1, 2, 3]
list2.extend([4, 5])
print(f"After extend: {list2}, len = {len(list2)}")  # 5
````

- Trả về các phần tử có trong **CẢ HAI** set
- set1 ∩ set2 = **{3, 4}**
- **Các toán tử set khác:**
  - `|` : hợp (union) - phần tử có trong ít nhất 1 set
  - `-` : hiệu (difference) - phần tử chỉ có trong set đầu
  - `^` : hiệu đối xứng - phần tử chỉ có trong 1 set

````python
# Demonstration
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
print(f"set1 & set2 (intersection): {set1 & set2}")  # {3, 4}
print(f"set1 | set2 (union): {set1 | set2}")          # {1, 2, 3, 4, 5, 6}
````

## Question 27 - Answer: B

**Correct Answer:** B. 0

**Explanation:**
- `dict.get(key, default)` returns the value for key if it exists
- If key doesn't exist, returns the default value (0 in this case)
- "score" is not in the dictionary, so returns 0

**Giải thích chi tiết (Tiếng Việt):**
- `dict.get(key, default)` trả về giá trị của key nếu tồn tại
- Nếu key **không tồn tại**, trả về giá trị mặc định (0 trong trường hợp này)
- "score" không có trong dictionary → trả về **0**
- **So sánh với `d[key]`:** Nếu key không tồn tại sẽ gây lỗi `KeyError`

## Question 28 - Answer: B

**Correct Answer:** B. 10

**Explanation:**
- `update()` merges another dictionary into the current one
- If a key already exists, its value is OVERWRITTEN
- "a" was 1, but after update it becomes 10

**Giải thích chi tiết (Tiếng Việt):**
- `update()` **gộp** dictionary khác vào dictionary hiện tại
- Nếu key đã tồn tại, giá trị của nó bị **GHI ĐÈ**
- "a" ban đầu là 1, sau update trở thành **10**
- **Lưu ý:**
  - Key mới ("c": 3) được thêm vào
  - Key cũ ("a") bị ghi đè giá trị mới

## Question 30 - Answer: B

**Correct Answer:** B. [1, 9, 25]

**Explanation:**
- List comprehension with condition
- `range(1, 6)` = 1, 2, 3, 4, 5
- `x % 2 != 0` filters odd numbers: 1, 3, 5
- `x**2` squares them: 1, 9, 25
- Result: [1, 9, 25]

**Giải thích chi tiết (Tiếng Việt):**
- **List comprehension** với điều kiện lọc
- `range(1, 6)` = 1, 2, 3, 4, 5
- `x % 2 != 0` lọc số lẻ: 1, 3, 5
- `x**2` bình phương: 1, 9, 25
- Kết quả: **[1, 9, 25]**
- **Cú pháp:** `[biểu_thức for biến in iterable if điều_kiện]`

````python
# Demonstration
squares = [x**2 for x in range(1, 6) if x % 2 != 0]
print(f"Odd number squares: {squares}")  # [1, 9, 25]
````
