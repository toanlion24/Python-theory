## 1. Giới thiệu về Câu lệnh Điều kiện

Câu lệnh điều kiện cho phép chương trình thực thi các đoạn code khác nhau dựa trên việc kiểm tra một điều kiện nào đó là đúng (True) hay sai (False).

Trong Python, câu lệnh điều kiện sử dụng từ khóa: `if`, `elif`, và `else`.

## 2. Câu lệnh IF đơn giản

Cú pháp cơ bản nhất của câu lệnh điều kiện là `if`.

**Cú pháp:**
```python
if điều_kiện:
    # Code thực thi nếu điều kiện là True
    câu_lệnh_1
    câu_lệnh_2
```

**Lưu ý quan trọng:**
- Sau dấu `:` phải có thụt lề (indentation)
- Python sử dụng thụt lề để xác định khối code, thường là 4 spaces hoặc 1 tab
- Nếu điều kiện là `False`, toàn bộ khối code trong `if` sẽ bị bỏ qua

````python
# Ví dụ 1: Câu lệnh if đơn giản
tuoi = 18

if tuoi >= 18:
    print("Bạn đã đủ tuổi trưởng thành!")
    print("Chúc mừng!")

print("Câu lệnh này luôn được thực thi")
````

````python
# Ví dụ 2: So sánh số
so = 10

if so > 5:
    print(f"{so} lớn hơn 5")

if so == 10:
    print(f"{so} bằng 10")

if so < 20:
    print(f"{so} nhỏ hơn 20")
````

## 3. Câu lệnh IF-ELSE

Khi muốn thực thi một đoạn code khác nếu điều kiện là `False`, ta sử dụng `else`.

**Cú pháp:**
```python
if điều_kiện:
    # Code thực thi nếu điều kiện là True
    câu_lệnh_1
else:
    # Code thực thi nếu điều kiện là False
    câu_lệnh_2
```

````python
# Ví dụ 3: if-else
tuoi = 16

if tuoi >= 18:
    print("Bạn đã đủ tuổi trưởng thành!")
else:
    print("Bạn chưa đủ tuổi trưởng thành!")

print("Chương trình kết thúc")
````

````python
# Ví dụ 4: Kiểm tra số chẵn/lẻ
so = 7

if so % 2 == 0:
    print(f"{so} là số chẵn")
else:
    print(f"{so} là số lẻ")
````

## 4. Câu lệnh IF-ELIF-ELSE

Khi có nhiều điều kiện cần kiểm tra, ta sử dụng `elif` (viết tắt của "else if").

**Cú pháp:**
```python
if điều_kiện_1:
    # Code thực thi nếu điều_kiện_1 là True
    câu_lệnh_1
elif điều_kiện_2:
    # Code thực thi nếu điều_kiện_1 là False và điều_kiện_2 là True
    câu_lệnh_2
elif điều_kiện_3:
    # Code thực thi nếu điều_kiện_1 và điều_kiện_2 đều False, điều_kiện_3 là True
    câu_lệnh_3
else:
    # Code thực thi nếu tất cả điều kiện trên đều False
    câu_lệnh_4
```

**Lưu ý:**
- Python sẽ kiểm tra các điều kiện theo thứ tự từ trên xuống
- Chỉ khối code đầu tiên có điều kiện `True` được thực thi
- Sau khi thực thi một khối, Python sẽ bỏ qua tất cả các `elif` và `else` còn lại

````python
# Ví dụ 5: Xếp loại điểm số
diem = 85

if diem >= 90:
    print("Xếp loại: Xuất sắc (A)")
elif diem >= 80:
    print("Xếp loại: Giỏi (B)")
elif diem >= 70:
    print("Xếp loại: Khá (C)")
elif diem >= 60:
    print("Xếp loại: Trung bình (D)")
else:
    print("Xếp loại: Yếu (F)")
````

````python
# Ví dụ 6: Xác định mùa trong năm
thang = 7

if thang in [12, 1, 2]:
    print("Mùa đông")
elif thang in [3, 4, 5]:
    print("Mùa xuân")
elif thang in [6, 7, 8]:
    print("Mùa hè")
elif thang in [9, 10, 11]:
    print("Mùa thu")
else:
    print("Tháng không hợp lệ!")
````

## 5. Các Toán tử So sánh

Python cung cấp các toán tử so sánh để tạo điều kiện:

| Toán tử | Ý nghĩa | Ví dụ | Kết quả |
|---------|---------|-------|---------|
| `==` | Bằng | `5 == 5` | `True` |
| `!=` | Khác | `5 != 3` | `True` |
| `<` | Nhỏ hơn | `3 < 5` | `True` |
| `>` | Lớn hơn | `5 > 3` | `True` |
| `<=` | Nhỏ hơn hoặc bằng | `5 <= 5` | `True` |
| `>=` | Lớn hơn hoặc bằng | `5 >= 3` | `True` |
| `is` | Kiểm tra cùng đối tượng | `a is b` | `True/False` |
| `is not` | Kiểm tra khác đối tượng | `a is not b` | `True/False` |
| `in` | Kiểm tra phần tử trong chuỗi/danh sách | `'a' in 'abc'` | `True` |
| `not in` | Kiểm tra phần tử không trong chuỗi/danh sách | `'d' not in 'abc'` | `True` |

````python
# Ví dụ 7: Các toán tử so sánh
a = 10
b = 20

print(f"a == b: {a == b}")  # False
print(f"a != b: {a != b}")  # True
print(f"a < b: {a < b}")    # True
print(f"a > b: {a > b}")    # False
print(f"a <= b: {a <= b}")  # True
print(f"a >= b: {a >= b}")  # False
````

````python
# Ví dụ 8: Toán tử in và not in
chuoi = "Python Programming"
danh_sach = [1, 2, 3, 4, 5]

print(f"'Python' in chuoi: {'Python' in chuoi}")  # True
print(f"'Java' in chuoi: {'Java' in chuoi}")      # False
print(f"3 in danh_sach: {3 in danh_sach}")       # True
print(f"10 in danh_sach: {10 in danh_sach}")      # False
print(f"'z' not in chuoi: {'z' not in chuoi}")    # True
````

## 6. Các Toán tử Logic

Để kết hợp nhiều điều kiện, ta sử dụng các toán tử logic:

| Toán tử | Ý nghĩa | Ví dụ | Kết quả |
|---------|---------|-------|---------|
| `and` | VÀ - cả hai điều kiện đều True | `True and False` | `False` |
| `or` | HOẶC - ít nhất một điều kiện True | `True or False` | `True` |
| `not` | PHỦ ĐỊNH - đảo ngược giá trị | `not True` | `False` |

**Bảng chân lý:**

| A | B | A and B | A or B | not A |
|---|---|---------|--------|-------|
| True | True | True | True | False |
| True | False | False | True | False |
| False | True | False | True | True |
| False | False | False | False | True |

````python
# Ví dụ 9: Toán tử logic
tuoi = 25
co_bang_lai = True

# Sử dụng and
if tuoi >= 18 and co_bang_lai:
    print("Bạn có thể lái xe!")
else:
    print("Bạn không thể lái xe!")

# Sử dụng or
diem_toan = 8.5
diem_ly = 7.0

if diem_toan >= 8.0 or diem_ly >= 8.0:
    print("Bạn có ít nhất một môn đạt điểm giỏi!")

# Sử dụng not
is_weekend = False

if not is_weekend:
    print("Hôm nay là ngày làm việc!")
````

````python
# Ví dụ 10: Kết hợp nhiều điều kiện
username = "admin"
password = "123456"
age = 20

# Kiểm tra đăng nhập với nhiều điều kiện
if username == "admin" and password == "123456" and age >= 18:
    print("Đăng nhập thành công!")
else:
    print("Đăng nhập thất bại!")

# Kiểm tra với điều kiện or
if username == "admin" or username == "user":
    print("Tên người dùng hợp lệ!")
````

## 7. Câu lệnh Điều kiện Lồng nhau (Nested Conditionals)

Ta có thể đặt câu lệnh điều kiện bên trong một câu lệnh điều kiện khác. Đây gọi là câu lệnh điều kiện lồng nhau.

**Cú pháp:**
```python
if điều_kiện_1:
    # Code thực thi nếu điều_kiện_1 là True
    if điều_kiện_2:
        # Code thực thi nếu cả điều_kiện_1 và điều_kiện_2 đều True
        câu_lệnh_1
    else:
        # Code thực thi nếu điều_kiện_1 là True nhưng điều_kiện_2 là False
        câu_lệnh_2
else:
    # Code thực thi nếu điều_kiện_1 là False
    câu_lệnh_3
```

````python
# Ví dụ 11: Câu lệnh điều kiện lồng nhau
so = 15

if so > 0:
    print(f"{so} là số dương")
    if so % 2 == 0:
        print(f"{so} là số chẵn")
    else:
        print(f"{so} là số lẻ")
else:
    print(f"{so} không phải số dương")
    if so == 0:
        print(f"{so} là số không")
    else:
        print(f"{so} là số âm")
````

````python
# Ví dụ 12: Kiểm tra điểm số chi tiết
diem = 85
co_thi_lai = False

if diem >= 60:
    print("Bạn đã đậu!")
    if diem >= 90:
        print("Xếp loại: Xuất sắc")
    elif diem >= 80:
        print("Xếp loại: Giỏi")
    elif diem >= 70:
        print("Xếp loại: Khá")
    else:
        print("Xếp loại: Trung bình")
else:
    print("Bạn chưa đậu!")
    if co_thi_lai:
        print("Bạn có thể thi lại!")
    else:
        print("Bạn không được thi lại!")
````

## 8. Toán tử Ternary (Toán tử 3 ngôi)

Toán tử ternary cho phép viết câu lệnh if-else ngắn gọn trong một dòng.

**Cú pháp:**
```python
bien = giá_trị_nếu_True if điều_kiện else giá_trị_nếu_False
```

**Tương đương với:**
```python
if điều_kiện:
    bien = giá_trị_nếu_True
else:
    bien = giá_trị_nếu_False
```

````python
# Ví dụ 13: Toán tử ternary
tuoi = 20

# Cách 1: Sử dụng if-else thông thường
if tuoi >= 18:
    trang_thai = "Trưởng thành"
else:
    trang_thai = "Chưa trưởng thành"

print(f"Cách 1: {trang_thai}")

# Cách 2: Sử dụng toán tử ternary (ngắn gọn hơn)
trang_thai = "Trưởng thành" if tuoi >= 18 else "Chưa trưởng thành"
print(f"Cách 2: {trang_thai}")
````

````python
# Ví dụ 14: Toán tử ternary với nhiều trường hợp
so = 7

# Tìm số lớn nhất giữa hai số
a, b = 10, 20
max_so = a if a > b else b
print(f"Số lớn nhất giữa {a} và {b} là: {max_so}")

# Kiểm tra số chẵn/lẻ
ket_qua = "Chẵn" if so % 2 == 0 else "Lẻ"
print(f"{so} là số {ket_qua}")

# Toán tử ternary lồng nhau
diem = 85
xep_loai = "Xuất sắc" if diem >= 90 else ("Giỏi" if diem >= 80 else ("Khá" if diem >= 70 else "Trung bình"))
print(f"Điểm {diem} - Xếp loại: {xep_loai}")
````

## 9. Giá trị Truthy và Falsy

Trong Python, mọi giá trị đều có thể được đánh giá là `True` hoặc `False` trong ngữ cảnh boolean.

**Falsy values** (giá trị được coi là False):
- `False`
- `None`
- `0` (số không)
- `0.0` (số thực không)
- `""` (chuỗi rỗng)
- `[]` (danh sách rỗng)
- `{}` (từ điển rỗng)
- `()` (tuple rỗng)
- `set()` (tập hợp rỗng)

**Truthy values** (giá trị được coi là True):
- Tất cả các giá trị khác không nằm trong danh sách Falsy

````python
# Ví dụ 15: Truthy và Falsy values
# Kiểm tra các giá trị Falsy
if False:
    print("False")
else:
    print("False là Falsy")

if None:
    print("None")
else:
    print("None là Falsy")

if 0:
    print("0")
else:
    print("0 là Falsy")

if "":
    print("Chuỗi rỗng")
else:
    print("Chuỗi rỗng là Falsy")

if []:
    print("Danh sách rỗng")
else:
    print("Danh sách rỗng là Falsy")

# Kiểm tra các giá trị Truthy
if 1:
    print("1 là Truthy")

if "Hello":
    print("Chuỗi không rỗng là Truthy")

if [1, 2, 3]:
    print("Danh sách không rỗng là Truthy")
````

````python
# Ví dụ 16: Ứng dụng Truthy/Falsy trong thực tế
ten = "Nguyễn Văn A"
danh_sach = [1, 2, 3]

# Kiểm tra biến có giá trị hay không
if ten:  # Tương đương với if ten != ""
    print(f"Tên: {ten}")

if danh_sach:  # Tương đương với if len(danh_sach) > 0
    print(f"Danh sách có {len(danh_sach)} phần tử")

# Kiểm tra danh sách rỗng
danh_sach_rong = []
if not danh_sach_rong:
    print("Danh sách rỗng!")
````

## 10. So sánh Chuỗi (String Comparison)

Python có thể so sánh chuỗi theo thứ tự từ điển (lexicographic order).

**Quy tắc so sánh:**
- So sánh từng ký tự từ trái sang phải
- Ký tự có mã ASCII nhỏ hơn sẽ nhỏ hơn
- Chữ hoa có mã ASCII nhỏ hơn chữ thường (A < a)
- Nếu ký tự đầu giống nhau, so sánh ký tự tiếp theo

````python
# Ví dụ 17: So sánh chuỗi
chuoi1 = "apple"
chuoi2 = "banana"

if chuoi1 < chuoi2:
    print(f"'{chuoi1}' đứng trước '{chuoi2}' trong từ điển")

if chuoi1 == "apple":
    print("Hai chuỗi bằng nhau")

# So sánh không phân biệt hoa thường
chuoi3 = "Apple"
chuoi4 = "apple"

if chuoi3.lower() == chuoi4.lower():
    print("Hai chuỗi giống nhau (không phân biệt hoa thường)")
````

## 11. Các Ví dụ Thực tế

### Ví dụ 1: Máy tính BMI (Body Mass Index)

````python
# Tính chỉ số BMI và đánh giá
can_nang = 70  # kg
chieu_cao = 1.75  # m

bmi = can_nang / (chieu_cao ** 2)
print(f"Chỉ số BMI: {bmi:.2f}")

if bmi < 18.5:
    danh_gia = "Gầy"
elif bmi < 25:
    danh_gia = "Bình thường"
elif bmi < 30:
    danh_gia = "Thừa cân"
else:
    danh_gia = "Béo phì"

print(f"Đánh giá: {danh_gia}")
````

### Ví dụ 2: Kiểm tra năm nhuận

````python
# Kiểm tra năm nhuận
# Năm nhuận là năm chia hết cho 4, trừ khi chia hết cho 100 nhưng không chia hết cho 400
nam = 2024

if (nam % 4 == 0 and nam % 100 != 0) or (nam % 400 == 0):
    print(f"{nam} là năm nhuận")
else:
    print(f"{nam} không phải năm nhuận")
````

### Ví dụ 3: Máy tính đơn giản

````python
# Máy tính đơn giản
so1 = 10
so2 = 5
phep_toan = "+"  # Có thể là: +, -, *, /, %

if phep_toan == "+":
    ket_qua = so1 + so2
elif phep_toan == "-":
    ket_qua = so1 - so2
elif phep_toan == "*":
    ket_qua = so1 * so2
elif phep_toan == "/":
    if so2 != 0:
        ket_qua = so1 / so2
    else:
        ket_qua = "Không thể chia cho 0!"
elif phep_toan == "%":
    if so2 != 0:
        ket_qua = so1 % so2
    else:
        ket_qua = "Không thể chia cho 0!"
else:
    ket_qua = "Phép toán không hợp lệ!"

print(f"{so1} {phep_toan} {so2} = {ket_qua}")
````

### Ví dụ 4: Kiểm tra mật khẩu

````python
# Kiểm tra độ mạnh của mật khẩu
mat_khau = "MyPass123!"

do_dai_ok = len(mat_khau) >= 8
co_chu_hoa = any(c.isupper() for c in mat_khau)
co_chu_thuong = any(c.islower() for c in mat_khau)
co_so = any(c.isdigit() for c in mat_khau)
co_ky_tu_dac_biet = any(c in "!@#$%^&*()_+-=[]{}|;:,.<>?" for c in mat_khau)

if do_dai_ok and co_chu_hoa and co_chu_thuong and co_so and co_ky_tu_dac_biet:
    print("Mật khẩu mạnh!")
elif do_dai_ok and co_chu_hoa and co_chu_thuong and co_so:
    print("Mật khẩu trung bình (thiếu ký tự đặc biệt)")
elif do_dai_ok:
    print("Mật khẩu yếu (thiếu nhiều yêu cầu)")
else:
    print("Mật khẩu quá ngắn (cần ít nhất 8 ký tự)")
````

## 12. Lưu ý và Best Practices

```python
if bien == True:
    ...
```

```python
    print("B")
```

## 13. Bài tập Thực hành

### Bài tập 1: Kiểm tra số nguyên tố
Viết chương trình kiểm tra một số có phải số nguyên tố hay không.

````python
# Giải bài tập 1: Kiểm tra số nguyên tố
so = 17

if so < 2:
    print(f"{so} không phải số nguyên tố")
elif so == 2:
    print(f"{so} là số nguyên tố")
else:
    la_nguyen_to = True
    for i in range(2, int(so ** 0.5) + 1):
        if so % i == 0:
            la_nguyen_to = False
            break
    
    if la_nguyen_to:
        print(f"{so} là số nguyên tố")
    else:
        print(f"{so} không phải số nguyên tố")
````

### Bài tập 2: Tìm số lớn nhất trong 3 số
Viết chương trình tìm số lớn nhất trong 3 số.

````python
# Giải bài tập 2: Tìm số lớn nhất trong 3 số
a, b, c = 15, 25, 10

if a >= b and a >= c:
    max_so = a
elif b >= a and b >= c:
    max_so = b
else:
    max_so = c

print(f"Số lớn nhất trong {a}, {b}, {c} là: {max_so}")
````

### Bài tập 3: Kiểm tra tam giác
Viết chương trình kiểm tra 3 cạnh có tạo thành tam giác và xác định loại tam giác.

````python
# Giải bài tập 3: Kiểm tra tam giác
a, b, c = 3, 4, 5

# Kiểm tra có phải tam giác không
if a + b > c and a + c > b and b + c > a:
    print("Ba cạnh tạo thành tam giác")
    
    # Xác định loại tam giác
    if a == b == c:
        print("Tam giác đều")
    elif a == b or a == c or b == c:
        print("Tam giác cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giác vuông")
    else:
        print("Tam giác thường")
else:
    print("Ba cạnh không tạo thành tam giác")
````
