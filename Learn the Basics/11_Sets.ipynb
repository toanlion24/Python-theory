# Sets (T·∫≠p h·ª£p) trong Python

## 1. Kh√°i ni·ªám v·ªÅ Sets

**Set** (t·∫≠p h·ª£p) l√† m·ªôt ki·ªÉu d·ªØ li·ªáu t·∫≠p h·ª£p (collection) trong Python, ƒë∆∞·ª£c s·ª≠ d·ª•ng ƒë·ªÉ l∆∞u tr·ªØ nhi·ªÅu ph·∫ßn t·ª≠ duy nh·∫•t kh√¥ng c√≥ th·ª© t·ª±. Sets c√≥ c√°c ƒë·∫∑c ƒëi·ªÉm:

- **Kh√¥ng c√≥ th·ª© t·ª± (Unordered)**: C√°c ph·∫ßn t·ª≠ kh√¥ng ƒë∆∞·ª£c s·∫Øp x·∫øp theo th·ª© t·ª± c·ª• th·ªÉ v√† th·ª© t·ª± c√≥ th·ªÉ thay ƒë·ªïi
- **C√≥ th·ªÉ thay ƒë·ªïi (Mutable)**: C√≥ th·ªÉ th√™m, x√≥a ph·∫ßn t·ª≠ sau khi t·∫°o
- **KH√îNG cho ph√©p tr√πng l·∫∑p**: M·ªói ph·∫ßn t·ª≠ ch·ªâ xu·∫•t hi·ªán m·ªôt l·∫ßn duy nh·∫•t
- **H·ªó tr·ª£ nhi·ªÅu ki·ªÉu d·ªØ li·ªáu**: C√≥ th·ªÉ ch·ª©a s·ªë, chu·ªói, tuple (immutable objects)
- **KH√îNG ƒë∆∞·ª£c ƒë√°nh ch·ªâ s·ªë (Unindexed)**: Kh√¥ng th·ªÉ truy c·∫≠p ph·∫ßn t·ª≠ b·∫±ng index nh∆∞ list hay tuple
- **Hi·ªáu su·∫•t cao**: T√¨m ki·∫øm v√† ki·ªÉm tra ph·∫ßn t·ª≠ r·∫•t nhanh (O(1))

### T·∫°i sao s·ª≠ d·ª•ng Sets?

- **Lo·∫°i b·ªè ph·∫ßn t·ª≠ tr√πng l·∫∑p**: T·ª± ƒë·ªông lo·∫°i b·ªè c√°c gi√° tr·ªã tr√πng l·∫∑p
- **Ki·ªÉm tra ph·∫ßn t·ª≠ nhanh**: T√¨m ki·∫øm ph·∫ßn t·ª≠ c√≥ trong set hay kh√¥ng r·∫•t nhanh
- **Th·ª±c hi·ªán c√°c ph√©p to√°n t·∫≠p h·ª£p**: Union, intersection, difference, v.v.
- **Hi·ªáu su·∫•t cao**: C√°c thao t√°c t√¨m ki·∫øm v√† ki·ªÉm tra nhanh h∆°n list/tuple

### So s√°nh Set vs List vs Tuple:

| ƒê·∫∑c ƒëi·ªÉm | Set | List | Tuple |
|----------|-----|------|-------|
| Khai b√°o | D√πng `{}` ho·∫∑c `set()` | D√πng `[]` ho·∫∑c `list()` | D√πng `()` ho·∫∑c `tuple()` |
| Th·ª© t·ª± | **Unordered** (kh√¥ng c√≥ th·ª© t·ª±) | **Ordered** (c√≥ th·ª© t·ª±) | **Ordered** (c√≥ th·ª© t·ª±) |
| Tr√πng l·∫∑p | ‚ùå **KH√îNG cho ph√©p** | ‚úÖ Cho ph√©p | ‚úÖ Cho ph√©p |
| Indexing | ‚ùå **KH√îNG h·ªó tr·ª£** | ‚úÖ H·ªó tr·ª£ | ‚úÖ H·ªó tr·ª£ |
| T√≠nh bi·∫øn ƒë·ªïi | **Mutable** (c√≥ th·ªÉ thay ƒë·ªïi) | **Mutable** (c√≥ th·ªÉ thay ƒë·ªïi) | **Immutable** (b·∫•t bi·∫øn) |
| T·ªëc ƒë·ªô t√¨m ki·∫øm | ‚ö° R·∫•t nhanh (O(1)) | üêå Ch·∫≠m (O(n)) | üêå Ch·∫≠m (O(n)) |

## 2. T·∫°o Sets

C√≥ nhi·ªÅu c√°ch ƒë·ªÉ t·∫°o m·ªôt set trong Python:

### 2.1. T·∫°o set r·ªóng

```python
# ‚ö†Ô∏è L∆ØU √ù QUAN TR·ªåNG: 
# Kh√¥ng th·ªÉ t·∫°o set r·ªóng b·∫±ng {} v√¨ {} t·∫°o dictionary r·ªóng!

# ‚ùå SAI: T·∫°o dictionary r·ªóng, kh√¥ng ph·∫£i set
dict_rong = {}
print(f"{{}} t·∫°o: {dict_rong}, ki·ªÉu: {type(dict_rong)}")

# ‚úÖ ƒê√öNG: S·ª≠ d·ª•ng h√†m set() ƒë·ªÉ t·∫°o set r·ªóng
set_rong = set()
print(f"set() t·∫°o: {set_rong}, ki·ªÉu: {type(set_rong)}")
print(f"Set r·ªóng c√≥ {len(set_rong)} ph·∫ßn t·ª≠")
```

### 2.2. T·∫°o set c√≥ ph·∫ßn t·ª≠

```python
# C√°ch 1: S·ª≠ d·ª•ng d·∫•u ngo·∫∑c nh·ªçn {}
fruits = {"t√°o", "chu·ªëi", "cam", "nho"}
print(f"Set fruits: {fruits}")
print(f"Ki·ªÉu d·ªØ li·ªáu: {type(fruits)}")

# Set t·ª± ƒë·ªông lo·∫°i b·ªè ph·∫ßn t·ª≠ tr√πng l·∫∑p
numbers = {1, 2, 3, 2, 4, 3, 5}
print(f"Set numbers (c√≥ tr√πng l·∫∑p trong khai b√°o): {numbers}")
# Ch·ªâ c√≤n: {1, 2, 3, 4, 5}

# C√°ch 2: S·ª≠ d·ª•ng h√†m set() v·ªõi iterable
set_tu_list = set([1, 2, 3, 4, 5, 5, 5])
print(f"Set t·ª´ list [1, 2, 3, 4, 5, 5, 5]: {set_tu_list}")

set_tu_chuoi = set("Python")
print(f"Set t·ª´ chu·ªói 'Python': {set_tu_chuoi}")
# {'P', 'y', 't', 'h', 'o', 'n'} - m·ªói k√Ω t·ª± l√† m·ªôt ph·∫ßn t·ª≠

set_tu_tuple = set((1, 2, 3, 2, 4))
print(f"Set t·ª´ tuple (1, 2, 3, 2, 4): {set_tu_tuple}")

# C√°ch 3: S·ª≠ d·ª•ng set comprehension (xem ph·∫ßn sau)
set_comprehension = {x for x in range(1, 6)}
print(f"Set t·ª´ comprehension: {set_comprehension}")
```

### 2.3. L∆∞u √Ω v·ªÅ ki·ªÉu d·ªØ li·ªáu trong Set

Set ch·ªâ c√≥ th·ªÉ ch·ª©a c√°c **immutable (b·∫•t bi·∫øn)** objects nh∆∞ s·ªë, chu·ªói, tuple. Kh√¥ng th·ªÉ ch·ª©a list, dictionary, hay set kh√°c.

```python
# ‚úÖ ƒê√öNG: Set ch·ª©a s·ªë
set_so = {1, 2, 3, 4, 5}
print(f"Set s·ªë: {set_so}")

# ‚úÖ ƒê√öNG: Set ch·ª©a chu·ªói
set_chuoi = {"t√°o", "chu·ªëi", "cam"}
print(f"Set chu·ªói: {set_chuoi}")

# ‚úÖ ƒê√öNG: Set ch·ª©a tuple (v√¨ tuple l√† immutable)
set_tuple = {(1, 2), (3, 4), (5, 6)}
print(f"Set tuple: {set_tuple}")

# ‚ùå SAI: Set kh√¥ng th·ªÉ ch·ª©a list (v√¨ list l√† mutable)
try:
    set_list = {[1, 2], [3, 4]}  # TypeError
except TypeError as e:
    print(f"L·ªói: {e}")
    print("Set kh√¥ng th·ªÉ ch·ª©a list!")

# ‚ùå SAI: Set kh√¥ng th·ªÉ ch·ª©a dictionary
try:
    set_dict = {{"a": 1}, {"b": 2}}  # TypeError
except TypeError as e:
    print(f"L·ªói: {e}")
    print("Set kh√¥ng th·ªÉ ch·ª©a dictionary!")

# ‚ùå SAI: Set kh√¥ng th·ªÉ ch·ª©a set kh√°c
try:
    set_in_set = {{1, 2}, {3, 4}}  # TypeError
except TypeError as e:
    print(f"L·ªói: {e}")
    print("Set kh√¥ng th·ªÉ ch·ª©a set kh√°c!")

# ‚úÖ NH∆ØNG: C√≥ th·ªÉ d√πng frozenset (immutable set) trong set
frozenset1 = frozenset([1, 2, 3])
frozenset2 = frozenset([4, 5, 6])
set_frozenset = {frozenset1, frozenset2}
print(f"Set ch·ª©a frozenset: {set_frozenset}")
```

## 4. Th√™m ph·∫ßn t·ª≠ v√†o Set

C√≥ nhi·ªÅu c√°ch ƒë·ªÉ th√™m ph·∫ßn t·ª≠ v√†o set:

### 4.1. add() - Th√™m m·ªôt ph·∫ßn t·ª≠

```python
fruits = {"t√°o", "chu·ªëi", "cam"}
print(f"Set ban ƒë·∫ßu: {fruits}")

# Th√™m m·ªôt ph·∫ßn t·ª≠ m·ªõi
fruits.add("nho")
print(f"Sau khi add('nho'): {fruits}")

# Th√™m ph·∫ßn t·ª≠ ƒë√£ t·ªìn t·∫°i (kh√¥ng c√≥ t√°c d·ª•ng, set kh√¥ng thay ƒë·ªïi)
fruits.add("t√°o")
print(f"Sau khi add('t√°o') l·∫°i: {fruits}")  # Kh√¥ng thay ƒë·ªïi

# Th√™m nhi·ªÅu ph·∫ßn t·ª≠ ri√™ng l·∫ª
fruits.add("xo√†i")
fruits.add("d∆∞a h·∫•u")
print(f"Sau khi th√™m nhi·ªÅu ph·∫ßn t·ª≠: {fruits}")
```

### 4.2. update() - Th√™m nhi·ªÅu ph·∫ßn t·ª≠ t·ª´ iterable

```python
fruits = {"t√°o", "chu·ªëi", "cam"}
print(f"Set ban ƒë·∫ßu: {fruits}")

# update() v·ªõi list
fruits.update(["nho", "xo√†i", "d∆∞a h·∫•u"])
print(f"Sau khi update(['nho', 'xo√†i', 'd∆∞a h·∫•u']): {fruits}")

# update() v·ªõi tuple
fruits.update(("m·∫≠n", "·ªïi"))
print(f"Sau khi update(('m·∫≠n', '·ªïi')): {fruits}")

# update() v·ªõi set kh√°c
more_fruits = {"s·∫ßu ri√™ng", "mƒÉng c·ª•t"}
fruits.update(more_fruits)
print(f"Sau khi update(set kh√°c): {fruits}")

# update() v·ªõi chu·ªói (m·ªói k√Ω t·ª± l√† m·ªôt ph·∫ßn t·ª≠)
fruits.update("ABC")
print(f"Sau khi update('ABC'): {fruits}")

# C√≥ th·ªÉ truy·ªÅn nhi·ªÅu iterable c√πng l√∫c
fruits = {"t√°o", "chu·ªëi"}
fruits.update([1, 2, 3], (4, 5), {6, 7})
print(f"Set h·ªón h·ª£p: {fruits}")
```

## 5. X√≥a ph·∫ßn t·ª≠ kh·ªèi Set

C√≥ nhi·ªÅu c√°ch ƒë·ªÉ x√≥a ph·∫ßn t·ª≠ kh·ªèi set:

### 5.1. remove() - X√≥a ph·∫ßn t·ª≠ (n·∫øu kh√¥ng t·ªìn t·∫°i s·∫Ω g√¢y l·ªói)

```python
fruits = {"t√°o", "chu·ªëi", "cam", "nho", "xo√†i"}
print(f"Set ban ƒë·∫ßu: {fruits}")

# X√≥a ph·∫ßn t·ª≠ t·ªìn t·∫°i
fruits.remove("cam")
print(f"Sau khi remove('cam'): {fruits}")

# N·∫øu ph·∫ßn t·ª≠ kh√¥ng t·ªìn t·∫°i, remove() s·∫Ω g√¢y l·ªói KeyError
try:
    fruits.remove("m·∫≠n")  # KeyError
except KeyError as e:
    print(f"L·ªói: Ph·∫ßn t·ª≠ 'm·∫≠n' kh√¥ng c√≥ trong set")
    print(f"Set hi·ªán t·∫°i: {fruits}")
```

### 5.3. pop() - X√≥a v√† tr·∫£ v·ªÅ ph·∫ßn t·ª≠ ng·∫´u nhi√™n

```python
fruits = {"t√°o", "chu·ªëi", "cam", "nho", "xo√†i"}
print(f"Set ban ƒë·∫ßu: {fruits}")

# pop() x√≥a v√† tr·∫£ v·ªÅ m·ªôt ph·∫ßn t·ª≠ ng·∫´u nhi√™n (v√¨ set kh√¥ng c√≥ th·ª© t·ª±)
removed = fruits.pop()
print(f"pop() tr·∫£ v·ªÅ: {removed}")
print(f"Set sau pop(): {fruits}")

# pop() v·ªõi set r·ªóng s·∫Ω g√¢y l·ªói
empty_set = set()
try:
    empty_set.pop()  # KeyError
except KeyError:
    print("L·ªói: Kh√¥ng th·ªÉ pop() t·ª´ set r·ªóng")

# pop() r·∫•t h·ªØu √≠ch khi c·∫ßn l·∫•y v√† x√≥a m·ªôt ph·∫ßn t·ª≠ b·∫•t k·ª≥
```

### 5.4. clear() - X√≥a to√†n b·ªô ph·∫ßn t·ª≠

```python
fruits = {"t√°o", "chu·ªëi", "cam", "nho", "xo√†i"}
print(f"Set ban ƒë·∫ßu: {fruits}")

# clear() x√≥a t·∫•t c·∫£ ph·∫ßn t·ª≠, set tr·ªü th√†nh r·ªóng
fruits.clear()
print(f"Sau khi clear(): {fruits}")
print(f"Set v·∫´n t·ªìn t·∫°i nh∆∞ng r·ªóng: {len(fruits)} ph·∫ßn t·ª≠")
print(f"Ki·ªÉu d·ªØ li·ªáu: {type(fruits)}")
```

## 6. C√°c ph√©p to√°n t·∫≠p h·ª£p (Set Operations)

Sets h·ªó tr·ª£ c√°c ph√©p to√°n t·∫≠p h·ª£p to√°n h·ªçc: Union, Intersection, Difference, Symmetric Difference.

### 6.1. Union (H·ª£p) - T·∫•t c·∫£ ph·∫ßn t·ª≠ t·ª´ c·∫£ hai sets

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# C√°ch 1: S·ª≠ d·ª•ng to√°n t·ª≠ |
union1 = set1 | set2
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set1 | set2: {union1}")
# K·∫øt qu·∫£: {1, 2, 3, 4, 5, 6, 7, 8}

# C√°ch 2: S·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c union()
union2 = set1.union(set2)
print(f"set1.union(set2): {union2}")

# union() v·ªõi nhi·ªÅu sets
set3 = {9, 10}
union3 = set1.union(set2, set3)
print(f"union(set1, set2, set3): {union3}")

# L∆∞u √Ω: union() kh√¥ng thay ƒë·ªïi set g·ªëc, t·∫°o set m·ªõi
print(f"set1 sau union: {set1}")  # Kh√¥ng thay ƒë·ªïi
```

### 6.2. Intersection (Giao) - Ph·∫ßn t·ª≠ chung c·ªßa c·∫£ hai sets

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# C√°ch 1: S·ª≠ d·ª•ng to√°n t·ª≠ &
intersection1 = set1 & set2
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set1 & set2: {intersection1}")
# K·∫øt qu·∫£: {4, 5}

# C√°ch 2: S·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c intersection()
intersection2 = set1.intersection(set2)
print(f"set1.intersection(set2): {intersection2}")

# intersection() v·ªõi nhi·ªÅu sets
set3 = {5, 6, 7}
intersection3 = set1.intersection(set2, set3)
print(f"intersection(set1, set2, set3): {intersection3}")
# K·∫øt qu·∫£: {5} (ph·∫ßn t·ª≠ c√≥ trong c·∫£ 3 sets)

# N·∫øu kh√¥ng c√≥ ph·∫ßn t·ª≠ chung, tr·∫£ v·ªÅ set r·ªóng
set4 = {10, 11, 12}
intersection4 = set1.intersection(set4)
print(f"intersection(set1, set4): {intersection4}")  # set()
```

### 6.3. Difference (Hi·ªáu) - Ph·∫ßn t·ª≠ c√≥ trong set1 nh∆∞ng kh√¥ng c√≥ trong set2

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# C√°ch 1: S·ª≠ d·ª•ng to√°n t·ª≠ -
difference1 = set1 - set2
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set1 - set2: {difference1}")
# K·∫øt qu·∫£: {1, 2, 3} (ph·∫ßn t·ª≠ c√≥ trong set1 nh∆∞ng kh√¥ng c√≥ trong set2)

# C√°ch 2: S·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c difference()
difference2 = set1.difference(set2)
print(f"set1.difference(set2): {difference2}")

# L∆∞u √Ω: Ph√©p tr·ª´ kh√¥ng c√≥ t√≠nh giao ho√°n
difference3 = set2 - set1
print(f"set2 - set1: {difference3}")
# K·∫øt qu·∫£: {6, 7, 8} (kh√°c v·ªõi set1 - set2)

# difference() v·ªõi nhi·ªÅu sets
set3 = {2, 9, 10}
difference4 = set1.difference(set2, set3)
print(f"difference(set1, set2, set3): {difference4}")
# Ph·∫ßn t·ª≠ c√≥ trong set1 nh∆∞ng kh√¥ng c√≥ trong set2 v√† set3: {1, 3}
```

### 6.4. Symmetric Difference (Hi·ªáu ƒë·ªëi x·ª©ng) - Ph·∫ßn t·ª≠ c√≥ trong m·ªôt trong hai sets, kh√¥ng c√≥ trong c·∫£ hai

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# C√°ch 1: S·ª≠ d·ª•ng to√°n t·ª≠ ^
symmetric_diff1 = set1 ^ set2
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set1 ^ set2: {symmetric_diff1}")
# K·∫øt qu·∫£: {1, 2, 3, 6, 7, 8}
# (ph·∫ßn t·ª≠ c√≥ trong set1 ho·∫∑c set2 nh∆∞ng kh√¥ng c√≥ trong c·∫£ hai)

# C√°ch 2: S·ª≠ d·ª•ng ph∆∞∆°ng th·ª©c symmetric_difference()
symmetric_diff2 = set1.symmetric_difference(set2)
print(f"set1.symmetric_difference(set2): {symmetric_diff2}")

# C√≥ th·ªÉ t√≠nh b·∫±ng: (set1 - set2) | (set2 - set1)
symmetric_diff3 = (set1 - set2) | (set2 - set1)
print(f"(set1 - set2) | (set2 - set1): {symmetric_diff3}")

# Ph√©p symmetric difference c√≥ t√≠nh giao ho√°n
symmetric_diff4 = set2 ^ set1
print(f"set2 ^ set1: {symmetric_diff4}")
# K·∫øt qu·∫£ gi·ªëng set1 ^ set2
```

### 6.5. T√≥m t·∫Øt c√°c ph√©p to√°n t·∫≠p h·ª£p

| Ph√©p to√°n | To√°n t·ª≠ | Ph∆∞∆°ng th·ª©c | K·∫øt qu·∫£ |
|-----------|---------|-------------|---------|
| Union (H·ª£p) | `\|` | `union()` | T·∫•t c·∫£ ph·∫ßn t·ª≠ t·ª´ c·∫£ hai sets |
| Intersection (Giao) | `&` | `intersection()` | Ph·∫ßn t·ª≠ chung c·ªßa c·∫£ hai sets |
| Difference (Hi·ªáu) | `-` | `difference()` | Ph·∫ßn t·ª≠ c√≥ trong set1, kh√¥ng c√≥ trong set2 |
| Symmetric Difference | `^` | `symmetric_difference()` | Ph·∫ßn t·ª≠ c√≥ trong m·ªôt trong hai sets, kh√¥ng c√≥ trong c·∫£ hai |

### 6.6. C√°c ph∆∞∆°ng th·ª©c c·∫≠p nh·∫≠t t·∫°i ch·ªó (In-place Updates)

C√°c ph∆∞∆°ng th·ª©c n√†y thay ƒë·ªïi set g·ªëc thay v√¨ t·∫°o set m·ªõi:

```python
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

# update() - T∆∞∆°ng ƒë∆∞∆°ng v·ªõi union() nh∆∞ng thay ƒë·ªïi set g·ªëc
set1_copy = set1.copy()
set1_copy.update(set2)
print(f"set1.update(set2): {set1_copy}")

# intersection_update() - Thay ƒë·ªïi set1 th√†nh intersection
set1_copy = set1.copy()
set1_copy.intersection_update(set2)
print(f"set1.intersection_update(set2): {set1_copy}")

# difference_update() - Thay ƒë·ªïi set1 th√†nh difference
set1_copy = set1.copy()
set1_copy.difference_update(set2)
print(f"set1.difference_update(set2): {set1_copy}")

# symmetric_difference_update() - Thay ƒë·ªïi set1 th√†nh symmetric difference
set1_copy = set1.copy()
set1_copy.symmetric_difference_update(set2)
print(f"set1.symmetric_difference_update(set2): {set1_copy}")

# L∆∞u √Ω: set1 g·ªëc kh√¥ng thay ƒë·ªïi v√¨ ta d√πng copy()
print(f"\nset1 g·ªëc: {set1}")  # Kh√¥ng thay ƒë·ªïi
```

### 7.4. isdisjoint() - Ki·ªÉm tra hai sets c√≥ r·ªùi nhau kh√¥ng

```python
set1 = {1, 2, 3}
set2 = {4, 5, 6}
set3 = {3, 4, 5}

# isdisjoint() tr·∫£ v·ªÅ True n·∫øu hai sets kh√¥ng c√≥ ph·∫ßn t·ª≠ chung
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set3: {set3}")

print(f"\nset1 v√† set2 c√≥ r·ªùi nhau? {set1.isdisjoint(set2)}")  # True
print(f"set1 v√† set3 c√≥ r·ªùi nhau? {set1.isdisjoint(set3)}")  # False (c√≥ ph·∫ßn t·ª≠ chung: 3)

# ·ª®ng d·ª•ng: Ki·ªÉm tra xem hai nh√≥m c√≥ th√†nh vi√™n chung kh√¥ng
hoc_sinh_lop_a = {"An", "B√¨nh", "Chi"}
hoc_sinh_lop_b = {"D≈©ng", "Em", "Ph√∫c"}
hoc_sinh_lop_c = {"Chi", "D≈©ng", "Giang"}

print(f"\nL·ªõp A v√† B kh√¥ng c√≥ h·ªçc sinh chung? {hoc_sinh_lop_a.isdisjoint(hoc_sinh_lop_b)}")
print(f"L·ªõp A v√† C kh√¥ng c√≥ h·ªçc sinh chung? {hoc_sinh_lop_a.isdisjoint(hoc_sinh_lop_c)}")
```

### 7.5. issubset() / issuperset() - Ki·ªÉm tra t·∫≠p con / t·∫≠p cha

```python
set1 = {1, 2, 3}
set2 = {1, 2, 3, 4, 5}
set3 = {2, 3}

# issubset() - Ki·ªÉm tra set1 c√≥ l√† t·∫≠p con c·ªßa set2 kh√¥ng
print(f"set1: {set1}")
print(f"set2: {set2}")
print(f"set3: {set3}")

print(f"\nset1 l√† t·∫≠p con c·ªßa set2? {set1.issubset(set2)}")  # True
print(f"set3 l√† t·∫≠p con c·ªßa set1? {set3.issubset(set1)}")  # True
print(f"set1 l√† t·∫≠p con c·ªßa set3? {set1.issubset(set3)}")  # False

# C√≥ th·ªÉ d√πng to√°n t·ª≠ <=
print(f"set1 <= set2: {set1 <= set2}")  # True
print(f"set1 < set2: {set1 < set2}")   # True (t·∫≠p con th·ª±c s·ª±)
print(f"set1 < set1: {set1 < set1}")   # False (kh√¥ng ph·∫£i t·∫≠p con th·ª±c s·ª±)

# issuperset() - Ki·ªÉm tra set1 c√≥ l√† t·∫≠p cha c·ªßa set2 kh√¥ng
print(f"\nset2 l√† t·∫≠p cha c·ªßa set1? {set2.issuperset(set1)}")  # True
print(f"set1 l√† t·∫≠p cha c·ªßa set2? {set1.issuperset(set2)}")  # False

# C√≥ th·ªÉ d√πng to√°n t·ª≠ >=
print(f"set2 >= set1: {set2 >= set1}")  # True
print(f"set2 > set1: {set2 > set1}")   # True (t·∫≠p cha th·ª±c s·ª±)

# ·ª®ng d·ª•ng: Ki·ªÉm tra quy·ªÅn truy c·∫≠p
admin_permissions = {"read", "write", "delete", "manage"}
user_permissions = {"read", "write"}

print(f"\nUser c√≥ ƒë·ªß quy·ªÅn admin? {user_permissions.issuperset(admin_permissions)}")  # False
print(f"User c√≥ quy·ªÅn con c·ªßa admin? {user_permissions.issubset(admin_permissions)}")  # True
```

## 8. Set Comprehension

Set Comprehension cho ph√©p t·∫°o set m·ªôt c√°ch ng·∫Øn g·ªçn, t∆∞∆°ng t·ª± nh∆∞ List Comprehension.

```python
# C√∫ ph√°p: {expression for item in iterable [if condition]}

# V√≠ d·ª• 1: T·∫°o set t·ª´ range
squares = {x**2 for x in range(1, 6)}
print(f"B√¨nh ph∆∞∆°ng t·ª´ 1 ƒë·∫øn 5: {squares}")
# K·∫øt qu·∫£: {1, 4, 9, 16, 25}

# V√≠ d·ª• 2: T·∫°o set v·ªõi ƒëi·ªÅu ki·ªán
even_squares = {x**2 for x in range(1, 11) if x % 2 == 0}
print(f"B√¨nh ph∆∞∆°ng s·ªë ch·∫µn t·ª´ 1 ƒë·∫øn 10: {even_squares}")
# K·∫øt qu·∫£: {16, 4, 36, 64, 100}

# V√≠ d·ª• 3: T·∫°o set t·ª´ chu·ªói (chuy·ªÉn th√†nh ch·ªØ hoa)
words = {"hello", "world", "python", "programming"}
upper_words = {word.upper() for word in words}
print(f"Ch·ªØ hoa: {upper_words}")

# V√≠ d·ª• 4: Lo·∫°i b·ªè tr√πng l·∫∑p v√† chuy·ªÉn ƒë·ªïi
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]
unique_even = {x for x in numbers if x % 2 == 0}
print(f"S·ªë ch·∫µn duy nh·∫•t: {unique_even}")

# V√≠ d·ª• 5: T·∫°o set t·ª´ nhi·ªÅu iterable
combined = {x for x in range(5)} | {x*2 for x in range(3)}
print(f"Set k·∫øt h·ª£p: {combined}")
```

## 9. ·ª®ng d·ª•ng th·ª±c t·∫ø c·ªßa Sets

Sets r·∫•t h·ªØu √≠ch trong nhi·ªÅu t√¨nh hu·ªëng th·ª±c t·∫ø:

### 9.1. Lo·∫°i b·ªè ph·∫ßn t·ª≠ tr√πng l·∫∑p t·ª´ list

```python
# C√°ch nhanh nh·∫•t ƒë·ªÉ lo·∫°i b·ªè tr√πng l·∫∑p
numbers = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]
unique_numbers = list(set(numbers))
print(f"List ban ƒë·∫ßu: {numbers}")
print(f"List sau khi lo·∫°i tr√πng: {unique_numbers}")

# L∆∞u √Ω: Set l√† unordered n√™n th·ª© t·ª± c√≥ th·ªÉ thay ƒë·ªïi
# N·∫øu c·∫ßn gi·ªØ th·ª© t·ª±, d√πng c√°ch kh√°c:
def remove_duplicates_keep_order(lst):
    seen = set()
    result = []
    for item in lst:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

ordered_unique = remove_duplicates_keep_order(numbers)
print(f"List lo·∫°i tr√πng (gi·ªØ th·ª© t·ª±): {ordered_unique}")
```

### 9.2. T√¨m ph·∫ßn t·ª≠ chung gi·ªØa hai danh s√°ch

```python
# T√¨m h·ªçc sinh h·ªçc c·∫£ hai m√¥n
hoc_sinh_toan = ["An", "B√¨nh", "Chi", "D≈©ng", "Em"]
hoc_sinh_ly = ["Chi", "D≈©ng", "Em", "Ph√∫c", "Giang"]

# Chuy·ªÉn sang set v√† t√¨m intersection
set_toan = set(hoc_sinh_toan)
set_ly = set(hoc_sinh_ly)
hoc_ca_hai_mon = set_toan & set_ly

print(f"H·ªçc sinh To√°n: {set_toan}")
print(f"H·ªçc sinh L√Ω: {set_ly}")
print(f"H·ªçc c·∫£ hai m√¥n: {hoc_ca_hai_mon}")

# T√¨m h·ªçc sinh ch·ªâ h·ªçc To√°n
chi_hoc_toan = set_toan - set_ly
print(f"Ch·ªâ h·ªçc To√°n: {chi_hoc_toan}")

# T√¨m h·ªçc sinh ch·ªâ h·ªçc L√Ω
chi_hoc_ly = set_ly - set_toan
print(f"Ch·ªâ h·ªçc L√Ω: {chi_hoc_ly}")
```

### 9.3. Ki·ªÉm tra t∆∞ c√°ch th√†nh vi√™n nhanh ch√≥ng

```python
# Danh s√°ch quy·ªÅn truy c·∫≠p
admin_users = {"admin", "root", "superuser"}
normal_users = {"user1", "user2", "user3"}

def check_access(username):
    if username in admin_users:
        return "Admin access"
    elif username in normal_users:
        return "Normal access"
    else:
        return "No access"

# Ki·ªÉm tra nhanh (O(1))
print(f"admin access: {check_access('admin')}")
print(f"user1 access: {check_access('user1')}")
print(f"guest access: {check_access('guest')}")

# So s√°nh v·ªõi list (ch·∫≠m h∆°n v·ªõi danh s√°ch l·ªõn)
admin_list = ["admin", "root", "superuser"]

# V·ªõi list, ph·∫£i duy·ªát qua t·ª´ng ph·∫ßn t·ª≠ (O(n))
def check_access_list(username):
    if username in admin_list:  # Ch·∫≠m h∆°n v·ªõi list l·ªõn
        return "Admin access"
    else:
        return "No access"
```

## 10. Frozenset - Immutable Set

Frozenset l√† phi√™n b·∫£n b·∫•t bi·∫øn (immutable) c·ªßa set. C√≥ th·ªÉ ƒë∆∞·ª£c d√πng nh∆∞ key trong dictionary ho·∫∑c ph·∫ßn t·ª≠ trong set kh√°c.

```python
# T·∫°o frozenset
frozen = frozenset([1, 2, 3, 4, 5])
print(f"Frozenset: {frozen}")
print(f"Ki·ªÉu: {type(frozen)}")

# Frozenset h·ªó tr·ª£ c√°c ph√©p to√°n t·∫≠p h·ª£p (union, intersection, etc.)
frozen1 = frozenset([1, 2, 3])
frozen2 = frozenset([3, 4, 5])
union = frozen1 | frozen2
print(f"Union: {union}")

# ‚ùå Nh∆∞ng KH√îNG th·ªÉ th√™m, x√≥a ph·∫ßn t·ª≠
try:
    frozen.add(6)  # AttributeError
except AttributeError:
    print("Frozenset kh√¥ng c√≥ ph∆∞∆°ng th·ª©c add()")

try:
    frozen.remove(1)  # AttributeError
except AttributeError:
    print("Frozenset kh√¥ng c√≥ ph∆∞∆°ng th·ª©c remove()")

# ‚úÖ C√≥ th·ªÉ d√πng frozenset l√†m key trong dictionary
fs1 = frozenset([1, 2, 3])
fs2 = frozenset([4, 5, 6])
my_dict = {fs1: "Set 1", fs2: "Set 2"}
print(f"Dictionary v·ªõi frozenset keys: {my_dict}")

# ‚úÖ C√≥ th·ªÉ d√πng frozenset trong set kh√°c
set_of_frozensets = {fs1, fs2}
print(f"Set ch·ª©a frozensets: {set_of_frozensets}")
```

## 11. L∆∞u √Ω v√† Best Practices

### 11.1. Khi n√†o n√™n d√πng Set?

- ‚úÖ C·∫ßn lo·∫°i b·ªè ph·∫ßn t·ª≠ tr√πng l·∫∑p
- ‚úÖ C·∫ßn ki·ªÉm tra ph·∫ßn t·ª≠ c√≥ t·ªìn t·∫°i hay kh√¥ng (membership testing)
- ‚úÖ C·∫ßn th·ª±c hi·ªán c√°c ph√©p to√°n t·∫≠p h·ª£p (union, intersection, etc.)
- ‚úÖ L√†m vi·ªác v·ªõi d·ªØ li·ªáu l·ªõn v√† c·∫ßn hi·ªáu su·∫•t cao
- ‚ùå KH√îNG d√πng khi c·∫ßn gi·ªØ th·ª© t·ª± ph·∫ßn t·ª≠
- ‚ùå KH√îNG d√πng khi c·∫ßn truy c·∫≠p ph·∫ßn t·ª≠ b·∫±ng index
- ‚ùå KH√îNG d√πng khi c·∫ßn ch·ª©a mutable objects (list, dict, set)

### 11.2. So s√°nh hi·ªáu su·∫•t

| Thao t√°c | Set | List | Time Complexity |
