# Dictionaries (Tá»« Ä‘iá»ƒn) trong Python

## 1. KhÃ¡i niá»‡m vá» Dictionaries

**Dictionary** (tá»« Ä‘iá»ƒn) lÃ  má»™t kiá»ƒu dá»¯ liá»‡u táº­p há»£p (collection) trong Python, Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ lÆ°u trá»¯ dá»¯ liá»‡u dÆ°á»›i dáº¡ng cáº·p **key-value** (khÃ³a-giÃ¡ trá»‹). Dictionaries cÃ³ cÃ¡c Ä‘áº·c Ä‘iá»ƒm:

- **CÃ³ thá»© tá»± (Ordered)** - Tá»« Python 3.7+: CÃ¡c pháº§n tá»­ Ä‘Æ°á»£c sáº¯p xáº¿p theo thá»© tá»± chÃ¨n vÃ o vÃ  giá»¯ nguyÃªn thá»© tá»± Ä‘Ã³
- **CÃ³ thá»ƒ thay Ä‘á»•i (Mutable)**: CÃ³ thá»ƒ thÃªm, xÃ³a, sá»­a Ä‘á»•i pháº§n tá»­ sau khi táº¡o
- **KHÃ”NG cho phÃ©p trÃ¹ng láº·p key**: Má»—i key chá»‰ xuáº¥t hiá»‡n má»™t láº§n duy nháº¥t (nhÆ°ng value cÃ³ thá»ƒ trÃ¹ng)
- **Há»— trá»£ nhiá»u kiá»ƒu dá»¯ liá»‡u**: Key pháº£i lÃ  immutable (sá»‘, chuá»—i, tuple), value cÃ³ thá»ƒ lÃ  báº¥t ká»³ kiá»ƒu dá»¯ liá»‡u nÃ o
- **Truy cáº­p báº±ng key**: Sá»­ dá»¥ng key Ä‘á»ƒ truy cáº­p value, khÃ´ng dÃ¹ng index nhÆ° list
- **Hiá»‡u suáº¥t cao**: TÃ¬m kiáº¿m vÃ  truy cáº­p value báº±ng key ráº¥t nhanh (O(1))

### Táº¡i sao sá»­ dá»¥ng Dictionaries?

- **LÆ°u trá»¯ dá»¯ liá»‡u cÃ³ cáº¥u trÃºc**: ThÃ´ng tin há»c sinh, sáº£n pháº©m, cáº¥u hÃ¬nh, v.v.
- **Truy cáº­p nhanh**: TÃ¬m kiáº¿m vÃ  truy cáº­p value báº±ng key ráº¥t nhanh
- **Dá»… dÃ ng thao tÃ¡c**: ThÃªm, xÃ³a, sá»­a Ä‘á»•i pháº§n tá»­ dá»… dÃ ng
- **Linh hoáº¡t**: Value cÃ³ thá»ƒ lÃ  báº¥t ká»³ kiá»ƒu dá»¯ liá»‡u nÃ o (sá»‘, chuá»—i, list, dict, v.v.)
- **MÃ´ phá»ng cáº¥u trÃºc dá»¯ liá»‡u thá»±c táº¿**: NhÆ° JSON, cÆ¡ sá»Ÿ dá»¯ liá»‡u, API responses

### So sÃ¡nh Dictionary vs List vs Tuple vs Set:

| Äáº·c Ä‘iá»ƒm | Dictionary | List | Tuple | Set |
|----------|------------|------|-------|-----|
| Khai bÃ¡o | DÃ¹ng `{}` hoáº·c `dict()` | DÃ¹ng `[]` hoáº·c `list()` | DÃ¹ng `()` hoáº·c `tuple()` | DÃ¹ng `{}` hoáº·c `set()` |
| Cáº¥u trÃºc | **Key-Value pairs** | Chá»‰ cÃ³ value | Chá»‰ cÃ³ value | Chá»‰ cÃ³ value |
| Thá»© tá»± | âœ… **Ordered** (tá»« 3.7+) | âœ… Ordered | âœ… Ordered | âŒ Unordered |
| TrÃ¹ng láº·p key | âŒ **KHÃ”NG cho phÃ©p** | âœ… Cho phÃ©p value | âœ… Cho phÃ©p value | âŒ KhÃ´ng cho phÃ©p |
| Indexing | âŒ DÃ¹ng **key** thay vÃ¬ index | âœ… Há»— trá»£ index | âœ… Há»— trá»£ index | âŒ KhÃ´ng há»— trá»£ |
| TÃ­nh biáº¿n Ä‘á»•i | **Mutable** | **Mutable** | **Immutable** | **Mutable** |
| Tá»‘c Ä‘á»™ tÃ¬m kiáº¿m | âš¡ Ráº¥t nhanh (O(1)) | ğŸŒ Cháº­m (O(n)) | ğŸŒ Cháº­m (O(n)) | âš¡ Ráº¥t nhanh (O(1)) |

## 2. Táº¡o Dictionaries

CÃ³ nhiá»u cÃ¡ch Ä‘á»ƒ táº¡o má»™t dictionary trong Python:

### 2.1. Táº¡o dictionary rá»—ng

```python
# CÃ¡ch 1: Sá»­ dá»¥ng dáº¥u ngoáº·c nhá»n {}
dict_rong1 = {}
print(f"Dictionary rá»—ng 1: {dict_rong1}, kiá»ƒu: {type(dict_rong1)}")

# CÃ¡ch 2: Sá»­ dá»¥ng hÃ m dict()
dict_rong2 = dict()
print(f"Dictionary rá»—ng 2: {dict_rong2}, kiá»ƒu: {type(dict_rong2)}")

# LÆ°u Ã½: {} táº¡o dictionary rá»—ng, khÃ´ng pháº£i set rá»—ng!
# Set rá»—ng pháº£i dÃ¹ng set()
```

### 2.2. Táº¡o dictionary cÃ³ pháº§n tá»­

```python
# CÃ¡ch 1: Sá»­ dá»¥ng dáº¥u ngoáº·c nhá»n {} vá»›i key:value
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A"
}
print(f"Dictionary student: {student}")

# CÃ¡ch 2: Sá»­ dá»¥ng hÃ m dict() vá»›i keyword arguments
student2 = dict(name="Tráº§n Thá»‹ BÃ¬nh", age=19, grade="B")
print(f"Dictionary student2: {student2}")

# CÃ¡ch 3: Sá»­ dá»¥ng hÃ m dict() vá»›i list/tuple chá»©a cÃ¡c tuple (key, value)
student3 = dict([("name", "LÃª VÄƒn Chi"), ("age", 21), ("grade", "A")])
print(f"Dictionary student3: {student3}")

# Dictionary cÃ³ thá»ƒ chá»©a nhiá»u kiá»ƒu dá»¯ liá»‡u
mixed_dict = {
    "string": "Hello",
    "number": 42,
    "float": 3.14,
    "boolean": True,
    "list": [1, 2, 3],
    "nested_dict": {"key": "value"}
}
print(f"Dictionary há»—n há»£p: {mixed_dict}")
```

### 2.3. LÆ°u Ã½ vá» Key trong Dictionary

- **Key pháº£i lÃ  immutable (báº¥t biáº¿n)**: Sá»‘, chuá»—i, tuple
- **Key KHÃ”NG thá»ƒ lÃ **: List, dictionary, set (vÃ¬ chÃºng lÃ  mutable)
- **Key lÃ  duy nháº¥t**: Náº¿u trÃ¹ng key, giÃ¡ trá»‹ sau sáº½ ghi Ä‘Ã¨ giÃ¡ trá»‹ trÆ°á»›c
- **Value cÃ³ thá»ƒ lÃ  báº¥t ká»³ kiá»ƒu dá»¯ liá»‡u nÃ o**

```python
# âœ… Key há»£p lá»‡: sá»‘, chuá»—i, tuple
valid_keys = {
    "name": "Python",           # Chuá»—i
    1: "Má»™t",                   # Sá»‘ nguyÃªn
    3.14: "Pi",                 # Sá»‘ thá»±c
    (1, 2): "Tuple key",        # Tuple
    True: "Boolean key"         # Boolean (lÃ  immutable)
}
print(f"Dictionary vá»›i cÃ¡c key há»£p lá»‡: {valid_keys}")

# âŒ Key khÃ´ng há»£p lá»‡: list, dict, set (mutable)
try:
    invalid_dict = {
        [1, 2]: "List key"  # TypeError: unhashable type: 'list'
    }
except TypeError as e:
    print(f"Lá»—i khi dÃ¹ng list lÃ m key: {e}")

# Key trÃ¹ng láº·p: giÃ¡ trá»‹ sau ghi Ä‘Ã¨ giÃ¡ trá»‹ trÆ°á»›c
duplicate_keys = {
    "name": "GiÃ¡ trá»‹ Ä‘áº§u tiÃªn",
    "age": 20,
    "name": "GiÃ¡ trá»‹ sau (ghi Ä‘Ã¨)"  # Ghi Ä‘Ã¨ "name" Ä‘áº§u tiÃªn
}
print(f"Dictionary vá»›i key trÃ¹ng láº·p: {duplicate_keys}")
# Chá»‰ cÃ²n: {'name': 'GiÃ¡ trá»‹ sau (ghi Ä‘Ã¨)', 'age': 20}
```

## 3. Truy cáº­p pháº§n tá»­ trong Dictionary

Dictionaries sá»­ dá»¥ng **key** Ä‘á»ƒ truy cáº­p **value**, khÃ´ng dÃ¹ng index nhÆ° list hay tuple.

### 3.1. Truy cáº­p báº±ng key vá»›i dáº¥u ngoáº·c vuÃ´ng []

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "subjects": ["Math", "Physics", "Chemistry"]
}

# Truy cáº­p value báº±ng key
print(f"TÃªn: {student['name']}")
print(f"Tuá»•i: {student['age']}")
print(f"Äiá»ƒm: {student['grade']}")
print(f"MÃ´n há»c: {student['subjects']}")

# Truy cáº­p pháº§n tá»­ trong nested list
print(f"MÃ´n há»c Ä‘áº§u tiÃªn: {student['subjects'][0]}")

# âš ï¸ LÆ°u Ã½: Náº¿u key khÃ´ng tá»“n táº¡i, sáº½ gÃ¢y lá»—i KeyError
try:
    print(student['address'])  # KeyError: 'address'
except KeyError as e:
    print(f"Lá»—i: Key '{e}' khÃ´ng tá»“n táº¡i trong dictionary")
```

### 3.2. Truy cáº­p báº±ng get() - An toÃ n hÆ¡n

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A"
}

# get() tráº£ vá» value náº¿u key tá»“n táº¡i
print(f"TÃªn: {student.get('name')}")

# get() tráº£ vá» None náº¿u key khÃ´ng tá»“n táº¡i (khÃ´ng gÃ¢y lá»—i)
print(f"Äá»‹a chá»‰: {student.get('address')}")  # None

# get() vá»›i giÃ¡ trá»‹ máº·c Ä‘á»‹nh
print(f"Äá»‹a chá»‰ (máº·c Ä‘á»‹nh): {student.get('address', 'ChÆ°a cÃ³ Ä‘á»‹a chá»‰')}")
print(f"Email (máº·c Ä‘á»‹nh): {student.get('email', 'ChÆ°a cÃ³ email')}")

# So sÃ¡nh [] vs get()
# []: GÃ¢y lá»—i náº¿u key khÃ´ng tá»“n táº¡i
# get(): Tráº£ vá» None hoáº·c giÃ¡ trá»‹ máº·c Ä‘á»‹nh, khÃ´ng gÃ¢y lá»—i
```

### 3.3. Kiá»ƒm tra key cÃ³ tá»“n táº¡i

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A"
}

# Sá»­ dá»¥ng toÃ¡n tá»­ 'in'
print(f"'name' cÃ³ trong dict: {'name' in student}")
print(f"'address' cÃ³ trong dict: {'address' in student}")

# Sá»­ dá»¥ng toÃ¡n tá»­ 'not in'
print(f"'email' khÃ´ng cÃ³ trong dict: {'email' not in student}")

# Sá»­ dá»¥ng vá»›i if statement
if "age" in student:
    print(f"Tuá»•i cá»§a há»c sinh: {student['age']}")
else:
    print("KhÃ´ng cÃ³ thÃ´ng tin tuá»•i")

# Kiá»ƒm tra trÆ°á»›c khi truy cáº­p (best practice)
key = "phone"
if key in student:
    print(f"Sá»‘ Ä‘iá»‡n thoáº¡i: {student[key]}")
else:
    print(f"KhÃ´ng cÃ³ thÃ´ng tin {key}")
```

## 4. Thay Ä‘á»•i vÃ  thÃªm pháº§n tá»­ vÃ o Dictionary

### 4.1. Thay Ä‘á»•i giÃ¡ trá»‹ (Update)

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A"
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# Thay Ä‘á»•i giÃ¡ trá»‹ báº±ng cÃ¡ch gÃ¡n láº¡i
student["age"] = 21
print(f"Sau khi thay Ä‘á»•i age: {student}")

student["grade"] = "A+"
print(f"Sau khi thay Ä‘á»•i grade: {student}")

# Thay Ä‘á»•i nhiá»u giÃ¡ trá»‹ cÃ¹ng lÃºc
student["age"] = 22
student["grade"] = "B"
print(f"Sau khi thay Ä‘á»•i nhiá»u giÃ¡ trá»‹: {student}")
```

### 4.2. ThÃªm pháº§n tá»­ má»›i

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# ThÃªm pháº§n tá»­ má»›i báº±ng cÃ¡ch gÃ¡n key má»›i
student["grade"] = "A"
student["email"] = "an@example.com"
student["phone"] = "0123456789"
print(f"Sau khi thÃªm pháº§n tá»­: {student}")

# Náº¿u key Ä‘Ã£ tá»“n táº¡i, sáº½ ghi Ä‘Ã¨ giÃ¡ trá»‹ cÅ©
student["age"] = 21  # Ghi Ä‘Ã¨ age cÅ©
print(f"Sau khi ghi Ä‘Ã¨ age: {student}")
```

### 4.3. update() - Cáº­p nháº­t nhiá»u pháº§n tá»­ cÃ¹ng lÃºc

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# update() vá»›i dictionary khÃ¡c
new_info = {
    "grade": "A",
    "email": "an@example.com",
    "phone": "0123456789"
}
student.update(new_info)
print(f"Sau khi update(new_info): {student}")

# update() vá»›i keyword arguments
student.update(city="HÃ  Ná»™i", country="Viá»‡t Nam")
print(f"Sau khi update(city, country): {student}")

# update() vá»›i list/tuple chá»©a cÃ¡c tuple (key, value)
student.update([("hobby", "Reading"), ("sport", "Football")])
print(f"Sau khi update([tuples]): {student}")

# update() ghi Ä‘Ã¨ giÃ¡ trá»‹ cÅ© náº¿u key Ä‘Ã£ tá»“n táº¡i
student.update({"age": 21, "grade": "A+"})
print(f"Sau khi update (ghi Ä‘Ã¨): {student}")
```

## 5. XÃ³a pháº§n tá»­ khá»i Dictionary

CÃ³ nhiá»u cÃ¡ch Ä‘á»ƒ xÃ³a pháº§n tá»­ khá»i dictionary:

### 5.1. del - XÃ³a pháº§n tá»­ báº±ng key

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# XÃ³a pháº§n tá»­ báº±ng del
del student["email"]
print(f"Sau khi del student['email']: {student}")

# Náº¿u key khÃ´ng tá»“n táº¡i, sáº½ gÃ¢y lá»—i KeyError
try:
    del student["phone"]  # KeyError
except KeyError as e:
    print(f"Lá»—i: Key '{e}' khÃ´ng tá»“n táº¡i")
    print(f"Dictionary hiá»‡n táº¡i: {student}")
```

### 5.2. pop() - XÃ³a vÃ  tráº£ vá» giÃ¡ trá»‹

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# pop() xÃ³a vÃ  tráº£ vá» value
removed_value = student.pop("email")
print(f"pop('email') tráº£ vá»: {removed_value}")
print(f"Dictionary sau pop(): {student}")

# pop() vá»›i giÃ¡ trá»‹ máº·c Ä‘á»‹nh (khÃ´ng gÃ¢y lá»—i náº¿u key khÃ´ng tá»“n táº¡i)
removed_phone = student.pop("phone", "KhÃ´ng cÃ³ sá»‘ Ä‘iá»‡n thoáº¡i")
print(f"pop('phone', default) tráº£ vá»: {removed_phone}")
print(f"Dictionary sau pop('phone'): {student}")

# pop() khÃ´ng cÃ³ default vÃ  key khÃ´ng tá»“n táº¡i sáº½ gÃ¢y lá»—i
try:
    student.pop("address")  # KeyError
except KeyError:
    print("Lá»—i: Key 'address' khÃ´ng tá»“n táº¡i")
```

### 5.3. popitem() - XÃ³a vÃ  tráº£ vá» cáº·p (key, value) cuá»‘i cÃ¹ng

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# popitem() xÃ³a vÃ  tráº£ vá» cáº·p (key, value) cuá»‘i cÃ¹ng (tá»« Python 3.7+)
removed_item = student.popitem()
print(f"popitem() tráº£ vá»: {removed_item}")
print(f"Dictionary sau popitem(): {student}")

# popitem() vá»›i dictionary rá»—ng sáº½ gÃ¢y lá»—i
empty_dict = {}
try:
    empty_dict.popitem()  # KeyError
except KeyError:
    print("Lá»—i: KhÃ´ng thá»ƒ popitem() tá»« dictionary rá»—ng")
```

### 5.4. clear() - XÃ³a toÃ n bá»™ pháº§n tá»­

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A"
}
print(f"Dictionary ban Ä‘áº§u: {student}")

# clear() xÃ³a táº¥t cáº£ pháº§n tá»­, dictionary trá»Ÿ thÃ nh rá»—ng
student.clear()
print(f"Sau khi clear(): {student}")
print(f"Dictionary váº«n tá»“n táº¡i nhÆ°ng rá»—ng: {len(student)} pháº§n tá»­")
print(f"Kiá»ƒu dá»¯ liá»‡u: {type(student)}")
```

### 5.5. So sÃ¡nh cÃ¡c phÆ°Æ¡ng thá»©c xÃ³a

| PhÆ°Æ¡ng thá»©c | XÃ³a pháº§n tá»­ | GÃ¢y lá»—i náº¿u khÃ´ng tá»“n táº¡i? | Tráº£ vá» giÃ¡ trá»‹? |
|-------------|-------------|---------------------------|-----------------|
| `del dict[key]` | XÃ³a key | âœ… CÃ³ (KeyError) | âŒ KhÃ´ng |
| `pop(key)` | XÃ³a key | âœ… CÃ³ (náº¿u khÃ´ng cÃ³ default) | âœ… CÃ³ (value) |
| `pop(key, default)` | XÃ³a key | âŒ KhÃ´ng (tráº£ vá» default) | âœ… CÃ³ (value hoáº·c default) |
| `popitem()` | XÃ³a item cuá»‘i | âœ… CÃ³ (náº¿u dict rá»—ng) | âœ… CÃ³ ((key, value)) |
| `clear()` | XÃ³a táº¥t cáº£ | âŒ KhÃ´ng | âŒ KhÃ´ng |

## 6. CÃ¡c phÆ°Æ¡ng thá»©c quan trá»ng cá»§a Dictionary

### 6.1. keys() - Láº¥y táº¥t cáº£ cÃ¡c key

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}

# keys() tráº£ vá» dict_keys object (cÃ³ thá»ƒ convert sang list)
keys = student.keys()
print(f"keys(): {keys}")
print(f"Kiá»ƒu: {type(keys)}")

# Convert sang list
keys_list = list(student.keys())
print(f"List cÃ¡c keys: {keys_list}")

# Duyá»‡t qua keys
print("CÃ¡c keys trong dictionary:")
for key in student.keys():
    print(f"  - {key}")

# keys() lÃ  dynamic view - thay Ä‘á»•i khi dict thay Ä‘á»•i
keys_view = student.keys()
print(f"Keys ban Ä‘áº§u: {list(keys_view)}")
student["phone"] = "0123456789"
print(f"Keys sau khi thÃªm 'phone': {list(keys_view)}")
```

### 6.2. values() - Láº¥y táº¥t cáº£ cÃ¡c value

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}

# values() tráº£ vá» dict_values object
values = student.values()
print(f"values(): {values}")
print(f"Kiá»ƒu: {type(values)}")

# Convert sang list
values_list = list(student.values())
print(f"List cÃ¡c values: {values_list}")

# Duyá»‡t qua values
print("CÃ¡c values trong dictionary:")
for value in student.values():
    print(f"  - {value}")

# values() cÅ©ng lÃ  dynamic view
values_view = student.values()
print(f"Values ban Ä‘áº§u: {list(values_view)}")
student["age"] = 21
print(f"Values sau khi thay Ä‘á»•i age: {list(values_view)}")
```

### 6.3. items() - Láº¥y táº¥t cáº£ cÃ¡c cáº·p (key, value)

```python
student = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "grade": "A",
    "email": "an@example.com"
}

# items() tráº£ vá» dict_items object chá»©a cÃ¡c tuple (key, value)
items = student.items()
print(f"items(): {items}")
print(f"Kiá»ƒu: {type(items)}")

# Convert sang list
items_list = list(student.items())
print(f"List cÃ¡c items: {items_list}")

# Duyá»‡t qua items (cÃ¡ch phá»• biáº¿n nháº¥t)
print("CÃ¡c cáº·p key-value trong dictionary:")
for key, value in student.items():
    print(f"  {key}: {value}")

# Unpacking trong vÃ²ng láº·p
for item in student.items():
    key, value = item
    print(f"Key: {key}, Value: {value}")

# items() cÅ©ng lÃ  dynamic view
items_view = student.items()
print(f"Items ban Ä‘áº§u: {list(items_view)}")
student["phone"] = "0123456789"
print(f"Items sau khi thÃªm 'phone': {list(items_view)}")
```

### 6.4. copy() - Sao chÃ©p dictionary

```python
original = {
    "name": "Nguyá»…n VÄƒn An",
    "age": 20,
    "subjects": ["Math", "Physics"]
}

# copy() táº¡o shallow copy (báº£n sao nÃ´ng)
copied = original.copy()
print(f"Original: {original}")
print(f"Copied: {copied}")

# Thay Ä‘á»•i trong copied khÃ´ng áº£nh hÆ°á»Ÿng original (cho immutable values)
copied["age"] = 21
print(f"\nSau khi thay Ä‘á»•i copied['age']:")
print(f"Original: {original}")
print(f"Copied: {copied}")

# âš ï¸ LÆ°u Ã½: Vá»›i mutable values (list, dict), shallow copy váº«n tham chiáº¿u
copied["subjects"].append("Chemistry")
print(f"\nSau khi thÃªm vÃ o copied['subjects']:")
print(f"Original: {original}")  # âš ï¸ CÅ©ng bá»‹ thay Ä‘á»•i!
print(f"Copied: {copied}")

# Äá»ƒ táº¡o deep copy (báº£n sao sÃ¢u), dÃ¹ng copy.deepcopy()
import copy
original2 = {
    "name": "Nguyá»…n VÄƒn An",
    "subjects": ["Math", "Physics"]
}
deep_copied = copy.deepcopy(original2)
deep_copied["subjects"].append("Chemistry")
print(f"\nVá»›i deepcopy():")
print(f"Original: {original2}")
print(f"Deep copied: {deep_copied}")
```

## 8. Dictionary Comprehension

Dictionary Comprehension lÃ  cÃ¡ch ngáº¯n gá»n Ä‘á»ƒ táº¡o dictionary tá»« iterable hoáº·c Ä‘iá»u kiá»‡n.

### 8.1. Dictionary Comprehension cÆ¡ báº£n

```python
# CÃº phÃ¡p: {key: value for item in iterable}

# Táº¡o dictionary tá»« range
squares = {x: x**2 for x in range(1, 6)}
print(f"BÃ¬nh phÆ°Æ¡ng tá»« 1-5: {squares}")

# Táº¡o dictionary tá»« list
names = ["An", "BÃ¬nh", "Chi"]
name_lengths = {name: len(name) for name in names}
print(f"Äá»™ dÃ i tÃªn: {name_lengths}")

# Táº¡o dictionary tá»« chuá»—i
char_counts = {char: char.upper() for char in "python"}
print(f"KÃ½ tá»± vÃ  chá»¯ hoa: {char_counts}")
```

### 8.2. Dictionary Comprehension vá»›i Ä‘iá»u kiá»‡n

```python
# Chá»‰ láº¥y sá»‘ cháºµn
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}
print(f"BÃ¬nh phÆ°Æ¡ng sá»‘ cháºµn: {even_squares}")

# Lá»c theo giÃ¡ trá»‹
students = {
    "An": 85,
    "BÃ¬nh": 92,
    "Chi": 78,
    "DÅ©ng": 95
}
excellent = {name: score for name, score in students.items() if score >= 90}
print(f"Há»c sinh xuáº¥t sáº¯c (>=90): {excellent}")

# Äiá»u kiá»‡n phá»©c táº¡p
grades = {name: "A" if score >= 90 else "B" if score >= 80 else "C" 
          for name, score in students.items()}
print(f"Xáº¿p loáº¡i: {grades}")
```

### 8.3. Dictionary Comprehension lá»“ng nhau

```python
# Táº¡o báº£ng cá»­u chÆ°Æ¡ng
multiplication_table = {
    f"{i}x{j}": i * j 
    for i in range(1, 4) 
    for j in range(1, 4)
}
print("Báº£ng cá»­u chÆ°Æ¡ng 3x3:")
for key, value in sorted(multiplication_table.items()):
    print(f"  {key} = {value}")

# Dictionary lá»“ng nhau
nested_dict = {
    f"row_{i}": {f"col_{j}": i * j for j in range(1, 4)}
    for i in range(1, 4)
}
print(f"\nDictionary lá»“ng nhau: {nested_dict}")
```

```python
students = {
    "student1": {
        "name": "Nguyá»…n VÄƒn An",
        "age": 20,
        "grades": {"Math": 85, "Physics": 90}
    },
    "student2": {
        "name": "Tráº§n Thá»‹ BÃ¬nh",
        "age": 19,
        "grades": {"Math": 92, "Physics": 88}
    }
}

# Duyá»‡t qua cÃ¡c student
print("ThÃ´ng tin cÃ¡c há»c sinh:")
for student_id, info in students.items():
    print(f"\n{student_id}:")
    print(f"  TÃªn: {info['name']}")
    print(f"  Tuá»•i: {info['age']}")
    print(f"  Äiá»ƒm:")
    for subject, score in info['grades'].items():
        print(f"    {subject}: {score}")

# Duyá»‡t lá»“ng nhau
print("\nDuyá»‡t lá»“ng nhau:")
for student_id in students:
    for key, value in students[student_id].items():
        if isinstance(value, dict):
            print(f"  {key}:")
            for k, v in value.items():
                print(f"    {k}: {v}")
        else:
            print(f"  {key}: {value}")
```

## 11. So sÃ¡nh vÃ  há»£p nháº¥t Dictionary

### 11.1. So sÃ¡nh Dictionary

```python
dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"a": 1, "b": 2, "c": 3}
dict3 = {"a": 1, "b": 2, "c": 4}

# So sÃ¡nh báº±ng (==)
print(f"dict1 == dict2: {dict1 == dict2}")  # True
print(f"dict1 == dict3: {dict1 == dict3}")  # False

# So sÃ¡nh khÃ´ng báº±ng (!=)
print(f"dict1 != dict3: {dict1 != dict3}")  # True

# âš ï¸ LÆ°u Ã½: Tá»« Python 3.7+, thá»© tá»± quan trá»ng
dict4 = {"a": 1, "b": 2, "c": 3}
dict5 = {"c": 3, "a": 1, "b": 2}
print(f"dict4 == dict5 (khÃ¡c thá»© tá»±): {dict4 == dict5}")  # True (chá»‰ so sÃ¡nh key-value, khÃ´ng so thá»© tá»±)
```

### 11.2. Há»£p nháº¥t Dictionary (Merge)

```python
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"b": 5, "e": 6}

# CÃ¡ch 1: Sá»­ dá»¥ng update() (thay Ä‘á»•i dict1)
dict1_copy = dict1.copy()
dict1_copy.update(dict2)
print(f"update(): {dict1_copy}")

# CÃ¡ch 2: Sá»­ dá»¥ng ** unpacking (Python 3.5+)
merged = {**dict1, **dict2}
print(f"** unpacking: {merged}")

# Há»£p nháº¥t nhiá»u dict
merged_all = {**dict1, **dict2, **dict3}
print(f"Há»£p nháº¥t nhiá»u dict: {merged_all}")
# âš ï¸ LÆ°u Ã½: dict3 cÃ³ key 'b' ghi Ä‘Ã¨ 'b' tá»« dict1

# CÃ¡ch 3: Sá»­ dá»¥ng | operator (Python 3.9+)
merged_pipe = dict1 | dict2
print(f"| operator: {merged_pipe}")

# |= operator (Python 3.9+)
dict1_copy2 = dict1.copy()
dict1_copy2 |= dict2
print(f"|= operator: {dict1_copy2}")
```

## 12. á»¨ng dá»¥ng thá»±c táº¿

### 12.1. Quáº£n lÃ½ thÃ´ng tin há»c sinh

```python
# Há»‡ thá»‘ng quáº£n lÃ½ há»c sinh
students = {
    "HS001": {
        "name": "Nguyá»…n VÄƒn An",
        "age": 20,
        "class": "12A1",
        "scores": {"Math": 85, "Physics": 90, "Chemistry": 88}
    },
    "HS002": {
        "name": "Tráº§n Thá»‹ BÃ¬nh",
        "age": 19,
        "class": "12A1",
        "scores": {"Math": 92, "Physics": 88, "Chemistry": 95}
    },
    "HS003": {
        "name": "LÃª VÄƒn Chi",
        "age": 20,
        "class": "12A2",
        "scores": {"Math": 78, "Physics": 85, "Chemistry": 82}
    }
}

# TÃ­nh Ä‘iá»ƒm trung bÃ¬nh cho má»—i há»c sinh
print("Äiá»ƒm trung bÃ¬nh cá»§a cÃ¡c há»c sinh:")
for student_id, info in students.items():
    scores = info["scores"]
    average = sum(scores.values()) / len(scores)
    print(f"{info['name']} ({student_id}): {average:.2f}")

# TÃ¬m há»c sinh cÃ³ Ä‘iá»ƒm cao nháº¥t mÃ´n ToÃ¡n
math_scores = {sid: info["scores"]["Math"] for sid, info in students.items()}
top_math = max(math_scores.items(), key=lambda x: x[1])
print(f"\nHá»c sinh Ä‘iá»ƒm ToÃ¡n cao nháº¥t: {students[top_math[0]]['name']} ({top_math[1]} Ä‘iá»ƒm)")

# Thá»‘ng kÃª theo lá»›p
class_stats = {}
for sid, info in students.items():
    class_name = info["class"]
    if class_name not in class_stats:
        class_stats[class_name] = []
    avg = sum(info["scores"].values()) / len(info["scores"])
    class_stats[class_name].append(avg)

print("\nÄiá»ƒm trung bÃ¬nh theo lá»›p:")
for class_name, averages in class_stats.items():
    class_avg = sum(averages) / len(averages)
    print(f"{class_name}: {class_avg:.2f}")
```

### 12.2. Äáº¿m táº§n suáº¥t xuáº¥t hiá»‡n

```python
# Äáº¿m táº§n suáº¥t kÃ½ tá»± trong chuá»—i
text = "hello world"
char_count = {}
for char in text:
    if char != " ":  # Bá» qua khoáº£ng tráº¯ng
        char_count[char] = char_count.get(char, 0) + 1

print("Táº§n suáº¥t kÃ½ tá»±:")
for char, count in sorted(char_count.items()):
    print(f"  '{char}': {count}")

# Äáº¿m táº§n suáº¥t tá»« trong danh sÃ¡ch
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
word_count = {}
for word in words:
    word_count[word] = word_count.get(word, 0) + 1

print("\nTáº§n suáº¥t tá»«:")
for word, count in word_count.items():
    print(f"  '{word}': {count}")

# Sá»­ dá»¥ng Counter tá»« collections (cÃ¡ch nhanh hÆ¡n)
from collections import Counter
word_count2 = Counter(words)
print(f"\nSá»­ dá»¥ng Counter: {dict(word_count2)}")
```

### 12.3. NhÃ³m dá»¯ liá»‡u theo tiÃªu chÃ­

```python
# NhÃ³m há»c sinh theo lá»›p
students_list = [
    {"name": "An", "class": "12A1", "score": 85},
    {"name": "BÃ¬nh", "class": "12A1", "score": 92},
    {"name": "Chi", "class": "12A2", "score": 78},
    {"name": "DÅ©ng", "class": "12A2", "score": 95},
    {"name": "Em", "class": "12A1", "score": 88}
]

# NhÃ³m theo lá»›p
students_by_class = {}
for student in students_list:
    class_name = student["class"]
    if class_name not in students_by_class:
        students_by_class[class_name] = []
    students_by_class[class_name].append(student)

print("Há»c sinh theo lá»›p:")
for class_name, students in students_by_class.items():
    print(f"\n{class_name}:")
    for student in students:
        print(f"  - {student['name']}: {student['score']} Ä‘iá»ƒm")

# NhÃ³m theo Ä‘iá»ƒm (>=90: Xuáº¥t sáº¯c, >=80: Giá»i, cÃ²n láº¡i: KhÃ¡)
students_by_grade = {"Xuáº¥t sáº¯c": [], "Giá»i": [], "KhÃ¡": []}
for student in students_list:
    score = student["score"]
    if score >= 90:
        students_by_grade["Xuáº¥t sáº¯c"].append(student)
    elif score >= 80:
        students_by_grade["Giá»i"].append(student)
    else:
        students_by_grade["KhÃ¡"].append(student)

print("\nHá»c sinh theo xáº¿p loáº¡i:")
for grade, students in students_by_grade.items():
    if students:
        print(f"\n{grade}:")
        for student in students:
            print(f"  - {student['name']}: {student['score']} Ä‘iá»ƒm")
```

### 12.4. Táº¡o tá»« Ä‘iá»ƒn Ã¡nh xáº¡ (Mapping)

```python
# Tá»« Ä‘iá»ƒn chuyá»ƒn Ä‘á»•i sá»‘ thÃ nh chá»¯
number_to_word = {
    0: "khÃ´ng", 1: "má»™t", 2: "hai", 3: "ba", 4: "bá»‘n",
    5: "nÄƒm", 6: "sÃ¡u", 7: "báº£y", 8: "tÃ¡m", 9: "chÃ­n"
}

def number_to_vietnamese(num):
    """Chuyá»ƒn sá»‘ thÃ nh chá»¯ tiáº¿ng Viá»‡t (0-9)"""
    return number_to_word.get(num, "khÃ´ng há»£p lá»‡")

print("Chuyá»ƒn Ä‘á»•i sá»‘ thÃ nh chá»¯:")
for i in range(10):
    print(f"  {i}: {number_to_vietnamese(i)}")

# Tá»« Ä‘iá»ƒn mÃ£ lá»—i
error_codes = {
    404: "Not Found",
    500: "Internal Server Error",
    403: "Forbidden",
    200: "OK",
    201: "Created"
}

def get_error_message(code):
    """Láº¥y thÃ´ng bÃ¡o lá»—i tá»« mÃ£ lá»—i"""
    return error_codes.get(code, "Unknown Error")

print("\nThÃ´ng bÃ¡o lá»—i:")
print(f"  404: {get_error_message(404)}")
print(f"  500: {get_error_message(500)}")
print(f"  999: {get_error_message(999)}")
```

## 13. LÆ°u Ã½ quan trá»ng vÃ  Best Practices

### 13.1. Key pháº£i lÃ  Immutable

- Key chá»‰ cÃ³ thá»ƒ lÃ : sá»‘, chuá»—i, tuple, boolean
- Key KHÃ”NG thá»ƒ lÃ : list, dict, set (vÃ¬ chÃºng lÃ  mutable)

### 13.2. Tham chiáº¿u vs Sao chÃ©p

- GÃ¡n trá»±c tiáº¿p (`dict2 = dict1`) táº¡o tham chiáº¿u, khÃ´ng pháº£i báº£n sao
- DÃ¹ng `copy()` Ä‘á»ƒ táº¡o shallow copy
- DÃ¹ng `copy.deepcopy()` Ä‘á»ƒ táº¡o deep copy (vá»›i nested dict)

```python
# Tham chiáº¿u
dict1 = {"a": 1, "b": [1, 2, 3]}
dict2 = dict1  # Tham chiáº¿u, khÃ´ng pháº£i báº£n sao
dict2["a"] = 10
print(f"dict1 sau khi thay Ä‘á»•i dict2: {dict1}")  # dict1 cÅ©ng thay Ä‘á»•i!

# Shallow copy
dict3 = {"a": 1, "b": [1, 2, 3]}
dict4 = dict3.copy()  # Shallow copy
dict4["a"] = 10
print(f"dict3 sau khi thay Ä‘á»•i dict4['a']: {dict3}")  # dict3 khÃ´ng Ä‘á»•i
dict4["b"].append(4)
print(f"dict3 sau khi thay Ä‘á»•i dict4['b']: {dict3}")  # âš ï¸ dict3['b'] cÅ©ng thay Ä‘á»•i!

# Deep copy
import copy
dict5 = {"a": 1, "b": [1, 2, 3]}
dict6 = copy.deepcopy(dict5)
dict6["b"].append(4)
print(f"dict5 sau khi thay Ä‘á»•i dict6['b']: {dict5}")  # dict5 khÃ´ng Ä‘á»•i
```

### 13.3. Sá»­ dá»¥ng get() thay vÃ¬ [] khi khÃ´ng cháº¯c key tá»“n táº¡i

- `dict[key]` gÃ¢y lá»—i KeyError náº¿u key khÃ´ng tá»“n táº¡i
- `dict.get(key, default)` an toÃ n hÆ¡n, tráº£ vá» default náº¿u key khÃ´ng tá»“n táº¡i

### 13.4. Sá»­ dá»¥ng items() Ä‘á»ƒ duyá»‡t qua dictionary

- `for key, value in dict.items()` lÃ  cÃ¡ch tá»‘t nháº¥t vÃ  Pythonic nháº¥t
- Hiá»‡u quáº£ hÆ¡n `for key in dict: value = dict[key]`

### 13.5. Dictionary Comprehension vs VÃ²ng láº·p

- Dictionary Comprehension ngáº¯n gá»n vÃ  nhanh hÆ¡n cho cÃ¡c trÆ°á»ng há»£p Ä‘Æ¡n giáº£n
- VÃ²ng láº·p thÃ´ng thÆ°á»ng dá»… Ä‘á»c hÆ¡n cho logic phá»©c táº¡p

```python
# Dictionary Comprehension (ngáº¯n gá»n)
squares_comp = {x: x**2 for x in range(1, 6)}
print(f"Comprehension: {squares_comp}")

# VÃ²ng láº·p (dá»… Ä‘á»c cho logic phá»©c táº¡p)
squares_loop = {}
for x in range(1, 6):
    squares_loop[x] = x**2
print(f"VÃ²ng láº·p: {squares_loop}")

# Logic phá»©c táº¡p - nÃªn dÃ¹ng vÃ²ng láº·p
complex_dict = {}
for x in range(1, 11):
    if x % 2 == 0:
        complex_dict[x] = x**2
    else:
        complex_dict[x] = x**3
print(f"Logic phá»©c táº¡p: {complex_dict}")
```
