## 1.1. Kh√°i ni·ªám v√† ƒê·ªãnh nghƒ©a

**Array (M·∫£ng)** l√† m·ªôt c·∫•u tr√∫c d·ªØ li·ªáu tuy·∫øn t√≠nh ch·ª©a m·ªôt t·∫≠p h·ª£p c√°c ph·∫ßn t·ª≠ c√πng ki·ªÉu d·ªØ li·ªáu, ƒë∆∞·ª£c l∆∞u tr·ªØ t·∫°i c√°c v·ªã tr√≠ b·ªô nh·ªõ li√™n ti·∫øp nhau.

### ƒê·∫∑c ƒëi·ªÉm c∆° b·∫£n:

1. **B·ªô nh·ªõ li√™n ti·∫øp (Contiguous Memory)**
   - C√°c ph·∫ßn t·ª≠ ƒë∆∞·ª£c l∆∞u tr·ªØ t·∫°i c√°c ƒë·ªãa ch·ªâ b·ªô nh·ªõ k·∫ø ti·∫øp nhau
   - Cho ph√©p t√≠nh to√°n ƒë·ªãa ch·ªâ nhanh: `address[i] = base_address + i * element_size`

2. **Ki·ªÉu d·ªØ li·ªáu ƒë·ªìng nh·∫•t (Homogeneous)**
   - T·∫•t c·∫£ ph·∫ßn t·ª≠ ph·∫£i c√πng ki·ªÉu d·ªØ li·ªáu (trong c√°c ng√¥n ng·ªØ tƒ©nh nh∆∞ C, Java)
   - Python List linh ho·∫°t h∆°n, c√≥ th·ªÉ ch·ª©a nhi·ªÅu ki·ªÉu

3. **Truy c·∫≠p b·∫±ng ch·ªâ s·ªë (Index-based Access)**
   - M·ªói ph·∫ßn t·ª≠ c√≥ ch·ªâ s·ªë duy nh·∫•t b·∫Øt ƒë·∫ßu t·ª´ 0 (ho·∫∑c 1 trong m·ªôt s·ªë ng√¥n ng·ªØ)
   - Truy c·∫≠p trong th·ªùi gian h·∫±ng s·ªë O(1)

### K√≠ch th∆∞·ªõc m·∫£ng:

- **Fixed-size Array**: K√≠ch th∆∞·ªõc ƒë∆∞·ª£c x√°c ƒë·ªãnh khi khai b√°o, kh√¥ng thay ƒë·ªïi
- **Dynamic Array**: K√≠ch th∆∞·ªõc c√≥ th·ªÉ thay ƒë·ªïi trong qu√° tr√¨nh th·ª±c thi (Python List, C++ vector, Java ArrayList)

## 1.3. C√°c Thao T√°c C∆° B·∫£n v√† ƒê·ªô Ph·ª©c T·∫°p

### 1.3.1. Truy c·∫≠p (Access) - O(1)

Truy c·∫≠p ph·∫ßn t·ª≠ t·∫°i index `i` lu√¥n l√† O(1) v√¨:
- Bi·∫øt ƒë·ªãa ch·ªâ c∆° s·ªü (base address)
- Bi·∫øt k√≠ch th∆∞·ªõc m·ªói ph·∫ßn t·ª≠
- T√≠nh ƒë·ªãa ch·ªâ: `base + i √ó element_size`

```python

```

### 1.3.2. T√¨m ki·∫øm (Search)

#### **Linear Search - O(n)**
Duy·ªát qua t·ª´ng ph·∫ßn t·ª≠ cho ƒë·∫øn khi t√¨m th·∫•y:

```python
def linear_search(arr, target):
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1
```

#### **Binary Search - O(log n)** (v·ªõi m·∫£ng ƒë√£ s·∫Øp x·∫øp)

```python
def binary_search(arr, target):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

### 1.3.3. Ch√®n (Insertion) - O(n)

ƒê·ªÉ ch√®n ph·∫ßn t·ª≠ v√†o v·ªã tr√≠ `i`:
1. D·ªãch chuy·ªÉn t·∫•t c·∫£ ph·∫ßn t·ª≠ t·ª´ `i` ƒë·∫øn cu·ªëi sang ph·∫£i 1 v·ªã tr√≠
2. G√°n gi√° tr·ªã m·ªõi v√†o v·ªã tr√≠ `i`

**Tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát:**
- Ch√®n v√†o cu·ªëi: O(1) amortized (v·ªõi dynamic array)
- Ch√®n v√†o ƒë·∫ßu: O(n)

### 1.3.4. X√≥a (Deletion) - O(n)

ƒê·ªÉ x√≥a ph·∫ßn t·ª≠ t·∫°i v·ªã tr√≠ `i`:
1. D·ªãch chuy·ªÉn t·∫•t c·∫£ ph·∫ßn t·ª≠ t·ª´ `i+1` ƒë·∫øn cu·ªëi sang tr√°i 1 v·ªã tr√≠
2. Gi·∫£m size c·ªßa m·∫£ng

**Tr∆∞·ªùng h·ª£p ƒë·∫∑c bi·ªát:**
- X√≥a ph·∫ßn t·ª≠ cu·ªëi: O(1)
- X√≥a ph·∫ßn t·ª≠ ƒë·∫ßu: O(n)

## 1.4. Dynamic Array - C∆° ch·∫ø Resize

Dynamic Array (nh∆∞ Python List) t·ª± ƒë·ªông thay ƒë·ªïi k√≠ch th∆∞·ªõc khi c·∫ßn.

### C∆° ch·∫ø ho·∫°t ƒë·ªông:

1. **Kh·ªüi t·∫°o**: C·∫•p ph√°t b·ªô nh·ªõ v·ªõi capacity ban ƒë·∫ßu (v√≠ d·ª•: 4 ho·∫∑c 8 ph·∫ßn t·ª≠)

2. **Khi th√™m ph·∫ßn t·ª≠ (append)**:
   - N·∫øu c√≤n ch·ªó: O(1)
   - N·∫øu ƒë·∫ßy: 
     - C·∫•p ph√°t m·∫£ng m·ªõi v·ªõi size l·ªõn h∆°n (th∆∞·ªùng g·∫•p ƒë√¥i)
     - Copy t·∫•t c·∫£ ph·∫ßn t·ª≠ sang m·∫£ng m·ªõi
     - Gi·∫£i ph√≥ng m·∫£ng c≈©
     - ƒê·ªô ph·ª©c t·∫°p: O(n) nh∆∞ng √≠t x·∫£y ra

3. **Amortized Analysis**:
   - M·∫∑c d√π resize l√† O(n), nh∆∞ng x·∫£y ra √≠t (m·ªói khi capacity g·∫•p ƒë√¥i)
   - **Amortized cost** c·ªßa append l√† **O(1)**

### V√≠ d·ª• qu√° tr√¨nh resize:

```
Size: 0, Capacity: 4
[_, _, _, _]

Size: 4, Capacity: 4
[1, 2, 3, 4]

Append 5 ‚Üí Resize!
Size: 5, Capacity: 8
[1, 2, 3, 4, 5, _, _, _]

Size: 8, Capacity: 8
[1, 2, 3, 4, 5, 6, 7, 8]

Append 9 ‚Üí Resize!
Size: 9, Capacity: 16
[1, 2, 3, 4, 5, 6, 7, 8, 9, _, _, _, _, _, _, _]
```

### C√¥ng th·ª©c tƒÉng capacity:

- **Doubling strategy**: `new_capacity = old_capacity √ó 2` (ph·ªï bi·∫øn nh·∫•t)
- **Fixed increment**: `new_capacity = old_capacity + k` (√≠t hi·ªáu qu·∫£ h∆°n)

### Trade-offs c·ªßa Resize Strategy:

| Strategy | Insertion | Wasted Space | Reallocation Frequency |
|----------|-----------|--------------|------------------------|
| Doubling | O(1) amortized | ~50% | Th·∫•p |
| Fixed increment | O(n) | Th·∫•p | Cao |
| 1.5x multiplier | O(1) amortized | ~33% | Trung b√¨nh |

## 2.2. Tri·ªÉn khai Dynamic Array t·ª´ ƒë·∫ßu

````python
class DynamicArray:
    """
    Tri·ªÉn khai Dynamic Array t∆∞∆°ng t·ª± Python List
    T·ª± ƒë·ªông resize khi c·∫ßn thi·∫øt
    """
    
    def __init__(self, capacity: int = 4):
        """Kh·ªüi t·∫°o v·ªõi capacity ban ƒë·∫ßu"""
        self.capacity = capacity
        self.size = 0
        self.data = [None] * capacity
        self.resize_factor = 2  # TƒÉng g·∫•p ƒë√¥i khi resize
    
    def __len__(self) -> int:
        """Tr·∫£ v·ªÅ s·ªë ph·∫ßn t·ª≠ hi·ªán t·∫°i"""
        return self.size
    
    def __getitem__(self, index: int) -> Any:
        """Truy c·∫≠p ph·∫ßn t·ª≠ - O(1)"""
        if index < 0:
            index = self.size + index
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} out of range [0, {self.size-1}]")
        return self.data[index]
    
    def __setitem__(self, index: int, value: Any):
        """G√°n gi√° tr·ªã - O(1)"""
        if index < 0:
            index = self.size + index
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} out of range")
        self.data[index] = value
    
    def append(self, value: Any):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(1) amortized"""
        if self.size >= self.capacity:
            self._resize()
        self.data[self.size] = value
        self.size += 1
    
    def insert(self, index: int, value: Any):
        """Ch√®n ph·∫ßn t·ª≠ v√†o v·ªã tr√≠ index - O(n)"""
        if index < 0:
            index = self.size + index
        if index < 0 or index > self.size:
            raise IndexError(f"Index {index} out of range")
        
        if self.size >= self.capacity:
            self._resize()
        
        # D·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠ sang ph·∫£i
        for i in range(self.size, index, -1):
            self.data[i] = self.data[i - 1]
        
        self.data[index] = value
        self.size += 1
    
    def delete(self, index: int) -> Any:
        """X√≥a v√† tr·∫£ v·ªÅ ph·∫ßn t·ª≠ t·∫°i index - O(n)"""
        if index < 0:
            index = self.size + index
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} out of range")
        
        value = self.data[index]
        
        # D·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠ sang tr√°i
        for i in range(index, self.size - 1):
            self.data[i] = self.data[i + 1]
        
        self.size -= 1
        self.data[self.size] = None
        
        # Shrink n·∫øu c·∫ßn (optional optimization)
        if self.size > 0 and self.size < self.capacity // 4:
            self._shrink()
        
        return value
    
    def pop(self) -> Any:
        """X√≥a v√† tr·∫£ v·ªÅ ph·∫ßn t·ª≠ cu·ªëi - O(1)"""
        if self.size == 0:
            raise IndexError("Cannot pop from empty array")
        self.size -= 1
        value = self.data[self.size]
        self.data[self.size] = None
        return value
    
    def _resize(self):
        """TƒÉng capacity - O(n)"""
        old_capacity = self.capacity
        self.capacity = int(self.capacity * self.resize_factor)
        new_data = [None] * self.capacity
        
        for i in range(self.size):
            new_data[i] = self.data[i]
        
        self.data = new_data
        print(f"  üìà Resize: {old_capacity} ‚Üí {self.capacity}")
    
    def _shrink(self):
        """Gi·∫£m capacity ƒë·ªÉ ti·∫øt ki·ªám b·ªô nh·ªõ - O(n)"""
        old_capacity = self.capacity
        self.capacity = max(4, self.capacity // 2)
        new_data = [None] * self.capacity
        
        for i in range(self.size):
            new_data[i] = self.data[i]
        
        self.data = new_data
        print(f"  üìâ Shrink: {old_capacity} ‚Üí {self.capacity}")
    
    def index(self, value: Any) -> int:
        """T√¨m index c·ªßa gi√° tr·ªã ƒë·∫ßu ti√™n - O(n)"""
        for i in range(self.size):
            if self.data[i] == value:
                return i
        raise ValueError(f"{value} not in array")
    
    def __contains__(self, value: Any) -> bool:
        """Ki·ªÉm tra xem value c√≥ trong array - O(n)"""
        return value in self.data[:self.size]
    
    def __str__(self) -> str:
        """Hi·ªÉn th·ªã array"""
        elements = [str(self.data[i]) for i in range(self.size)]
        return f"DynamicArray([{', '.join(elements)}], size={self.size}, capacity={self.capacity})"
    
    def __repr__(self) -> str:
        return self.__str__()

# Demo DynamicArray
print("=== DEMO: Dynamic Array ===\n")
arr = DynamicArray(capacity=4)

# Test append
print("1. Th√™m ph·∫ßn t·ª≠ (append):")
for i in range(10):
    arr.append(i * 10)
    print(f"   Th√™m {i*10}: {arr}")

print("\n2. Ch√®n ph·∫ßn t·ª≠ (insert):")
arr.insert(3, 999)
print(f"   Ch√®n 999 v√†o index 3: {arr}")

print("\n3. X√≥a ph·∫ßn t·ª≠ (delete):")
removed = arr.delete(3)
print(f"   X√≥a ph·∫ßn t·ª≠ t·∫°i index 3 (gi√° tr·ªã {removed}): {arr}")

print("\n4. Pop ph·∫ßn t·ª≠ cu·ªëi:")
popped = arr.pop()
print(f"   Pop: {popped}, Array: {arr}")

print("\n5. T√¨m ki·∫øm:")
print(f"   Index c·ªßa 40: {arr.index(40)}")
print(f"   50 c√≥ trong array? {50 in arr}")
print(f"   999 c√≥ trong array? {999 in arr}")
````

## 2.3. So s√°nh Performance: Array vs List

H√£y so s√°nh hi·ªáu su·∫•t gi·ªØa DynamicArray t·ª± tri·ªÉn khai v√† Python List:

````python
# So s√°nh hi·ªáu su·∫•t
def benchmark_append(n: int):
    """So s√°nh th·ªùi gian append"""
    # Python List
    start = time.time()
    py_list = []
    for i in range(n):
        py_list.append(i)
    py_time = time.time() - start
    
    # DynamicArray
    start = time.time()
    dyn_arr = DynamicArray()
    for i in range(n):
        dyn_arr.append(i)
    dyn_time = time.time() - start
    
    return py_time, dyn_time

# Ch·∫°y benchmark
sizes = [1000, 10000, 100000]
py_times = []
dyn_times = []

print("=== BENCHMARK: Append Performance ===\n")
print(f"{'Size':<10} {'Python List':<15} {'DynamicArray':<15} {'Ratio':<10}")
print("-" * 55)

for n in sizes:
    py_time, dyn_time = benchmark_append(n)
    py_times.append(py_time)
    dyn_times.append(dyn_time)
    ratio = dyn_time / py_time if py_time > 0 else 0
    print(f"{n:<10} {py_time:<15.6f} {dyn_time:<15.6f} {ratio:<10.2f}x")

# Visualize
plt.figure(figsize=(12, 6))
x = range(len(sizes))
width = 0.35

plt.bar([i - width/2 for i in x], py_times, width, label='Python List', alpha=0.8)
plt.bar([i + width/2 for i in x], dyn_times, width, label='DynamicArray', alpha=0.8)

plt.xlabel('Array Size')
plt.ylabel('Time (seconds)')
plt.title('Append Performance Comparison')
plt.xticks(x, [str(s) for s in sizes])
plt.legend()
plt.yscale('log')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

# Ph·∫ßn 3: Linked Lists - L√Ω Thuy·∫øt Chi Ti·∫øt {#linked-lists-theory}

## 3.1. Kh√°i ni·ªám v√† ƒê·ªãnh nghƒ©a

**Linked List (Danh s√°ch li√™n k·∫øt)** l√† m·ªôt c·∫•u tr√∫c d·ªØ li·ªáu tuy·∫øn t√≠nh, trong ƒë√≥ c√°c ph·∫ßn t·ª≠ (nodes) kh√¥ng ƒë∆∞·ª£c l∆∞u tr·ªØ t·∫°i c√°c v·ªã tr√≠ b·ªô nh·ªõ li√™n ti·∫øp nhau. Thay v√†o ƒë√≥, m·ªói node ch·ª©a:
- **Data**: D·ªØ li·ªáu c·ªßa ph·∫ßn t·ª≠
- **Next (Pointer)**: Tham chi·∫øu ƒë·∫øn node ti·∫øp theo

### ƒê·∫∑c ƒëi·ªÉm c∆° b·∫£n:

1. **B·ªô nh·ªõ kh√¥ng li√™n ti·∫øp (Non-contiguous Memory)**
   - C√°c node c√≥ th·ªÉ n·∫±m ·ªü b·∫•t k·ª≥ ƒë√¢u trong b·ªô nh·ªõ
   - Kh√¥ng c·∫ßn c·∫•p ph√°t b·ªô nh·ªõ tr∆∞·ªõc

2. **Dynamic Size**
   - K√≠ch th∆∞·ªõc thay ƒë·ªïi ƒë·ªông trong runtime
   - Kh√¥ng l√£ng ph√≠ b·ªô nh·ªõ nh∆∞ array c√≥ capacity l·ªõn

# Ph·∫ßn 4: Linked Lists - Tri·ªÉn khai v√† ·ª®ng d·ª•ng {#linked-lists-implementation}

## 4.1. Tri·ªÉn khai Singly Linked List

````python
class ListNode:
    """Node trong Singly Linked List"""
    def __init__(self, data: Any = None):
        self.data = data
        self.next: Optional['ListNode'] = None
    
    def __repr__(self):
        return f"ListNode({self.data})"


class SinglyLinkedList:
    """Tri·ªÉn khai Singly Linked List"""
    
    def __init__(self):
        self.head: Optional[ListNode] = None
        self.size = 0
    
    def __len__(self) -> int:
        """Tr·∫£ v·ªÅ s·ªë ph·∫ßn t·ª≠ - O(1)"""
        return self.size
    
    def is_empty(self) -> bool:
        """Ki·ªÉm tra list c√≥ r·ªóng kh√¥ng - O(1)"""
        return self.head is None
    
    def prepend(self, data: Any):
        """Th√™m v√†o ƒë·∫ßu - O(1)"""
        new_node = ListNode(data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1
    
    def append(self, data: Any):
        """Th√™m v√†o cu·ªëi - O(n)"""
        new_node = ListNode(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
        self.size += 1
    
    def insert_at(self, index: int, data: Any):
        """Ch√®n v√†o v·ªã tr√≠ index - O(n)"""
        if index < 0 or index > self.size:
            raise IndexError(f"Index {index} out of range [0, {self.size}]")
        
        if index == 0:
            self.prepend(data)
            return
        
        new_node = ListNode(data)
        current = self.head
        for _ in range(index - 1):
            current = current.next
        
        new_node.next = current.next
        current.next = new_node
        self.size += 1
    
    def delete_first(self) -> Any:
        """X√≥a ph·∫ßn t·ª≠ ƒë·∫ßu - O(1)"""
        if self.head is None:
            raise ValueError("Cannot delete from empty list")
        
        data = self.head.data
        self.head = self.head.next
        self.size -= 1
        return data
    
    def delete_last(self) -> Any:
        """X√≥a ph·∫ßn t·ª≠ cu·ªëi - O(n)"""
        if self.head is None:
            raise ValueError("Cannot delete from empty list")
        
        if self.head.next is None:
            data = self.head.data
            self.head = None
            self.size -= 1
            return data
        
        current = self.head
        while current.next.next is not None:
            current = current.next
        
        data = current.next.data
        current.next = None
        self.size -= 1
        return data
    
    def delete_at(self, index: int) -> Any:
        """X√≥a ph·∫ßn t·ª≠ t·∫°i index - O(n)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} out of range [0, {self.size-1}]")
        
        if index == 0:
            return self.delete_first()
        
        current = self.head
        for _ in range(index - 1):
            current = current.next
        
        data = current.next.data
        current.next = current.next.next
        self.size -= 1
        return data
    
    def get(self, index: int) -> Any:
        """L·∫•y gi√° tr·ªã t·∫°i index - O(n)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} out of range [0, {self.size-1}]")
        
        current = self.head
        for _ in range(index):
            current = current.next
        return current.data
    
    def index_of(self, data: Any) -> int:
        """T√¨m index c·ªßa data - O(n)"""
        current = self.head
        index = 0
        while current is not None:
            if current.data == data:
                return index
            current = current.next
            index += 1
        raise ValueError(f"{data} not in list")
    
    def contains(self, data: Any) -> bool:
        """Ki·ªÉm tra data c√≥ trong list - O(n)"""
        current = self.head
        while current is not None:
            if current.data == data:
                return True
            current = current.next
        return False
    
    def reverse(self):
        """ƒê·∫£o ng∆∞·ª£c linked list - O(n)"""
        prev = None
        current = self.head
        
        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        
        self.head = prev
    
    def to_list(self) -> List[Any]:
        """Chuy·ªÉn linked list th√†nh Python list - O(n)"""
        result = []
        current = self.head
        while current is not None:
            result.append(current.data)
            current = current.next
        return result
    
    def __str__(self) -> str:
        """Hi·ªÉn th·ªã linked list"""
        if self.head is None:
            return "SinglyLinkedList([])"
        
        nodes = []
        current = self.head
        while current is not None:
            nodes.append(str(current.data))
            current = current.next
        
        return f"SinglyLinkedList([{' -> '.join(nodes)}])"
    
    def __repr__(self) -> str:
        return self.__str__()

# Demo SinglyLinkedList
print("=== DEMO: Singly Linked List ===\n")
ll = SinglyLinkedList()

# Test append v√† prepend
print("1. Th√™m ph·∫ßn t·ª≠:")
ll.append(10)
ll.append(20)
ll.append(30)
ll.prepend(5)
print(f"   {ll}")

# Test insert
print("\n2. Ch√®n ph·∫ßn t·ª≠:")
ll.insert_at(2, 15)
print(f"   Ch√®n 15 v√†o index 2: {ll}")

# Test get
print("\n3. Truy c·∫≠p ph·∫ßn t·ª≠:")
print(f"   Ph·∫ßn t·ª≠ t·∫°i index 0: {ll.get(0)}")
print(f"   Ph·∫ßn t·ª≠ t·∫°i index 2: {ll.get(2)}")

# Test delete
print("\n4. X√≥a ph·∫ßn t·ª≠:")
removed = ll.delete_at(2)
print(f"   X√≥a ph·∫ßn t·ª≠ t·∫°i index 2 (gi√° tr·ªã {removed}): {ll}")

# Test reverse
print("\n5. ƒê·∫£o ng∆∞·ª£c:")
print(f"   Tr∆∞·ªõc: {ll}")
ll.reverse()
print(f"   Sau: {ll}")

# Test search
print("\n6. T√¨m ki·∫øm:")
print(f"   Index c·ªßa 10: {ll.index_of(10)}")
print(f"   20 c√≥ trong list? {ll.contains(20)}")
print(f"   999 c√≥ trong list? {ll.contains(999)}")
````

## 4.2. Tri·ªÉn khai Doubly Linked List

````python
class DoublyListNode:
    """Node trong Doubly Linked List"""
    def __init__(self, data: Any = None):
        self.data = data
        self.prev: Optional['DoublyListNode'] = None
        self.next: Optional['DoublyListNode'] = None
    
    def __repr__(self):
        return f"DoublyListNode({self.data})"


class DoublyLinkedList:
    """Tri·ªÉn khai Doubly Linked List v·ªõi head v√† tail"""
    
    def __init__(self):
        self.head: Optional[DoublyListNode] = None
        self.tail: Optional[DoublyListNode] = None
        self.size = 0
    
    def __len__(self) -> int:
        return self.size
    
    def is_empty(self) -> bool:
        return self.head is None
    
    def prepend(self, data: Any):
        """Th√™m v√†o ƒë·∫ßu - O(1)"""
        new_node = DoublyListNode(data)
        
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        
        self.size += 1
    
    def append(self, data: Any):
        """Th√™m v√†o cu·ªëi - O(1)"""
        new_node = DoublyListNode(data)
        
        if self.tail is None:
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
        
        self.size += 1
    
    def delete_first(self) -> Any:
        """X√≥a ph·∫ßn t·ª≠ ƒë·∫ßu - O(1)"""
        if self.head is None:
            raise ValueError("Cannot delete from empty list")
        
        data = self.head.data
        
        if self.head == self.tail:  # Ch·ªâ c√≥ 1 ph·∫ßn t·ª≠
            self.head = self.tail = None
        else:
            self.head = self.head.next
            self.head.prev = None
        
        self.size -= 1
        return data
    
    def delete_last(self) -> Any:
        """X√≥a ph·∫ßn t·ª≠ cu·ªëi - O(1)"""
        if self.tail is None:
            raise ValueError("Cannot delete from empty list")
        
        data = self.tail.data
        
        if self.head == self.tail:  # Ch·ªâ c√≥ 1 ph·∫ßn t·ª≠
            self.head = self.tail = None
        else:
            self.tail = self.tail.prev
            self.tail.next = None
        
        self.size -= 1
        return data
    
    def delete_node(self, node: DoublyListNode) -> Any:
        """X√≥a node c·ª• th·ªÉ - O(1) (n·∫øu c√≥ reference)"""
        if node is None:
            raise ValueError("Node cannot be None")
        
        data = node.data
        
        if node.prev is not None:
            node.prev.next = node.next
        else:
            self.head = node.next
        
        if node.next is not None:
            node.next.prev = node.prev
        else:
            self.tail = node.prev
        
        self.size -= 1
        return data
    
    def to_list_forward(self) -> List[Any]:
        """Chuy·ªÉn th√†nh list (duy·ªát t·ª´ ƒë·∫ßu) - O(n)"""
        result = []
        current = self.head
        while current is not None:
            result.append(current.data)
            current = current.next
        return result
    
    def to_list_backward(self) -> List[Any]:
        """Chuy·ªÉn th√†nh list (duy·ªát t·ª´ cu·ªëi) - O(n)"""
        result = []
        current = self.tail
        while current is not None:
            result.append(current.data)
            current = current.prev
        return result
    
    def __str__(self) -> str:
        """Hi·ªÉn th·ªã linked list"""
        forward = self.to_list_forward()
        if not forward:
            return "DoublyLinkedList([])"
        return f"DoublyLinkedList({forward})"
    
    def __repr__(self):
        return self.__str__()

# Demo DoublyLinkedList
print("=== DEMO: Doubly Linked List ===\n")
dll = DoublyLinkedList()

# Test append v√† prepend
print("1. Th√™m ph·∫ßn t·ª≠:")
dll.append(10)
dll.append(20)
dll.append(30)
dll.prepend(5)
print(f"   {dll}")

# Test duy·ªát ng∆∞·ª£c
print("\n2. Duy·ªát t·ª´ cu·ªëi:")
print(f"   Forward: {dll.to_list_forward()}")
print(f"   Backward: {dll.to_list_backward()}")

# Test delete
print("\n3. X√≥a ph·∫ßn t·ª≠:")
removed_first = dll.delete_first()
print(f"   X√≥a ƒë·∫ßu ({removed_first}): {dll}")

removed_last = dll.delete_last()
print(f"   X√≥a cu·ªëi ({removed_last}): {dll}")
````

# Ph·∫ßn 5: So s√°nh Arrays vs Linked Lists {#comparison}

## 5.1. B·∫£ng So s√°nh Chi ti·∫øt

| Ti√™u ch√≠ | Array | Linked List |
|----------|-------|-------------|
| **B·ªô nh·ªõ** | Contiguous (li√™n ti·∫øp) | Non-contiguous (kh√¥ng li√™n ti·∫øp) |
| **Truy c·∫≠p** | O(1) - Random access | O(n) - Ph·∫£i duy·ªát t·ª´ ƒë·∫ßu |
| **T√¨m ki·∫øm** | O(n) linear, O(log n) n·∫øu sorted | O(n) |
| **Ch√®n ·ªü ƒë·∫ßu** | O(n) | O(1) |
| **Ch√®n ·ªü cu·ªëi** | O(1) amortized | O(n) singly, O(1) doubly |
| **Ch√®n ·ªü gi·ªØa** | O(n) | O(n) - nh∆∞ng kh√¥ng c·∫ßn shift |
| **X√≥a ·ªü ƒë·∫ßu** | O(n) | O(1) |
| **X√≥a ·ªü cu·ªëi** | O(1) | O(n) singly, O(1) doubly |
| **X√≥a ·ªü gi·ªØa** | O(n) | O(n) - nh∆∞ng kh√¥ng c·∫ßn shift |
| **B·ªô nh·ªõ overhead** | Th·∫•p | Cao (c·∫ßn pointer) |
| **Cache locality** | T·ªët (n·∫±m g·∫ßn nhau) | K√©m (ph√¢n t√°n) |

## 5.4. Visual Comparison: Performance Benchmark

H√£y so s√°nh hi·ªáu su·∫•t th·ª±c t·∫ø gi·ªØa Array v√† Linked List:

````python
# Benchmark: Insertion at beginning
def benchmark_insert_beginning(n: int):
    """So s√°nh ch√®n ·ªü ƒë·∫ßu"""
    # Python List (Array)
    start = time.time()
    arr = []
    for i in range(n):
        arr.insert(0, i)  # Ch√®n ·ªü ƒë·∫ßu
    arr_time = time.time() - start
    
    # Linked List
    start = time.time()
    ll = SinglyLinkedList()
    for i in range(n):
        ll.prepend(i)  # Ch√®n ·ªü ƒë·∫ßu - O(1)
    ll_time = time.time() - start
    
    return arr_time, ll_time

# Benchmark: Access by index
def benchmark_access(n: int):
    """So s√°nh truy c·∫≠p theo index"""
    # T·∫°o d·ªØ li·ªáu
    arr = list(range(n))
    ll = SinglyLinkedList()
    for i in range(n):
        ll.append(i)
    
    # Array access
    start = time.time()
    for i in range(min(1000, n)):
        _ = arr[i]
    arr_time = time.time() - start
    
    # Linked List access
    start = time.time()
    for i in range(min(1000, n)):
        _ = ll.get(i)
    ll_time = time.time() - start
    
    return arr_time, ll_time

# Ch·∫°y benchmarks
sizes = [1000, 5000, 10000]

print("=== BENCHMARK: Insert at Beginning ===\n")
print(f"{'Size':<10} {'Array (insert(0))':<20} {'Linked List (prepend)':<25} {'Speedup':<10}")
print("-" * 70)

insert_arr_times = []
insert_ll_times = []

for n in sizes:
    arr_time, ll_time = benchmark_insert_beginning(n)
    insert_arr_times.append(arr_time)
    insert_ll_times.append(ll_time)
    speedup = arr_time / ll_time if ll_time > 0 else 0
    print(f"{n:<10} {arr_time:<20.6f} {ll_time:<25.6f} {speedup:<10.2f}x")

print("\n=== BENCHMARK: Access by Index ===\n")
print(f"{'Size':<10} {'Array [i]':<20} {'Linked List get(i)':<25} {'Slowdown':<10}")
print("-" * 70)

access_arr_times = []
access_ll_times = []

for n in sizes:
    arr_time, ll_time = benchmark_access(n)
    access_arr_times.append(arr_time)
    access_ll_times.append(ll_time)
    slowdown = ll_time / arr_time if arr_time > 0 else 0
    print(f"{n:<10} {arr_time:<20.6f} {ll_time:<25.6f} {slowdown:<10.2f}x")

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Insert at beginning
x = range(len(sizes))
width = 0.35
ax1.bar([i - width/2 for i in x], insert_arr_times, width, label='Array (insert(0))', alpha=0.8, color='red')
ax1.bar([i + width/2 for i in x], insert_ll_times, width, label='Linked List (prepend)', alpha=0.8, color='green')
ax1.set_xlabel('List Size')
ax1.set_ylabel('Time (seconds)')
ax1.set_title('Insert at Beginning Performance')
ax1.set_xticks(x)
ax1.set_xticklabels([str(s) for s in sizes])
ax1.legend()
ax1.set_yscale('log')
ax1.grid(True, alpha=0.3)

# Access by index
ax2.bar([i - width/2 for i in x], access_arr_times, width, label='Array [i]', alpha=0.8, color='blue')
ax2.bar([i + width/2 for i in x], access_ll_times, width, label='Linked List get(i)', alpha=0.8, color='orange')
ax2.set_xlabel('List Size')
ax2.set_ylabel('Time (seconds)')
ax2.set_title('Access by Index Performance')
ax2.set_xticks(x)
ax2.set_xticklabels([str(s) for s in sizes])
ax2.legend()
ax2.set_yscale('log')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
````

# Ph·∫ßn 6: C·∫•u tr√∫c d·ªØ li·ªáu n√¢ng cao {#advanced-structures}

## 6.1. Circular Linked List

### 6.1.1. Tri·ªÉn khai Circular Singly Linked List

````python
class CircularSinglyLinkedList:
    """Circular Singly Linked List - con tr·ªè cu·ªëi tr·ªè v·ªÅ ƒë·∫ßu"""
    
    def __init__(self):
        self.tail: Optional[ListNode] = None  # L∆∞u tail thay v√¨ head ƒë·ªÉ append O(1)
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.tail is None
    
    def append(self, data: Any):
        """Th√™m v√†o cu·ªëi - O(1)"""
        new_node = ListNode(data)
        
        if self.tail is None:
            new_node.next = new_node  # T·ª± tr·ªè v√†o ch√≠nh n√≥
            self.tail = new_node
        else:
            new_node.next = self.tail.next  # Tr·ªè v·ªÅ ƒë·∫ßu
            self.tail.next = new_node  # Node cu·ªëi c≈© tr·ªè v·ªÅ node m·ªõi
            self.tail = new_node  # Node m·ªõi l√† tail m·ªõi
        
        self.size += 1
    
    def prepend(self, data: Any):
        """Th√™m v√†o ƒë·∫ßu - O(1)"""
        new_node = ListNode(data)
        
        if self.tail is None:
            new_node.next = new_node
            self.tail = new_node
        else:
            new_node.next = self.tail.next  # Tr·ªè v·ªÅ ƒë·∫ßu hi·ªán t·∫°i
            self.tail.next = new_node  # Tail tr·ªè v·ªÅ node m·ªõi (ƒë·∫ßu m·ªõi)
        
        self.size += 1
    
    def get_head(self) -> Optional[ListNode]:
        """L·∫•y node ƒë·∫ßu"""
        return self.tail.next if self.tail else None
    
    def to_list(self) -> List[Any]:
        """Chuy·ªÉn th√†nh list - O(n)"""
        if self.tail is None:
            return []
        
        result = []
        current = self.tail.next  # B·∫Øt ƒë·∫ßu t·ª´ head
        while True:
            result.append(current.data)
            if current == self.tail:
                break
            current = current.next
        return result
    
    def __str__(self):
        if self.tail is None:
            return "CircularSinglyLinkedList([])"
        items = self.to_list()
        return f"CircularSinglyLinkedList({items})"

# Demo Circular Linked List
print("=== DEMO: Circular Singly Linked List ===\n")
cll = CircularSinglyLinkedList()

print("1. Th√™m ph·∫ßn t·ª≠:")
cll.append(10)
cll.append(20)
cll.append(30)
cll.prepend(5)
print(f"   {cll}")
print(f"   Head: {cll.get_head().data}, Tail: {cll.tail.data}")
print(f"   Tail.next tr·ªè v·ªÅ: {cll.tail.next.data} (ƒë·∫ßu)")
````

## 6.2. ·ª®ng d·ª•ng th·ª±c t·∫ø

### 6.2.1. Stack s·ª≠ d·ª•ng Linked List

Stack (NgƒÉn x·∫øp) - LIFO (Last In First Out):

````python
class Stack:
    """Stack tri·ªÉn khai b·∫±ng Linked List - push/pop ·ªü ƒë·∫ßu O(1)"""
    
    def __init__(self):
        self.top: Optional[ListNode] = None
        self.size = 0
    
    def push(self, data: Any):
        """Th√™m v√†o ƒë·ªânh - O(1)"""
        new_node = ListNode(data)
        new_node.next = self.top
        self.top = new_node
        self.size += 1
    
    def pop(self) -> Any:
        """X√≥a v√† tr·∫£ v·ªÅ ph·∫ßn t·ª≠ ƒë·ªânh - O(1)"""
        if self.top is None:
            raise ValueError("Stack is empty")
        
        data = self.top.data
        self.top = self.top.next
        self.size -= 1
        return data
    
    def peek(self) -> Any:
        """Xem ph·∫ßn t·ª≠ ƒë·ªânh kh√¥ng x√≥a - O(1)"""
        if self.top is None:
            raise ValueError("Stack is empty")
        return self.top.data
    
    def is_empty(self) -> bool:
        return self.top is None
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        items = []
        current = self.top
        while current is not None:
            items.append(str(current.data))
            current = current.next
        return f"Stack([{' -> '.join(items)}])"

# Demo Stack
print("=== DEMO: Stack (Linked List Implementation) ===\n")
stack = Stack()

print("1. Push c√°c ph·∫ßn t·ª≠:")
for i in [10, 20, 30, 40]:
    stack.push(i)
    print(f"   Push {i}: {stack}")

print(f"\n2. Peek (xem ƒë·ªânh): {stack.peek()}")

print("\n3. Pop c√°c ph·∫ßn t·ª≠:")
while not stack.is_empty():
    popped = stack.pop()
    print(f"   Pop: {popped}, Stack: {stack}")
````

### 6.2.2. Queue s·ª≠ d·ª•ng Linked List

Queue (H√†ng ƒë·ª£i) - FIFO (First In First Out):

````python
class Queue:
    """Queue tri·ªÉn khai b·∫±ng Linked List - enqueue ·ªü cu·ªëi, dequeue ·ªü ƒë·∫ßu"""
    
    def __init__(self):
        self.head: Optional[ListNode] = None
        self.tail: Optional[ListNode] = None
        self.size = 0
    
    def enqueue(self, data: Any):
        """Th√™m v√†o cu·ªëi - O(1)"""
        new_node = ListNode(data)
        
        if self.tail is None:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        
        self.size += 1
    
    def dequeue(self) -> Any:
        """X√≥a v√† tr·∫£ v·ªÅ ph·∫ßn t·ª≠ ƒë·∫ßu - O(1)"""
        if self.head is None:
            raise ValueError("Queue is empty")
        
        data = self.head.data
        self.head = self.head.next
        
        if self.head is None:
            self.tail = None
        
        self.size -= 1
        return data
    
    def front(self) -> Any:
        """Xem ph·∫ßn t·ª≠ ƒë·∫ßu kh√¥ng x√≥a - O(1)"""
        if self.head is None:
            raise ValueError("Queue is empty")
        return self.head.data
    
    def is_empty(self) -> bool:
        return self.head is None
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        items = []
        current = self.head
        while current is not None:
            items.append(str(current.data))
            current = current.next
        return f"Queue([{' -> '.join(items)}])"

# Demo Queue
print("=== DEMO: Queue (Linked List Implementation) ===\n")
queue = Queue()

print("1. Enqueue c√°c ph·∫ßn t·ª≠:")
for i in [10, 20, 30, 40]:
    queue.enqueue(i)
    print(f"   Enqueue {i}: {queue}")

print(f"\n2. Front (xem ƒë·∫ßu): {queue.front()}")

print("\n3. Dequeue c√°c ph·∫ßn t·ª≠:")
while not queue.is_empty():
    dequeued = queue.dequeue()
    print(f"   Dequeue: {dequeued}, Queue: {queue}")
````

# Ph·∫ßn 7: B√†i t·∫≠p v√† ·ª®ng d·ª•ng th·ª±c t·∫ø {#exercises}

## 7.1. C√°c b√†i to√°n ph·ªï bi·∫øn v·ªõi Linked List

### 7.1.1. T√¨m ph·∫ßn t·ª≠ gi·ªØa (Middle Element)

````python
def find_middle(head: Optional[ListNode]) -> Optional[ListNode]:
    """
    T√¨m node gi·ªØa c·ªßa linked list
    S·ª≠ d·ª•ng k·ªπ thu·∫≠t Slow-Fast Pointer (Tortoise and Hare)
    Time: O(n), Space: O(1)
    """
    if head is None or head.next is None:
        return head
    
    slow = head
    fast = head
    
    # Fast ƒëi 2 b∆∞·ªõc, slow ƒëi 1 b∆∞·ªõc
    # Khi fast ƒë·∫øn cu·ªëi, slow ·ªü gi·ªØa
    while fast is not None and fast.next is not None:
        slow = slow.next
        fast = fast.next.next
    
    return slow

# Demo
ll = SinglyLinkedList()
for i in range(1, 6):
    ll.append(i * 10)

print("=== T√¨m ph·∫ßn t·ª≠ gi·ªØa ===")
print(f"List: {ll}")
middle = find_middle(ll.head)
print(f"Ph·∫ßn t·ª≠ gi·ªØa: {middle.data if middle else None}")
````

### 7.1.2. Ph√°t hi·ªán v√≤ng l·∫∑p (Cycle Detection)

S·ª≠ d·ª•ng Floyd's Cycle Detection Algorithm (Tortoise and Hare):

````python
def has_cycle(head: Optional[ListNode]) -> bool:
    """
    Ki·ªÉm tra linked list c√≥ v√≤ng l·∫∑p kh√¥ng
    S·ª≠ d·ª•ng Floyd's Cycle Detection
    Time: O(n), Space: O(1)
    """
    if head is None or head.next is None:
        return False
    
    slow = head
    fast = head.next
    
    while fast is not None and fast.next is not None:
        if slow == fast:
            return True
        slow = slow.next
        fast = fast.next.next
    
    return False

# Demo: T·∫°o linked list c√≥ cycle
print("=== Ph√°t hi·ªán v√≤ng l·∫∑p ===")

# List kh√¥ng c√≥ cycle
ll1 = SinglyLinkedList()
for i in [1, 2, 3, 4, 5]:
    ll1.append(i)
print(f"List 1: {ll1}")
print(f"C√≥ cycle? {has_cycle(ll1.head)}")

# List c√≥ cycle (t·∫°o b·∫±ng c√°ch n·ªëi tail v·ªÅ node gi·ªØa)
ll2 = SinglyLinkedList()
for i in [1, 2, 3, 4, 5]:
    ll2.append(i)
# T·∫°o cycle: node 5 -> node 3
current = ll2.head
while current.next is not None:
    current = current.next
# current gi·ªù l√† node cu·ªëi (5)
middle_node = ll2.head.next.next  # node 3
current.next = middle_node  # 5 -> 3 (t·∫°o cycle)

print(f"\nList 2 c√≥ cycle (5 -> 3)")
print(f"C√≥ cycle? {has_cycle(ll2.head)}")
````

### 7.1.3. Merge Two Sorted Linked Lists

H·ª£p nh·∫•t hai linked list ƒë√£ s·∫Øp x·∫øp:

````python
def merge_sorted_lists(l1: Optional[ListNode], l2: Optional[ListNode]) -> Optional[ListNode]:
    """
    H·ª£p nh·∫•t hai sorted linked list th√†nh m·ªôt sorted list
    Time: O(n + m), Space: O(1)
    """
    dummy = ListNode(0)  # Dummy node ƒë·ªÉ d·ªÖ x·ª≠ l√Ω
    current = dummy
    
    while l1 is not None and l2 is not None:
        if l1.data <= l2.data:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next
    
    # N·ªëi ph·∫ßn c√≤n l·∫°i
    if l1 is not None:
        current.next = l1
    elif l2 is not None:
        current.next = l2
    
    return dummy.next

# Helper function ƒë·ªÉ hi·ªÉn th·ªã
def list_to_string(head: Optional[ListNode]) -> str:
    items = []
    current = head
    while current is not None:
        items.append(str(current.data))
        current = current.next
    return ' -> '.join(items) if items else 'None'

# Demo
print("=== Merge Two Sorted Linked Lists ===")

# T·∫°o list 1: [1, 3, 5, 7]
l1 = ListNode(1)
l1.next = ListNode(3)
l1.next.next = ListNode(5)
l1.next.next.next = ListNode(7)

# T·∫°o list 2: [2, 4, 6, 8]
l2 = ListNode(2)
l2.next = ListNode(4)
l2.next.next = ListNode(6)
l2.next.next.next = ListNode(8)

print(f"List 1: {list_to_string(l1)}")
print(f"List 2: {list_to_string(l2)}")

merged = merge_sorted_lists(l1, l2)
print(f"Merged: {list_to_string(merged)}")
````

## 7.2. C√°c b√†i to√°n ph·ªï bi·∫øn v·ªõi Array

### 7.2.1. T√¨m hai s·ªë c√≥ t·ªïng b·∫±ng target (Two Sum)

````python
def two_sum(arr: List[int], target: int) -> List[int]:
    """
    T√¨m hai s·ªë trong array c√≥ t·ªïng b·∫±ng target
    Tr·∫£ v·ªÅ indices c·ªßa hai s·ªë ƒë√≥
    Time: O(n), Space: O(n)
    """
    seen = {}  # value -> index
    
    for i, num in enumerate(arr):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    
    return []

# Demo
print("=== Two Sum Problem ===")
arr = [2, 7, 11, 15]
target = 9
result = two_sum(arr, target)
print(f"Array: {arr}")
print(f"Target: {target}")
print(f"Indices: {result}")
print(f"Values: [{arr[result[0]]}, {arr[result[1]]}]")
````

### 7.2.2. T√¨m ph·∫ßn t·ª≠ l·ªõn nh·∫•t trong m·∫£ng con (Sliding Window Maximum)

·ª®ng d·ª•ng: T√¨m maximum trong m·ªói window k√≠ch th∆∞·ªõc k

````python
def max_sliding_window(nums: List[int], k: int) -> List[int]:
    """
    T√¨m maximum trong m·ªói sliding window k√≠ch th∆∞·ªõc k
    S·ª≠ d·ª•ng deque ƒë·ªÉ t·ªëi ∆∞u
    Time: O(n), Space: O(k)
    """
    from collections import deque
    
    result = []
    dq = deque()  # L∆∞u indices
    
    for i in range(len(nums)):
        # Lo·∫°i b·ªè indices ngo√†i window
        while dq and dq[0] < i - k + 1:
            dq.popleft()
        
        # Lo·∫°i b·ªè c√°c ph·∫ßn t·ª≠ nh·ªè h∆°n nums[i]
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # Th√™m maximum v√†o result khi window ƒë·ªß k√≠ch th∆∞·ªõc
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

# Demo
print("=== Sliding Window Maximum ===")
nums = [1, 3, -1, -3, 5, 3, 6, 7]
k = 3
result = max_sliding_window(nums, k)
print(f"Array: {nums}")
print(f"Window size: {k}")
print(f"Maximums: {result}")
print(f"Gi·∫£i th√≠ch: [3, 3, 5, 5, 6, 7] l√† max trong c√°c window:")
for i in range(len(nums) - k + 1):
    window = nums[i:i+k]
    print(f"  Window {i}: {window} -> max = {max(window)}")
````
