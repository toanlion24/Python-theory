## 1.1. Kh√°i ni·ªám v·ªÅ Arrays

**Array (M·∫£ng)** l√† m·ªôt c·∫•u tr√∫c d·ªØ li·ªáu tuy·∫øn t√≠nh (linear data structure) l∆∞u tr·ªØ c√°c ph·∫ßn t·ª≠ c√πng ki·ªÉu d·ªØ li·ªáu trong c√°c v·ªã tr√≠ b·ªô nh·ªõ li√™n ti·∫øp nhau.

### ƒê·∫∑c ƒëi·ªÉm c·ªßa Arrays:
- **Truy c·∫≠p ng·∫´u nhi√™n (Random Access)**: C√≥ th·ªÉ truy c·∫≠p ph·∫ßn t·ª≠ b·∫•t k·ª≥ b·∫±ng index trong th·ªùi gian O(1)
- **K√≠ch th∆∞·ªõc c·ªë ƒë·ªãnh ho·∫∑c ƒë·ªông**: T√πy v√†o ng√¥n ng·ªØ v√† c√°ch tri·ªÉn khai
- **B·ªô nh·ªõ li√™n ti·∫øp**: C√°c ph·∫ßn t·ª≠ ƒë∆∞·ª£c l∆∞u tr·ªØ li√™n ti·∫øp trong b·ªô nh·ªõ
- **Hi·ªáu qu·∫£ v·ªÅ b·ªô nh·ªõ**: Kh√¥ng c·∫ßn l∆∞u tr·ªØ th√™m con tr·ªè nh∆∞ linked list
- **D·ªÖ d√†ng thao t√°c**: Th√™m, x√≥a, s·ª≠a ƒë·ªïi ph·∫ßn t·ª≠ ƒë∆°n gi·∫£n

### ƒê·ªô ph·ª©c t·∫°p th·ªùi gian (Time Complexity):
- **Truy c·∫≠p**: O(1)
- **T√¨m ki·∫øm**: O(n) - t√¨m ki·∫øm tuy·∫øn t√≠nh
- **Ch√®n**: O(n) - ph·∫£i d·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠
- **X√≥a**: O(n) - ph·∫£i d·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠

### 1.2.3. Tri·ªÉn khai Array t√πy ch·ªânh (Custom Array Implementation)

Tri·ªÉn khai m·ªôt l·ªõp Array ƒë∆°n gi·∫£n ƒë·ªÉ hi·ªÉu r√µ c√°ch ho·∫°t ƒë·ªông b√™n trong.

```python
class CustomArray:
    """Tri·ªÉn khai m·∫£ng t√πy ch·ªânh v·ªõi c√°c thao t√°c c∆° b·∫£n"""
    
    def __init__(self, capacity: int = 10):
        """Kh·ªüi t·∫°o m·∫£ng v·ªõi dung l∆∞·ª£ng ban ƒë·∫ßu"""
        self.capacity = capacity
        self.size = 0
        self.data = [None] * capacity
    
    def __len__(self):
        """Tr·∫£ v·ªÅ s·ªë ph·∫ßn t·ª≠ hi·ªán t·∫°i"""
        return self.size
    
    def __getitem__(self, index: int):
        """Truy c·∫≠p ph·∫ßn t·ª≠ b·∫±ng index - O(1)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} n·∫±m ngo√†i ph·∫°m vi [0, {self.size-1}]")
        return self.data[index]
    
    def __setitem__(self, index: int, value):
        """G√°n gi√° tr·ªã cho ph·∫ßn t·ª≠ - O(1)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} n·∫±m ngo√†i ph·∫°m vi [0, {self.size-1}]")
        self.data[index] = value
    
    def append(self, value):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(1) amortized"""
        if self.size >= self.capacity:
            self._resize()
        self.data[self.size] = value
        self.size += 1
    
    def insert(self, index: int, value):
        """Ch√®n ph·∫ßn t·ª≠ v√†o v·ªã tr√≠ index - O(n)"""
        if index < 0 or index > self.size:
            raise IndexError(f"Index {index} kh√¥ng h·ª£p l·ªá")
        
        if self.size >= self.capacity:
            self._resize()
        
        # D·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠ sang ph·∫£i
        for i in range(self.size, index, -1):
            self.data[i] = self.data[i-1]
        
        self.data[index] = value
        self.size += 1
    
    def delete(self, index: int):
        """X√≥a ph·∫ßn t·ª≠ t·∫°i v·ªã tr√≠ index - O(n)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} n·∫±m ngo√†i ph·∫°m vi")
        
        # D·ªãch chuy·ªÉn c√°c ph·∫ßn t·ª≠ sang tr√°i
        for i in range(index, self.size - 1):
            self.data[i] = self.data[i + 1]
        
        self.size -= 1
        self.data[self.size] = None
    
    def _resize(self):
        """TƒÉng g·∫•p ƒë√¥i dung l∆∞·ª£ng khi m·∫£ng ƒë·∫ßy - O(n)"""
        old_capacity = self.capacity
        self.capacity *= 2
        new_data = [None] * self.capacity
        
        for i in range(old_capacity):
            new_data[i] = self.data[i]
        
        self.data = new_data
        print(f"  ‚Üí M·∫£ng ƒë√£ ƒë∆∞·ª£c resize t·ª´ {old_capacity} l√™n {self.capacity}")
    
    def search(self, value):
        """T√¨m ki·∫øm gi√° tr·ªã - O(n)"""
        for i in range(self.size):
            if self.data[i] == value:
                return i
        return -1
    
    def __str__(self):
        """Hi·ªÉn th·ªã m·∫£ng"""
        elements = [str(self.data[i]) for i in range(self.size)]
        return f"CustomArray([{', '.join(elements)}], size={self.size}, capacity={self.capacity})"
    
    def __repr__(self):
        return self.__str__()

# S·ª≠ d·ª•ng CustomArray
print("=== Tri·ªÉn khai Custom Array ===\n")
arr = CustomArray(capacity=5)

# Th√™m ph·∫ßn t·ª≠
for i in range(7):
    arr.append(i * 10)
    print(f"Sau khi th√™m {i*10}: {arr}")

# Ch√®n ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi ch√®n: {arr}")
arr.insert(2, 99)
print(f"Sau khi ch√®n 99 v√†o v·ªã tr√≠ 2: {arr}")

# X√≥a ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi x√≥a: {arr}")
arr.delete(3)
print(f"Sau khi x√≥a ph·∫ßn t·ª≠ t·∫°i index 3: {arr}")

# T√¨m ki·∫øm
index = arr.search(40)
print(f"\nT√¨m ki·∫øm 40: {'T√¨m th·∫•y t·∫°i index ' + str(index) if index != -1 else 'Kh√¥ng t√¨m th·∫•y'}")

# Truy c·∫≠p ph·∫ßn t·ª≠
print(f"\nPh·∫ßn t·ª≠ t·∫°i index 0: {arr[0]}")
print(f"Ph·∫ßn t·ª≠ t·∫°i index 2: {arr[2]}")
```

```python
import matplotlib.patches as mpatches

def visualize_search(arr, target, current_idx, found_idx=None, search_type="Linear"):
    """V·∫Ω m·ªôt b∆∞·ªõc c·ªßa thu·∫≠t to√°n t√¨m ki·∫øm"""
    fig, ax = plt.subplots(figsize=(12, 6))
    colors = ['lightblue'] * len(arr)
    
    # T√¥ m√†u ph·∫ßn t·ª≠ ƒëang x√©t
    if 0 <= current_idx < len(arr):
        if found_idx is not None and current_idx == found_idx:
            colors[current_idx] = 'green'  # T√¨m th·∫•y
        else:
            colors[current_idx] = 'yellow'  # ƒêang x√©t
    
    # T√¥ m√†u c√°c ph·∫ßn t·ª≠ ƒë√£ x√©t (n·∫øu l√† linear search)
    if search_type == "Linear" and current_idx > 0:
        for i in range(current_idx):
            colors[i] = 'lightgray'
    
    bars = ax.bar(range(len(arr)), arr, color=colors, edgecolor='black', linewidth=1.5)
    
    # Th√™m gi√° tr·ªã v√† index
    for i, (bar, val) in enumerate(zip(bars, arr)):
        height = bar.get_height()
        # Gi√° tr·ªã
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val}', ha='center', va='bottom', fontsize=10, fontweight='bold')
        # Index
        ax.text(bar.get_x() + bar.get_width()/2., -max(arr)*0.05,
                f'[{i}]', ha='center', va='top', fontsize=9)
    
    # Highlight target
    if found_idx is not None:
        ax.text(found_idx, arr[found_idx] + max(arr)*0.1, 
               '‚úì T√¨m th·∫•y!', ha='center', fontsize=12, fontweight='bold', 
               color='green', bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))
    
    title = f'{search_type} Search - T√¨m ki·∫øm {target}'
    if found_idx is not None:
        title += f' (T√¨m th·∫•y t·∫°i index {found_idx})'
    elif current_idx < len(arr):
        title += f' - ƒêang x√©t index {current_idx}'
    
    ax.set_xlabel('Index', fontsize=12)
    ax.set_ylabel('Gi√° tr·ªã', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xticks(range(len(arr)))
    ax.grid(True, alpha=0.3, axis='y')
    
    # Legend
    legend_elements = [
        mpatches.Patch(facecolor='yellow', label='ƒêang x√©t'),
        mpatches.Patch(facecolor='lightgray', label='ƒê√£ x√©t'),
        mpatches.Patch(facecolor='green', label='T√¨m th·∫•y'),
        mpatches.Patch(facecolor='lightblue', label='Ch∆∞a x√©t')
    ]
    ax.legend(handles=legend_elements, loc='upper right')
    
    plt.tight_layout()
    plt.show()

def linear_search_visualized(arr, target):
    """Linear Search v·ªõi visualization t·ª´ng b∆∞·ªõc"""
    steps = []
    for i in range(len(arr)):
        steps.append({
            'current_idx': i,
            'found': None if arr[i] != target else i
        })
        if arr[i] == target:
            break
    
    return steps

def binary_search_visualized(arr, target):
    """Binary Search v·ªõi visualization t·ª´ng b∆∞·ªõc"""
    steps = []
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        steps.append({
            'left': left,
            'right': right,
            'mid': mid,
            'found': None
        })
        
        if arr[mid] == target:
            steps[-1]['found'] = mid
            break
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return steps

# V√≠ d·ª• minh h·ªça Linear Search
print("=== Visualization Thu·∫≠t to√°n T√¨m ki·∫øm ===\n")
search_array = [10, 20, 30, 40, 50, 60, 70, 80, 90]
target_value = 50

print(f"üîç LINEAR SEARCH - T√¨m ki·∫øm {target_value} trong m·∫£ng {search_array}\n")
linear_steps = linear_search_visualized(search_array, target_value)

for i, step in enumerate(linear_steps):
    visualize_search(search_array, target_value, step['current_idx'], 
                    step['found'], "Linear")
    if step['found'] is not None:
        break
```

```python
# V√≠ d·ª• minh h·ªça Binary Search
print(f"\nüîç BINARY SEARCH - T√¨m ki·∫øm {target_value} trong m·∫£ng ƒë√£ s·∫Øp x·∫øp\n")
sorted_search_array = sorted(search_array)
binary_steps = binary_search_visualized(sorted_search_array, target_value)

def visualize_binary_search(arr, left, right, mid, target, found_idx=None):
    """V·∫Ω m·ªôt b∆∞·ªõc c·ªßa Binary Search"""
    fig, ax = plt.subplots(figsize=(14, 6))
    colors = ['lightblue'] * len(arr)
    
    # T√¥ m√†u v√πng ƒëang t√¨m ki·∫øm
    for i in range(len(arr)):
        if i < left or i > right:
            colors[i] = 'lightgray'  # Ngo√†i ph·∫°m vi
        elif i == mid:
            colors[i] = 'yellow'  # Ph·∫ßn t·ª≠ gi·ªØa ƒëang x√©t
        elif found_idx is not None and i == found_idx:
            colors[i] = 'green'  # T√¨m th·∫•y
    
    bars = ax.bar(range(len(arr)), arr, color=colors, edgecolor='black', linewidth=1.5)
    
    # Th√™m gi√° tr·ªã v√† index
    for i, (bar, val) in enumerate(zip(bars, arr)):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val}', ha='center', va='bottom', fontsize=10, fontweight='bold')
        ax.text(bar.get_x() + bar.get_width()/2., -max(arr)*0.05,
                f'[{i}]', ha='center', va='top', fontsize=9)
    
    # V·∫Ω m≈©i t√™n ch·ªâ left, right, mid
    if left <= right:
        ax.annotate('left', xy=(left, arr[left]), xytext=(left, max(arr)*1.2),
                   arrowprops=dict(arrowstyle='->', color='blue', lw=2),
                   fontsize=11, fontweight='bold', color='blue')
        ax.annotate('right', xy=(right, arr[right]), xytext=(right, max(arr)*1.2),
                   arrowprops=dict(arrowstyle='->', color='red', lw=2),
                   fontsize=11, fontweight='bold', color='red')
        ax.annotate('mid', xy=(mid, arr[mid]), xytext=(mid, max(arr)*1.4),
                   arrowprops=dict(arrowstyle='->', color='orange', lw=2),
                   fontsize=11, fontweight='bold', color='orange')
    
    if found_idx is not None:
        ax.text(found_idx, arr[found_idx] + max(arr)*0.1, 
               '‚úì T√¨m th·∫•y!', ha='center', fontsize=12, fontweight='bold', 
               color='green', bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.7))
    
    title = f'Binary Search - T√¨m ki·∫øm {target}'
    if found_idx is not None:
        title += f' (T√¨m th·∫•y t·∫°i index {found_idx})'
    else:
        title += f' - left={left}, right={right}, mid={mid}'
    
    ax.set_xlabel('Index', fontsize=12)
    ax.set_ylabel('Gi√° tr·ªã', fontsize=12)
    ax.set_title(title, fontsize=14, fontweight='bold')
    ax.set_xticks(range(len(arr)))
    ax.grid(True, alpha=0.3, axis='y')
    
    # Legend
    legend_elements = [
        mpatches.Patch(facecolor='yellow', label='mid (ƒëang x√©t)'),
        mpatches.Patch(facecolor='lightblue', label='V√πng t√¨m ki·∫øm [left, right]'),
        mpatches.Patch(facecolor='lightgray', label='ƒê√£ lo·∫°i b·ªè'),
        mpatches.Patch(facecolor='green', label='T√¨m th·∫•y')
    ]
    ax.legend(handles=legend_elements, loc='upper right')
    
    plt.tight_layout()
    plt.show()

for i, step in enumerate(binary_steps):
    visualize_binary_search(sorted_search_array, step['left'], step['right'], 
                           step['mid'], target_value, step['found'])
    if step['found'] is not None:
        break
```

```python
def linear_search(arr: List, target) -> int:
    """T√¨m ki·∫øm tuy·∫øn t√≠nh - O(n)"""
    for i in range(len(arr)):
        if arr[i] == target:
            return i
    return -1

def binary_search(arr: List, target: int) -> int:
    """T√¨m ki·∫øm nh·ªã ph√¢n - O(log n) - Y√™u c·∫ßu m·∫£ng ƒë√£ s·∫Øp x·∫øp"""
    left, right = 0, len(arr) - 1
    
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    
    return -1

# V√≠ d·ª• s·ª≠ d·ª•ng
arr_unsorted = [64, 34, 25, 12, 22, 11, 90]
arr_sorted = sorted(arr_unsorted)

target = 22

print(f"M·∫£ng ch∆∞a s·∫Øp x·∫øp: {arr_unsorted}")
result_linear = linear_search(arr_unsorted, target)
print(f"T√¨m ki·∫øm tuy·∫øn t√≠nh {target}: {'T√¨m th·∫•y t·∫°i index ' + str(result_linear) if result_linear != -1 else 'Kh√¥ng t√¨m th·∫•y'}")

print(f"\nM·∫£ng ƒë√£ s·∫Øp x·∫øp: {arr_sorted}")
result_binary = binary_search(arr_sorted, target)
print(f"T√¨m ki·∫øm nh·ªã ph√¢n {target}: {'T√¨m th·∫•y t·∫°i index ' + str(result_binary) if result_binary != -1 else 'Kh√¥ng t√¨m th·∫•y'}")

# So s√°nh hi·ªáu su·∫•t
import time

large_arr = list(range(1000000))
target = 999999

# Linear search
start = time.time()
linear_search(large_arr, target)
linear_time = time.time() - start

# Binary search
start = time.time()
binary_search(large_arr, target)
binary_time = time.time() - start

print(f"\n=== So s√°nh hi·ªáu su·∫•t (m·∫£ng 1 tri·ªáu ph·∫ßn t·ª≠) ===")
print(f"Linear Search: {linear_time*1000:.4f} ms")
print(f"Binary Search: {binary_time*1000:.4f} ms")

# Tr√°nh chia cho 0
if binary_time > 0:
    ratio = linear_time / binary_time
    print(f"Binary Search nhanh h∆°n {ratio:.0f} l·∫ßn")
else:
    print("Binary Search qu√° nhanh (th·ªùi gian < 0.0001 ms), kh√¥ng th·ªÉ t√≠nh t·ª∑ l·ªá ch√≠nh x√°c")
```

```python
def bubble_sort(arr: List) -> List:
    """S·∫Øp x·∫øp n·ªïi b·ªçt - O(n¬≤)"""
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

def selection_sort(arr: List) -> List:
    """S·∫Øp x·∫øp ch·ªçn - O(n¬≤)"""
    arr = arr.copy()
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

def insertion_sort(arr: List) -> List:
    """S·∫Øp x·∫øp ch√®n - O(n¬≤)"""
    arr = arr.copy()
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key
    return arr

def quick_sort(arr: List) -> List:
    """S·∫Øp x·∫øp nhanh - O(n log n) average, O(n¬≤) worst case"""
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# V√≠ d·ª• s·ª≠ d·ª•ng
test_arr = [64, 34, 25, 12, 22, 11, 90]
print(f"M·∫£ng ban ƒë·∫ßu: {test_arr}\n")

print(f"Bubble Sort:    {bubble_sort(test_arr)}")
print(f"Selection Sort: {selection_sort(test_arr)}")
print(f"Insertion Sort: {insertion_sort(test_arr)}")
print(f"Quick Sort:     {quick_sort(test_arr)}")

# So s√°nh hi·ªáu su·∫•t
import random

sizes = [100, 500, 1000, 2000]
algorithms = {
    "Bubble Sort": bubble_sort,
    "Selection Sort": selection_sort,
    "Insertion Sort": insertion_sort,
    "Quick Sort": quick_sort
}

print("\n=== So s√°nh hi·ªáu su·∫•t c√°c thu·∫≠t to√°n s·∫Øp x·∫øp ===")
print(f"{'K√≠ch th∆∞·ªõc':<12} {'Bubble':<12} {'Selection':<12} {'Insertion':<12} {'Quick':<12}")
print("-" * 60)

for size in sizes:
    arr = [random.randint(1, 1000) for _ in range(size)]
    times = []
    
    for name, func in algorithms.items():
        start = time.time()
        func(arr)
        elapsed = (time.time() - start) * 1000
        times.append(f"{elapsed:.2f}ms")
    
    print(f"{size:<12} {times[0]:<12} {times[1]:<12} {times[2]:<12} {times[3]:<12}")
```

# 2. Linked Lists (Danh s√°ch Li√™n k·∫øt) {#linked-lists}

## 2.1. Kh√°i ni·ªám v·ªÅ Linked Lists

**Linked List (Danh s√°ch li√™n k·∫øt)** l√† m·ªôt c·∫•u tr√∫c d·ªØ li·ªáu tuy·∫øn t√≠nh trong ƒë√≥ c√°c ph·∫ßn t·ª≠ (nodes) kh√¥ng ƒë∆∞·ª£c l∆∞u tr·ªØ ·ªü c√°c v·ªã tr√≠ b·ªô nh·ªõ li√™n ti·∫øp, m√† m·ªói node ch·ª©a d·ªØ li·ªáu v√† m·ªôt con tr·ªè (pointer) tr·ªè ƒë·∫øn node ti·∫øp theo.

### ƒê·∫∑c ƒëi·ªÉm c·ªßa Linked Lists:
- **B·ªô nh·ªõ ƒë·ªông**: K√≠ch th∆∞·ªõc c√≥ th·ªÉ thay ƒë·ªïi trong qu√° tr√¨nh th·ª±c thi
- **B·ªô nh·ªõ kh√¥ng li√™n ti·∫øp**: C√°c node c√≥ th·ªÉ n·∫±m ·ªü c√°c v·ªã tr√≠ kh√°c nhau trong b·ªô nh·ªõ
- **Th√™m/x√≥a hi·ªáu qu·∫£**: Th√™m ho·∫∑c x√≥a ph·∫ßn t·ª≠ ·ªü ƒë·∫ßu danh s√°ch l√† O(1)
- **Kh√¥ng truy c·∫≠p ng·∫´u nhi√™n**: Ph·∫£i duy·ªát t·ª´ ƒë·∫ßu ƒë·ªÉ ƒë·∫øn ph·∫ßn t·ª≠ c·∫ßn t√¨m - O(n)
- **T·ªën b·ªô nh·ªõ h∆°n**: M·ªói node c·∫ßn l∆∞u th√™m con tr·ªè

### C√°c lo·∫°i Linked Lists:
1. **Singly Linked List**: M·ªói node ch·ªâ c√≥ con tr·ªè ƒë·∫øn node ti·∫øp theo
2. **Doubly Linked List**: M·ªói node c√≥ con tr·ªè ƒë·∫øn node tr∆∞·ªõc v√† node sau
3. **Circular Linked List**: Node cu·ªëi tr·ªè v·ªÅ node ƒë·∫ßu

### ƒê·ªô ph·ª©c t·∫°p th·ªùi gian:
- **Truy c·∫≠p**: O(n)
- **T√¨m ki·∫øm**: O(n)
- **Ch√®n ·ªü ƒë·∫ßu**: O(1)
- **Ch√®n ·ªü cu·ªëi**: O(1) n·∫øu c√≥ tail pointer, O(n) n·∫øu kh√¥ng
- **X√≥a ·ªü ƒë·∫ßu**: O(1)
- **X√≥a ·ªü cu·ªëi**: O(n) cho singly, O(1) cho doubly n·∫øu c√≥ tail pointer

### ƒê·ªô ph·ª©c t·∫°p kh√¥ng gian: O(n)

## 2.2. Tri·ªÉn khai Singly Linked List

```python
class Node:
    """Node ƒë∆°n gi·∫£n cho Singly Linked List"""
    def __init__(self, data):
        self.data = data
        self.next = None
    
    def __repr__(self):
        return f"Node({self.data})"

class SinglyLinkedList:
    """Tri·ªÉn khai Singly Linked List"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        """Ki·ªÉm tra danh s√°ch c√≥ r·ªóng kh√¥ng"""
        return self.head is None
    
    def prepend(self, data):
        """Th√™m ph·∫ßn t·ª≠ v√†o ƒë·∫ßu - O(1)"""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        self.size += 1
    
    def append(self, data):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(n)"""
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node
        self.size += 1
    
    def insert_after(self, prev_data, data):
        """Ch√®n ph·∫ßn t·ª≠ sau node c√≥ gi√° tr·ªã prev_data - O(n)"""
        new_node = Node(data)
        current = self.head
        
        while current is not None:
            if current.data == prev_data:
                new_node.next = current.next
                current.next = new_node
                self.size += 1
                return
            current = current.next
        
        raise ValueError(f"Kh√¥ng t√¨m th·∫•y node c√≥ gi√° tr·ªã {prev_data}")
    
    def delete(self, data):
        """X√≥a node c√≥ gi√° tr·ªã data - O(n)"""
        if self.head is None:
            raise ValueError("Danh s√°ch r·ªóng")
        
        # X√≥a node ƒë·∫ßu
        if self.head.data == data:
            self.head = self.head.next
            self.size -= 1
            return
        
        # T√¨m v√† x√≥a node
        current = self.head
        while current.next is not None:
            if current.next.data == data:
                current.next = current.next.next
                self.size -= 1
                return
            current = current.next
        
        raise ValueError(f"Kh√¥ng t√¨m th·∫•y node c√≥ gi√° tr·ªã {data}")
    
    def search(self, data):
        """T√¨m ki·∫øm node c√≥ gi√° tr·ªã data - O(n)"""
        current = self.head
        index = 0
        while current is not None:
            if current.data == data:
                return index
            current = current.next
            index += 1
        return -1
    
    def get(self, index):
        """L·∫•y gi√° tr·ªã t·∫°i index - O(n)"""
        if index < 0 or index >= self.size:
            raise IndexError(f"Index {index} n·∫±m ngo√†i ph·∫°m vi")
        
        current = self.head
        for _ in range(index):
            current = current.next
        return current.data
    
    def reverse(self):
        """ƒê·∫£o ng∆∞·ª£c linked list - O(n)"""
        prev = None
        current = self.head
        
        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        
        self.head = prev
    
    def to_list(self):
        """Chuy·ªÉn linked list th√†nh Python list"""
        result = []
        current = self.head
        while current is not None:
            result.append(current.data)
            current = current.next
        return result
    
    def __str__(self):
        """Hi·ªÉn th·ªã linked list"""
        if self.head is None:
            return "SinglyLinkedList([])"
        
        nodes = []
        current = self.head
        while current is not None:
            nodes.append(str(current.data))
            current = current.next
        
        return f"SinglyLinkedList([{' -> '.join(nodes)}])"
    
    def __repr__(self):
        return self.__str__()

# S·ª≠ d·ª•ng Singly Linked List
print("=== Singly Linked List ===\n")
sll = SinglyLinkedList()

# Th√™m ph·∫ßn t·ª≠
print("Th√™m ph·∫ßn t·ª≠ v√†o ƒë·∫ßu:")
for i in [3, 2, 1]:
    sll.prepend(i)
    print(f"  {sll}")

print("\nTh√™m ph·∫ßn t·ª≠ v√†o cu·ªëi:")
for i in [4, 5, 6]:
    sll.append(i)
    print(f"  {sll}")

# Ch√®n ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi ch√®n: {sll}")
sll.insert_after(3, 3.5)
print(f"Sau khi ch√®n 3.5 sau 3: {sll}")

# T√¨m ki·∫øm
index = sll.search(5)
print(f"\nT√¨m ki·∫øm 5: {'T√¨m th·∫•y t·∫°i index ' + str(index) if index != -1 else 'Kh√¥ng t√¨m th·∫•y'}")

# L·∫•y gi√° tr·ªã
print(f"Gi√° tr·ªã t·∫°i index 2: {sll.get(2)}")

# X√≥a ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi x√≥a: {sll}")
sll.delete(3.5)
print(f"Sau khi x√≥a 3.5: {sll}")

# ƒê·∫£o ng∆∞·ª£c
print(f"\nTr∆∞·ªõc khi ƒë·∫£o ng∆∞·ª£c: {sll}")
sll.reverse()
print(f"Sau khi ƒë·∫£o ng∆∞·ª£c: {sll}")
```

```python
class DoublyNode:
    """Node cho Doubly Linked List"""
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
    
    def __repr__(self):
        return f"DoublyNode({self.data})"

class DoublyLinkedList:
    """Tri·ªÉn khai Doubly Linked List"""
    
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def is_empty(self):
        return self.head is None
    
    def prepend(self, data):
        """Th√™m ph·∫ßn t·ª≠ v√†o ƒë·∫ßu - O(1)"""
        new_node = DoublyNode(data)
        if self.head is None:
            self.head = self.tail = new_node
        else:
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
        self.size += 1
    
    def append(self, data):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(1)"""
        new_node = DoublyNode(data)
        if self.tail is None:
            self.head = self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1
    
    def delete(self, data):
        """X√≥a node c√≥ gi√° tr·ªã data - O(n)"""
        current = self.head
        
        while current is not None:
            if current.data == data:
                # X√≥a node ƒë·∫ßu
                if current.prev is None:
                    self.head = current.next
                    if self.head:
                        self.head.prev = None
                    else:
                        self.tail = None
                # X√≥a node cu·ªëi
                elif current.next is None:
                    self.tail = current.prev
                    self.tail.next = None
                # X√≥a node ·ªü gi·ªØa
                else:
                    current.prev.next = current.next
                    current.next.prev = current.prev
                
                self.size -= 1
                return
            current = current.next
        
        raise ValueError(f"Kh√¥ng t√¨m th·∫•y node c√≥ gi√° tr·ªã {data}")
    
    def traverse_forward(self):
        """Duy·ªát t·ª´ ƒë·∫ßu ƒë·∫øn cu·ªëi"""
        result = []
        current = self.head
        while current is not None:
            result.append(current.data)
            current = current.next
        return result
    
    def traverse_backward(self):
        """Duy·ªát t·ª´ cu·ªëi v·ªÅ ƒë·∫ßu"""
        result = []
        current = self.tail
        while current is not None:
            result.append(current.data)
            current = current.prev
        return result
    
    def __str__(self):
        """Hi·ªÉn th·ªã linked list"""
        if self.head is None:
            return "DoublyLinkedList([])"
        
        nodes = []
        current = self.head
        while current is not None:
            nodes.append(str(current.data))
            current = current.next
        
        return f"DoublyLinkedList([{' <-> '.join(nodes)}])"
    
    def __repr__(self):
        return self.__str__()

# S·ª≠ d·ª•ng Doubly Linked List
print("=== Doubly Linked List ===\n")
dll = DoublyLinkedList()

# Th√™m ph·∫ßn t·ª≠
print("Th√™m ph·∫ßn t·ª≠ v√†o ƒë·∫ßu:")
for i in [3, 2, 1]:
    dll.prepend(i)
    print(f"  {dll}")

print("\nTh√™m ph·∫ßn t·ª≠ v√†o cu·ªëi:")
for i in [4, 5]:
    dll.append(i)
    print(f"  {dll}")

# Duy·ªát
print(f"\nDuy·ªát t·ª´ ƒë·∫ßu ƒë·∫øn cu·ªëi: {dll.traverse_forward()}")
print(f"Duy·ªát t·ª´ cu·ªëi v·ªÅ ƒë·∫ßu: {dll.traverse_backward()}")

# X√≥a ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi x√≥a: {dll}")
dll.delete(3)
print(f"Sau khi x√≥a 3: {dll}")
dll.delete(5)
print(f"Sau khi x√≥a 5: {dll}")
```

```python
class CircularLinkedList:
    """Tri·ªÉn khai Circular Linked List"""
    
    def __init__(self):
        self.head = None
        self.size = 0
    
    def __len__(self):
        return self.size
    
    def append(self, data):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(n)"""
        new_node = Node(data)
        
        if self.head is None:
            self.head = new_node
            new_node.next = self.head  # Tr·ªè v·ªÅ ch√≠nh n√≥
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head
        
        self.size += 1
    
    def delete(self, data):
        """X√≥a node c√≥ gi√° tr·ªã data - O(n)"""
        if self.head is None:
            raise ValueError("Danh s√°ch r·ªóng")
        
        current = self.head
        prev = None
        
        # T√¨m node c·∫ßn x√≥a
        while True:
            if current.data == data:
                # X√≥a node ƒë·∫ßu
                if current == self.head:
                    if self.size == 1:
                        self.head = None
                    else:
                        # T√¨m node cu·ªëi v√† c·∫≠p nh·∫≠t
                        last = self.head
                        while last.next != self.head:
                            last = last.next
                        self.head = current.next
                        last.next = self.head
                else:
                    prev.next = current.next
                
                self.size -= 1
                return
            
            prev = current
            current = current.next
            
            if current == self.head:
                break
        
        raise ValueError(f"Kh√¥ng t√¨m th·∫•y node c√≥ gi√° tr·ªã {data}")
    
    def to_list(self):
        """Chuy·ªÉn th√†nh Python list"""
        if self.head is None:
            return []
        
        result = []
        current = self.head
        while True:
            result.append(current.data)
            current = current.next
            if current == self.head:
                break
        return result
    
    def __str__(self):
        """Hi·ªÉn th·ªã circular linked list"""
        if self.head is None:
            return "CircularLinkedList([])"
        
        nodes = []
        current = self.head
        while True:
            nodes.append(str(current.data))
            current = current.next
            if current == self.head:
                break
        
        return f"CircularLinkedList([{' -> '.join(nodes)} -> ...])"
    
    def __repr__(self):
        return self.__str__()

# S·ª≠ d·ª•ng Circular Linked List
print("=== Circular Linked List ===\n")
cll = CircularLinkedList()

# Th√™m ph·∫ßn t·ª≠
for i in [1, 2, 3, 4, 5]:
    cll.append(i)
    print(f"  {cll}")

# X√≥a ph·∫ßn t·ª≠
print(f"\nTr∆∞·ªõc khi x√≥a: {cll}")
cll.delete(3)
print(f"Sau khi x√≥a 3: {cll}")

print(f"\nDanh s√°ch: {cll.to_list()}")
```

## 4.1. Stack (NgƒÉn x·∫øp) - S·ª≠ d·ª•ng Array ho·∫∑c Linked List

```python
class StackArray:
    """Stack tri·ªÉn khai b·∫±ng Array"""
    def __init__(self):
        self.items = []
    
    def push(self, item):
        """Th√™m ph·∫ßn t·ª≠ v√†o ƒë·ªânh - O(1)"""
        self.items.append(item)
    
    def pop(self):
        """L·∫•y v√† x√≥a ph·∫ßn t·ª≠ ·ªü ƒë·ªânh - O(1)"""
        if self.is_empty():
            raise IndexError("Stack r·ªóng")
        return self.items.pop()
    
    def peek(self):
        """Xem ph·∫ßn t·ª≠ ·ªü ƒë·ªânh - O(1)"""
        if self.is_empty():
            raise IndexError("Stack r·ªóng")
        return self.items[-1]
    
    def is_empty(self):
        return len(self.items) == 0
    
    def size(self):
        return len(self.items)
    
    def __str__(self):
        return f"Stack({self.items})"

class StackLinkedList:
    """Stack tri·ªÉn khai b·∫±ng Linked List"""
    def __init__(self):
        self.head = None
        self.size = 0
    
    def push(self, item):
        """Th√™m ph·∫ßn t·ª≠ v√†o ƒë·ªânh - O(1)"""
        new_node = Node(item)
        new_node.next = self.head
        self.head = new_node
        self.size += 1
    
    def pop(self):
        """L·∫•y v√† x√≥a ph·∫ßn t·ª≠ ·ªü ƒë·ªânh - O(1)"""
        if self.is_empty():
            raise IndexError("Stack r·ªóng")
        data = self.head.data
        self.head = self.head.next
        self.size -= 1
        return data
    
    def peek(self):
        """Xem ph·∫ßn t·ª≠ ·ªü ƒë·ªânh - O(1)"""
        if self.is_empty():
            raise IndexError("Stack r·ªóng")
        return self.head.data
    
    def is_empty(self):
        return self.head is None
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        items = []
        current = self.head
        while current:
            items.append(str(current.data))
            current = current.next
        return f"Stack([{' -> '.join(reversed(items))}])"

# ·ª®ng d·ª•ng: Ki·ªÉm tra d·∫•u ngo·∫∑c ƒë√∫ng
def check_parentheses(expression):
    """Ki·ªÉm tra bi·ªÉu th·ª©c c√≥ d·∫•u ngo·∫∑c ƒë√∫ng kh√¥ng"""
    stack = StackArray()
    pairs = {')': '(', '}': '{', ']': '['}
    
    for char in expression:
        if char in pairs.values():  # M·ªü ngo·∫∑c
            stack.push(char)
        elif char in pairs.keys():  # ƒê√≥ng ngo·∫∑c
            if stack.is_empty() or stack.pop() != pairs[char]:
                return False
    
    return stack.is_empty()

# V√≠ d·ª•
expressions = [
    "()",
    "()[]{}",
    "([{}])",
    "([)]",
    "(((",
    "())"
]

print("=== Ki·ªÉm tra d·∫•u ngo·∫∑c ===\n")
for expr in expressions:
    result = check_parentheses(expr)
    print(f"{expr:15} -> {'‚úÖ H·ª£p l·ªá' if result else '‚ùå Kh√¥ng h·ª£p l·ªá'}")

# So s√°nh Stack Array vs Linked List
print("\n=== So s√°nh Stack Array vs Linked List ===\n")
stack_arr = StackArray()
stack_ll = StackLinkedList()

for i in range(5):
    stack_arr.push(i)
    stack_ll.push(i)

print(f"Stack Array: {stack_arr}")
print(f"Stack Linked List: {stack_ll}")

print(f"\nPop t·ª´ Stack Array: {stack_arr.pop()}")
print(f"Pop t·ª´ Stack Linked List: {stack_ll.pop()}")
print(f"\nSau khi pop:")
print(f"Stack Array: {stack_arr}")
print(f"Stack Linked List: {stack_ll}")
```

## 4.2. Queue (H√†ng ƒë·ª£i) - S·ª≠ d·ª•ng Linked List

```python
class QueueLinkedList:
    """Queue tri·ªÉn khai b·∫±ng Linked List"""
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0
    
    def enqueue(self, item):
        """Th√™m ph·∫ßn t·ª≠ v√†o cu·ªëi - O(1)"""
        new_node = Node(item)
        if self.tail is None:
            self.head = self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1
    
    def dequeue(self):
        """L·∫•y v√† x√≥a ph·∫ßn t·ª≠ ·ªü ƒë·∫ßu - O(1)"""
        if self.is_empty():
            raise IndexError("Queue r·ªóng")
        data = self.head.data
        self.head = self.head.next
        if self.head is None:
            self.tail = None
        self.size -= 1
        return data
    
    def front(self):
        """Xem ph·∫ßn t·ª≠ ·ªü ƒë·∫ßu - O(1)"""
        if self.is_empty():
            raise IndexError("Queue r·ªóng")
        return self.head.data
    
    def is_empty(self):
        return self.head is None
    
    def __len__(self):
        return self.size
    
    def __str__(self):
        items = []
        current = self.head
        while current:
            items.append(str(current.data))
            current = current.next
        return f"Queue([{' -> '.join(items)}])"

# ·ª®ng d·ª•ng: M√¥ ph·ªèng h√†ng ƒë·ª£i
print("=== Queue (H√†ng ƒë·ª£i) ===\n")
queue = QueueLinkedList()

# Th√™m v√†o h√†ng ƒë·ª£i
print("Th√™m v√†o h√†ng ƒë·ª£i:")
for person in ["An", "B√¨nh", "Chi", "D≈©ng"]:
    queue.enqueue(person)
    print(f"  {queue}")

# Ph·ª•c v·ª• t·ª´ h√†ng ƒë·ª£i
print("\nPh·ª•c v·ª• t·ª´ h√†ng ƒë·ª£i:")
while not queue.is_empty():
    served = queue.dequeue()
    print(f"  ƒêang ph·ª•c v·ª•: {served}")
    print(f"  H√†ng ƒë·ª£i c√≤n l·∫°i: {queue}")
```

```python
def visualize_linked_list(ll: SinglyLinkedList, title="Singly Linked List"):
    """V·∫Ω bi·ªÉu ƒë·ªì minh h·ªça c·∫•u tr√∫c Linked List"""
    if ll.head is None:
        print("Linked List r·ªóng")
        return
    
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.set_xlim(-1, len(ll) * 2)
    ax.set_ylim(-1, 2)
    ax.axis('off')
    ax.set_title(title, fontsize=14, fontweight='bold', pad=20)
    
    current = ll.head
    x_pos = 0
    
    while current is not None:
        # V·∫Ω node (h√¨nh ch·ªØ nh·∫≠t)
        rect = plt.Rectangle((x_pos, 0.5), 1.5, 1, 
                           linewidth=2, edgecolor='black', 
                           facecolor='lightblue', zorder=2)
        ax.add_patch(rect)
        
        # V·∫Ω d·ªØ li·ªáu
        ax.text(x_pos + 0.75, 1, str(current.data), 
               ha='center', va='center', fontsize=12, fontweight='bold', zorder=3)
        
        # V·∫Ω m≈©i t√™n (n·∫øu kh√¥ng ph·∫£i node cu·ªëi)
        if current.next is not None:
            arrow = plt.Arrow(x_pos + 1.5, 1, 0.3, 0, 
                            width=0.1, color='red', zorder=1)
            ax.add_patch(arrow)
            # V·∫Ω "next"
            ax.text(x_pos + 1.65, 1.3, 'next', 
                   ha='center', va='center', fontsize=9, color='red', zorder=3)
        
        x_pos += 2
    
    # V·∫Ω "head" pointer
    if ll.head is not None:
        ax.annotate('head', xy=(0, 0.5), xytext=(-0.5, -0.5),
                   arrowprops=dict(arrowstyle='->', color='green', lw=2),
                   fontsize=10, fontweight='bold', color='green')
    
    plt.tight_layout()
    plt.show()

# T·∫°o v√† visualize linked list
print("=== Visualization Linked List ===\n")
sll = SinglyLinkedList()
for i in [10, 20, 30, 40, 50]:
    sll.append(i)

print(f"Linked List: {sll}")
visualize_linked_list(sll, "Singly Linked List - V√≠ d·ª•")
```

## 5.4. Ph√¢n t√≠ch b·ªô nh·ªõ: Array vs Linked List

```python
def get_size(obj):
    """T√≠nh k√≠ch th∆∞·ªõc b·ªô nh·ªõ c·ªßa object"""
    return sys.getsizeof(obj)

# So s√°nh b·ªô nh·ªõ
sizes = [10, 100, 1000, 10000]
memory_comparison = {
    "Array": [],
    "Linked List": []
}

print("=== Ph√¢n t√≠ch b·ªô nh·ªõ ===\n")
print(f"{'K√≠ch th∆∞·ªõc':<12} {'Array (bytes)':<18} {'Linked List (bytes)':<22} {'T·ª∑ l·ªá':<10}")
print("-" * 62)

for size in sizes:
    # Array
    arr = list(range(size))
    arr_memory = get_size(arr) + sum(get_size(item) for item in arr)
    
    # Linked List
    sll = SinglyLinkedList()
    for i in range(size):
        sll.append(i)
    
    # T√≠nh b·ªô nh·ªõ linked list (∆∞·ªõc t√≠nh)
    # M·ªói node: data (24 bytes) + next pointer (8 bytes) + overhead
    node_size = get_size(Node(0))  # Kho·∫£ng 56 bytes m·ªói node
    ll_memory = node_size * size + get_size(sll)
    
    memory_comparison["Array"].append(arr_memory)
    memory_comparison["Linked List"].append(ll_memory)
    
    ratio = ll_memory / arr_memory if arr_memory > 0 else 0
    print(f"{size:<12} {arr_memory:<18} {ll_memory:<22} {ratio:<10.2f}x")

# V·∫Ω bi·ªÉu ƒë·ªì
plt.figure(figsize=(10, 6))
plt.plot(sizes, memory_comparison["Array"], 'o-', label='Array (Python List)', linewidth=2, markersize=8)
plt.plot(sizes, memory_comparison["Linked List"], 's-', label='Linked List', linewidth=2, markersize=8)
plt.xlabel('S·ªë ph·∫ßn t·ª≠', fontsize=12)
plt.ylabel('B·ªô nh·ªõ (bytes)', fontsize=12)
plt.title('So s√°nh b·ªô nh·ªõ: Array vs Linked List', fontsize=14, fontweight='bold')
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.yscale('log')
plt.tight_layout()
plt.show()

print("\nüí° Nh·∫≠n x√©t: Linked List t·ªën b·ªô nh·ªõ h∆°n Array do c·∫ßn l∆∞u tr·ªØ con tr·ªè (pointer)")
```
