## 1. Giá»›i thiá»‡u vá» File Handling

**File Handling** lÃ  ká»¹ thuáº­t Ä‘á»c vÃ  ghi dá»¯ liá»‡u tá»«/vÃ o file trÃªn á»• Ä‘Ä©a.

### Táº¡i sao cáº§n xá»­ lÃ½ file?
- LÆ°u trá»¯ dá»¯ liá»‡u vÄ©nh viá»…n (persistent storage)
- Äá»c dá»¯ liá»‡u tá»« file cáº¥u hÃ¬nh
- Ghi log hoáº¡t Ä‘á»™ng
- Xá»­ lÃ½ dá»¯ liá»‡u lá»›n (CSV, JSON, XML)
- Import/Export dá»¯ liá»‡u

### CÃ¡c loáº¡i file:

| Loáº¡i | MÃ´ táº£ | VÃ­ dá»¥ |
|------|-------|-------|
| **Text file** | Chá»©a vÄƒn báº£n, cÃ³ thá»ƒ Ä‘á»c Ä‘Æ°á»£c | .txt, .csv, .json, .py |
| **Binary file** | Chá»©a dá»¯ liá»‡u nhá»‹ phÃ¢n | .jpg, .mp3, .pdf, .exe |

### Quy trÃ¬nh xá»­ lÃ½ file:
```
1. Má»Ÿ file (open)
2. Äá»c/Ghi (read/write)
3. ÄÃ³ng file (close)
```

````python
# ========================================
# Váº¤N Äá»€ KHI QUÃŠN ÄÃ“NG FILE
# ========================================

# Náº¿u cÃ³ lá»—i xáº£y ra, file cÃ³ thá»ƒ khÃ´ng Ä‘Æ°á»£c Ä‘Ã³ng
try:
    file = open('samples/sample.txt', 'r', encoding='utf-8')
    content = file.read()
    # Giáº£ sá»­ cÃ³ lá»—i xáº£y ra á»Ÿ Ä‘Ã¢y
    result = 1 / 0  # ZeroDivisionError
    file.close()  # DÃ²ng nÃ y khÃ´ng Ä‘Æ°á»£c thá»±c thi!
except ZeroDivisionError:
    print("âŒ Lá»—i xáº£y ra!")
    print(f"File Ä‘Ã£ Ä‘Ã³ng chÆ°a? {file.closed}")
finally:
    if not file.closed:
        file.close()
        print("âœ… ÄÃ£ Ä‘Ã³ng file trong finally")
````

````python
# ========================================
# SEEK() VÃ€ TELL() - Di chuyá»ƒn con trá»
# ========================================

with open('samples/multiline.txt', 'r', encoding='utf-8') as f:
    print("=== seek() vÃ  tell() ===")
    
    # Vá»‹ trÃ­ ban Ä‘áº§u
    print(f"Vá»‹ trÃ­ ban Ä‘áº§u: {f.tell()}")
    
    # Äá»c 10 kÃ½ tá»±
    print(f"Äá»c 10 kÃ½ tá»±: '{f.read(10)}'")
    print(f"Vá»‹ trÃ­ sau Ä‘á»c: {f.tell()}")
    
    # Di chuyá»ƒn vá» Ä‘áº§u file
    f.seek(0)
    print(f"\nSau seek(0): {f.tell()}")
    print(f"Äá»c láº¡i: '{f.read(10)}'")
    
    # Di chuyá»ƒn Ä‘áº¿n vá»‹ trÃ­ cá»¥ thá»ƒ
    f.seek(50)
    print(f"\nSau seek(50): '{f.read(20)}'")
````

````python
# ========================================
# Äá»ŒC FILE Lá»šN Tá»ªNG CHUNK
# ========================================

def read_in_chunks(file_path, chunk_size=100):
    """Äá»c file lá»›n tá»«ng chunk Ä‘á»ƒ tiáº¿t kiá»‡m bá»™ nhá»›"""
    with open(file_path, 'r', encoding='utf-8') as f:
        chunk_num = 0
        while True:
            chunk = f.read(chunk_size)
            if not chunk:  # Háº¿t file
                break
            chunk_num += 1
            yield chunk_num, chunk

print("Äá»c file theo chunks (100 kÃ½ tá»±):")
print("=" * 50)
for num, chunk in read_in_chunks('samples/multiline.txt', 100):
    print(f"\nChunk {num}:")
    print(repr(chunk))
````

````python
# ========================================
# COPY FILE
# ========================================

def copy_file(source, destination):
    """Copy file tá»« source sang destination"""
    with open(source, 'r', encoding='utf-8') as src, \
         open(destination, 'w', encoding='utf-8') as dst:
        # Äá»c vÃ  ghi tá»«ng dÃ²ng (hiá»‡u quáº£ vá»›i file lá»›n)
        for line in src:
            dst.write(line)
    return True

# Test
copy_file('samples/sample.txt', 'samples/sample_copy.txt')
print("âœ… ÄÃ£ copy file!")

# Kiá»ƒm tra
with open('samples/sample_copy.txt', 'r', encoding='utf-8') as f:
    print("\nNá»™i dung file copy:")
    print(f.read())
````

````python
# ========================================
# á»¨NG Dá»¤NG 1: Äáº¿m tá»« trong file
# ========================================

def count_words(file_path):
    """Äáº¿m sá»‘ tá»«, dÃ²ng, kÃ½ tá»± trong file"""
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()
        
    lines = content.split('\n')
    words = content.split()
    chars = len(content)
    
    return {
        'lines': len(lines),
        'words': len(words),
        'characters': chars
    }

# Test
stats = count_words('samples/multiline.txt')
print("Thá»‘ng kÃª file:")
for key, value in stats.items():
    print(f"  {key}: {value}")
````

````python
# ========================================
# á»¨NG Dá»¤NG 3: TÃ¬m vÃ  thay tháº¿ trong file
# ========================================

def find_replace(file_path, find_text, replace_text):
    """TÃ¬m vÃ  thay tháº¿ text trong file"""
    # Äá»c
    with open(file_path, 'r', encoding='utf-8') as f:
        content = f.read()
    
    # Äáº¿m sá»‘ láº§n xuáº¥t hiá»‡n
    count = content.count(find_text)
    
    # Thay tháº¿
    new_content = content.replace(find_text, replace_text)
    
    # Ghi láº¡i
    with open(file_path, 'w', encoding='utf-8') as f:
        f.write(new_content)
    
    return count

# Táº¡o file test
with open('samples/replace_test.txt', 'w', encoding='utf-8') as f:
    f.write("Hello World! Hello Python! Hello Everyone!")

print("TrÆ°á»›c:")
with open('samples/replace_test.txt', 'r', encoding='utf-8') as f:
    print(f.read())

count = find_replace('samples/replace_test.txt', 'Hello', 'Hi')
print(f"\nÄÃ£ thay tháº¿ {count} láº§n\n")

print("Sau:")
with open('samples/replace_test.txt', 'r', encoding='utf-8') as f:
    print(f.read())
````

````python
# ========================================
# á»¨NG Dá»¤NG 4: Äá»c file cáº¥u hÃ¬nh
# ========================================

def load_config(config_file):
    """Äá»c file config dáº¡ng key=value"""
    config = {}
    
    with open(config_file, 'r', encoding='utf-8') as f:
        for line in f:
            line = line.strip()
            # Bá» qua comment vÃ  dÃ²ng trá»‘ng
            if not line or line.startswith('#'):
                continue
            # Parse key=value
            if '=' in line:
                key, value = line.split('=', 1)
                config[key.strip()] = value.strip()
    
    return config

# Táº¡o file config
config_content = """# Database configuration
db_host=localhost
db_port=5432
db_name=myapp

# App settings
debug=true
log_level=INFO
"""

with open('samples/app.config', 'w', encoding='utf-8') as f:
    f.write(config_content)

# Äá»c config
config = load_config('samples/app.config')
print("Config:")
for key, value in config.items():
    print(f"  {key} = {value}")
````

````python
# ========================================
# á»¨NG Dá»¤NG 5: Todo List Manager
# ========================================
import json
from datetime import datetime

class TodoManager:
    def __init__(self, file_path='samples/todos.json'):
        self.file_path = file_path
        self.todos = self._load()
    
    def _load(self):
        try:
            with open(self.file_path, 'r', encoding='utf-8') as f:
                return json.load(f)
        except FileNotFoundError:
            return []
    
    def _save(self):
        with open(self.file_path, 'w', encoding='utf-8') as f:
            json.dump(self.todos, f, indent=2, ensure_ascii=False)
    
    def add(self, task):
        todo = {
            "id": len(self.todos) + 1,
            "task": task,
            "done": False,
            "created": datetime.now().isoformat()
        }
        self.todos.append(todo)
        self._save()
        return todo
    
    def complete(self, todo_id):
        for todo in self.todos:
            if todo["id"] == todo_id:
                todo["done"] = True
                self._save()
                return True
        return False
    
    def list_all(self):
        return self.todos

# Demo
manager = TodoManager()
manager.add("Há»c Python")
manager.add("LÃ m bÃ i táº­p")
manager.add("Äá»c sÃ¡ch")
manager.complete(1)

print("Todo List:")
for todo in manager.list_all():
    status = "âœ…" if todo["done"] else "â¬œ"
    print(f"  {status} [{todo['id']}] {todo['task']}")
````

````python
# ========================================
# BÃ€I Táº¬P 1: Äá»c file vÃ  Ä‘áº¿m dÃ²ng chá»©a tá»« khÃ³a
# ========================================

def count_lines_with_keyword(file_path, keyword):
    """Äáº¿m sá»‘ dÃ²ng chá»©a tá»« khÃ³a"""
    count = 0
    matching_lines = []
    
    with open(file_path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f, 1):
            if keyword.lower() in line.lower():
                count += 1
                matching_lines.append((i, line.strip()))
    
    return count, matching_lines

# Test
count, lines = count_lines_with_keyword('samples/multiline.txt', 'Python')
print(f"TÃ¬m tháº¥y {count} dÃ²ng chá»©a 'Python':")
for line_num, content in lines:
    print(f"  DÃ²ng {line_num}: {content}")
````

````python
# ========================================
# BÃ€I Táº¬P 2: Merge nhiá»u file
# ========================================
from pathlib import Path

def merge_files(input_files, output_file):
    """Gá»™p nhiá»u file thÃ nh 1 file"""
    with open(output_file, 'w', encoding='utf-8') as out_f:
        for file_path in input_files:
            out_f.write(f"\n=== {file_path} ===\n")
            with open(file_path, 'r', encoding='utf-8') as in_f:
                out_f.write(in_f.read())
            out_f.write("\n")

# Test
files = list(Path('samples').glob('*.txt'))[:3]
merge_files(files, 'samples/merged.txt')

print("ÄÃ£ merge cÃ¡c file:")
for f in files:
    print(f"  - {f}")

print(f"\nKáº¿t quáº£: samples/merged.txt")
````

````python
# ========================================
# BÃ€I Táº¬P 3: Backup file vá»›i timestamp
# ========================================
from datetime import datetime
from pathlib import Path
import shutil

def backup_file(file_path):
    """Táº¡o báº£n sao backup vá»›i timestamp"""
    path = Path(file_path)
    
    if not path.exists():
        return None
    
    timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
    backup_name = f"{path.stem}_backup_{timestamp}{path.suffix}"
    backup_path = path.parent / backup_name
    
    shutil.copy2(file_path, backup_path)
    return backup_path

# Test
backup = backup_file('samples/sample.txt')
print(f"âœ… ÄÃ£ backup: {backup}")
````

````python
# ========================================
# BÃ€I Táº¬P 4: PhÃ¢n tÃ­ch log file
# ========================================

def analyze_log(log_file):
    """PhÃ¢n tÃ­ch file log"""
    stats = {"INFO": 0, "WARNING": 0, "ERROR": 0}
    errors = []
    
    with open(log_file, 'r', encoding='utf-8') as f:
        for line in f:
            for level in stats.keys():
                if f"[{level}]" in line:
                    stats[level] += 1
                    if level == "ERROR":
                        errors.append(line.strip())
    
    return stats, errors

# Test
stats, errors = analyze_log('samples/app.log')
print("Thá»‘ng kÃª log:")
for level, count in stats.items():
    print(f"  {level}: {count}")

if errors:
    print("\nCÃ¡c lá»—i:")
    for err in errors:
        print(f"  âŒ {err}")
````
