## 1. Gi·ªõi thi·ªáu v·ªÅ Exception

### Exception l√† g√¨?

**Exception (Ngo·∫°i l·ªá)** l√† l·ªói x·∫£y ra trong qu√° tr√¨nh ch·∫°y ch∆∞∆°ng tr√¨nh (runtime), kh√°c v·ªõi **Syntax Error** (l·ªói c√∫ ph√°p) x·∫£y ra khi bi√™n d·ªãch.

### T·∫°i sao c·∫ßn x·ª≠ l√Ω Exception?

- **NgƒÉn ch∆∞∆°ng tr√¨nh crash** ƒë·ªôt ng·ªôt
- **X·ª≠ l√Ω l·ªói gracefully** (thanh l·ªãch)
- **Th√¥ng b√°o l·ªói r√µ r√†ng** cho user
- **D·ªçn d·∫πp t√†i nguy√™n** (ƒë√≥ng file, database connection)
- **Debug d·ªÖ d√†ng h∆°n**

````python
# ========================================
# TRY-EXCEPT C∆† B·∫¢N
# ========================================

print("=== try-except c∆° b·∫£n ===")

# V√≠ d·ª• 1: Chia cho 0
try:
    result = 10 / 0
    print(f"K·∫øt qu·∫£: {result}")
except ZeroDivisionError:
    print("‚ùå L·ªói: Kh√¥ng th·ªÉ chia cho 0!")

print("\n‚úÖ Ch∆∞∆°ng tr√¨nh v·∫´n ti·∫øp t·ª•c ch·∫°y!")
````

````python
# ========================================
# B·∫ÆT T·∫§T C·∫¢ EXCEPTION (Kh√¥ng khuy·∫øn kh√≠ch)
# ========================================

print("=== B·∫Øt t·∫•t c·∫£ Exception ===")

# C√°ch 1: except kh√¥ng c√≥ lo·∫°i c·ª• th·ªÉ
try:
    x = 1 / 0
except:
    print("C√≥ l·ªói x·∫£y ra! (kh√¥ng bi·∫øt lo·∫°i g√¨)")

# C√°ch 2: except Exception (t·ªët h∆°n)
try:
    x = int("abc")
except Exception as e:
    print(f"L·ªói: {type(e).__name__}: {e}")

print("\n‚ö†Ô∏è N√™n b·∫Øt exception C·ª§ TH·ªÇ thay v√¨ b·∫Øt t·∫•t c·∫£!")
````

````python
# ========================================
# V√ç D·ª§ TH·ª∞C T·∫æ: Nh·∫≠p s·ªë t·ª´ user
# ========================================

def get_integer(prompt):
    """Nh·∫≠p s·ªë nguy√™n v·ªõi x·ª≠ l√Ω l·ªói"""
    while True:
        try:
            value = int(input(prompt))
            return value
        except ValueError:
            print("‚ùå Vui l√≤ng nh·∫≠p m·ªôt s·ªë nguy√™n!")

# Gi·∫£ l·∫≠p input
test_inputs = ["abc", "12.5", "42"]
print("Gi·∫£ l·∫≠p nh·∫≠p s·ªë:")
for inp in test_inputs:
    try:
        result = int(inp)
        print(f"  '{inp}' ‚Üí {result} ‚úÖ")
    except ValueError:
        print(f"  '{inp}' ‚Üí Kh√¥ng ph·∫£i s·ªë! ‚ùå")
````

## 3. B·∫Øt nhi·ªÅu Exception

````python
# ========================================
# NHI·ªÄU KH·ªêI EXCEPT RI√äNG BI·ªÜT
# ========================================

def process_data(data, index, divisor):
    """X·ª≠ l√Ω d·ªØ li·ªáu v·ªõi nhi·ªÅu lo·∫°i l·ªói c√≥ th·ªÉ x·∫£y ra"""
    try:
        value = data[index]
        result = value / divisor
        return result
    except IndexError:
        return "‚ùå Index v∆∞·ª£t qu√° ph·∫°m vi list!"
    except ZeroDivisionError:
        return "‚ùå Kh√¥ng th·ªÉ chia cho 0!"
    except TypeError:
        return "‚ùå Ki·ªÉu d·ªØ li·ªáu kh√¥ng h·ª£p l·ªá!"

data = [10, 20, 30]

print("=== Nhi·ªÅu kh·ªëi except ===")
print(f"data[1] / 2 = {process_data(data, 1, 2)}")
print(f"data[10] / 2 = {process_data(data, 10, 2)}")
print(f"data[0] / 0 = {process_data(data, 0, 0)}")
print(f"data[0] / 'a' = {process_data(data, 0, 'a')}")
````

````python
# ========================================
# G·ªòP NHI·ªÄU EXCEPTION TRONG 1 KH·ªêI
# ========================================

def safe_access(data, key):
    """Truy c·∫≠p an to√†n cho c·∫£ list v√† dict"""
    try:
        return data[key]
    except (IndexError, KeyError, TypeError) as e:
        return f"‚ùå L·ªói truy c·∫≠p: {type(e).__name__}"

print("=== G·ªôp nhi·ªÅu exception ===")

my_list = [1, 2, 3]
my_dict = {"a": 1, "b": 2}

print(f"list[0] = {safe_access(my_list, 0)}")
print(f"list[10] = {safe_access(my_list, 10)}")
print(f"dict['a'] = {safe_access(my_dict, 'a')}")
print(f"dict['z'] = {safe_access(my_dict, 'z')}")
print(f"None[0] = {safe_access(None, 0)}")
````

````python
# ========================================
# TH·ª® T·ª∞ EXCEPTION (T·ª´ c·ª• th·ªÉ ƒë·∫øn chung)
# ========================================

print("=== Th·ª© t·ª± exception ===")

def demo_exception_order():
    try:
        x = 1 / 0
    except ArithmeticError:
        print("  B·∫Øt ƒë∆∞·ª£c ArithmeticError (l·ªõp cha)")
    except ZeroDivisionError:
        # Kh·ªëi n√†y KH√îNG BAO GI·ªú ƒë∆∞·ª£c th·ª±c thi!
        print("  B·∫Øt ƒë∆∞·ª£c ZeroDivisionError (l·ªõp con)")

print("Th·ª© t·ª± SAI (cha tr∆∞·ªõc con):")
demo_exception_order()

def demo_correct_order():
    try:
        x = 1 / 0
    except ZeroDivisionError:
        print("  B·∫Øt ƒë∆∞·ª£c ZeroDivisionError (c·ª• th·ªÉ) ‚úÖ")
    except ArithmeticError:
        print("  B·∫Øt ƒë∆∞·ª£c ArithmeticError (chung)")

print("\nTh·ª© t·ª± ƒê√öNG (con tr∆∞·ªõc cha):")
demo_correct_order()

print("\nüí° Lu√¥n ƒë·∫∑t exception C·ª§ TH·ªÇ tr∆∞·ªõc, CHUNG sau!")
````

## 4. Kh·ªëi else v√† finally

### C·∫•u tr√∫c ƒë·∫ßy ƒë·ªß:

```python
try:
    # Code c√≥ th·ªÉ g√¢y l·ªói
except ExceptionType:
    # X·ª≠ l√Ω khi C√ì l·ªói
else:
    # Ch·∫°y khi KH√îNG c√≥ l·ªói
finally:
    # LU√îN ch·∫°y (c√≥ l·ªói hay kh√¥ng)
```

````python
# ========================================
# KH·ªêI ELSE - Ch·∫°y khi kh√¥ng c√≥ l·ªói
# ========================================

def divide_with_else(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("  ‚ùå Kh√¥ng th·ªÉ chia cho 0!")
    else:
        # Ch·ªâ ch·∫°y khi try TH√ÄNH C√îNG
        print(f"  ‚úÖ K·∫øt qu·∫£: {result}")
        return result

print("=== Kh·ªëi else ===")
print("10 / 2:")
divide_with_else(10, 2)

print("\n10 / 0:")
divide_with_else(10, 0)
````

````python
# ========================================
# KH·ªêI FINALLY - Lu√¥n ch·∫°y
# ========================================

def demo_finally(should_fail):
    try:
        print("  ‚Üí B·∫Øt ƒë·∫ßu try")
        if should_fail:
            raise ValueError("L·ªói gi·∫£ l·∫≠p!")
        print("  ‚Üí K·∫øt th√∫c try (th√†nh c√¥ng)")
    except ValueError as e:
        print(f"  ‚Üí B·∫Øt ƒë∆∞·ª£c l·ªói: {e}")
    finally:
        print("  ‚Üí Finally: LU√îN ch·∫°y!")

print("=== Kh·ªëi finally ===")

print("\nKhi KH√îNG c√≥ l·ªói:")
demo_finally(False)

print("\nKhi C√ì l·ªói:")
demo_finally(True)
````

````python
# ========================================
# FINALLY CH·∫†Y NGAY C·∫¢ KHI C√ì RETURN
# ========================================

def demo_finally_with_return():
    try:
        print("  ‚Üí Trong try")
        return "Return t·ª´ try"
    finally:
        print("  ‚Üí Finally v·∫´n ch·∫°y!")

print("=== Finally v·ªõi return ===")
result = demo_finally_with_return()
print(f"  ‚Üí K·∫øt qu·∫£: {result}")
````

````python
# ========================================
# C·∫§U TR√öC ƒê·∫¶Y ƒê·ª¶: try-except-else-finally
# ========================================

def full_exception_handling(filename):
    """V√≠ d·ª• ƒë·ªçc file v·ªõi ƒë·∫ßy ƒë·ªß c·∫•u tr√∫c"""
    file = None
    try:
        print(f"  1. TRY: M·ªü file '{filename}'")
        file = open(filename, 'r')
        content = file.read()
    except FileNotFoundError:
        print(f"  2. EXCEPT: File kh√¥ng t·ªìn t·∫°i!")
        content = None
    else:
        print(f"  2. ELSE: ƒê·ªçc th√†nh c√¥ng ({len(content)} bytes)")
    finally:
        print("  3. FINALLY: D·ªçn d·∫πp t√†i nguy√™n")
        if file:
            file.close()
            print("     ‚Üí ƒê√£ ƒë√≥ng file")
    
    return content

print("=== C·∫•u tr√∫c ƒë·∫ßy ƒë·ªß ===")

# T·∫°o file test
with open('test_file.txt', 'w') as f:
    f.write("Hello, World!")

print("\nFile T·ªíN T·∫†I:")
full_exception_handling('test_file.txt')

print("\nFile KH√îNG t·ªìn t·∫°i:")
full_exception_handling('not_exist.txt')

# X√≥a file test
import os
os.remove('test_file.txt')
````

````python
# ========================================
# FINALLY CHO D·ªåN D·∫∏P T√ÄI NGUY√äN
# ========================================

class DatabaseConnection:
    """Gi·∫£ l·∫≠p database connection"""
    def __init__(self, name):
        self.name = name
        self.connected = False
    
    def connect(self):
        print(f"  üì° K·∫øt n·ªëi database '{self.name}'...")
        self.connected = True
    
    def query(self, sql):
        if not self.connected:
            raise ConnectionError("Ch∆∞a k·∫øt n·ªëi!")
        print(f"  üìã Th·ª±c thi: {sql}")
        if "error" in sql.lower():
            raise Exception("SQL Error!")
        return [{"id": 1, "name": "Test"}]
    
    def close(self):
        print(f"  üîå ƒê√≥ng k·∫øt n·ªëi '{self.name}'")
        self.connected = False

def execute_query(sql):
    db = DatabaseConnection("mydb")
    try:
        db.connect()
        result = db.query(sql)
        return result
    except Exception as e:
        print(f"  ‚ùå L·ªói: {e}")
        return None
    finally:
        # LU√îN ƒë√≥ng connection!
        db.close()

print("=== Finally cho d·ªçn d·∫πp ===")

print("\nQuery th√†nh c√¥ng:")
execute_query("SELECT * FROM users")

print("\nQuery l·ªói:")
execute_query("SELECT error FROM users")
````

````python
# ========================================
# X·ª¨ L√ù C·ª§ TH·ªÇ T·ª™NG LO·∫†I
# ========================================

def safe_convert(value, target_type):
    """Chuy·ªÉn ƒë·ªïi ki·ªÉu d·ªØ li·ªáu an to√†n"""
    try:
        if target_type == int:
            return int(value)
        elif target_type == float:
            return float(value)
        elif target_type == str:
            return str(value)
        elif target_type == list:
            return list(value)
        else:
            return target_type(value)
    except ValueError:
        return f"‚ùå Kh√¥ng th·ªÉ convert '{value}' sang {target_type.__name__}"
    except TypeError:
        return f"‚ùå Ki·ªÉu '{type(value).__name__}' kh√¥ng th·ªÉ convert"

print("=== Safe Convert ===")
print(f"'123' ‚Üí int: {safe_convert('123', int)}")
print(f"'abc' ‚Üí int: {safe_convert('abc', int)}")
print(f"'3.14' ‚Üí float: {safe_convert('3.14', float)}")
print(f"123 ‚Üí list: {safe_convert(123, list)}")
````

````python
# ========================================
# CUSTOM EXCEPTION V·ªöI NHI·ªÄU TH√îNG TIN
# ========================================

class ValidationError(Exception):
    """Exception cho validation v·ªõi th√¥ng tin chi ti·∫øt"""
    
    def __init__(self, field, value, message):
        self.field = field
        self.value = value
        self.message = message
        super().__init__(self.format_message())
    
    def format_message(self):
        return f"Validation failed for '{self.field}': {self.message} (got: {self.value!r})"
    
    def to_dict(self):
        return {
            "field": self.field,
            "value": self.value,
            "message": self.message
        }

def validate_email(email):
    if not isinstance(email, str):
        raise ValidationError("email", email, "ph·∫£i l√† string")
    if "@" not in email:
        raise ValidationError("email", email, "ph·∫£i ch·ª©a @")
    if len(email) < 5:
        raise ValidationError("email", email, "qu√° ng·∫Øn")
    return True

print("=== Custom Exception chi ti·∫øt ===")

test_emails = ["user@email.com", "invalid", 12345, "a@b"]
for email in test_emails:
    try:
        validate_email(email)
        print(f"‚úÖ '{email}' h·ª£p l·ªá")
    except ValidationError as e:
        print(f"‚ùå {e}")
        print(f"   Dict: {e.to_dict()}")
````

````python
# ========================================
# EXCEPTION HIERARCHY CHO ·ª®NG D·ª§NG
# ========================================

# Base exception cho ·ª©ng d·ª•ng
class AppError(Exception):
    """Base exception cho ·ª©ng d·ª•ng"""
    pass

# C√°c exception c·ª• th·ªÉ
class DatabaseError(AppError):
    """L·ªói li√™n quan database"""
    pass

class ConnectionError(DatabaseError):
    """Kh√¥ng th·ªÉ k·∫øt n·ªëi database"""
    pass

class QueryError(DatabaseError):
    """L·ªói khi th·ª±c thi query"""
    pass

class AuthenticationError(AppError):
    """L·ªói x√°c th·ª±c"""
    pass

class AuthorizationError(AppError):
    """L·ªói ph√¢n quy·ªÅn"""
    pass

# S·ª≠ d·ª•ng
def simulate_error(error_type):
    errors = {
        "connection": ConnectionError("Kh√¥ng th·ªÉ k·∫øt n·ªëi MySQL"),
        "query": QueryError("Syntax error in SQL"),
        "auth": AuthenticationError("Invalid credentials"),
        "permission": AuthorizationError("Access denied")
    }
    raise errors.get(error_type, AppError("Unknown error"))

print("=== Exception Hierarchy ===")

for error_type in ["connection", "query", "auth", "permission"]:
    try:
        simulate_error(error_type)
    except DatabaseError as e:
        print(f"üî¥ Database Error: {e}")
    except AuthenticationError as e:
        print(f"üü† Auth Error: {e}")
    except AppError as e:
        print(f"‚ö™ App Error: {e}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 1: Retry Pattern
# ========================================
import random
import time

def retry(max_attempts=3, delay=1):
    """Decorator ƒë·ªÉ retry khi c√≥ l·ªói"""
    def decorator(func):
        def wrapper(*args, **kwargs):
            last_exception = None
            for attempt in range(1, max_attempts + 1):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    last_exception = e
                    print(f"  Attempt {attempt} failed: {e}")
                    if attempt < max_attempts:
                        print(f"  Retrying in {delay}s...")
                        time.sleep(delay)
            raise last_exception
        return wrapper
    return decorator

@retry(max_attempts=3, delay=0.5)
def unstable_api_call():
    """Gi·∫£ l·∫≠p API kh√¥ng ·ªïn ƒë·ªãnh"""
    if random.random() < 0.7:  # 70% th·∫•t b·∫°i
        raise ConnectionError("Server unavailable")
    return {"status": "success", "data": [1, 2, 3]}

print("=== Retry Pattern ===")
try:
    result = unstable_api_call()
    print(f"‚úÖ Success: {result}")
except ConnectionError:
    print("‚ùå All attempts failed!")
````

````python
# ========================================
# ·ª®NG D·ª§NG 2: Input Validation
# ========================================

class Validator:
    """Validator v·ªõi nhi·ªÅu rules"""
    
    @staticmethod
    def required(value, field_name):
        if value is None or value == "":
            raise ValueError(f"{field_name} is required")
        return value
    
    @staticmethod
    def min_length(value, min_len, field_name):
        if len(value) < min_len:
            raise ValueError(f"{field_name} must be at least {min_len} characters")
        return value
    
    @staticmethod
    def email(value, field_name="email"):
        if "@" not in value or "." not in value:
            raise ValueError(f"Invalid {field_name} format")
        return value
    
    @staticmethod
    def in_range(value, min_val, max_val, field_name):
        if not (min_val <= value <= max_val):
            raise ValueError(f"{field_name} must be between {min_val} and {max_val}")
        return value

def register_user(username, email, age):
    """ƒêƒÉng k√Ω user v·ªõi validation"""
    errors = []
    
    try:
        Validator.required(username, "Username")
        Validator.min_length(username, 3, "Username")
    except ValueError as e:
        errors.append(str(e))
    
    try:
        Validator.required(email, "Email")
        Validator.email(email)
    except ValueError as e:
        errors.append(str(e))
    
    try:
        Validator.in_range(age, 18, 100, "Age")
    except ValueError as e:
        errors.append(str(e))
    
    if errors:
        return {"success": False, "errors": errors}
    
    return {"success": True, "user": {"username": username, "email": email, "age": age}}

print("=== Input Validation ===")

# Test cases
test_cases = [
    ("john_doe", "john@email.com", 25),
    ("ab", "invalid", 15),
    ("", "", 200),
]

for username, email, age in test_cases:
    result = register_user(username, email, age)
    if result["success"]:
        print(f"‚úÖ Registered: {result['user']}")
    else:
        print(f"‚ùå Errors: {result['errors']}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 3: File Processing Pipeline
# ========================================

class FileProcessingError(Exception):
    """Base error for file processing"""
    pass

class FileReadError(FileProcessingError):
    pass

class FileParseError(FileProcessingError):
    pass

class FileValidationError(FileProcessingError):
    pass

def process_data_file(filename):
    """Pipeline x·ª≠ l√Ω file v·ªõi error handling"""
    
    # Step 1: Read file
    try:
        with open(filename, 'r') as f:
            content = f.read()
        print(f"  1. ‚úÖ Read file ({len(content)} bytes)")
    except FileNotFoundError:
        raise FileReadError(f"File not found: {filename}")
    except PermissionError:
        raise FileReadError(f"Permission denied: {filename}")
    
    # Step 2: Parse content
    try:
        lines = content.strip().split('\n')
        data = [line.split(',') for line in lines]
        print(f"  2. ‚úÖ Parsed ({len(data)} rows)")
    except Exception as e:
        raise FileParseError(f"Failed to parse: {e}")
    
    # Step 3: Validate
    try:
        if len(data) == 0:
            raise ValueError("Empty file")
        for i, row in enumerate(data):
            if len(row) < 2:
                raise ValueError(f"Row {i+1} has insufficient columns")
        print(f"  3. ‚úÖ Validated")
    except ValueError as e:
        raise FileValidationError(str(e))
    
    return data

# T·∫°o file test
with open('test_data.csv', 'w') as f:
    f.write("name,age\nAn,20\nBinh,21")

print("=== File Processing Pipeline ===")

# Test th√†nh c√¥ng
print("\nProcessing 'test_data.csv':")
try:
    result = process_data_file('test_data.csv')
    print(f"  Result: {result}")
except FileProcessingError as e:
    print(f"  ‚ùå {type(e).__name__}: {e}")

# Test file kh√¥ng t·ªìn t·∫°i
print("\nProcessing 'not_exist.csv':")
try:
    result = process_data_file('not_exist.csv')
except FileProcessingError as e:
    print(f"  ‚ùå {type(e).__name__}: {e}")

# Cleanup
import os
os.remove('test_data.csv')
````

````python
# ========================================
# ·ª®NG D·ª§NG 4: Safe Dictionary Access
# ========================================

def safe_get(data, *keys, default=None):
    """Truy c·∫≠p nested dict an to√†n"""
    try:
        result = data
        for key in keys:
            result = result[key]
        return result
    except (KeyError, IndexError, TypeError):
        return default

# Data ph·ª©c t·∫°p
user_data = {
    "user": {
        "profile": {
            "name": "An",
            "addresses": [
                {"city": "Hanoi", "zip": "100000"},
                {"city": "HCM", "zip": "700000"}
            ]
        }
    }
}

print("=== Safe Dict Access ===")

# Truy c·∫≠p th√†nh c√¥ng
print(f"Name: {safe_get(user_data, 'user', 'profile', 'name')}")
print(f"First city: {safe_get(user_data, 'user', 'profile', 'addresses', 0, 'city')}")

# Truy c·∫≠p th·∫•t b·∫°i (tr·∫£ v·ªÅ default)
print(f"Email: {safe_get(user_data, 'user', 'email', default='N/A')}")
print(f"Phone: {safe_get(user_data, 'user', 'profile', 'phone', default='Not set')}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 5: Calculator v·ªõi Error Handling
# ========================================

class CalculatorError(Exception):
    pass

class Calculator:
    """Calculator v·ªõi ƒë·∫ßy ƒë·ªß error handling"""
    
    def __init__(self):
        self.history = []
    
    def _validate_numbers(self, *args):
        for arg in args:
            if not isinstance(arg, (int, float)):
                raise TypeError(f"Expected number, got {type(arg).__name__}")
    
    def _record(self, operation, result):
        self.history.append({"operation": operation, "result": result})
    
    def add(self, a, b):
        self._validate_numbers(a, b)
        result = a + b
        self._record(f"{a} + {b}", result)
        return result
    
    def divide(self, a, b):
        self._validate_numbers(a, b)
        if b == 0:
            raise ZeroDivisionError("Division by zero")
        result = a / b
        self._record(f"{a} / {b}", result)
        return result
    
    def sqrt(self, a):
        self._validate_numbers(a)
        if a < 0:
            raise CalculatorError("Cannot calculate square root of negative number")
        result = a ** 0.5
        self._record(f"‚àö{a}", result)
        return result

calc = Calculator()

print("=== Calculator ===")

operations = [
    lambda: calc.add(10, 5),
    lambda: calc.divide(10, 2),
    lambda: calc.divide(10, 0),
    lambda: calc.sqrt(16),
    lambda: calc.sqrt(-4),
    lambda: calc.add("5", 3),
]

for op in operations:
    try:
        result = op()
        print(f"‚úÖ Result: {result}")
    except (ZeroDivisionError, CalculatorError, TypeError) as e:
        print(f"‚ùå {type(e).__name__}: {e}")

print(f"\nHistory: {calc.history}")
````

````python
# ========================================
# B√ÄI T·∫¨P 1: Safe Division
# ========================================

def safe_divide(a, b):
    """
    Chia a cho b v·ªõi x·ª≠ l√Ω l·ªói ƒë·∫ßy ƒë·ªß.
    - ZeroDivisionError: tr·∫£ v·ªÅ None
    - TypeError: tr·∫£ v·ªÅ th√¥ng b√°o l·ªói
    """
    try:
        return a / b
    except ZeroDivisionError:
        return None
    except TypeError:
        return f"Cannot divide {type(a).__name__} by {type(b).__name__}"

# Test
print("=== B√†i t·∫≠p 1: Safe Division ===")
print(f"10 / 2 = {safe_divide(10, 2)}")
print(f"10 / 0 = {safe_divide(10, 0)}")
print(f"'10' / 2 = {safe_divide('10', 2)}")
````

````python
# ========================================
# B√ÄI T·∫¨P 2: Password Validator
# ========================================

class PasswordError(Exception):
    pass

def validate_password(password):
    """
    Validate password:
    - √çt nh·∫•t 8 k√Ω t·ª±
    - C√≥ ch·ªØ hoa
    - C√≥ ch·ªØ th∆∞·ªùng
    - C√≥ s·ªë
    """
    if len(password) < 8:
        raise PasswordError("Password ph·∫£i c√≥ √≠t nh·∫•t 8 k√Ω t·ª±")
    if not any(c.isupper() for c in password):
        raise PasswordError("Password ph·∫£i c√≥ √≠t nh·∫•t 1 ch·ªØ hoa")
    if not any(c.islower() for c in password):
        raise PasswordError("Password ph·∫£i c√≥ √≠t nh·∫•t 1 ch·ªØ th∆∞·ªùng")
    if not any(c.isdigit() for c in password):
        raise PasswordError("Password ph·∫£i c√≥ √≠t nh·∫•t 1 s·ªë")
    return True

# Test
print("=== B√†i t·∫≠p 2: Password Validator ===")

passwords = ["short", "nouppercase1", "NOLOWERCASE1", "NoNumber", "ValidPass1"]
for pwd in passwords:
    try:
        validate_password(pwd)
        print(f"‚úÖ '{pwd}' h·ª£p l·ªá")
    except PasswordError as e:
        print(f"‚ùå '{pwd}': {e}")
````

````python
# ========================================
# B√ÄI T·∫¨P 3: JSON Config Loader
# ========================================
import json

def load_config(filename, required_keys=None):
    """
    Load v√† validate JSON config.
    Raise exception n·∫øu:
    - File kh√¥ng t·ªìn t·∫°i
    - JSON kh√¥ng h·ª£p l·ªá
    - Thi·∫øu required keys
    """
    try:
        with open(filename, 'r') as f:
            config = json.load(f)
    except FileNotFoundError:
        raise FileNotFoundError(f"Config file not found: {filename}")
    except json.JSONDecodeError as e:
        raise ValueError(f"Invalid JSON: {e}")
    
    if required_keys:
        missing = [key for key in required_keys if key not in config]
        if missing:
            raise KeyError(f"Missing required keys: {missing}")
    
    return config

# T·∫°o file test
with open('config.json', 'w') as f:
    json.dump({"host": "localhost", "port": 8080}, f)

print("=== B√†i t·∫≠p 3: Config Loader ===")

# Test th√†nh c√¥ng
try:
    config = load_config('config.json', required_keys=['host', 'port'])
    print(f"‚úÖ Loaded: {config}")
except Exception as e:
    print(f"‚ùå {type(e).__name__}: {e}")

# Test thi·∫øu key
try:
    config = load_config('config.json', required_keys=['host', 'database'])
except Exception as e:
    print(f"‚ùå {type(e).__name__}: {e}")

# Cleanup
import os
os.remove('config.json')
````

````python
# ========================================
# B√ÄI T·∫¨P 4: Bank Account v·ªõi Exception
# ========================================

class InsufficientFundsError(Exception):
    def __init__(self, balance, amount):
        self.balance = balance
        self.amount = amount
        super().__init__(
            f"Insufficient funds: balance={balance}, requested={amount}"
        )

class InvalidAmountError(Exception):
    pass

class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self._balance = balance
    
    @property
    def balance(self):
        return self._balance
    
    def deposit(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Deposit amount must be positive")
        self._balance += amount
        return self._balance
    
    def withdraw(self, amount):
        if amount <= 0:
            raise InvalidAmountError("Withdrawal amount must be positive")
        if amount > self._balance:
            raise InsufficientFundsError(self._balance, amount)
        self._balance -= amount
        return self._balance

# Test
print("=== B√†i t·∫≠p 4: Bank Account ===")

account = BankAccount("An", 1000)
print(f"Initial balance: {account.balance}")

# Deposit
try:
    account.deposit(500)
    print(f"After deposit 500: {account.balance}")
except InvalidAmountError as e:
    print(f"‚ùå {e}")

# Withdraw success
try:
    account.withdraw(200)
    print(f"After withdraw 200: {account.balance}")
except InsufficientFundsError as e:
    print(f"‚ùå {e}")

# Withdraw fail
try:
    account.withdraw(5000)
except InsufficientFundsError as e:
    print(f"‚ùå {e}")
    print(f"   Available: {e.balance}, Requested: {e.amount}")
````
