## 1. Gi·ªõi thi·ªáu v·ªÅ Dictionaries

**Dictionary** l√† c·∫•u tr√∫c d·ªØ li·ªáu l∆∞u tr·ªØ theo c·∫∑p **key-value** (kh√≥a-gi√° tr·ªã).

### ƒê·∫∑c ƒëi·ªÉm c·ªßa Dictionary:

| ƒê·∫∑c ƒëi·ªÉm | M√¥ t·∫£ |
|----------|-------|
| **Key-Value pairs** | M·ªói ph·∫ßn t·ª≠ g·ªìm 1 key v√† 1 value |
| **Ordered** (Python 3.7+) | Gi·ªØ th·ª© t·ª± ch√®n |
| **Mutable** | C√≥ th·ªÉ thay ƒë·ªïi sau khi t·∫°o |
| **Keys unique** | Key kh√¥ng ƒë∆∞·ª£c tr√πng l·∫∑p |
| **Keys hashable** | Key ph·∫£i l√† immutable (str, int, tuple) |
| **Values any type** | Value c√≥ th·ªÉ l√† b·∫•t k·ª≥ ki·ªÉu n√†o |

### C·∫•u tr√∫c:
```python
my_dict = {
    "key1": "value1",
    "key2": "value2",
    "key3": "value3"
}
```

### Khi n√†o d√πng Dictionary?
- L∆∞u tr·ªØ d·ªØ li·ªáu c√≥ c·∫•u tr√∫c (nh∆∞ JSON)
- √Ånh x·∫° (mapping) gi·ªØa c√°c gi√° tr·ªã
- ƒê·∫øm t·∫ßn su·∫•t xu·∫•t hi·ªán
- Cache v√† lookup tables
- C·∫•u h√¨nh ·ª©ng d·ª•ng

## 2. T·∫°o Dictionaries

````python
# ========================================
# C√ÅCH 1: T·∫°o dictionary r·ªóng
# ========================================

# C√°ch 1: D√πng {}
empty_dict1 = {}
print("Dict r·ªóng (c√°ch 1):", empty_dict1)
print("Ki·ªÉu:", type(empty_dict1))

# C√°ch 2: D√πng dict()
empty_dict2 = dict()
print("Dict r·ªóng (c√°ch 2):", empty_dict2)
````

````python
# ========================================
# C√ÅCH 2: T·∫°o dictionary v·ªõi c√°c ph·∫ßn t·ª≠
# ========================================

# Th√¥ng tin sinh vi√™n
student = {
    "name": "Nguyen Van An",
    "age": 20,
    "major": "Computer Science",
    "gpa": 8.5
}
print("Sinh vi√™n:")
for key, value in student.items():
    print(f"  {key}: {value}")
````

````python
# ========================================
# C√ÅCH 3: T·∫°o t·ª´ dict() v·ªõi keyword arguments
# ========================================

# L∆∞u √Ω: Key ph·∫£i l√† string h·ª£p l·ªá (nh∆∞ t√™n bi·∫øn)
person = dict(name="John", age=30, city="New York")
print("Person:", person)

# Kh√¥ng th·ªÉ d√πng s·ªë ho·∫∑c k√Ω t·ª± ƒë·∫∑c bi·ªát l√†m key theo c√°ch n√†y
# dict(1="one")  # SyntaxError!
````

````python
# ========================================
# C√ÅCH 4: T·∫°o t·ª´ list c√°c tuples
# ========================================

# List of tuples (key, value)
pairs = [("a", 1), ("b", 2), ("c", 3)]
dict_from_pairs = dict(pairs)
print("T·ª´ list of tuples:", dict_from_pairs)

# T·ª´ 2 lists (zip)
keys = ["name", "age", "city"]
values = ["Alice", 25, "London"]
dict_from_zip = dict(zip(keys, values))
print("T·ª´ zip:", dict_from_zip)
````

````python
# ========================================
# C√ÅCH 5: dict.fromkeys() - T·∫°o v·ªõi gi√° tr·ªã m·∫∑c ƒë·ªãnh
# ========================================

# T·∫•t c·∫£ keys c√≥ c√πng gi√° tr·ªã
keys = ["a", "b", "c", "d"]
dict_default = dict.fromkeys(keys, 0)
print("V·ªõi gi√° tr·ªã m·∫∑c ƒë·ªãnh 0:", dict_default)

# Kh√¥ng c√≥ gi√° tr·ªã m·∫∑c ƒë·ªãnh ‚Üí None
dict_none = dict.fromkeys(["x", "y", "z"])
print("Kh√¥ng c√≥ gi√° tr·ªã:", dict_none)
````

````python
# ========================================
# C√ÅC LO·∫†I KEY H·ª¢P L·ªÜ
# ========================================

# Key c√≥ th·ªÉ l√† b·∫•t k·ª≥ ki·ªÉu hashable n√†o
valid_keys = {
    "string_key": "value1",     # String
    42: "value2",                # Integer
    3.14: "value3",              # Float
    (1, 2): "value4",            # Tuple
    True: "value5",              # Boolean
    None: "value6"               # None
}

print("Dict v·ªõi nhi·ªÅu ki·ªÉu key:")
for key, value in valid_keys.items():
    print(f"  {repr(key):15} ‚Üí {value}")
````

````python
# ========================================
# ‚ö†Ô∏è KEY KH√îNG H·ª¢P L·ªÜ (Unhashable)
# ========================================

# List kh√¥ng th·ªÉ l√†m key
try:
    invalid = {[1, 2, 3]: "value"}
except TypeError as e:
    print(f"‚ùå List l√†m key: {e}")

# Dict kh√¥ng th·ªÉ l√†m key
try:
    invalid = {{"a": 1}: "value"}
except TypeError as e:
    print(f"‚ùå Dict l√†m key: {e}")

# Set kh√¥ng th·ªÉ l√†m key
try:
    invalid = {{1, 2, 3}: "value"}
except TypeError as e:
    print(f"‚ùå Set l√†m key: {e}")
````

````python
# ========================================
# KEY TR√ôNG L·∫∂P - Key sau ghi ƒë√® key tr∆∞·ªõc
# ========================================

# Key tr√πng ‚Üí gi√° tr·ªã cu·ªëi c√πng ƒë∆∞·ª£c gi·ªØ
duplicate_keys = {
    "name": "Alice",
    "age": 25,
    "name": "Bob",      # Ghi ƒë√® "Alice"
    "name": "Charlie"   # Ghi ƒë√® "Bob"
}
print("K·∫øt qu·∫£:", duplicate_keys)

# True v√† 1 ƒë∆∞·ª£c coi l√† c√πng key
weird = {True: "True", 1: "One"}
print("\n{True: 'True', 1: 'One'}:", weird)
````

## 3. Truy c·∫≠p v√† thay ƒë·ªïi gi√° tr·ªã

````python
# ========================================
# TRUY C·∫¨P GI√Å TR·ªä B·∫∞NG KEY
# ========================================

student = {
    "name": "Nguyen Van An",
    "age": 20,
    "major": "Computer Science",
    "gpa": 8.5
}

# C√°ch 1: D√πng []
print("T√™n:", student["name"])
print("Tu·ªïi:", student["age"])
print("Ng√†nh:", student["major"])
````

````python
# ========================================
# L·ªñI KHI KEY KH√îNG T·ªíN T·∫†I
# ========================================

student = {"name": "An", "age": 20}

try:
    print(student["email"])  # Key kh√¥ng t·ªìn t·∫°i
except KeyError as e:
    print(f"‚ùå KeyError: {e}")
    print("‚Üí Key 'email' kh√¥ng t·ªìn t·∫°i trong dictionary!")
````

````python
# ========================================
# GET() - Truy c·∫≠p an to√†n
# ========================================

student = {"name": "An", "age": 20, "gpa": 8.5}

# get() tr·∫£ v·ªÅ None n·∫øu key kh√¥ng t·ªìn t·∫°i
print("name:", student.get("name"))      # An
print("email:", student.get("email"))    # None (kh√¥ng l·ªói)

# get() v·ªõi gi√° tr·ªã m·∫∑c ƒë·ªãnh
print("\nemail (default):", student.get("email", "N/A"))
print("phone (default):", student.get("phone", "Kh√¥ng c√≥"))
````

````python
# ========================================
# SO S√ÅNH [] vs get()
# ========================================

data = {"a": 1, "b": 2}

print("Comparison: [] vs get()")
print("="*40)

# [] - Nhanh nh∆∞ng c√≥ th·ªÉ l·ªói
print("data['a']:", data['a'])  # OK
# print(data['c'])  # KeyError!

# get() - An to√†n h∆°n
print("data.get('a'):", data.get('a'))      # 1
print("data.get('c'):", data.get('c'))      # None
print("data.get('c', 0):", data.get('c', 0))  # 0

print("\nüí° Tip: D√πng get() khi kh√¥ng ch·∫Øc key c√≥ t·ªìn t·∫°i!")
````

````python
# ========================================
# KI·ªÇM TRA KEY T·ªíN T·∫†I
# ========================================

student = {"name": "An", "age": 20, "gpa": 8.5}

# D√πng 'in'
print("'name' in student:", "name" in student)      # True
print("'email' in student:", "email" in student)    # False
print("'phone' not in student:", "phone" not in student)  # True

# ·ª®ng d·ª•ng
if "email" in student:
    print(f"Email: {student['email']}")
else:
    print("Kh√¥ng c√≥ th√¥ng tin email")
````

````python
# ========================================
# THAY ƒê·ªîI GI√Å TR·ªä
# ========================================

student = {"name": "An", "age": 20, "gpa": 8.5}
print("Ban ƒë·∫ßu:", student)

# Thay ƒë·ªïi gi√° tr·ªã
student["age"] = 21
print("Sau khi s·ª≠a age:", student)

student["gpa"] = 9.0
print("Sau khi s·ª≠a gpa:", student)
````

## 4. Th√™m, s·ª≠a, x√≥a ph·∫ßn t·ª≠

````python
# ========================================
# TH√äM PH·∫¶N T·ª¨ M·ªöI
# ========================================

student = {"name": "An", "age": 20}
print("Ban ƒë·∫ßu:", student)

# Th√™m key m·ªõi
student["email"] = "an@example.com"
print("Sau th√™m email:", student)

student["phone"] = "0123456789"
print("Sau th√™m phone:", student)
````

````python
# ========================================
# UPDATE() - C·∫≠p nh·∫≠t nhi·ªÅu ph·∫ßn t·ª≠
# ========================================

student = {"name": "An", "age": 20}
print("Ban ƒë·∫ßu:", student)

# Update v·ªõi dict kh√°c
student.update({"age": 21, "major": "CS", "gpa": 8.5})
print("Sau update:", student)

# Update v·ªõi keyword arguments
student.update(email="an@email.com", phone="0123")
print("Sau update 2:", student)
````

````python
# ========================================
# SETDEFAULT() - Th√™m n·∫øu ch∆∞a c√≥
# ========================================

student = {"name": "An", "age": 20}
print("Ban ƒë·∫ßu:", student)

# Key ch∆∞a t·ªìn t·∫°i ‚Üí th√™m m·ªõi
result1 = student.setdefault("major", "Unknown")
print(f"setdefault('major', 'Unknown') = {result1}")
print("Dict:", student)

# Key ƒë√£ t·ªìn t·∫°i ‚Üí kh√¥ng thay ƒë·ªïi, tr·∫£ v·ªÅ gi√° tr·ªã c≈©
result2 = student.setdefault("name", "Bob")
print(f"\nsetdefault('name', 'Bob') = {result2}")
print("Dict:", student)  # name v·∫´n l√† 'An'
````

````python
# ========================================
# DEL - X√≥a ph·∫ßn t·ª≠
# ========================================

student = {"name": "An", "age": 20, "major": "CS", "gpa": 8.5}
print("Ban ƒë·∫ßu:", student)

# X√≥a 1 key
del student["gpa"]
print("Sau del 'gpa':", student)

# X√≥a key kh√¥ng t·ªìn t·∫°i ‚Üí KeyError
try:
    del student["email"]
except KeyError as e:
    print(f"\n‚ùå KeyError: Key {e} kh√¥ng t·ªìn t·∫°i!")
````

````python
# ========================================
# POP() - X√≥a v√† tr·∫£ v·ªÅ gi√° tr·ªã
# ========================================

student = {"name": "An", "age": 20, "major": "CS"}
print("Ban ƒë·∫ßu:", student)

# Pop key t·ªìn t·∫°i
age = student.pop("age")
print(f"pop('age') = {age}")
print("Dict:", student)

# Pop key kh√¥ng t·ªìn t·∫°i v·ªõi default
email = student.pop("email", "N/A")
print(f"\npop('email', 'N/A') = {email}")

# Pop key kh√¥ng t·ªìn t·∫°i kh√¥ng c√≥ default ‚Üí KeyError
try:
    student.pop("phone")
except KeyError as e:
    print(f"‚ùå KeyError: {e}")
````

````python
# ========================================
# POPITEM() - X√≥a c·∫∑p key-value cu·ªëi c√πng
# ========================================

student = {"name": "An", "age": 20, "major": "CS"}
print("Ban ƒë·∫ßu:", student)

# X√≥a v√† tr·∫£ v·ªÅ c·∫∑p cu·ªëi (Python 3.7+: LIFO order)
item = student.popitem()
print(f"popitem() = {item}")
print("Dict:", student)

item = student.popitem()
print(f"\npopitem() = {item}")
print("Dict:", student)
````

````python
# ========================================
# CLEAR() - X√≥a t·∫•t c·∫£
# ========================================

student = {"name": "An", "age": 20, "major": "CS"}
print("Ban ƒë·∫ßu:", student)
print("ƒê·ªô d√†i:", len(student))

student.clear()
print("\nSau clear():", student)
print("ƒê·ªô d√†i:", len(student))
````

## 5. Dictionary Methods

### B·∫£ng t·ªïng h·ª£p:

| Method | M√¥ t·∫£ | Return |
|--------|-------|--------|
| `get(key, default)` | L·∫•y gi√° tr·ªã, tr·∫£ default n·∫øu kh√¥ng c√≥ | Value ho·∫∑c default |
| `keys()` | L·∫•y t·∫•t c·∫£ keys | dict_keys |
| `values()` | L·∫•y t·∫•t c·∫£ values | dict_values |
| `items()` | L·∫•y t·∫•t c·∫£ c·∫∑p (key, value) | dict_items |
| `update(other)` | C·∫≠p nh·∫≠t/th√™m t·ª´ dict kh√°c | None |
| `setdefault(key, val)` | L·∫•y ho·∫∑c set n·∫øu ch∆∞a c√≥ | Value |
| `pop(key, default)` | X√≥a key v√† tr·∫£ v·ªÅ value | Value |
| `popitem()` | X√≥a c·∫∑p cu·ªëi c√πng | (key, value) |
| `clear()` | X√≥a t·∫•t c·∫£ | None |
| `copy()` | T·∫°o b·∫£n sao n√¥ng | dict |
| `fromkeys(keys, val)` | T·∫°o dict t·ª´ keys | dict |

````python
# ========================================
# KEYS(), VALUES(), ITEMS()
# ========================================

student = {
    "name": "Nguyen Van An",
    "age": 20,
    "major": "Computer Science",
    "gpa": 8.5
}

# keys() - L·∫•y t·∫•t c·∫£ keys
keys = student.keys()
print("Keys:", keys)
print("Type:", type(keys))

# values() - L·∫•y t·∫•t c·∫£ values
values = student.values()
print("\nValues:", values)

# items() - L·∫•y t·∫•t c·∫£ c·∫∑p (key, value)
items = student.items()
print("\nItems:", items)
````

````python
# ========================================
# CHUY·ªÇN ƒê·ªîI SANG LIST
# ========================================

student = {"name": "An", "age": 20, "gpa": 8.5}

# Chuy·ªÉn keys/values/items sang list
keys_list = list(student.keys())
values_list = list(student.values())
items_list = list(student.items())

print("Keys list:", keys_list)
print("Values list:", values_list)
print("Items list:", items_list)
````

````python
# ========================================
# COPY() - Shallow Copy
# ========================================

original = {"name": "An", "scores": [85, 90, 95]}
print("Original:", original)

# Shallow copy
copied = original.copy()

# Thay ƒë·ªïi gi√° tr·ªã ƒë∆°n gi·∫£n
copied["name"] = "B√¨nh"
print("\nSau khi s·ª≠a copied['name']:")
print("Original:", original)  # Kh√¥ng thay ƒë·ªïi
print("Copied:", copied)

# Thay ƒë·ªïi gi√° tr·ªã nested (mutable)
copied["scores"].append(100)
print("\nSau khi s·ª≠a copied['scores']:")
print("Original:", original)  # B·ªä thay ƒë·ªïi!
print("Copied:", copied)
````

````python
# ========================================
# DEEP COPY cho nested dict
# ========================================
import copy

original = {
    "name": "An",
    "scores": [85, 90, 95],
    "info": {"city": "HN", "year": 2024}
}

# Deep copy
deep_copied = copy.deepcopy(original)

# Thay ƒë·ªïi nested
deep_copied["scores"].append(100)
deep_copied["info"]["city"] = "HCM"

print("Original:", original)
print("Deep copied:", deep_copied)
print("\n‚úÖ Original kh√¥ng b·ªã ·∫£nh h∆∞·ªüng!")
````

## 6. Duy·ªát Dictionary

````python
# ========================================
# DUY·ªÜT QUA KEYS
# ========================================

student = {"name": "An", "age": 20, "major": "CS", "gpa": 8.5}

# C√°ch 1: M·∫∑c ƒë·ªãnh (duy·ªát keys)
print("Duy·ªát keys (m·∫∑c ƒë·ªãnh):")
for key in student:
    print(f"  {key}")

# C√°ch 2: D√πng keys()
print("\nDuy·ªát keys (explicit):")
for key in student.keys():
    print(f"  {key}: {student[key]}")
````

````python
# ========================================
# DUY·ªÜT QUA VALUES
# ========================================

student = {"name": "An", "age": 20, "major": "CS", "gpa": 8.5}

print("Duy·ªát values:")
for value in student.values():
    print(f"  {value}")
````

````python
# ========================================
# DUY·ªÜT QUA ITEMS (key-value pairs)
# ========================================

student = {"name": "An", "age": 20, "major": "CS", "gpa": 8.5}

# Unpacking trong v√≤ng l·∫∑p
print("Duy·ªát items:")
for key, value in student.items():
    print(f"  {key}: {value}")
````

````python
# ========================================
# DUY·ªÜT V·ªöI ENUMERATE
# ========================================

student = {"name": "An", "age": 20, "major": "CS", "gpa": 8.5}

print("Duy·ªát v·ªõi index:")
for index, (key, value) in enumerate(student.items(), 1):
    print(f"  {index}. {key} = {value}")
````

````python
# ========================================
# DUY·ªÜT THEO TH·ª® T·ª∞ S·∫ÆP X·∫æP
# ========================================

scores = {"An": 85, "B√¨nh": 92, "C∆∞·ªùng": 78, "D≈©ng": 88}

# S·∫Øp x·∫øp theo key
print("S·∫Øp x·∫øp theo t√™n (key):")
for name in sorted(scores.keys()):
    print(f"  {name}: {scores[name]}")

# S·∫Øp x·∫øp theo value
print("\nS·∫Øp x·∫øp theo ƒëi·ªÉm (value):")
for name, score in sorted(scores.items(), key=lambda x: x[1], reverse=True):
    print(f"  {name}: {score}")
````

````python
# ========================================
# V√ç D·ª§: In b·∫£ng ƒë·∫πp
# ========================================

products = {
    "Laptop": 1200,
    "Mouse": 25,
    "Keyboard": 75,
    "Monitor": 300,
    "Headphones": 150
}

print("B·∫¢NG GI√Å S·∫¢N PH·∫®M")
print("=" * 30)
print(f"{'S·∫£n ph·∫©m':<15} {'Gi√° ($)':>10}")
print("-" * 30)

total = 0
for product, price in products.items():
    print(f"{product:<15} {price:>10,}")
    total += price

print("=" * 30)
print(f"{'T·ªîNG':<15} {total:>10,}")
````

## 7. Nested Dictionaries

````python
# ========================================
# T·∫†O NESTED DICTIONARY
# ========================================

# Dictionary ch·ª©a dictionary
students = {
    "SV001": {
        "name": "Nguyen Van An",
        "age": 20,
        "scores": {"math": 85, "physics": 90, "chemistry": 88}
    },
    "SV002": {
        "name": "Tran Thi Binh",
        "age": 21,
        "scores": {"math": 92, "physics": 85, "chemistry": 90}
    },
    "SV003": {
        "name": "Le Van Cuong",
        "age": 19,
        "scores": {"math": 78, "physics": 82, "chemistry": 85}
    }
}

# In ra
import json
print(json.dumps(students, indent=2, ensure_ascii=False))
````

````python
# ========================================
# ·ª®NG D·ª§NG 1: ƒê·∫øm t·∫ßn su·∫•t k√Ω t·ª±
# ========================================

def count_chars(text):
    """ƒê·∫øm t·∫ßn su·∫•t k√Ω t·ª± trong chu·ªói"""
    char_count = {}
    for char in text.lower():
        if char.isalpha():
            char_count[char] = char_count.get(char, 0) + 1
    return char_count

text = "Hello World Python Programming"
result = count_chars(text)

# S·∫Øp x·∫øp theo t·∫ßn su·∫•t
sorted_chars = sorted(result.items(), key=lambda x: x[1], reverse=True)
print(f"Text: {text}")
print("\nT·∫ßn su·∫•t k√Ω t·ª±:")
for char, count in sorted_chars:
    print(f"  '{char}': {count}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 2: Nh√≥m d·ªØ li·ªáu
# ========================================

students = [
    {"name": "An", "class": "10A", "score": 85},
    {"name": "Binh", "class": "10B", "score": 92},
    {"name": "Cuong", "class": "10A", "score": 78},
    {"name": "Dung", "class": "10B", "score": 88},
    {"name": "Em", "class": "10A", "score": 90},
]

# Nh√≥m theo l·ªõp
by_class = {}
for student in students:
    class_name = student["class"]
    if class_name not in by_class:
        by_class[class_name] = []
    by_class[class_name].append(student)

print("Sinh vi√™n theo l·ªõp:")
for class_name, class_students in by_class.items():
    print(f"\n{class_name}:")
    for s in class_students:
        print(f"  - {s['name']}: {s['score']}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 3: Cache / Memoization
# ========================================

# Cache k·∫øt qu·∫£ t√≠nh to√°n
fibonacci_cache = {}

def fibonacci(n):
    """Fibonacci v·ªõi cache"""
    if n in fibonacci_cache:
        return fibonacci_cache[n]
    
    if n <= 1:
        result = n
    else:
        result = fibonacci(n-1) + fibonacci(n-2)
    
    fibonacci_cache[n] = result
    return result

# Test
print("Fibonacci sequence:")
for i in range(15):
    print(f"F({i}) = {fibonacci(i)}")

print(f"\nCache: {fibonacci_cache}")
````

````python
# ========================================
# ·ª®NG D·ª§NG 6: Gi·ªè h√†ng (Shopping Cart)
# ========================================

class ShoppingCart:
    def __init__(self):
        self.items = {}  # {product_name: {"price": x, "quantity": y}}
    
    def add_item(self, name, price, quantity=1):
        if name in self.items:
            self.items[name]["quantity"] += quantity
        else:
            self.items[name] = {"price": price, "quantity": quantity}
    
    def remove_item(self, name):
        if name in self.items:
            del self.items[name]
    
    def get_total(self):
        return sum(item["price"] * item["quantity"] 
                   for item in self.items.values())
    
    def display(self):
        print("\n" + "=" * 50)
        print("GI·ªé H√ÄNG")
        print("=" * 50)
        print(f"{'S·∫£n ph·∫©m':<20} {'Gi√°':>10} {'SL':>5} {'T·ªïng':>10}")
        print("-" * 50)
        
        for name, info in self.items.items():
            subtotal = info["price"] * info["quantity"]
            print(f"{name:<20} {info['price']:>10,} {info['quantity']:>5} {subtotal:>10,}")
        
        print("=" * 50)
        print(f"{'T·ªîNG C·ªòNG':>37} {self.get_total():>10,}")

# Demo
cart = ShoppingCart()
cart.add_item("Laptop", 15000000)
cart.add_item("Mouse", 500000, 2)
cart.add_item("Keyboard", 1200000)
cart.add_item("Mouse", 500000, 1)  # Th√™m 1 mouse n·ªØa
cart.display()
````

````python
# ========================================
# ·ª®NG D·ª§NG 7: Counter (from collections)
# ========================================
from typing_extensions import LiteralString


from collections import Counter

# ƒê·∫øm t·ª´
text = "apple banana apple cherry banana apple cherry cherry cherry"
word_counter = Counter[LiteralString](text.split())

print("Counter:", word_counter)
print("\nTop 2:", word_counter.most_common(2))
print("S·ªë l·∫ßn 'apple':", word_counter["apple"])

# ƒê·∫øm k√Ω t·ª±
char_counter = Counter("abracadabra")
print("\nK√Ω t·ª± trong 'abracadabra':", char_counter)
````

````python
# ========================================
# ·ª®NG D·ª§NG 8: DefaultDict
# ========================================
from collections import defaultdict

# Nh√≥m t·ª´ theo ch·ªØ c√°i ƒë·∫ßu
words = ["apple", "banana", "avocado", "blueberry", "cherry", "apricot"]

# C√°ch th√¥ng th∆∞·ªùng
grouped_normal = {}
for word in words:
    first_letter = word[0]
    if first_letter not in grouped_normal:
        grouped_normal[first_letter] = []
    grouped_normal[first_letter].append(word)

# V·ªõi defaultdict
grouped_default = defaultdict(list)
for word in words:
    grouped_default[word[0]].append(word)

print("Nh√≥m theo ch·ªØ c√°i ƒë·∫ßu:")
for letter, word_list in sorted(grouped_default.items()):
    print(f"  {letter}: {word_list}")
````

````python
# ========================================
# B√ÄI T·∫¨P 1: H·ª£p nh·∫•t 2 dictionaries
# ========================================

dict1 = {"a": 1, "b": 2, "c": 3}
dict2 = {"c": 30, "d": 4, "e": 5}

# C√°ch 1: update()
merged1 = dict1.copy()
merged1.update(dict2)
print("C√°ch 1 (update):", merged1)

# C√°ch 2: unpacking (Python 3.5+)
merged2 = {**dict1, **dict2}
print("C√°ch 2 (**unpack):", merged2)

# C√°ch 3: | operator (Python 3.9+)
merged3 = dict1 | dict2
print("C√°ch 3 (|):", merged3)
````

````python
# ========================================
# B√ÄI T·∫¨P 2: T√¨m value l·ªõn nh·∫•t
# ========================================

scores = {"An": 85, "Binh": 92, "Cuong": 78, "Dung": 92, "Em": 88}

# T√¨m max value
max_score = max(scores.values())
print(f"ƒêi·ªÉm cao nh·∫•t: {max_score}")

# T√¨m key c√≥ max value
top_student = max(scores, key=scores.get)
print(f"H·ªçc sinh ƒëi·ªÉm cao nh·∫•t: {top_student}")

# N·∫øu nhi·ªÅu ng∆∞·ªùi c√πng ƒëi·ªÉm cao nh·∫•t
top_students = [name for name, score in scores.items() if score == max_score]
print(f"T·∫•t c·∫£ h·ªçc sinh ƒëi·ªÉm cao nh·∫•t: {top_students}")
````

````python
# ========================================
# B√ÄI T·∫¨P 6: Flatten nested dict
# ========================================

def flatten_dict(d, parent_key='', sep='.'):
    """L√†m ph·∫≥ng nested dictionary"""
    items = []
    for k, v in d.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.extend(flatten_dict(v, new_key, sep).items())
        else:
            items.append((new_key, v))
    return dict(items)

nested = {
    "user": {
        "name": "An",
        "address": {
            "city": "Hanoi",
            "country": "Vietnam"
        }
    },
    "active": True
}

flat = flatten_dict(nested)
print("Nested:", nested)
print("\nFlattened:", flat)
````

## T·ªïng k·∫øt

### C√°c ƒëi·ªÉm quan tr·ªçng c·∫ßn nh·ªõ:

1. **Key-Value pairs** - M·ªói ph·∫ßn t·ª≠ g·ªìm key v√† value
2. **Keys ph·∫£i unique v√† hashable** - Kh√¥ng ƒë∆∞·ª£c tr√πng, ph·∫£i l√† immutable
3. **Truy c·∫≠p b·∫±ng key** - `dict[key]` ho·∫∑c `dict.get(key)`
4. **Ordered** (Python 3.7+) - Gi·ªØ th·ª© t·ª± ch√®n
5. **Mutable** - C√≥ th·ªÉ th√™m, s·ª≠a, x√≥a

### Methods quan tr·ªçng:

| Method | M√¥ t·∫£ |
|--------|-------|
| `get()` | Truy c·∫≠p an to√†n |
| `keys()`, `values()`, `items()` | L·∫•y keys/values/items |
| `update()` | C·∫≠p nh·∫≠t nhi·ªÅu ph·∫ßn t·ª≠ |
| `pop()` | X√≥a v√† tr·∫£ v·ªÅ value |
| `setdefault()` | L·∫•y ho·∫∑c set n·∫øu ch∆∞a c√≥ |

### Best Practices:
- D√πng `get()` thay `[]` khi kh√¥ng ch·∫Øc key t·ªìn t·∫°i
- D√πng dict comprehension ƒë·ªÉ t·∫°o dict nhanh
- D√πng `collections.Counter` ƒë·ªÉ ƒë·∫øm
- D√πng `collections.defaultdict` ƒë·ªÉ tr√°nh ki·ªÉm tra key t·ªìn t·∫°i
- D√πng `copy.deepcopy()` cho nested dict
