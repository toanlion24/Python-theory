# Tuples (B·ªô d·ªØ li·ªáu) trong Python

## 1. Kh√°i ni·ªám v·ªÅ Tuples

**Tuple** (b·ªô d·ªØ li·ªáu) l√† m·ªôt ki·ªÉu d·ªØ li·ªáu t·∫≠p h·ª£p (collection) trong Python, t∆∞∆°ng t·ª± nh∆∞ List nh∆∞ng c√≥ m·ªôt ƒëi·ªÉm kh√°c bi·ªát quan tr·ªçng:

- **C√≥ th·ª© t·ª± (Ordered)**: C√°c ph·∫ßn t·ª≠ ƒë∆∞·ª£c s·∫Øp x·∫øp theo th·ª© t·ª± v√† gi·ªØ nguy√™n th·ª© t·ª± ƒë√≥
- **B·∫•t bi·∫øn (Immutable)**: **KH√îNG th·ªÉ** thay ƒë·ªïi, th√™m, x√≥a ph·∫ßn t·ª≠ sau khi t·∫°o
- **Cho ph√©p tr√πng l·∫∑p**: C√≥ th·ªÉ ch·ª©a c√°c ph·∫ßn t·ª≠ gi·ªëng nhau
- **H·ªó tr·ª£ nhi·ªÅu ki·ªÉu d·ªØ li·ªáu**: C√≥ th·ªÉ ch·ª©a s·ªë, chu·ªói, boolean, list, tuple kh√°c
- **ƒê∆∞·ª£c ƒë√°nh ch·ªâ s·ªë (Indexed)**: M·ªói ph·∫ßn t·ª≠ c√≥ m·ªôt ch·ªâ s·ªë (index) b·∫Øt ƒë·∫ßu t·ª´ 0
- **Nhanh h∆°n List**: Do t√≠nh b·∫•t bi·∫øn, tuple ho·∫°t ƒë·ªông nhanh h∆°n list

### T·∫°i sao s·ª≠ d·ª•ng Tuples?

- **B·∫£o v·ªá d·ªØ li·ªáu**: ƒê·∫£m b·∫£o d·ªØ li·ªáu kh√¥ng b·ªã thay ƒë·ªïi v√¥ t√¨nh
- **Hi·ªáu su·∫•t t·ªët**: Nhanh h∆°n list trong m·ªôt s·ªë tr∆∞·ªùng h·ª£p
- **S·ª≠ d·ª•ng nh∆∞ keys trong dictionary**: Tuples c√≥ th·ªÉ d√πng l√†m keys (list kh√¥ng th·ªÉ)
- **Unpacking d·ªÖ d√†ng**: H·ªó tr·ª£ unpacking ƒë·ªÉ g√°n nhi·ªÅu bi·∫øn c√πng l√∫c
- **Tr·∫£ v·ªÅ nhi·ªÅu gi√° tr·ªã t·ª´ h√†m**: M·ªôt h√†m c√≥ th·ªÉ tr·∫£ v·ªÅ nhi·ªÅu gi√° tr·ªã d∆∞·ªõi d·∫°ng tuple

### So s√°nh Tuple vs List:

| ƒê·∫∑c ƒëi·ªÉm | Tuple | List |
|----------|-------|------|
| Khai b√°o | D√πng `()` ho·∫∑c `tuple()` | D√πng `[]` ho·∫∑c `list()` |
| T√≠nh bi·∫øn ƒë·ªïi | **Immutable** (b·∫•t bi·∫øn) | **Mutable** (c√≥ th·ªÉ thay ƒë·ªïi) |
| T·ªëc ƒë·ªô | Nhanh h∆°n | Ch·∫≠m h∆°n |
| B·ªô nh·ªõ | √çt h∆°n | Nhi·ªÅu h∆°n |
| D√πng l√†m key | ‚úÖ C√≥ th·ªÉ | ‚ùå Kh√¥ng th·ªÉ |
| Ph∆∞∆°ng th·ª©c | √çt ph∆∞∆°ng th·ª©c | Nhi·ªÅu ph∆∞∆°ng th·ª©c |


## 2. T·∫°o Tuples

C√≥ nhi·ªÅu c√°ch ƒë·ªÉ t·∫°o m·ªôt tuple trong Python:


### 2.1. T·∫°o tuple r·ªóng


```python
# C√°ch 1: S·ª≠ d·ª•ng d·∫•u ngo·∫∑c tr√≤n r·ªóng
tuple_rong1 = ()
print(f"Tuple r·ªóng 1: {tuple_rong1}, ki·ªÉu: {type(tuple_rong1)}")

# C√°ch 2: S·ª≠ d·ª•ng h√†m tuple()
tuple_rong2 = tuple()
print(f"Tuple r·ªóng 2: {tuple_rong2}, ki·ªÉu: {type(tuple_rong2)}")

# L∆∞u √Ω: KH√îNG th·ªÉ t·∫°o tuple 1 ph·∫ßn t·ª≠ b·∫±ng c√°ch tuple_rong1 = (5)
# Ph·∫£i c√≥ d·∫•u ph·∫©y: tuple_1_phan_tu = (5,)
```

### 2.2. T·∫°o tuple c√≥ ph·∫ßn t·ª≠


```python
# C√°ch 1: S·ª≠ d·ª•ng d·∫•u ngo·∫∑c tr√≤n
diem_so = (85, 90, 78, 92, 88)
print(f"ƒêi·ªÉm s·ªë: {diem_so}")

thong_tin = ("Nguy·ªÖn VƒÉn A", 25, "H√† N·ªôi")
print(f"Th√¥ng tin: {thong_tin}")

# C√°ch 2: Kh√¥ng c·∫ßn d·∫•u ngo·∫∑c tr√≤n (nh∆∞ng n√™n d√πng ƒë·ªÉ r√µ r√†ng)
diem_so2 = 85, 90, 78, 92, 88
print(f"ƒêi·ªÉm s·ªë 2: {diem_so2}, ki·ªÉu: {type(diem_so2)}")

# C√°ch 3: S·ª≠ d·ª•ng h√†m tuple() v·ªõi iterable
tuple_tu_list = tuple([1, 2, 3, 4, 5])
print(f"Tuple t·ª´ list: {tuple_tu_list}")

tuple_tu_chuoi = tuple("Python")
print(f"Tuple t·ª´ chu·ªói: {tuple_tu_chuoi}")

# C√°ch 4: T·∫°o tuple t·ª´ range()
tuple_tu_range = tuple(range(1, 6))
print(f"Tuple t·ª´ range: {tuple_tu_range}")
```

### 2.3. T·∫°o tuple c√≥ 1 ph·∫ßn t·ª≠ (quan tr·ªçng!)


```python
# ‚ùå SAI: ƒê√¢y KH√îNG ph·∫£i l√† tuple, m√† l√† int
sai = (5)
print(f"Gi√° tr·ªã: {sai}, ki·ªÉu: {type(sai)}")  # Ki·ªÉu: <class 'int'>

# ‚úÖ ƒê√öNG: Ph·∫£i c√≥ d·∫•u ph·∫©y sau ph·∫ßn t·ª≠
dung = (5,)
print(f"Gi√° tr·ªã: {dung}, ki·ªÉu: {type(dung)}")  # Ki·ªÉu: <class 'tuple'>

# C√°ch kh√°c (kh√¥ng d·∫•u ngo·∫∑c tr√≤n, nh∆∞ng v·∫´n c·∫ßn d·∫•u ph·∫©y)
dung2 = 5,
print(f"Gi√° tr·ªã: {dung2}, ki·ªÉu: {type(dung2)}")

# V√≠ d·ª• th·ª±c t·∫ø
ten_mon = ("To√°n",)  # Tuple 1 ph·∫ßn t·ª≠
print(f"M√¥n h·ªçc: {ten_mon}")
```

### 2.4. Tuple h·ªón h·ª£p (ch·ª©a nhi·ªÅu ki·ªÉu d·ªØ li·ªáu)


```python
# Tuple ch·ª©a nhi·ªÅu ki·ªÉu d·ªØ li·ªáu
thong_tin_hoc_sinh = ("Nguy·ªÖn VƒÉn A", 18, 8.5, True, "L·ªõp 12A1")
print(f"Th√¥ng tin h·ªçc sinh: {thong_tin_hoc_sinh}")

# Tuple ch·ª©a list v√† tuple kh√°c
thong_tin_phuc_tap = ("Python", [1, 2, 3], (4, 5, 6), {"key": "value"})
print(f"Th√¥ng tin ph·ª©c t·∫°p: {thong_tin_phuc_tap}")

# Tuple ch·ª©a tuple (nested tuple)
toa_do = ((0, 0), (10, 20), (30, 40))
print(f"T·ªça ƒë·ªô: {toa_do}")
```

## 3. Truy c·∫≠p ph·∫ßn t·ª≠ trong Tuple

### 3.1. Indexing (Truy c·∫≠p theo ch·ªâ s·ªë)


```python
diem_so = (85, 90, 78, 92, 88)

# Truy c·∫≠p theo index d∆∞∆°ng (t·ª´ tr√°i sang ph·∫£i, b·∫Øt ƒë·∫ßu t·ª´ 0)
print(f"Ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n (index 0): {diem_so[0]}")
print(f"Ph·∫ßn t·ª≠ th·ª© hai (index 1): {diem_so[1]}")
print(f"Ph·∫ßn t·ª≠ cu·ªëi c√πng (index 4): {diem_so[4]}")

# Truy c·∫≠p theo index √¢m (t·ª´ ph·∫£i sang tr√°i, b·∫Øt ƒë·∫ßu t·ª´ -1)
print(f"Ph·∫ßn t·ª≠ cu·ªëi c√πng (index -1): {diem_so[-1]}")
print(f"Ph·∫ßn t·ª≠ √°p ch√≥t (index -2): {diem_so[-2]}")
print(f"Ph·∫ßn t·ª≠ ƒë·∫ßu ti√™n (index -5): {diem_so[-5]}")

# L·ªói n·∫øu index v∆∞·ª£t qu√°
try:
    print(diem_so[10])  # IndexError
except IndexError as e:
    print(f"L·ªói: {e}")
```

### 3.2. Slicing (C·∫Øt tuple)


```python
diem_so = (85, 90, 78, 92, 88, 95, 87)

# C√∫ ph√°p: tuple[start:stop:step]

# L·∫•y ph·∫ßn t·ª≠ t·ª´ index 1 ƒë·∫øn 4 (kh√¥ng bao g·ªìm 4)
print(f"T·ª´ index 1 ƒë·∫øn 4: {diem_so[1:4]}")  # (90, 78, 92)

# L·∫•y t·ª´ ƒë·∫ßu ƒë·∫øn index 3
print(f"T·ª´ ƒë·∫ßu ƒë·∫øn index 3: {diem_so[:3]}")  # (85, 90, 78)

# L·∫•y t·ª´ index 3 ƒë·∫øn cu·ªëi
print(f"T·ª´ index 3 ƒë·∫øn cu·ªëi: {diem_so[3:]}")  # (92, 88, 95, 87)

# L·∫•y t·∫•t c·∫£
print(f"T·∫•t c·∫£: {diem_so[:]}")  # (85, 90, 78, 92, 88, 95, 87)

# B∆∞·ªõc nh·∫£y (step)
print(f"C√°ch 2 ph·∫ßn t·ª≠: {diem_so[::2]}")  # (85, 78, 88, 87)
print(f"ƒê·∫£o ng∆∞·ª£c: {diem_so[::-1]}")  # (87, 95, 88, 92, 78, 90, 85)

# Slicing v·ªõi index √¢m
print(f"3 ph·∫ßn t·ª≠ cu·ªëi: {diem_so[-3:]}")  # (88, 95, 87)
print(f"T·ª´ -5 ƒë·∫øn -2: {diem_so[-5:-2]}")  # (78, 92, 88)
```

### 3.3. Truy c·∫≠p tuple l·ªìng nhau (Nested Tuples)


```python
# Tuple ch·ª©a tuple
diem_cac_mon = (("To√°n", 8.5), ("L√Ω", 9.0), ("H√≥a", 7.5))

# Truy c·∫≠p tuple b√™n trong
print(f"Tuple ƒë·∫ßu ti√™n: {diem_cac_mon[0]}")  # ("To√°n", 8.5)

# Truy c·∫≠p ph·∫ßn t·ª≠ trong tuple l·ªìng nhau
print(f"M√¥n To√°n: {diem_cac_mon[0][0]}")  # "To√°n"
print(f"ƒêi·ªÉm To√°n: {diem_cac_mon[0][1]}")  # 8.5

# Tuple l·ªìng nhau nhi·ªÅu c·∫•p
toa_do_3D = ((0, 0, 0), (10, 20, 30), (5, 15, 25))
print(f"T·ªça ƒë·ªô X c·ªßa ƒëi·ªÉm 2: {toa_do_3D[1][0]}")  # 10
print(f"T·ªça ƒë·ªô Y c·ªßa ƒëi·ªÉm 2: {toa_do_3D[1][1]}")  # 20
print(f"T·ªça ƒë·ªô Z c·ªßa ƒëi·ªÉm 2: {toa_do_3D[1][2]}")  # 30
```

## 4. T√≠nh b·∫•t bi·∫øn (Immutable) c·ªßa Tuple

**ƒê·∫∑c ƒëi·ªÉm quan tr·ªçng nh·∫•t c·ªßa tuple**: Kh√¥ng th·ªÉ thay ƒë·ªïi sau khi t·∫°o.

### 4.1. Kh√¥ng th·ªÉ s·ª≠a ƒë·ªïi ph·∫ßn t·ª≠


```python
diem_so = (85, 90, 78, 92, 88)

# ‚ùå SAI: Kh√¥ng th·ªÉ thay ƒë·ªïi ph·∫ßn t·ª≠
try:
    diem_so[0] = 100  # TypeError
except TypeError as e:
    print(f"L·ªói: {e}")
    print("Tuple kh√¥ng cho ph√©p s·ª≠a ƒë·ªïi ph·∫ßn t·ª≠!")

# ‚úÖ Nh∆∞ng c√≥ th·ªÉ t·∫°o tuple m·ªõi t·ª´ tuple c≈©
diem_moi = (100,) + diem_so[1:]  # T·∫°o tuple m·ªõi
print(f"Tuple c≈©: {diem_so}")
print(f"Tuple m·ªõi: {diem_moi}")
```

### 4.2. Kh√¥ng th·ªÉ th√™m ho·∫∑c x√≥a ph·∫ßn t·ª≠


```python
diem_so = (85, 90, 78)

# ‚ùå SAI: Kh√¥ng c√≥ ph∆∞∆°ng th·ª©c append(), insert(), remove()
# diem_so.append(100)  # AttributeError: 'tuple' object has no attribute 'append'
# diem_so.remove(90)   # AttributeError: 'tuple' object has no attribute 'remove'

# ‚úÖ ƒê√öNG: T·∫°o tuple m·ªõi b·∫±ng c√°ch n·ªëi (concatenate)
diem_moi = diem_so + (100,)  # Th√™m ph·∫ßn t·ª≠ cu·ªëi
print(f"Tuple sau khi 'th√™m': {diem_moi}")

diem_moi2 = (95,) + diem_so  # Th√™m ph·∫ßn t·ª≠ ƒë·∫ßu
print(f"Tuple sau khi 'th√™m' ƒë·∫ßu: {diem_moi2}")

# Lo·∫°i b·ªè ph·∫ßn t·ª≠ b·∫±ng c√°ch t·∫°o tuple m·ªõi (kh√¥ng ch·ª©a ph·∫ßn t·ª≠ ƒë√≥)
diem_khong_co_90 = tuple(x for x in diem_so if x != 90)
print(f"Tuple sau khi 'x√≥a' 90: {diem_khong_co_90}")
```

### 4.3. L∆∞u √Ω v·ªÅ t√≠nh b·∫•t bi·∫øn

L∆∞u √Ω quan tr·ªçng: Tuple ch·ªâ b·∫•t bi·∫øn ·ªü c·∫•p ƒë·ªô ph·∫ßn t·ª≠, kh√¥ng ph·∫£i ·ªü n·ªôi dung c·ªßa ph·∫ßn t·ª≠ (n·∫øu ph·∫ßn t·ª≠ ƒë√≥ l√† mutable object).


```python
# Tuple ch·ª©a list (list l√† mutable)
tuple_co_list = ([1, 2, 3], [4, 5, 6])

# ‚ùå Kh√¥ng th·ªÉ thay ƒë·ªïi list (kh√¥ng th·ªÉ g√°n list m·ªõi)
# tuple_co_list[0] = [7, 8, 9]  # TypeError

# ‚úÖ NH∆ØNG c√≥ th·ªÉ thay ƒë·ªïi n·ªôi dung b√™n trong list
tuple_co_list[0].append(4)  # C√≥ th·ªÉ!
print(f"Tuple sau khi s·ª≠a list: {tuple_co_list}")

tuple_co_list[1][0] = 99  # C√≥ th·ªÉ!
print(f"Tuple sau khi s·ª≠a ph·∫ßn t·ª≠ list: {tuple_co_list}")

# V·ªõi immutable objects (nh∆∞ s·ªë, chu·ªói, tuple kh√°c), kh√¥ng th·ªÉ s·ª≠a
tuple_binh_thuong = (1, 2, 3)
# tuple_binh_thuong[0] = 10  # TypeError
```

## 5. C√°c thao t√°c v·ªõi Tuple

### 5.1. N·ªëi tuple (Concatenation)


```python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

# N·ªëi tuple b·∫±ng to√°n t·ª≠ +
tuple_ket_hop = tuple1 + tuple2
print(f"Tuple k·∫øt h·ª£p: {tuple_ket_hop}")

# N·ªëi nhi·ªÅu tuple
tuple3 = (7, 8, 9)
tuple_tat_ca = tuple1 + tuple2 + tuple3
print(f"T·∫•t c·∫£ tuple: {tuple_tat_ca}")

# L∆∞u √Ω: Ph·∫£i l√† tuple v·ªõi tuple, kh√¥ng th·ªÉ n·ªëi v·ªõi list
tuple4 = (10, 11)
# tuple_ket_hop2 = tuple1 + [12, 13]  # TypeError
```

### 5.2. L·∫∑p l·∫°i tuple (Repetition)


```python
tuple_goc = (1, 2, 3)

# L·∫∑p l·∫°i tuple b·∫±ng to√°n t·ª≠ *
tuple_lap = tuple_goc * 3
print(f"Tuple l·∫∑p l·∫°i 3 l·∫ßn: {tuple_lap}")

tuple_lap2 = (0,) * 5
print(f"Tuple (0,) l·∫∑p 5 l·∫ßn: {tuple_lap2}")
```

### 5.3. Ki·ªÉm tra ph·∫ßn t·ª≠ trong tuple


```python
diem_so = (85, 90, 78, 92, 88)

# Ki·ªÉm tra b·∫±ng to√°n t·ª≠ in v√† not in
print(f"C√≥ ƒëi·ªÉm 90 kh√¥ng? {90 in diem_so}")  # True
print(f"C√≥ ƒëi·ªÉm 100 kh√¥ng? {100 in diem_so}")  # False
print(f"Kh√¥ng c√≥ ƒëi·ªÉm 50? {50 not in diem_so}")  # True

# Ki·ªÉm tra tuple con
tuple_lon = (1, 2, 3, 4, 5)
tuple_nho = (2, 3, 4)
print(f"Tuple nh·ªè c√≥ trong tuple l·ªõn? {all(x in tuple_lon for x in tuple_nho)}")  # True
```

### 5.4. ƒê·∫øm s·ªë l∆∞·ª£ng ph·∫ßn t·ª≠


```python
diem_so = (85, 90, 78, 90, 88, 90)

# H√†m len() - t·ªïng s·ªë ph·∫ßn t·ª≠
print(f"S·ªë l∆∞·ª£ng ph·∫ßn t·ª≠: {len(diem_so)}")

# Ph∆∞∆°ng th·ª©c count() - ƒë·∫øm s·ªë l·∫ßn xu·∫•t hi·ªán
print(f"S·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa 90: {diem_so.count(90)}")
print(f"S·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa 85: {diem_so.count(85)}")
print(f"S·ªë l·∫ßn xu·∫•t hi·ªán c·ªßa 100: {diem_so.count(100)}")  # 0
```

### 5.5. T√¨m v·ªã tr√≠ ph·∫ßn t·ª≠


```python
diem_so = (85, 90, 78, 90, 88)

# Ph∆∞∆°ng th·ª©c index() - t√¨m v·ªã tr√≠ ƒë·∫ßu ti√™n
vi_tri = diem_so.index(90)
print(f"V·ªã tr√≠ ƒë·∫ßu ti√™n c·ªßa 90: {vi_tri}")

# index() v·ªõi tham s·ªë start
vi_tri2 = diem_so.index(90, 2)  # T√¨m t·ª´ index 2
print(f"V·ªã tr√≠ c·ªßa 90 t·ª´ index 2: {vi_tri2}")

# index() v·ªõi start v√† stop
vi_tri3 = diem_so.index(90, 1, 4)  # T√¨m t·ª´ index 1 ƒë·∫øn 4 (kh√¥ng bao g·ªìm 4)
print(f"V·ªã tr√≠ c·ªßa 90 trong kho·∫£ng [1, 4): {vi_tri3}")

# L·ªói n·∫øu kh√¥ng t√¨m th·∫•y
try:
    diem_so.index(100)  # ValueError
except ValueError as e:
    print(f"L·ªói: {e}")
```

### 5.6. Duy·ªát qua tuple (Iteration)


```python
diem_so = (85, 90, 78, 92, 88)

# Duy·ªát b·∫±ng for loop
print("Duy·ªát b·∫±ng for loop:")
for diem in diem_so:
    print(f"ƒêi·ªÉm: {diem}")

# Duy·ªát v·ªõi enumerate (l·∫•y c·∫£ index)
print("\nDuy·ªát v·ªõi enumerate:")
for index, diem in enumerate(diem_so):
    print(f"Index {index}: ƒêi·ªÉm {diem}")

# Duy·ªát v·ªõi range() v√† len()
print("\nDuy·ªát v·ªõi range(len()):")
for i in range(len(diem_so)):
    print(f"Index {i}: ƒêi·ªÉm {diem_so[i]}")

# Duy·ªát ng∆∞·ª£c
print("\nDuy·ªát ng∆∞·ª£c:")
for diem in reversed(diem_so):
    print(f"ƒêi·ªÉm: {diem}")

# Duy·ªát tuple l·ªìng nhau
diem_cac_mon = (("To√°n", 8.5), ("L√Ω", 9.0), ("H√≥a", 7.5))
print("\nDuy·ªát tuple l·ªìng nhau:")
for mon, diem in diem_cac_mon:
    print(f"M√¥n {mon}: ƒêi·ªÉm {diem}")
```

## 6. Unpacking Tuple (Gi·∫£i n√©n tuple)

**Unpacking** l√† k·ªπ thu·∫≠t g√°n c√°c ph·∫ßn t·ª≠ c·ªßa tuple cho nhi·ªÅu bi·∫øn c√πng l√∫c.

### 6.1. Unpacking c∆° b·∫£n


```python
# Unpacking c∆° b·∫£n
thong_tin = ("Nguy·ªÖn VƒÉn A", 25, "H√† N·ªôi")
ten, tuoi, dia_chi = thong_tin

print(f"T√™n: {ten}")
print(f"Tu·ªïi: {tuoi}")
print(f"ƒê·ªãa ch·ªâ: {dia_chi}")

# Unpacking v·ªõi to√°n t·ª≠ *
so_lieu = (1, 2, 3, 4, 5)
a, b, *phan_con_lai = so_lieu
print(f"a = {a}, b = {b}, ph·∫ßn c√≤n l·∫°i = {phan_con_lai}")

# Unpacking - ph·∫ßn ƒë·∫ßu v√†o bi·∫øn, ph·∫ßn cu·ªëi v√†o bi·∫øn
diem_so = (85, 90, 78, 92, 88)
diem_dau, *diem_giua, diem_cuoi = diem_so
print(f"ƒêi·ªÉm ƒë·∫ßu: {diem_dau}, ƒêi·ªÉm gi·ªØa: {diem_giua}, ƒêi·ªÉm cu·ªëi: {diem_cuoi}")

# Unpacking - b·ªè qua ph·∫ßn t·ª≠ b·∫±ng _
diem_so2 = (85, 90, 78)
diem_dau2, _, diem_cuoi2 = diem_so2
print(f"ƒêi·ªÉm ƒë·∫ßu: {diem_dau2}, ƒêi·ªÉm cu·ªëi: {diem_cuoi2}")
```

### 6.2. Unpacking khi g·ªçi h√†m


```python
def tinh_tong(a, b, c):
    """T√≠nh t·ªïng 3 s·ªë"""
    return a + b + c

# Truy·ªÅn tuple v√†o h√†m b·∫±ng unpacking
so_lieu = (10, 20, 30)
tong = tinh_tong(*so_lieu)  # Unpacking v·ªõi *
print(f"T·ªïng: {tong}")

# T∆∞∆°ng ƒë∆∞∆°ng v·ªõi
tong2 = tinh_tong(10, 20, 30)
print(f"T·ªïng 2: {tong2}")

# Unpacking khi g·ªçi h√†m v·ªõi nhi·ªÅu tham s·ªë
def in_thong_tin(ten, tuoi, dia_chi, email):
    print(f"T√™n: {ten}, Tu·ªïi: {tuoi}, ƒê·ªãa ch·ªâ: {dia_chi}, Email: {email}")

thong_tin_day_du = ("Nguy·ªÖn VƒÉn A", 25, "H√† N·ªôi", "nguyenvana@example.com")
in_thong_tin(*thong_tin_day_du)
```

### 6.3. Unpacking tuple l·ªìng nhau

### 6.4. Ho√°n ƒë·ªïi gi√° tr·ªã (Swapping)

Unpacking r·∫•t h·ªØu √≠ch cho vi·ªác ho√°n ƒë·ªïi gi√° tr·ªã hai bi·∫øn.


```python
# Ho√°n ƒë·ªïi gi√° tr·ªã hai bi·∫øn
a = 10
b = 20

print(f"Tr∆∞·ªõc khi ho√°n ƒë·ªïi: a = {a}, b = {b}")

# C√°ch 1: D√πng tuple (Pythonic way)
a, b = b, a

print(f"Sau khi ho√°n ƒë·ªïi: a = {a}, b = {b}")

# Ho√°n ƒë·ªïi nhi·ªÅu bi·∫øn
x, y, z = 1, 2, 3
print(f"Tr∆∞·ªõc: x = {x}, y = {y}, z = {z}")

x, y, z = z, x, y
print(f"Sau: x = {x}, y = {y}, z = {z}")
```

### 9.3. all() v√† any()

```python
tuple1 = (1, 2, 3)
tuple2 = (1, 2, 3)
tuple3 = (1, 2, 4)

# So s√°nh b·∫±ng
print(f"tuple1 == tuple2: {tuple1 == tuple2}")  # True
print(f"tuple1 == tuple3: {tuple1 == tuple3}")  # False

# So s√°nh kh√°c
print(f"tuple1 != tuple3: {tuple1 != tuple3}")  # True

# So s√°nh v·ªõi list (False v√¨ kh√°c ki·ªÉu)
list1 = [1, 2, 3]
print(f"tuple1 == list1: {tuple1 == list1}")  # False
```

## 11. ·ª®ng d·ª•ng th·ª±c t·∫ø c·ªßa Tuple

### 11.1. Tr·∫£ v·ªÅ nhi·ªÅu gi√° tr·ªã t·ª´ h√†m


```python
def tinh_diem_trung_binh(diem_so):
    """T√≠nh ƒëi·ªÉm trung b√¨nh, cao nh·∫•t, th·∫•p nh·∫•t"""
    tb = sum(diem_so) / len(diem_so)
    cao_nhat = max(diem_so)
    thap_nhat = min(diem_so)
    return (tb, cao_nhat, thap_nhat)  # Tr·∫£ v·ªÅ tuple

# S·ª≠ d·ª•ng
diem = (85, 90, 78, 92, 88)
tb, cao, thap = tinh_diem_trung_binh(diem)  # Unpacking
print(f"ƒêi·ªÉm trung b√¨nh: {tb:.2f}")
print(f"ƒêi·ªÉm cao nh·∫•t: {cao}")
print(f"ƒêi·ªÉm th·∫•p nh·∫•t: {thap}")

# Ho·∫∑c gi·ªØ nguy√™n tuple
ket_qua = tinh_diem_trung_binh(diem)
print(f"K·∫øt qu·∫£ (tuple): {ket_qua}")
```

### 11.2. L∆∞u tr·ªØ t·ªça ƒë·ªô (Coordinates)


```python
# T·ªça ƒë·ªô 2D
diem_A = (10, 20)
diem_B = (30, 40)

def tinh_khoang_cach(p1, p2):
    """T√≠nh kho·∫£ng c√°ch gi·ªØa hai ƒëi·ªÉm"""
    x1, y1 = p1
    x2, y2 = p2
    return ((x2 - x1)**2 + (y2 - y1)**2)**0.5

khoang_cach = tinh_khoang_cach(diem_A, diem_B)
print(f"Kho·∫£ng c√°ch gi·ªØa A{diem_A} v√† B{diem_B}: {khoang_cach:.2f}")

# T·ªça ƒë·ªô 3D
diem_3D = (10, 20, 30)
x, y, z = diem_3D
print(f"T·ªça ƒë·ªô 3D: x={x}, y={y}, z={z}")
```

```python
# ‚úÖ ƒê√∫ng - s·ªë l∆∞·ª£ng kh·ªõp
a, b, c = (1, 2, 3)

# ‚ùå Sai - s·ªë l∆∞·ª£ng kh√¥ng kh·ªõp
try:
    x, y = (1, 2, 3)  # ValueError: too many values to unpack
except ValueError as e:
    print(f"L·ªói: {e}")

# ‚úÖ D√πng * ƒë·ªÉ gi·∫£i quy·∫øt
x, *y = (1, 2, 3, 4, 5)
print(f"x = {x}, y = {y}")  # x = 1, y = [2, 3, 4, 5]

*a, b, c = (1, 2, 3, 4, 5)
print(f"a = {a}, b = {b}, c = {c}")  # a = [1, 2, 3], b = 4, c = 5
```
